```python
import torch.nn as nn
import torch

class CNNBlock(nn.Module):
    def __init__(self, in_channels=4, out_channels=96):
        super().__init__()
        # The CNN structure is from your TelomereEncoder's cnn block
        self.cnn = nn.Sequential(
            nn.Conv1d(in_channels, 64, kernel_size=5, padding=2),
            nn.GELU(),
            nn.Conv1d(64, out_channels, kernel_size=5, padding=2),
            nn.GELU()
        )

    def forward(self, x):
        # x is expected to be [batch_size, in_channels, sequence_length]
        # If your input is [batch_size, sequence_length, in_channels], you'll need to permute it before passing to this block.
        return self.cnn(x)

# Example of how to use it:
# cnn_block_instance = CNNBlock(in_channels=4, out_channels=96)
# dummy_input = torch.randn(1, 4, 100) # Batch_size, Channels, Sequence_length
# output = cnn_block_instance(dummy_input)
# print(output.shape)
```

```python
import torch
import torch.nn as nn

# Ensure TelomereEncoder is defined (it's in cell _7zXeY9v1Hsv)
# If it's not run, you might need to run cell _7zXeY9v1Hsv first

# Define the VirtualCellModel as 'model'
# Assuming TelomereEncoder and other components are already defined or can be imported/run.
# For this example, I'll instantiate a placeholder TelomereEncoder.
# In a real scenario, you'd use the actual TelomereEncoder from your code.

# Placeholder for TelomereEncoder if it's not run yet in current session
class TelomereEncoder(nn.Module):
    def __init__(self, d_model=96, num_heads=6, depth=2):
        super().__init__()
        self.cnn = nn.Sequential(
            nn.Conv1d(4, 64, kernel_size=5, padding=2),
            nn.GELU(),
            nn.Conv1d(64, d_model, kernel_size=5, padding=2),
            nn.GELU()
        )
        encoder_layer = nn.TransformerEncoderLayer(d_model=d_model, nhead=num_heads, batch_first=True)
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=depth)
        self.d_model = d_model

    def forward(self, x):
        x = x.permute(0, 2, 1)
        x = self.cnn(x)
        x = x.permute(0, 2, 1)
        x = self.transformer(x)
        return x

# Placeholder for TelomeraseController
class TelomeraseController(nn.Module):
    def __init__(self, input_dim=96, output_dim=1):
        super().__init__()
        self.net = nn.Sequential(nn.Linear(input_dim, 64), nn.GELU(), nn.Linear(64, output_dim), nn.Sigmoid())
    def forward(self, x):
        # For this example, we'll assume the encoder output is pooled to match input_dim
        if x.dim() == 3: # if it's [B, L, D]
            x = x.mean(dim=1) # Pool to [B, D]
        return self.net(x)


class VirtualCellModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = TelomereEncoder(d_model=96) # Use default d_model from TelomereEncoder
        encoder_output_dim = self.encoder.d_model # Output dimension after pooling/mean

        # Changed output dimensions of heads to 1 to match the 5-task loss
        self.telomere_head = nn.Linear(encoder_output_dim, 1)
        self.telomerase_head = nn.Linear(encoder_output_dim, 1)
        self.phago_head = nn.Linear(encoder_output_dim, 1)
        self.stemcell_head = nn.Linear(encoder_output_dim, 1)
        # TelomeraseController already outputs 1, ensure its input_dim matches encoder output
        self.controller = TelomeraseController(input_dim=encoder_output_dim)

    def forward(self, x):
        h = self.encoder(x)
        # Pool the sequence dimension for the heads
        if h.dim() == 3:
            h_pooled = h.mean(dim=1) # [B, D]
        else:
            h_pooled = h

        # Collect predictions from each head (each will be [batch_size, 1])
        telo_pred = self.telomere_head(h_pooled)
        telomerase_pred = self.telomerase_head(h_pooled)
        phago_pred = self.phago_head(h_pooled)
        stem_pred = self.stemcell_head(h_pooled)
        controller_signal_pred = self.controller(h)

        # Concatenate all predictions into a single tensor of shape [batch_size, 5]
        combined_preds = torch.cat([telo_pred, telomerase_pred, phago_pred, stem_pred, controller_signal_pred], dim=1)

        return combined_preds

model = VirtualCellModel()

# Define EPOCHS
EPOCHS = 10

# Define train_loader (dummy data for demonstration)
# Assuming input shape for encoder is [batch_size, sequence_length, in_channels (4)]
batch_size = 32
sequence_length = 100
in_channels = 4

dummy_data = torch.randn(1000, sequence_length, in_channels) # N, L, C
dummy_labels = torch.randint(0, 5, (1000, 5)).float() # Dummy labels for 5 tasks

dummy_dataset = torch.utils.data.TensorDataset(dummy_data, dummy_labels)
train_loader = torch.utils.data.DataLoader(dummy_dataset, batch_size=batch_size, shuffle=True)
```

```python
import torch
import torch.nn as nn

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Define learnable log_vars for 5 tasks, initialized on the determined device
log_vars = torch.nn.Parameter(torch.zeros(5, device=device))

def compute_loss(preds, targets, log_vars=log_vars):
    # This function assumes 'preds' and 'targets' are tensors of shape [batch_size, 5],
    # where each column corresponds to a task.
    per_task_mse = (preds - targets).pow(2).mean(dim=0)  # [5]
    loss = 0.0
    for i in range(5):
        s = log_vars[i]
        loss = loss + 0.5 * (torch.exp(-s) * per_task_mse[i] + s)
    return loss
```

```python
# define in model or training wrapper
log_vars = torch.nn.Parameter(torch.zeros(5, device=device))  # 5 tasks

def multitask_loss(preds, targets, log_vars):
    # preds, targets: [B, 5]
    losses = (preds - targets).pow(2).mean(dim=0)  # per-task MSE
    weighted = 0.0
    for i in range(len(losses)):
        s = log_vars[i]
        weighted = weighted + 0.5 * (torch.exp(-s) * losses[i] + s)
    return weighted
```

```python
# ----------------------------
# Contrastive pretraining (SimCLR) for telomere sequences
# ----------------------------
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import random
import numpy as np

# --- Augmentations for one-hot sequences ---
def mutate_seq_onehot(onehot, p=0.04):
    # onehot: np.array [L,4]
    res = onehot.copy()
    L = res.shape[0]
    for i in range(L):
        if random.random() < p:
            # replace with a random nucleotide one-hot
            idx = random.randint(0,3)
            res[i,:] = 0.0
            res[i, idx] = 1.0
    return res

def insert_random_motif(onehot, motif_len=6, p=0.08):
    if random.random() > p: return onehot
    L = onehot.shape[0]
    pos = random.randint(0, max(0, L-motif_len))
    motif = np.zeros((motif_len,4), dtype=np.float32)
    for i in range(motif_len):
        motif[i, random.randint(0,3)] = 1.0
    out = onehot.copy()
    out[pos:pos+motif_len] = motif
    return out

def crop_seq(onehot, min_frac=0.5):
    L = onehot.shape[0]
    newL = random.randint(int(L*min_frac), L)
    start = random.randint(0, L-newL)
    out = np.zeros_like(onehot)
    out[:newL] = onehot[start:start+newL]
    return out

def kmask(onehot, k=4, p=0.05):
    res = onehot.copy()
    L = res.shape[0]
    for i in range(L-k+1):
        if random.random() < p:
            res[i:i+k] = 0.25  # soften to avoid all-zero
    return res

def random_augment(onehot):
    # compose 2 augmentations
    aug = onehot.copy()
    choices = [mutate_seq_onehot, insert_random_motif, crop_seq, kmask]
    funcs = random.sample(choices, 2)
    for f in funcs:
        aug = f(aug)
    return aug

# --- Contrastive Dataset: returns two views ---
class ContrastiveTelomereDataset(Dataset):
    def __init__(self, seq_tensor, context=None):
        # seq_tensor: np.array [N, L, 4]
        self.seq = seq_tensor
        self.context = context
    def __len__(self):
        return len(self.seq)
    def __getitem__(self, idx):
        s = self.seq[idx]
        v1 = random_augment(s)
        v2 = random_augment(s)
        # return views as float32 tensors
        return torch.tensor(v1, dtype=torch.float32), torch.tensor(v2, dtype=torch.float32)

# --- Projector head ---
class Projector(nn.Module):
    def __init__(self, input_dim, hidden_dim=256, out_dim=128):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.GELU(),
            nn.Linear(hidden_dim, out_dim)
        )
    def forward(self, x):
        return self.net(x)

# --- Contrastive trainer (SimCLR) ---
def info_nce_loss(z_i, z_j, temperature=0.1):
    # z_i, z_j: [B, D], normalized
    batch_size = z_i.shape[0]
    z = torch.cat([z_i, z_j], dim=0)  # [2B, D]
    sim = torch.matmul(z, z.T)  # [2B, 2B]
    sim = sim / temperature
    # mask out self-similarity
    labels = torch.arange(batch_size).to(z.device)
    labels = torch.cat([labels, labels], dim=0)
    # numerator: similarity between positive pairs
    exp_sim = torch.exp(sim)
    # mask to remove i==j
    mask = (~torch.eye(2*batch_size, dtype=torch.bool, device=z.device)).float()
    exp_sim = exp_sim * mask
    # for each i, positive is at index i+batch_size (if first half) or i-batch_size
    positives = torch.exp( (z_i * z_j).sum(dim=-1) / temperature )
    # compute denominators
    denom = exp_sim.sum(dim=1)
    # numerator for each sample: positive (two halves)
    num = torch.cat([positives, positives], dim=0)
    loss = - torch.log(num / denom)
    return loss.mean()

# --- Integrate with your encoder ---
# Assume `encoder` is your CNN+Transformer that maps [B, L, 4] -> [B, emb_dim]
# If your encoder expects [B, L, 4], ensure it's implemented accordingly.

# Example small training loop:
def pretrain_contrastive(encoder, projector, dataset, device,
                         epochs=30, batch_size=128, lr=1e-3, tau=0.1):
    loader = DataLoader(dataset, batch_size=batch_size, shuffle=True, drop_last=True)
    encoder.to(device)
    projector.to(device)
    # train both encoder and projector
    params = list(encoder.parameters()) + list(projector.parameters())
    optimizer = torch.optim.AdamW(params, lr=lr, weight_decay=1e-5)
    scaler = torch.cuda.amp.GradScaler()
    for epoch in range(epochs):
        epoch_loss = 0.0
        for v1, v2 in loader:
            v1 = v1.to(device); v2 = v2.to(device)
            optimizer.zero_grad()
            with torch.cuda.amp.autocast():
                h1 = encoder(v1)  # [B, emb]
                h2 = encoder(v2)
                # pool: mean pooling -> vector
                # assume encoder returns [B, emb], if returns [B, L, d] pool avg
                if h1.dim()==3:
                    h1 = h1.mean(dim=1)
                    h2 = h2.mean(dim=1)
                z1 = projector(h1)
                z2 = projector(h2)
                z1 = F.normalize(z1, dim=1)
                z2 = F.normalize(z2, dim=1)
                loss = info_nce_loss(z1, z2, temperature=tau)
            scaler.scale(loss).backward()
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(params, 1.0)
            scaler.step(optimizer)
            scaler.update()
            epoch_loss += loss.item() * v1.size(0)
        epoch_loss = epoch_loss / len(loader.dataset)
        print(f"[Contrastive] Epoch {epoch+1}/{epochs}  Loss: {epoch_loss:.4f}")
    # return encoder with pretrained weights (and projector)
    return encoder, projector

# ------------------------------
# How to call (example)
# ------------------------------
# from earlier code: seq_tensors shape [N, L, 4] numpy array
# dataset = ContrastiveTelomereDataset(seq_tensors)
# encoder = your_existing_encoder_model  # must accept [B, L, 4]
# projector = Projector(input_dim=encoder_output_dim)
# device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
# encoder, projector = pretrain_contrastive(encoder, projector, dataset, device,
#                                           epochs=40, batch_size=128, lr=3e-3, tau=0.1)
```

```python
# =============================================
# Embedding Evaluation Module (kNN + Linear Probe)
# =============================================
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import accuracy_score, f1_score, mean_absolute_error, r2_score
import numpy as np

# ---------------------------------------------
# Dataset wrapper for evaluation
# ---------------------------------------------
class EmbeddingEvalDataset(Dataset):
    """
    seqs: numpy array [N, L, 4]
    labels: numpy array [N] | float or int
    """
    def __init__(self, seqs, labels):
        self.seqs = seqs
        self.labels = labels
    def __len__(self):
        return len(self.seqs)
    def __getitem__(self, idx):
        x = torch.tensor(self.seqs[idx], dtype=torch.float32)
        y = torch.tensor(self.labels[idx])
        return x, y

# ---------------------------------------------
# Extract embeddings from encoder
# ---------------------------------------------
def extract_embeddings(encoder, dataloader, device):
    encoder.eval()
    all_embs = []
    all_labels = []
    with torch.no_grad():
        for x, y in dataloader:
            x = x.to(device)
            h = encoder(x)
            # if encoder outputs [B, L, D], pool:
            if h.dim() == 3:
                h = h.mean(dim=1)
            all_embs.append(h.cpu().numpy())
            all_labels.append(y.cpu().numpy())
    embs = np.concatenate(all_embs, axis=0)
    labels = np.concatenate(all_labels, axis=0)
    return embs, labels

# ---------------------------------------------
# k-NN Evaluation (highly diagnostic)
# ---------------------------------------------
def evaluate_knn(embs_train, y_train, embs_test, y_test, k=5, metric='cosine'):
    knn = KNeighborsClassifier(n_neighbors=k, metric=metric)
    knn.fit(embs_train, y_train)
    preds = knn.predict(embs_test)
    acc = accuracy_score(y_test, preds)
    f1 = f1_score(y_test, preds, average='weighted')
    return {"knn_acc": acc, "knn_f1": f1}

# ---------------------------------------------
# Linear Probe (classification)
# ---------------------------------------------
def evaluate_linear_probe_cls(embs_train, y_train, embs_test, y_test):
    # Logistic regression on top of frozen embeddings
    clf = LogisticRegression(max_iter=2000)
    clf.fit(embs_train, y_train)
    preds = clf.predict(embs_test)
    acc = accuracy_score(y_test, preds)
    f1 = f1_score(y_test, preds, average='weighted')
    return {"linear_acc": acc, "linear_f1": f1}

# ---------------------------------------------
# Linear Probe (regression)
# ---------------------------------------------
def evaluate_linear_probe_reg(embs_train, y_train, embs_test, y_test):
    reg = LinearRegression()
    reg.fit(embs_train, y_train)
    preds = reg.predict(embs_test)
    mae = mean_absolute_error(y_test, preds)
    r2 = r2_score(y_test, preds)
    return {"linear_mae": mae, "linear_r2": r2}

# ---------------------------------------------
# Full evaluation pipeline
# ---------------------------------------------
def evaluate_embeddings(
    encoder,
    seq_train, y_train,
    seq_test, y_test,
    device,
    task_type="classification",
    k=5
):
    # build datasets/loaders
    train_ds = EmbeddingEvalDataset(seq_train, y_train)
    test_ds  = EmbeddingEvalDataset(seq_test, y_test)

    train_loader = DataLoader(train_ds, batch_size=128, shuffle=False)
    test_loader  = DataLoader(test_ds, batch_size=128, shuffle=False)

    # extract embeddings
    print("Extracting embeddings...")
    emb_train, y_train = extract_embeddings(encoder, train_loader, device)
    emb_test,  y_test  = extract_embeddings(encoder, test_loader,  device)

    results = {}

    # 1. kNN
    print("Evaluating kNN...")
    knn_res = evaluate_knn(emb_train, y_train, emb_test, y_test, k=k)
    results.update(knn_res)

    # 2. Linear Probe
    print("Evaluating Linear Probe...")
    if task_type == "classification":
        lp_res = evaluate_linear_probe_cls(emb_train, y_train, emb_test, y_test)
    else:
        lp_res = evaluate_linear_probe_reg(emb_train, y_train, emb_test, y_test)
    results.update(lp_res)

    return results
```

```python
# telomere_pipeline_full.py / Notebook

import os
import random
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, random_split
from sklearn.model_selection import train_test_split
from Bio import SeqIO
import matplotlib.pyplot as plt
import umap
from sklearn.manifold import TSNE
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import accuracy_score, f1_score, mean_absolute_error, r2_score

# -----------------------------
# 1) CONFIGURACIN Y PARMETROS
# -----------------------------
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Paths para datos
SYN_SEQ_PATH = "synthetic_telomeres.npy"           # tu secuencias sint茅ticas one-hot
SYN_LABEL_PATH = "synthetic_labels.npy"             # tus etiquetas sint茅ticas
CHM13_TEL_FASTA = "CHM13_telomeres.fasta"            # fasta con regiones telom茅ricas reales

# Hyperpar谩metros contrastivo
CONTRAST_EPOCHS = 30
CONTRAST_BATCH = 128
CONTRAST_LR = 3e-3
TEMPERATURE = 0.1

# Fine-tuning multitarea
FT_EPOCHS = 20
FT_BATCH = 64
LR_ENCODER = 1e-4
LR_HEADS = 1e-3

# -----------------------------
# 2) FUNCIONES PARA AUGMENTACIN
# -----------------------------
def mutate_onehot(onehot, p=0.04):
    res = onehot.copy()
    L = res.shape[0]
    for i in range(L):
        if random.random() < p:
            idx = random.randint(0, 3)
            res[i, :] = 0.0
            res[i, idx] = 1.0
    return res

def insert_motif(onehot, motif_len=6, p=0.08):
    if random.random() > p:
        return onehot
    L = onehot.shape[0]
    pos = random.randint(0, max(0, L - motif_len))
    motif = np.zeros((motif_len, 4), dtype=np.float32)
    for i in range(motif_len):
        motif[i, random.randint(0, 3)] = 1.0
    out = onehot.copy()
    out[pos:pos + motif_len] = motif
    return out

def crop_onehot(onehot, min_frac=0.5):
    L = onehot.shape[0]
    newL = random.randint(int(L * min_frac), L)
    start = random.randint(0, L - newL)
    out = np.zeros_like(onehot)
    out[:newL] = onehot[start:start + newL]
    return out

def kmask(onehot, k=4, p=0.05):
    res = onehot.copy()
    L = res.shape[0]
    for i in range(L - k + 1):
        if random.random() < p:
            res[i:i + k] = 0.25
    return res

def random_augment(onehot):
    funcs = [mutate_onehot, insert_motif, crop_onehot, kmask]
    f1, f2 = random.sample(funcs, 2)
    aug = f1(onehot)
    aug = f2(aug)
    return aug

# -----------------------------
# 3) DATASETS
# -----------------------------
class ContrastiveTelomereDataset(Dataset):
    def __init__(self, seqs):
        self.seqs = seqs

    def __len__(self):
        return len(self.seqs)

    def __getitem__(self, idx):
        s = self.seqs[idx]
        v1 = random_augment(s)
        v2 = random_augment(s)
        return torch.tensor(v1, dtype=torch.float32), torch.tensor(v2, dtype=torch.float32)

class EvalDataset(Dataset):
    def __init__(self, seqs, labels):
        self.seqs = seqs
        self.labels = labels

    def __len__(self):
        return len(self.seqs)

    def __getitem__(self, idx):
        x = torch.tensor(self.seqs[idx], dtype=torch.float32)
        y = self.labels[idx]
        return x, y

# -----------------------------
# 4) MODELO: ENCODER + PROYECTOR + MULTITAREA
# -----------------------------
class TelomereEncoder(nn.Module):
    def __init__(self, d_model=96, num_heads=6, depth=2):
        super().__init__()
        self.cnn = nn.Sequential(
            nn.Conv1d(4, 64, kernel_size=5, padding=2),
            nn.GELU(),
            nn.Conv1d(64, d_model, kernel_size=5, padding=2),
            nn.GELU()
        )
        encoder_layer = nn.TransformerEncoderLayer(d_model=d_model, nhead=num_heads, batch_first=True)
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=depth)
        self.d_model = d_model

    def forward(self, x):
        # x: [B, L, 4]
        x = x.permute(0, 2, 1)  #  [B, 4, L]
        x = self.cnn(x)         # [B, d_model, L]
        x = x.permute(0, 2, 1)  # [B, L, d_model]
        x = self.transformer(x) # [B, L, d_model]
        return x  # secuencia embebida

class Projector(nn.Module):
    def __init__(self, input_dim, hidden_dim=256, out_dim=128):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.GELU(),
            nn.Linear(hidden_dim, out_dim)
        )

    def forward(self, x):
        return self.net(x)

class MultiTaskModel(nn.Module):
    def __init__(self, encoder, emb_dim):
        super().__init__()
        self.encoder = encoder
        self.stability_head = nn.Linear(emb_dim, 3)       # ejemplo: 3 clases
        self.telomerase_head = nn.Linear(emb_dim, 1)      # regresi贸n afinidad
        self.length_head = nn.Linear(emb_dim, 1)          # longitud real

    def forward(self, x):
        h = self.encoder(x)
        # pool over length
        if h.dim() == 3:
            h = h.mean(dim=1)
        return {
            "stability": self.stability_head(h),
            "telomerase": self.telomerase_head(h).squeeze(-1),
            "length": self.length_head(h).squeeze(-1)
        }

# -----------------------------
# 5) CONTRASTIVE PRDIDA
# -----------------------------
def info_nce_loss(z1, z2, temperature=0.1):
    batch_size = z1.size(0)
    z = torch.cat([z1, z2], dim=0)  # [2B, D]
    sim = torch.matmul(z, z.T) / temperature
    # Mascar self
    mask = (~torch.eye(2*batch_size, dtype=torch.bool, device=z.device)).float()
    exp_sim = torch.exp(sim) * mask
    denom = exp_sim.sum(dim=1)
    positives = torch.exp((z1 * z2).sum(dim=1) / temperature)
    positives = torch.cat([positives, positives], dim=0)
    loss = -torch.log(positives / denom)
    return loss.mean()

# -----------------------------
# 6) CARGA DE DATOS
# -----------------------------
# 6.1 Cargar secuencias sint茅ticas
seq_synth = np.load(SYN_SEQ_PATH)  # shape [N, L, 4]
labels_synth = np.load(SYN_LABEL_PATH)

# 6.2 Cargar secuencias reales desde FASTA
seq_real_list = []
for rec in SeqIO.parse(CHM13_TEL_FASTA, "fasta"):
    seq = rec.seq
    L = len(seq)
    oh = np.zeros((L, 4), dtype=np.float32)
    mapping = {'A':0, 'C':1, 'G':2, 'T':3}
    for i, b in enumerate(seq):
        if b in mapping:
            oh[i, mapping[b]] = 1.0
    seq_real_list.append(oh)
seq_real = np.array(seq_real_list, dtype=object)  # secuencias de largo variable

# Para simplificar, recortamos/padding al m谩ximo largo entre todas
max_L = max([s.shape[0] for s in seq_real] + [seq_synth.shape[1]])
def pad_onehot(arr, L):
    out = np.zeros((L, 4), dtype=np.float32)
    out[:arr.shape[0], :] = arr
    return out

seq_real_padded = np.stack([pad_onehot(s, max_L) for s in seq_real])
seq_synth_padded = np.stack([pad_onehot(s, max_L) for s in seq_synth])

# 6.3 Combinar
seq_all = np.concatenate([seq_synth_padded, seq_real_padded], axis=0)
labels_all = np.concatenate([labels_synth, np.zeros(len(seq_real_padded))], axis=0)

# 6.4 Crear datasets
contrast_ds = ContrastiveTelomereDataset(seq_all)
train_eval_seq, test_eval_seq, train_eval_labels, test_eval_labels = train_test_split(
    seq_all, labels_all, test_size=0.2, random_state=SEED
)
eval_train_ds = EvalDataset(train_eval_seq, train_eval_labels)
eval_test_ds = EvalDataset(test_eval_seq, test_eval_labels)

# -----------------------------
# 7) PREENTRENAMIENTO CONTRASTIVO
# -----------------------------
encoder = TelomereEncoder().to(DEVICE)
projector = Projector(input_dim=encoder.d_model, hidden_dim=256, out_dim=128).to(DEVICE)

contrast_loader = DataLoader(contrast_ds, batch_size=CONTRAST_BATCH, shuffle=True, drop_last=True)
opt_con = torch.optim.AdamW(list(encoder.parameters()) + list(projector.parameters()), lr=CONTRAST_LR, weight_decay=1e-5)

for epoch in range(CONTRAST_EPOCHS):
    encoder.train()
    projector.train()
    running_loss = 0.0
    for v1, v2 in contrast_loader:
        v1 = v1.to(DEVICE); v2 = v2.to(DEVICE)
        opt_con.zero_grad()
        h1 = encoder(v1)
        h2 = encoder(v2)
        # pool
        h1p = h1.mean(dim=1)
        h2p = h2.mean(dim=1)
        z1 = projector(h1p)
        z2 = projector(h2p)
        z1 = F.normalize(z1, dim=1)
        z2 = F.normalize(z2, dim=1)
        loss = info_nce_loss(z1, z2, TEMPERATURE)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(list(encoder.parameters()) + list(projector.parameters()), 1.0)
        opt_con.step()
        running_loss += loss.item() * v1.size(0)
    epoch_loss = running_loss / len(contrast_loader.dataset)
    print(f"[Pretrain] Epoch {epoch+1}/{CONTRAST_EPOCHS} - Loss: {epoch_loss:.4f}")

# -----------------------------
# 8) EVALUACIN DE EMBEDDINGS (k-NN + Linear Probe)
# -----------------------------
def extract_embeddings(encoder, loader):
    encoder.eval()
    embs = []
    labs = []
    with torch.no_grad():
        for x, y in loader:
            x = x.to(DEVICE)
            h = encoder(x)
            h = h.mean(dim=1)
            embs.append(h.cpu().numpy())
            labs.append(y.numpy())
    return np.concatenate(embs, axis=0), np.concatenate(labs, axis=0)

train_loader_eval = DataLoader(eval_train_ds, batch_size=FT_BATCH, shuffle=False)
test_loader_eval = DataLoader(eval_test_ds, batch_size=FT_BATCH, shuffle=False)

emb_train, y_train = extract_embeddings(encoder, train_loader_eval)
emb_test, y_test = extract_embeddings(encoder, test_loader_eval)

# k-NN
knn = KNeighborsClassifier(n_neighbors=5, metric='cosine')
knn.fit(emb_train, y_train)
y_pred_knn = knn.predict(emb_test)
knn_acc = accuracy_score(y_test, y_pred_knn)
knn_f1 = f1_score(y_test, y_pred_knn, average='weighted')

# Linear probe (regresi贸n si las etiquetas son continuas, ac谩 supongo clasificaci贸n)
lin = LogisticRegression(max_iter=2000)
lin.fit(emb_train, y_train)
y_pred_lin = lin.predict(emb_test)
lin_acc = accuracy_score(y_test, y_pred_lin)
lin_f1 = f1_score(y_test, y_pred_lin, average='weighted')

print("k-NN Acc:", knn_acc, "F1:", knn_f1)
print("Linear Probe Acc:", lin_acc, "F1:", lin_f1)

# -----------------------------
# 9) VISUALIZACIN DE EMBEDDINGS (UMAP & t-SNE)
# -----------------------------
reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, metric='cosine', random_state=SEED)
emb_2d = reducer.fit_transform(np.concatenate([emb_train, emb_test], axis=0))
labels_2d = np.concatenate([y_train, y_test], axis=0)

plt.figure(figsize=(8,6))
plt.scatter(emb_2d[:,0], emb_2d[:,1], c=labels_2d, cmap="Spectral", s=10)
plt.colorbar()
plt.title("UMAP de embebidos telom茅ricos")
plt.xlabel("UMAP1"); plt.ylabel("UMAP2")
plt.show()

tsne = TSNE(n_components=2, perplexity=30, n_iter=1000, metric='cosine', random_state=SEED)
emb_tsne = tsne.fit_transform(np.concatenate([emb_train, emb_test], axis=0))
plt.figure(figsize=(8,6))
plt.scatter(emb_tsne[:,0], emb_tsne[:,1], c=labels_2d, cmap="Spectral", s=10)
plt.colorbar()
plt.title("t-SNE de embebidos telom茅ricos")
plt.xlabel("t-SNE1"); plt.ylabel("t-SNE2")
plt.show()

# -----------------------------
# 10) FINETUNING MULTITAREA
# -----------------------------
# Preparar dataset para fine-tune
class FT_Dataset(Dataset):
    def __init__(self, seqs, labels_stab, labels_bind, labels_len):
        self.seqs = seqs
        self.labels_stab = labels_stab
        self.labels_bind = labels_bind
        self.labels_len = labels_len

    def __len__(self):
        return len(self.seqs)

    def __getitem__(self, idx):
        x = torch.tensor(self.seqs[idx], dtype=torch.float32)
        return (x,
                torch.tensor(self.labels_stab[idx], dtype=torch.long),
                torch.tensor(self.labels_bind[idx], dtype=torch.float32),
                torch.tensor(self.labels_len[idx], dtype=torch.float32))

# Aqu铆 deber铆as tener etiquetas reales o sint茅ticas para tus tres tareas:
# - stability (ej: clases 0,1,2)
# - telomerase affinity (regresi贸n)
# - length (regresi贸n)
# Por ahora genero placeholders con ceros:
labels_stab = np.zeros(len(seq_all), dtype=int)
labels_bind = np.zeros(len(seq_all), dtype=float)
labels_len = np.zeros(len(seq_all), dtype=float)

ft_ds = FT_Dataset(seq_all, labels_stab, labels_bind, labels_len)
train_ds, val_ds = random_split(ft_ds, [int(len(ft_ds)*0.8), len(ft_ds) - int(len(ft_ds)*0.8)])

train_loader = DataLoader(train_ds, batch_size=FT_BATCH, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=FT_BATCH, shuffle=False)

# Instanciar modelo
mt_model = MultiTaskModel(encoder, emb_dim=encoder.d_model).to(DEVICE)
optimizer = torch.optim.AdamW([
    {'params': mt_model.encoder.parameters(), 'lr': LR_ENCODER},
    {'params': mt_model.stability_head.parameters(), 'lr': LR_HEADS},
    {'params': mt_model.telomerase_head.parameters(), 'lr': LR_HEADS},
    {'params': mt_model.length_head.parameters(), 'lr': LR_HEADS},
], weight_decay=1e-5)
criterion_stab = nn.CrossEntropyLoss()
criterion_reg = nn.MSELoss()

# Entrenamiento
for epoch in range(FT_EPOCHS):
    mt_model.train()
    total_loss = 0.0
    for x, y_stab, y_bind, y_len in train_loader:
        x = x.to(DEVICE)
        y_stab = y_stab.to(DEVICE)
        y_bind = y_bind.to(DEVICE)
        y_len = y_len.to(DEVICE)

        out = mt_model(x)
        loss1 = criterion_stab(out["stability"], y_stab)
        loss2 = criterion_reg(out["telomerase"], y_bind)
        loss3 = criterion_reg(out["length"], y_len)
        loss = loss1 + loss2 + loss3

        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(mt_model.parameters(), 1.0)
        optimizer.step()
        total_loss += loss.item() * x.size(0)

    avg_loss = total_loss / len(train_loader.dataset)
    print(f"[Fine-tune] Epoch {epoch+1}/{FT_EPOCHS} - Loss: {avg_loss:.4f}")

    # Validaci贸n simple
    mt_model.eval()
    val_loss = 0.0
    with torch.no_grad():
        for x, y_stab, y_bind, y_len in val_loader:
            x = x.to(DEVICE)
            y_stab = y_stab.to(DEVICE)
            y_bind = y_bind.to(DEVICE)
            y_len = y_len.to(DEVICE)
            out = mt_model(x)
            l1 = criterion_stab(out["stability"], y_stab)
            l2 = criterion_reg(out["telomerase"], y_bind)
            l3 = criterion_reg(out["length"], y_len)
            val_loss += (l1 + l2 + l3).item() * x.size(0)
    val_loss /= len(val_loader.dataset)
    print(f"  Validaci贸n Loss: {val_loss:.4f}")
```
