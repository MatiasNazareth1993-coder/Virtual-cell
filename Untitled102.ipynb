```python
"""
virtual_cell_foundational_model.py

Foundational, purely in-silico framework to simulate telomere dynamics, a "virtual telomerase"
control policy, generate synthetic "resistant" repetitive DNA sequences, and train a
deep-learning model (PyTorch) to predict when to activate telomerase.

This is a simulation + ML sandbox - intended for computational experiments only.
Do NOT use for wet-lab protocols or biological manipulation.

Requirements:
- Python 3.8+
- numpy, pandas, matplotlib, torch, scikit-learn

High-level components:
1. TelomereSimulator: stochastic model of telomere length across divisions, oxidative stress,
   and telomerase action.
2. SequenceGenerator: create synthetic repetitive telomeric-like sequences and score them for
   in-silico "stability" (heuristic score based on GC content, repeat periodicity and entropy).
3. Dataset builder: run many simulated cells with varied policies to build train / val sets.
4. PyTorch predictor: small LSTM/Transformer that predicts whether telomerase should activate
   next cycle given recent state features. Training loop with adjustable learning rate.
5. Example run at bottom demonstrating quick training and evaluation.

Adapt and expand: replace heuristics with biologically-derived features or experimental
simulations (in-silico chemistry engines) if available. Keep experiments computational only.
"""

import math
import random
import time
from dataclasses import dataclass
from typing import List, Tuple, Dict, Any

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, accuracy_score

# ----------------------- Simulation core ---------------------------------

@dataclass
class CellState:
    telomere_length: float  # abstract units (repeat units)
    age: int                # cell divisions experienced
    damage: float           # accumulated damage (0..1)


class TelomereSimulator:
    """Simple stochastic telomere model.

    - Each division shortens telomere by `shorten_mean` +/- `shorten_var`.
    - Oxidative damage increases damage; damage increases shortening rate.
    - Virtual telomerase when active adds repeats (with efficieny factor).
    - If telomere_length < senescence_thresh -> senescence event.
    - If telomerase_active too often or telomere exceeds cancer_thresh -> risk_of_uncontrolled_growth.
    """

    def __init__(self,
                 init_length: float = 200.0,
                 shorten_mean: float = 5.0,
                 shorten_std: float = 1.5,
                 damage_rate: float = 0.01,
                 damage_var: float = 0.005,
                 telomerase_add_mean: float = 50.0,
                 telomerase_add_std: float = 8.0,
                 senescence_thresh: float = 50.0,
                 cancer_thresh: float = 1000.0):
        self.init_length = init_length
        self.shorten_mean = shorten_mean
        self.shorten_std = shorten_std
        self.damage_rate = damage_rate
        self.damage_var = damage_var
        self.telomerase_add_mean = telomerase_add_mean
        self.telomerase_add_std = telomerase_add_std
        self.senescence_thresh = senescence_thresh
        self.cancer_thresh = cancer_thresh

    def step(self, state: CellState, telomerase_active: bool, oxidative_burst: float = 0.0) -> CellState:
        """Advance one division/cycle and return new state."""
        # Natural shortening influenced by damage
        shorten = np.random.normal(self.shorten_mean * (1.0 + state.damage), self.shorten_std)
        shorten += oxidative_burst  # occasional extra shortening
        shorten = max(0.0, shorten)

        new_length = state.telomere_length - shorten

        # Telomerase effect
        if telomerase_active:
            add = np.random.normal(self.telomerase_add_mean, self.telomerase_add_std)
            add = max(0.0, add)
            new_length += add

        # Damage accumulation
        damage_inc = np.random.normal(self.damage_rate, self.damage_var)
        damage_inc = max(0.0, damage_inc) + oxidative_burst * 0.001
        new_damage = min(1.0, state.damage + damage_inc)

        new_age = state.age + 1

        return CellState(telomere_length=new_length, age=new_age, damage=new_damage)

    def check_events(self, state: CellState, telomerase_activation_count: int) -> Dict[str, Any]:
        """Return events: senescence, uncontrolled_growth_risk (simple heuristic)."""
        senescent = state.telomere_length <= self.senescence_thresh
        # If telomere grows above cancer threshold OR telomerase used too frequently -> risk
        risk_score = 0.0
        if state.telomere_length >= self.cancer_thresh:
            risk_score += 1.0
        # scale risk with activation frequency and low damage control
        if telomerase_activation_count > max(1, state.age * 0.3):
            risk_score += 0.5
        return {"senescent": senescent, "risk_score": risk_score}


# ----------------------- Sequence heuristics ---------------------------------

class SequenceGenerator:
    """Generate repetitive 'telomeric-like' sequences and a heuristic stability score.

    Heuristic considers:
    - GC content (higher GC can increase stability in some contexts)
    - Repeat periodicity (clean repeats better)
    - Shannon entropy (lower sometimes more stable for repeats)
    """

    bases = ['A', 'C', 'G', 'T']

    def make_repeat(self, motif: str, n_repeat: int) -> str:
        return motif * n_repeat

    def random_motif(self, length: int = 3) -> str:
        return ''.join(random.choices(self.bases, k=length))

    def stability_score(self, seq: str) -> float:
        # GC content
        gc = (seq.count('G') + seq.count('C')) / max(1, len(seq))
        # entropy
        freqs = [seq.count(b) / len(seq) for b in self.bases]
        entropy = -sum([p * math.log2(p) for p in freqs if p > 0])
        # periodicity: check best small motif that reconstructs the string
        best_period = 0
        for p in range(1, min(8, len(seq))):
            motif = seq[:p]
            if motif * (len(seq) // p) == seq[:p * (len(seq) // p)]:
                best_period = p
                break
        # heuristic combination: higher GC, lower entropy, clear periodic motif preferred
        score = (gc * 0.6) + ((2.0 - entropy) * 0.2) + ((1.0 if best_period > 0 else 0.0) * 0.2)
        # normalize roughly 0..1
        return max(0.0, min(1.0, score))

    def generate_candidates(self, n: int = 100, motif_lengths: List[int] = [2,3,4]) -> List[Tuple[str,float]]:
        results = []
        for _ in range(n):
            mlen = random.choice(motif_lengths)
            motif = self.random_motif(mlen)
            repeats = random.randint(10, 200)
            seq = self.make_repeat(motif, repeats)
            score = self.stability_score(seq)
            results.append((seq, score))
        return results


# ----------------------- Data generation for ML -------------------------------

def generate_dataset(sim: TelomereSimulator,
                     n_cells: int = 2000,
                     max_cycles: int = 100,
                     policy: str = 'threshold',
                     threshold_on: float = 120.0) -> pd.DataFrame:
    """Simulate many cells using a simple policy to label when telomerase should be active.

    policy: 'threshold' (activate when telomere < threshold_on), 'heuristic' (activation when damage high),
    or 'random' (random activations). Output is a dataframe of per-cycle states with label = activate_next_cycle.
    """
    rows = []
    for _ in range(n_cells):
        state = CellState(telomere_length=sim.init_length * random.uniform(0.8, 1.2), age=0, damage=random.uniform(0.0,0.1))
        activation_count = 0
        # run cycles
        for cycle in range(max_cycles):
            # compute label for next cycle based on policy (we label whether we would activate now)
            if policy == 'threshold':
                label = 1 if state.telomere_length < threshold_on else 0
            elif policy == 'heuristic':
                label = 1 if (state.damage > 0.2 and state.telomere_length < sim.init_length*0.9) else 0
            else:
                label = 1 if random.random() < 0.05 else 0

            # optionally inject oxidative burst
            oxidative = 0.0
            if random.random() < 0.02:
                oxidative = random.uniform(0.0, 10.0)

            rows.append({
                'telomere_length': state.telomere_length,
                'age': state.age,
                'damage': state.damage,
                'cycle': cycle,
                'label': label
            })

            # apply step using label as activation to progress (simulate policy behavior)
            state = sim.step(state, telomerase_active=bool(label), oxidative_burst=oxidative)
            if label:
                activation_count += 1
            ev = sim.check_events(state, activation_count)
            if ev['senescent']:
                break

    df = pd.DataFrame(rows)
    return df


# ----------------------- PyTorch model ---------------------------------------

class SimpleLSTM(nn.Module):
    def __init__(self, input_dim=3, hidden_dim=64, n_layers=1, dropout=0.1):
        super().__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers=n_layers, batch_first=True, dropout=dropout)
        self.fc = nn.Sequential(
            nn.Linear(hidden_dim, hidden_dim//2),
            nn.ReLU(),
            nn.Linear(hidden_dim//2, 1),
        )

    def forward(self, x):
        # x: batch, seq_len, features
        out, _ = self.lstm(x)
        # take last timestep
        last = out[:, -1, :]
        return self.fc(last).squeeze(-1)


class TimeSeriesDataset(Dataset):
    def __init__(self, df: pd.DataFrame, seq_len: int = 8):
        # sort by synthetic trajectory: we only have flattened rows, so group by runs by index blocks
        self.seq_len = seq_len
        self.features = []
        self.labels = []

        # build sequences by sliding windows in each cell's run (we assume contiguous cycles)
        grouped = df.groupby((df.index // seq_len))  # crude grouping - but OK for synthetic dataset
        # More robust: if we had cell IDs, we'd group by that. For now, create sliding windows globally.
        arr = df[['telomere_length','age','damage','label']].values
        n = len(arr)
        for i in range(0, n - seq_len - 1):
            window = arr[i:i+seq_len]
            next_label = arr[i+seq_len][3]
            X = window[:, :3].astype(np.float32)
            y = float(next_label)
            self.features.append(X)
            self.labels.append(y)

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        return torch.from_numpy(self.features[idx]), torch.tensor(self.labels[idx], dtype=torch.float32)


def train_model(df: pd.DataFrame,
                seq_len: int = 8,
                lr: float = 1e-3,
                batch_size: int = 128,
                epochs: int = 8,
                device: str = 'cpu') -> Tuple[nn.Module, Dict[str, Any]]:
    train_df, val_df = train_test_split(df, test_size=0.2, random_state=42, shuffle=True)
    train_ds = TimeSeriesDataset(train_df, seq_len=seq_len)
    val_ds = TimeSeriesDataset(val_df, seq_len=seq_len)
    train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_ds, batch_size=batch_size, shuffle=False)

    model = SimpleLSTM(input_dim=3, hidden_dim=64).to(device)
    opt = torch.optim.Adam(model.parameters(), lr=lr)
    criterion = nn.BCEWithLogitsLoss()

    history = {'train_loss': [], 'val_loss': [], 'val_auc': []}
    for ep in range(epochs):
        model.train()
        running_loss = 0.0
        for Xb, yb in train_loader:
            Xb = Xb.to(device)
            yb = yb.to(device)
            logits = model(Xb)
            loss = criterion(logits, yb)
            opt.zero_grad()
            loss.backward()
            opt.step()
            running_loss += loss.item() * Xb.size(0)
        avg_train_loss = running_loss / len(train_loader.dataset)

        # validation
        model.eval()
        val_loss = 0.0
        preds, trues = [], []
        with torch.no_grad():
            for Xv, yv in val_loader:
                Xv = Xv.to(device)
                yv = yv.to(device)
                logits = model(Xv)
                loss = criterion(logits, yv)
                val_loss += loss.item() * Xv.size(0)
                probs = torch.sigmoid(logits).cpu().numpy()
                preds.extend(probs.tolist())
                trues.extend(yv.cpu().numpy().tolist())
        avg_val_loss = val_loss / len(val_loader.dataset)
        try:
            auc = roc_auc_score(trues, preds)
        except Exception:
            auc = 0.5

        history['train_loss'].append(avg_train_loss)
        history['val_loss'].append(avg_val_loss)
        history['val_auc'].append(auc)

        print(f"Epoch {ep+1}/{epochs}  train_loss={avg_train_loss:.4f}  val_loss={avg_val_loss:.4f}  val_auc={auc:.4f}")

    return model, history


# ----------------------- Quick demonstration / main --------------------------

if __name__ == '__main__':
    # Quick smoke-run to produce a tiny dataset and train quickly. Adjust parameters for real experiments.
    sim = TelomereSimulator()
    print("Generating synthetic dataset (this may take a moment)...")
    df = generate_dataset(sim, n_cells=600, max_cycles=60, policy='threshold', threshold_on=120.0)
    print("Dataset size:", len(df))

    # Inspect distribution
    print(df[['telomere_length','damage','label']].describe())

    # Train with a higher learning rate option
    lr = 5e-3  # "higher" learning rate - tune carefully
    model, history = train_model(df, seq_len=6, lr=lr, batch_size=64, epochs=6)

    # Plot training history
    plt.figure()
    plt.plot(history['train_loss'], label='train_loss')
    plt.plot(history['val_loss'], label='val_loss')
    plt.legend()
    plt.title('Training loss')
    plt.show()

    # Generate candidate sequences
    sg = SequenceGenerator()
    candidates = sg.generate_candidates(n=50)
    # sort by simulated stability
    candidates_sorted = sorted(candidates, key=lambda x: x[1], reverse=True)
    print("Top sequence stability scores (top 5):")
    for seq, score in candidates_sorted[:5]:
        print(score)

    # example inference on a single trajectory
    sample = df.sample(12)
    sample_sorted = sample.sort_values(['cycle']).head(8)[['telomere_length','age','damage']].values.astype(np.float32)
    X = torch.from_numpy(sample_sorted[np.newaxis, :, :])
    model.eval()
    with torch.no_grad():
        logit = model(X)
        prob = torch.sigmoid(logit).item()
    print(f"Model predicts telomerase activation probability next cycle: {prob:.3f}")

    print("""Done.\nNotes: this framework is intentionally heuristic; replace or extend subcomponents with\n          domain-specific simulators or more elaborate policies as you progress.""")
```

```python
import math
import random
import time
from dataclasses import dataclass
from typing import List, Tuple, Dict, Any

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, accuracy_score

# ----------------------- Simulation core ---------------------------------

@dataclass
class CellState:
    telomere_length: float  # abstract units (repeat units)
    age: int                # cell divisions experienced
    damage: float           # accumulated damage (0..1)


class TelomereSimulator:
    """Simple stochastic telomere model.

    - Each division shortens telomere by `shorten_mean` +/- `shorten_var`.
    - Oxidative damage increases damage; damage increases shortening rate.
    - Virtual telomerase when active adds repeats (with efficieny factor).
    - If telomere_length < senescence_thresh -> senescence event.
    - If telomerase_active too often or telomere exceeds cancer_thresh -> risk_of_uncontrolled_growth.
    """

    def __init__(self,
                 init_length: float = 200.0,
                 shorten_mean: float = 5.0,
                 shorten_std: float = 1.5,
                 damage_rate: float = 0.01,
                 damage_var: float = 0.005,
                 telomerase_add_mean: float = 50.0,
                 telomerase_add_std: float = 8.0,
                 senescence_thresh: float = 50.0,
                 cancer_thresh: float = 1000.0):
        self.init_length = init_length
        self.shorten_mean = shorten_mean
        self.shorten_std = shorten_std
        self.damage_rate = damage_rate
        self.damage_var = damage_var
        self.telomerase_add_mean = telomerase_add_mean
        self.telomerase_add_std = telomerase_add_std
        self.senescence_thresh = senescence_thresh
        self.cancer_thresh = cancer_thresh

    def step(self, state: CellState, telomerase_active: bool, oxidative_burst: float = 0.0) -> CellState:
        """Advance one division/cycle and return new state."""
        # Natural shortening influenced by damage
        shorten = np.random.normal(self.shorten_mean * (1.0 + state.damage), self.shorten_std)
        shorten += oxidative_burst  # occasional extra shortening
        shorten = max(0.0, shorten)

        new_length = state.telomere_length - shorten

        # Telomerase effect
        if telomerase_active:
            add = np.random.normal(self.telomerase_add_mean, self.telomerase_add_std)
            add = max(0.0, add)
            new_length += add

        # Damage accumulation
        damage_inc = np.random.normal(self.damage_rate, self.damage_var)
        damage_inc = max(0.0, damage_inc) + oxidative_burst * 0.001
        new_damage = min(1.0, state.damage + damage_inc)

        new_age = state.age + 1

        return CellState(telomere_length=new_length, age=new_age, damage=new_damage)

    def check_events(self, state: CellState, telomerase_activation_count: int) -> Dict[str, Any]:
        """Return events: senescence, uncontrolled_growth_risk (simple heuristic)."""
        senescent = state.telomere_length <= self.senescence_thresh
        # If telomere grows above cancer threshold OR telomerase used too frequently -> risk
        risk_score = 0.0
        if state.telomere_length >= self.cancer_thresh:
            risk_score += 1.0
        # scale risk with activation frequency and low damage control
        if telomerase_activation_count > max(1, state.age * 0.3):
            risk_score += 0.5
        return {"senescent": senescent, "risk_score": risk_score}


# ----------------------- Sequence heuristics ---------------------------------

class SequenceGenerator:
    """Generate repetitive 'telomeric-like' sequences and a heuristic stability score.

    Heuristic considers:
    - GC content (higher GC can increase stability in some contexts)
    - Repeat periodicity (clean repeats better)
    - Shannon entropy (lower sometimes more stable for repeats)
    """

    bases = ['A', 'C', 'G', 'T']

    def make_repeat(self, motif: str, n_repeat: int) -> str:
        return motif * n_repeat

    def random_motif(self, length: int = 3) -> str:
        return ''.join(random.choices(self.bases, k=length))

    def stability_score(self, seq: str) -> float:
        # GC content
        gc = (seq.count('G') + seq.count('C')) / max(1, len(seq))
        # entropy
        freqs = [seq.count(b) / len(seq) for b in self.bases]
        entropy = -sum([p * math.log2(p) for p in freqs if p > 0])
        # periodicity: check best small motif that reconstructs the string
        best_period = 0
        for p in range(1, min(8, len(seq))):
            motif = seq[:p]
            if motif * (len(seq) // p) == seq[:p * (len(seq) // p)]:
                best_period = p
                break
        # heuristic combination: higher GC, lower entropy, clear periodic motif preferred
        score = (gc * 0.6) + ((2.0 - entropy) * 0.2) + ((1.0 if best_period > 0 else 0.0) * 0.2)
        # normalize roughly 0..1
        return max(0.0, min(1.0, score))

    def generate_candidates(self, n: int = 100, motif_lengths: List[int] = [2,3,4]) -> List[Tuple[str,float]]:
        results = []
        for _ in range(n):
            mlen = random.choice(motif_lengths)
            motif = self.random_motif(mlen)
            repeats = random.randint(10, 200)
            seq = self.make_repeat(motif, repeats)
            score = self.stability_score(seq)
            results.append((seq, score))
        return results


# ----------------------- Data generation for ML -------------------------------

def generate_dataset(sim: TelomereSimulator,
                     n_cells: int = 2000,
                     max_cycles: int = 100,
                     policy: str = 'threshold',
                     threshold_on: float = 120.0,
                     lineage_size: int = 5) -> pd.DataFrame:
    """Expanded dataset generator.

    Additions:
    - cell_id: unique identifier for each simulated cell
    - lineage_id: simulate small families of related cells
    - oxidative_events: count of oxidative bursts per cycle
    - telomerase_activation_count: running count of activations
    - extra engineered feature: "stress_index" = damage * (initial_length / current_length)
    """
    rows = []
    cell_id = 0
    for lineage in range(n_cells // lineage_size):
        for _ in range(lineage_size):
            cell_id += 1
            state = CellState(telomere_length=sim.init_length * random.uniform(0.8, 1.2),
                               age=0,
                               damage=random.uniform(0.0,0.1))
            activation_count = 0
            oxidative_events = 0
            for cycle in range(max_cycles):
                # determine label
                if policy == 'threshold':
                    label = 1 if state.telomere_length < threshold_on else 0
                elif policy == 'heuristic':
                    label = 1 if (state.damage > 0.2 and state.telomere_length < sim.init_length*0.9) else 0
                else:
                    label = 1 if random.random() < 0.05 else 0

                oxidative = 0.0
                if random.random() < 0.03:
                    oxidative = random.uniform(0.0, 10.0)
                    oxidative_events += 1

                # engineered feature
                stress_index = state.damage * (sim.init_length / max(1e-3, state.telomere_length))

                rows.append({
                    'cell_id': cell_id,
                    'lineage_id': lineage,
                    'telomere_length': state.telomere_length,
                    'age': state.age,
                    'damage': state.damage,
                    'cycle': cycle,
                    'oxidative_events': oxidative_events,
                    'activation_count': activation_count,
                    'stress_index': stress_index,
                    'label': label
                })

                state = sim.step(state, telomerase_active=bool(label), oxidative_burst=oxidative)
                if label:
                    activation_count += 1
                ev = sim.check_events(state, activation_count)
                if ev['senescent']:
                    break

    df = pd.DataFrame(rows)
    return df

# ----------------------- Multiprocessing dataset generation --------------------
from multiprocessing import Pool, cpu_count

def _simulate_lineage(args):
    """Worker function to simulate a subset of lineages in parallel."""
    sim, lineage_start, lineage_end, max_cycles, policy, threshold_on, lineage_size = args
    rows = []
    cell_id = lineage_start * lineage_size
    for lineage in range(lineage_start, lineage_end):
        for _ in range(lineage_size):
            cell_id += 1
            state = CellState(telomere_length=sim.init_length * random.uniform(0.8, 1.2),
                               age=0,
                               damage=random.uniform(0.0,0.1))
            activation_count = 0
            oxidative_events = 0
            for cycle in range(max_cycles):
                if policy == 'threshold':
                    label = 1 if state.telomere_length < threshold_on else 0
                elif policy == 'heuristic':
                    label = 1 if (state.damage > 0.2 and state.telomere_length < sim.init_length*0.9) else 0
                else:
                    label = 1 if random.random() < 0.05 else 0

                oxidative = 0.0
                if random.random() < 0.03:
                    oxidative = random.uniform(0.0, 10.0)
                    oxidative_events += 1

                stress_index = state.damage * (sim.init_length / max(1e-3, state.telomere_length))

                rows.append({
                    'cell_id': cell_id,
                    'lineage_id': lineage,
                    'telomere_length': state.telomere_length,
                    'age': state.age,
                    'damage': state.damage,
                    'cycle': cycle,
                    'oxidative_events': oxidative_events,
                    'activation_count': activation_count,
                    'stress_index': stress_index,
                    'label': label
                })

                state = sim.step(state, telomerase_active=bool(label), oxidative_burst=oxidative)
                if label:
                    activation_count += 1
                if sim.check_events(state, activation_count)['senescent']:
                    break

    return rows

def generate_dataset_mp(sim: TelomereSimulator,
                         n_cells: int = 20000,
                         max_cycles: int = 100,
                         policy: str = 'threshold',
                         threshold_on: float = 120.0,
                         lineage_size: int = 5,
                         n_workers: int = None) -> pd.DataFrame:
    """Generate large datasets using multiprocessing.

    Splits the total number of lineages across available CPU cores.
    Returns a combined pandas DataFrame.
    """
    if n_workers is None:
        n_workers = max(1, cpu_count() - 1)

    n_lineages = n_cells // lineage_size
    chunk = math.ceil(n_lineages / n_workers)

    tasks = []
    for w in range(n_workers):
        start = w * chunk
        end = min((w + 1) * chunk, n_lineages)
        if start >= end:
            continue
        tasks.append((sim, start, end, max_cycles, policy, threshold_on, lineage_size))

    with Pool(processes=n_workers) as pool:
        all_rows = pool.map(_simulate_lineage, tasks)

    # flatten results
    flat = [item for sublist in all_rows for item in sublist]
    return pd.DataFrame(flat)


# ----------------------- PyTorch model ---------------------------------------

class SimpleLSTM(nn.Module):
    def __init__(self, input_dim=3, hidden_dim=64, n_layers=1, dropout=0.1):
        super().__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers=n_layers, batch_first=True, dropout=dropout)
        self.fc = nn.Sequential(
            nn.Linear(hidden_dim, hidden_dim//2),
            nn.ReLU(),
            nn.Linear(hidden_dim//2, 1),
        )

    def forward(self, x):
        # x: batch, seq_len, features
        out, _ = self.lstm(x)
        # take last timestep
        last = out[:, -1, :]
        return self.fc(last).squeeze(-1)


class TimeSeriesDataset(Dataset):
    def __init__(self, df: pd.DataFrame, seq_len: int = 8):
        # sort by synthetic trajectory: we only have flattened rows, so group by runs by cell_id
        self.seq_len = seq_len
        self.features = []
        self.labels = []

        # Iterate through each cell's trajectory
        for cell_id, cell_df in df.groupby('cell_id'):
            arr = cell_df[['telomere_length', 'age', 'damage', 'stress_index', 'oxidative_events', 'activation_count', 'label']].values
            n = len(arr)
            # Create sliding windows for each cell's trajectory
            for i in range(0, n - seq_len - 1):
                window = arr[i:i+seq_len]
                next_label = arr[i+seq_len][6] # 'label' is the 7th column (index 6)
                # Features for LSTM: telomere_length, age, damage, stress_index, oxidative_events, activation_count
                X = window[:, :6].astype(np.float32) # Take first 6 columns as features
                y = float(next_label)
                self.features.append(X)
                self.labels.append(y)

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        return torch.from_numpy(self.features[idx]), torch.tensor(self.labels[idx], dtype=torch.float32)


def train_model(df: pd.DataFrame,
                seq_len: int = 8,
                lr: float = 1e-3,
                batch_size: int = 128,
                epochs: int = 8,
                device: str = 'cpu') -> Tuple[nn.Module, Dict[str, Any]]:
    train_df, val_df = train_test_split(df, test_size=0.2, random_state=42, shuffle=True)
    train_ds = TimeSeriesDataset(train_df, seq_len=seq_len)
    val_ds = TimeSeriesDataset(val_df, seq_len=seq_len)
    train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_ds, batch_size=batch_size, shuffle=False)

    # Adjust input_dim based on the new features in TimeSeriesDataset
    model = SimpleLSTM(input_dim=6, hidden_dim=64).to(device)
    opt = torch.optim.Adam(model.parameters(), lr=lr)
    criterion = nn.BCEWithLogitsLoss()

    history = {'train_loss': [], 'val_loss': [], 'val_auc': []}
    for ep in range(epochs):
        model.train()
        running_loss = 0.0
        for Xb, yb in train_loader:
            Xb = Xb.to(device)
            yb = yb.to(device)
            logits = model(Xb)
            loss = criterion(logits, yb)
            opt.zero_grad()
            loss.backward()
            opt.step()
            running_loss += loss.item() * Xb.size(0)
        avg_train_loss = running_loss / len(train_loader.dataset)

        # validation
        model.eval()
        val_loss = 0.0
        preds, trues = [], []
        with torch.no_grad():
            for Xv, yv in val_loader:
                Xv = Xv.to(device)
                yv = yv.to(device)
                logits = model(Xv)
                loss = criterion(logits, yv)
                val_loss += loss.item() * Xv.size(0)
                probs = torch.sigmoid(logits).cpu().numpy()
                preds.extend(probs.tolist())
                trues.extend(yv.cpu().numpy().tolist())
        avg_val_loss = val_loss / len(val_loader.dataset)
        try:
            auc = roc_auc_score(trues, preds)
        except Exception:
            auc = 0.5

        history['train_loss'].append(avg_train_loss)
        history['val_loss'].append(avg_val_loss)
        history['val_auc'].append(auc)

        print(f"Epoch {ep+1}/{epochs}  train_loss={avg_train_loss:.4f}  val_loss={avg_val_loss:.4f}  val_auc={auc:.4f}")

    return model, history


# ----------------------- Quick demonstration / main --------------------------

if __name__ == '__main__':
    # Quick smoke-run to produce a tiny dataset and train quickly. Adjust parameters for real experiments.
    sim = TelomereSimulator()
    print("Generating synthetic dataset (this may take a moment)...")
    # Updated generate_dataset call
    df = generate_dataset(sim, n_cells=600, max_cycles=60, policy='threshold', threshold_on=120.0, lineage_size=5)
    print("Dataset size:", len(df))

    # Inspect distribution
    print(df[['telomere_length','damage','label']].describe())

    # Train with a higher learning rate option
    lr = 5e-3  # "higher" learning rate - tune carefully
    model, history = train_model(df, seq_len=6, lr=lr, batch_size=64, epochs=6)

    # Plot training history
    plt.figure()
    plt.plot(history['train_loss'], label='train_loss')
    plt.plot(history['val_loss'], label='val_loss')
    plt.legend()
    plt.title('Training loss')
    plt.show()

    # Generate candidate sequences
    sg = SequenceGenerator()
    candidates = sg.generate_candidates(n=50)
    # sort by simulated stability
    candidates_sorted = sorted(candidates, key=lambda x: x[1], reverse=True)
    print("Top sequence stability scores (top 5):")
    for seq, score in candidates_sorted[:5]:
        print(score)

    # example inference on a single trajectory
    # updated sample to include all features used for training
    sample = df.sample(12)
    sample_sorted = sample.sort_values(['cycle']).head(8)[['telomere_length','age','damage','stress_index','oxidative_events','activation_count']].values.astype(np.float32)
    X = torch.from_numpy(sample_sorted[np.newaxis, :, :])
    model.eval()
    with torch.no_grad():
        logit = model(X)
        prob = torch.sigmoid(logit).item()
    print(f"Model predicts telomerase activation probability next cycle: {prob:.3f}")

    print("""Done.\nNotes: this framework is intentionally heuristic; replace or extend subcomponents with\n          domain-specific simulators or more elaborate policies as you progress.""")
```

```python
# virtual_cell_pipeline.py
"""
Virtual Cell  in silico simulation + representation learning pipeline (safe, abstracted).
NOT biological instructions. All "sequences" and behaviors are synthetic proxies.
"""

import math
import random
import argparse
from typing import Tuple, List
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from tqdm import tqdm
from sklearn.metrics import roc_auc_score, accuracy_score
from sklearn.cluster import KMeans
import umap
import matplotlib.pyplot as plt
import os

# ----------------------------
# 1) Synthetic data generation
# ----------------------------
TOKEN_VOCAB = list("ACGTX")  # abstract tokens, not real DNA instruction
VOCAB_TO_IDX = {c: i+1 for i, c in enumerate(TOKEN_VOCAB)}  # reserve 0 for padding

def generate_abstract_sequence(length: int) -> str:
    # Create sequences where patterns correlate with "stability"
    # 'X' token acts as special repeat that increases simulated stability
    seq = []
    for i in range(length):
        p = random.random()
        if p < 0.05:
            seq.append('X')  # synthetic resistant motif
        else:
            seq.append(random.choice(['A','C','G','T']))
    return ''.join(seq)

def seq_to_indices(seq: str, maxlen: int):
    arr = [VOCAB_TO_IDX.get(c, 0) for c in seq][:maxlen]
    if len(arr) < maxlen:
        arr += [0] * (maxlen - len(arr))
    return arr

def simulate_telomere_dynamics(seq: str, rng_seed=None) -> Tuple[float, float, float, int]:
    # Abstract simulation returning:
    # - initial_length proxy
    # - oxidative_stress (0..1)
    # - repair_efficiency (0..1)
    # - label: 1 => healthy division / desirable; 0 => risk (uncontrolled or senescent)
    if rng_seed is not None:
        random.seed(rng_seed)
    # features derived from sequence:
    resistant_motif_count = seq.count('X')  # proxy for "resistant repeats"
    initial_length = 100 + resistant_motif_count * 15 + random.gauss(0, 5)
    oxidative_stress = min(max(random.betavariate(2,5) + 0.2*(1/(1+resistant_motif_count)), 0.0), 1.0)
    repair_eff = min(max(random.betavariate(5,2) + 0.1 * resistant_motif_count/10.0, 0.0), 1.0)
    # telomerase activation history (abstract)
    telomerase_activity_score = min(max(0.2 * resistant_motif_count + repair_eff - oxidative_stress + random.gauss(0,0.05), 0.0), 2.0)
    # label logic (abstract/rule-based + noise):
    score = 0.5* (initial_length/150.0) + 0.3*repair_eff - 0.4*oxidative_stress + 0.2*(telomerase_activity_score/2.0)
    prob_good = 1/(1+math.exp(-6*(score-0.4)))  # sigmoid with steeper slope
    label = 1 if random.random() < prob_good else 0
    return initial_length, oxidative_stress, repair_eff, label

class VirtualCellDataset(Dataset):
    def __init__(self, n_samples:int=5000, seq_len:int=60, maxlen:int=64, seed:int=42):
        self.samples = []
        random.seed(seed)
        np.random.seed(seed)
        for i in range(n_samples):
            s = generate_abstract_sequence(seq_len)
            initial_len, ox, repair, label = simulate_telomere_dynamics(s)
            numeric_feats = np.array([initial_len, ox, repair], dtype=np.float32)
            idxs = np.array(seq_to_indices(s, maxlen), dtype=np.int64)
            self.samples.append((idxs, numeric_feats, int(label)))
    def __len__(self): return len(self.samples)
    def __getitem__(self, i):
        return self.samples[i]

# ----------------------------
# 2) Model: little Transformer + numeric feature head
# ----------------------------
class SeqTransformer(nn.Module):
    def __init__(self, vocab_size:int, emb_dim:int=64, n_heads:int=4, num_layers:int=2, seq_maxlen:int=64, numeric_dim:int=3, hidden_dim:int=128, proj_dim:int=64):
        super().__init__()
        self.token_emb = nn.Embedding(vocab_size+1, emb_dim, padding_idx=0)  # +1 for padding idx=0
        self.pos_emb = nn.Embedding(seq_maxlen, emb_dim)
        encoder_layer = nn.TransformerEncoderLayer(d_model=emb_dim, nhead=n_heads, dim_feedforward=emb_dim*4, activation='gelu')
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        self.pool = nn.AdaptiveAvgPool1d(1)  # pool sequence to vector
        # numeric feature MLP
        self.num_mlp = nn.Sequential(
            nn.Linear(numeric_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, emb_dim)
        )
        # final projection
        self.project = nn.Sequential(
            nn.Linear(emb_dim*2, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, proj_dim)
        )
        # prediction head
        self.classifier = nn.Sequential(
            nn.Linear(proj_dim, hidden_dim//2),
            nn.ReLU(),
            nn.Linear(hidden_dim//2, 1)
        )

    def forward(self, token_idxs, numeric_feats):
        # token_idxs: [B, L]
        B, L = token_idxs.shape
        tok = self.token_emb(token_idxs)  # [B, L, E]
        pos = self.pos_emb(torch.arange(L, device=token_idxs.device).unsqueeze(0).expand(B,-1))
        x = tok + pos
        # transformer expects [L, B, E]
        x = x.permute(1,0,2)
        x = self.transformer(x)  # [L,B,E]
        x = x.permute(1,2,0)  # [B,E,L]
        x = self.pool(x).squeeze(-1)  # [B, E]
        num_v = self.num_mlp(numeric_feats)  # [B, E]
        joint = torch.cat([x, num_v], dim=1)  # [B, 2E]
        proj = self.project(joint)  # [B, proj_dim] -> embeddings
        logits = self.classifier(proj).squeeze(-1)
        return logits, proj  # return both for training and embedding extraction

# ----------------------------
# 3) Training loop & utilities
# ----------------------------
def train_one_epoch(model, loader, optim, device, criterion, scaler=None):
    model.train()
    losses = []
    for tokens, nums, labels in tqdm(loader, desc="train"):
        tokens = tokens.to(device)
        nums = nums.to(device)
        labels = labels.to(device).float()
        optim.zero_grad()
        logits, emb = model(tokens, nums)
        loss = criterion(logits, labels)
        loss.backward()
        optim.step()
        losses.append(loss.item())
    return float(np.mean(losses))

def eval_model(model, loader, device):
    model.eval()
    ys, preds, scores = [], [], []
    embs = []
    with torch.no_grad():
        for tokens, nums, labels in loader:
            tokens = tokens.to(device)
            nums = nums.to(device)
            labels = labels.numpy()
            logits, proj = model(tokens, nums)
            prob = torch.sigmoid(logits).cpu().numpy()
            preds.append((prob>0.5).astype(int))
            ys.append(labels)
            scores.append(prob)
            embs.append(proj.cpu().numpy())
    ys = np.concatenate(ys).ravel()
    preds = np.concatenate(preds).ravel()
    scores = np.concatenate(scores).ravel()
    embs = np.vstack(embs)
    auc = roc_auc_score(ys, scores)
    acc = accuracy_score(ys, preds)
    return {'auc': auc, 'acc': acc, 'embeddings': embs, 'labels': ys, 'scores': scores}

# ----------------------------
# 4) Main runnable pipeline
# ----------------------------
def main(args):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print("Device:", device)
    # create datasets
    ds = VirtualCellDataset(n_samples=args.n_samples, seq_len=args.seq_len, maxlen=args.maxlen, seed=args.seed)
    n = len(ds)
    idxs = list(range(n))
    random.shuffle(idxs)
    split1 = int(n*0.8)
    split2 = int(n*0.9)
    train_idx = idxs[:split1]
    val_idx = idxs[split1:split2]
    test_idx = idxs[split2:]
    def subset(ds, idxs):
        s = torch.utils.data.Subset(ds, idxs)
        return s
    train_dl = DataLoader(subset(ds, train_idx), batch_size=args.batch_size, shuffle=True, drop_last=True)
    val_dl = DataLoader(subset(ds, val_idx), batch_size=args.batch_size, shuffle=False)
    test_dl = DataLoader(subset(ds, test_idx), batch_size=args.batch_size, shuffle=False)

    # model
    model = SeqTransformer(vocab_size=len(TOKEN_VOCAB), emb_dim=args.emb_dim, n_heads=args.n_heads,
                           num_layers=args.num_layers, seq_maxlen=args.maxlen, numeric_dim=3,
                           hidden_dim=args.hidden_dim, proj_dim=args.proj_dim).to(device)

    # optimizer with high learning rate suggestion and scheduler
    optim = torch.optim.AdamW(model.parameters(), lr=args.learning_rate, weight_decay=1e-4)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optim, 'max', patience=3, factor=0.5)
    criterion = nn.BCEWithLogitsLoss()

    best_val_auc = -1.0
    ckpt_dir = args.ckpt_dir
    os.makedirs(ckpt_dir, exist_ok=True)

    for epoch in range(1, args.epochs+1):
        print(f"\n=== Epoch {epoch} ===")
        train_loss = train_one_epoch(model, train_dl, optim, device, criterion)
        val_res = eval_model(model, val_dl, device)
        print(f"Train loss {train_loss:.4f} | Val AUC {val_res['auc']:.4f} | Val ACC {val_res['acc']:.4f}")
        scheduler.step(val_res['auc'])
        if val_res['auc'] > best_val_auc:
            best_val_auc = val_res['auc']
            torch.save({'model_state': model.state_dict(), 'epoch': epoch}, os.path.join(ckpt_dir, 'best.pth'))
            print("Saved best checkpoint.")

    # final test eval
    test_res = eval_model(model, test_dl, device)
    print(f"\nTest AUC {test_res['auc']:.4f} | Test ACC {test_res['acc']:.4f}")

    # Step: dimensionality reduction + clustering for pattern discovery
    embeddings = test_res['embeddings']
    labels = test_res['labels']
    reducer = umap.UMAP(n_components=2, random_state=42)
    proj2 = reducer.fit_transform(embeddings)
    km = KMeans(n_clusters=3, random_state=42).fit(embeddings)
    cluster_ids = km.labels_

    # plot UMAP colored by cluster and by label
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    plt.scatter(proj2[:,0], proj2[:,1], c=cluster_ids, cmap='tab10', s=8)
    plt.title("UMAP of embeddings (clusters)")
    plt.subplot(1,2,2)
    plt.scatter(proj2[:,0], proj2[:,1], c=labels, cmap='coolwarm', s=8)
    plt.title("UMAP colored by true label (1=good)")
    plt.tight_layout()
    plt.savefig(os.path.join(ckpt_dir, 'umap_clusters.png'))
    print("Saved UMAP plot to", os.path.join(ckpt_dir, 'umap_clusters.png'))

    # simple nearest-neighbor lookup example for "drug discovery"
    # cluster centroids -> score, rank clusters by average predicted score
    cluster_scores = []
    for c in np.unique(cluster_ids):
        mask = cluster_ids == c
        cluster_scores.append((c, embeddings[mask].mean(axis=0).dot(np.ones(embeddings.shape[1]))))  # simple proxy
    print("Cluster scores (proxy):", cluster_scores)
    print("Pipeline complete. Check", ckpt_dir)

if __name__ == "__main__":
    parser = argparse.ArgumentParser()
    parser.add_argument("--n_samples", type=int, default=6000)
    parser.add_argument("--seq_len", type=int, default=60)
    parser.add_argument("--maxlen", type=int, default=64)
    parser.add_argument("--batch_size", type=int, default=64)
    parser.add_argument("--epochs", type=int, default=14)
    parser.add_argument("--learning_rate", type=float, default=2e-3)  # higher LR suggestion; tune with scheduler
    parser.add_argument("--emb_dim", type=int, default=64)
    parser.add_argument("--n_heads", type=int, default=4)
    parser.add_argument("--num_layers", type=int, default=2)
    parser.add_argument("--hidden_dim", type=int, default=128)
    parser.add_argument("--proj_dim", type=int, default=64)
    parser.add_argument("--seed", type=int, default=42)
    parser.add_argument("--ckpt_dir", type=str, default="checkpoints")
    args = parser.parse_args(args=[]) # Modified this line
    main(args)
```

```python
"""
virtual_cell_foundational_model.py

Foundational, purely in-silico framework to simulate telomere dynamics, a "virtual telomerase"
control policy, generate synthetic "resistant" repetitive DNA sequences, and train a
deep-learning model (PyTorch) to predict when to activate telomerase.

This is a simulation + ML sandbox - intended for computational experiments only.
Do NOT use for wet-lab protocols or biological manipulation.

Requirements:
- Python 3.8+
- numpy, pandas, matplotlib, torch, scikit-learn

High-level components:
1. TelomereSimulator: stochastic model of telomere length across divisions, oxidative stress,
   and telomerase action.
2. SequenceGenerator: create synthetic repetitive telomeric-like sequences and score them for
   in-silico "stability" (heuristic score based on GC content, repeat periodicity and entropy).
3. Dataset builder: run many simulated cells with varied policies to build train / val sets.
4. PyTorch predictor: small LSTM/Transformer that predicts whether telomerase should activate
   next cycle given recent state features. Training loop with adjustable learning rate.
5. Example run at bottom demonstrating quick training and evaluation.

Adapt and expand: replace heuristics with biologically-derived features or experimental
simulations (in-silico chemistry engines) if available. Keep experiments computational only.
"""

import math
import random
import time
from dataclasses import dataclass
from typing import List, Tuple, Dict, Any

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, accuracy_score

# ----------------------- Simulation core ---------------------------------

@dataclass
class CellState:
    telomere_length: float  # abstract units (repeat units)
    age: int                # cell divisions experienced
    damage: float           # accumulated damage (0..1)


class TelomereSimulator:
    """Simple stochastic telomere model.

    - Each division shortens telomere by `shorten_mean` +/- `shorten_var`.
    - Oxidative damage increases damage; damage increases shortening rate.
    - Virtual telomerase when active adds repeats (with efficieny factor).
    - If telomere_length < senescence_thresh -> senescence event.
    - If telomerase_active too often or telomere exceeds cancer_thresh -> risk_of_uncontrolled_growth.
    """

    def __init__(self,
                 init_length: float = 200.0,
                 shorten_mean: float = 5.0,
                 shorten_std: float = 1.5,
                 damage_rate: float = 0.01,
                 damage_var: float = 0.005,
                 telomerase_add_mean: float = 50.0,
                 telomerase_add_std: float = 8.0,
                 senescence_thresh: float = 50.0,
                 cancer_thresh: float = 1000.0):
        self.init_length = init_length
        self.shorten_mean = shorten_mean
        self.shorten_std = shorten_std
        self.damage_rate = damage_rate
        self.damage_var = damage_var
        self.telomerase_add_mean = telomerase_add_mean
        self.telomerase_add_std = telomerase_add_std
        self.senescence_thresh = senescence_thresh
        self.cancer_thresh = cancer_thresh

    def step(self, state: CellState, telomerase_active: bool, oxidative_burst: float = 0.0) -> CellState:
        """Advance one division/cycle and return new state."""
        # Natural shortening influenced by damage
        shorten = np.random.normal(self.shorten_mean * (1.0 + state.damage), self.shorten_std)
        shorten += oxidative_burst  # occasional extra shortening
        shorten = max(0.0, shorten)

        new_length = state.telomere_length - shorten

        # Telomerase effect
        if telomerase_active:
            add = np.random.normal(self.telomerase_add_mean, self.telomerase_add_std)
            add = max(0.0, add)
            new_length += add

        # Damage accumulation
        damage_inc = np.random.normal(self.damage_rate, self.damage_var)
        damage_inc = max(0.0, damage_inc) + oxidative_burst * 0.001
        new_damage = min(1.0, state.damage + damage_inc)

        new_age = state.age + 1

        return CellState(telomere_length=new_length, age=new_age, damage=new_damage)

    def check_events(self, state: CellState, telomerase_activation_count: int) -> Dict[str, Any]:
        """Return events: senescence, uncontrolled_growth_risk (simple heuristic)."""
        senescent = state.telomere_length <= self.senescence_thresh
        # If telomere grows above cancer threshold OR telomerase used too frequently -> risk
        risk_score = 0.0
        if state.telomere_length >= self.cancer_thresh:
            risk_score += 1.0
        # scale risk with activation frequency and low damage control
        if telomerase_activation_count > max(1, state.age * 0.3):
            risk_score += 0.5
        return {"senescent": senescent, "risk_score": risk_score}


# ----------------------- Sequence heuristics ---------------------------------

class SequenceGenerator:
    """Generate repetitive 'telomeric-like' sequences and a heuristic stability score.

    Heuristic considers:
    - GC content (higher GC can increase stability in some contexts)
    - Repeat periodicity (clean repeats better)
    - Shannon entropy (lower sometimes more stable for repeats)
    """

    bases = ['A', 'C', 'G', 'T']

    def make_repeat(self, motif: str, n_repeat: int) -> str:
        return motif * n_repeat

    def random_motif(self, length: int = 3) -> str:
        return ''.join(random.choices(self.bases, k=length))

    def stability_score(self, seq: str) -> float:
        # GC content
        gc = (seq.count('G') + seq.count('C')) / max(1, len(seq))
        # entropy
        freqs = [seq.count(b) / len(seq) for b in self.bases]
        entropy = -sum([p * math.log2(p) for p in freqs if p > 0])
        # periodicity: check best small motif that reconstructs the string
        best_period = 0
        for p in range(1, min(8, len(seq))):
            motif = seq[:p]
            if motif * (len(seq) // p) == seq[:p * (len(seq) // p)]:
                best_period = p
                break
        # heuristic combination: higher GC, lower entropy, clear periodic motif preferred
        score = (gc * 0.6) + ((2.0 - entropy) * 0.2) + ((1.0 if best_period > 0 else 0.0) * 0.2)
        # normalize roughly 0..1
        return max(0.0, min(1.0, score))

    def generate_candidates(self, n: int = 100, motif_lengths: List[int] = [2,3,4]) -> List[Tuple[str,float]]:
        results = []
        for _ in range(n):
            mlen = random.choice(motif_lengths)
            motif = self.random_motif(mlen)
            repeats = random.randint(10, 200)
            seq = self.make_repeat(motif, repeats)
            score = self.stability_score(seq)
            results.append((seq, score))
        return results


# ----------------------- Data generation for ML -------------------------------

def generate_dataset(sim: TelomereSimulator,
                     n_cells: int = 2000,
                     max_cycles: int = 100,
                     policy: str = 'threshold',
                     threshold_on: float = 120.0,
                     lineage_size: int = 5) -> pd.DataFrame:
    """Expanded dataset generator.

    Additions:
    - cell_id: unique identifier for each simulated cell
    - lineage_id: simulate small families of related cells
    - oxidative_events: count of oxidative bursts per cycle
    - telomerase_activation_count: running count of activations
    - extra engineered feature: "stress_index" = damage * (initial_length / current_length)
    """
    rows = []
    cell_id = 0
    for lineage in range(n_cells // lineage_size):
        for _ in range(lineage_size):
            cell_id += 1
            state = CellState(telomere_length=sim.init_length * random.uniform(0.8, 1.2),
                               age=0,
                               damage=random.uniform(0.0,0.1))
            activation_count = 0
            oxidative_events = 0
            for cycle in range(max_cycles):
                # determine label
                if policy == 'threshold':
                    label = 1 if state.telomere_length < threshold_on else 0
                elif policy == 'heuristic':
                    label = 1 if (state.damage > 0.2 and state.telomere_length < sim.init_length*0.9) else 0
                else:
                    label = 1 if random.random() < 0.05 else 0

                oxidative = 0.0
                if random.random() < 0.03:
                    oxidative = random.uniform(0.0, 10.0)
                    oxidative_events += 1

                # engineered feature
                stress_index = state.damage * (sim.init_length / max(1e-3, state.telomere_length))

                rows.append({
                    'cell_id': cell_id,
                    'lineage_id': lineage,
                    'telomere_length': state.telomere_length,
                    'age': state.age,
                    'damage': state.damage,
                    'cycle': cycle,
                    'oxidative_events': oxidative_events,
                    'activation_count': activation_count,
                    'stress_index': stress_index,
                    'label': label
                })

                state = sim.step(state, telomerase_active=bool(label), oxidative_burst=oxidative)
                if label:
                    activation_count += 1
                ev = sim.check_events(state, activation_count)
                if ev['senescent']:
                    break

    df = pd.DataFrame(rows)
    return df

# ----------------------- Multiprocessing dataset generation --------------------
from multiprocessing import Pool, cpu_count

def _simulate_lineage(args):
    """Worker function to simulate a subset of lineages in parallel."""
    sim, lineage_start, lineage_end, max_cycles, policy, threshold_on, lineage_size = args
    rows = []
    cell_id = lineage_start * lineage_size
    for lineage in range(lineage_start, lineage_end):
        for _ in range(lineage_size):
            cell_id += 1
            state = CellState(telomere_length=sim.init_length * random.uniform(0.8, 1.2),
                               age=0,
                               damage=random.uniform(0.0,0.1))
            activation_count = 0
            oxidative_events = 0
            for cycle in range(max_cycles):
                if policy == 'threshold':
                    label = 1 if state.telomere_length < threshold_on else 0
                elif policy == 'heuristic':
                    label = 1 if (state.damage > 0.2 and state.telomere_length < sim.init_length*0.9) else 0
                else:
                    label = 1 if random.random() < 0.05 else 0

                oxidative = 0.0
                if random.random() < 0.03:
                    oxidative = random.uniform(0.0, 10.0)
                    oxidative_events += 1

                stress_index = state.damage * (sim.init_length / max(1e-3, state.telomere_length))

                rows.append({
                    'cell_id': cell_id,
                    'lineage_id': lineage,
                    'telomere_length': state.telomere_length,
                    'age': state.age,
                    'damage': state.damage,
                    'cycle': cycle,
                    'oxidative_events': oxidative_events,
                    'activation_count': activation_count,
                    'stress_index': stress_index,
                    'label': label
                })

                state = sim.step(state, telomerase_active=bool(label), oxidative_burst=oxidative)
                if label:
                    activation_count += 1
                if sim.check_events(state, activation_count)['senescent']:
                    break

    return rows

def generate_dataset_mp(sim: TelomereSimulator,
                         n_cells: int = 20000,
                         max_cycles: int = 100,
                         policy: str = 'threshold',
                         threshold_on: float = 120.0,
                         lineage_size: int = 5,
                         n_workers: int = None) -> pd.DataFrame:
    """Generate large datasets using multiprocessing.

    Splits the total number of lineages across available CPU cores.
    Returns a combined pandas DataFrame.
    """
    if n_workers is None:
        n_workers = max(1, cpu_count() - 1)

    n_lineages = n_cells // lineage_size
    chunk = math.ceil(n_lineages / n_workers)

    tasks = []
    for w in range(n_workers):
        start = w * chunk
        end = min((w + 1) * chunk, n_lineages)
        if start >= end:
            continue
        tasks.append((sim, start, end, max_cycles, policy, threshold_on, lineage_size))

    with Pool(processes=n_workers) as pool:
        all_rows = pool.map(_simulate_lineage, tasks)

    # flatten results
    flat = [item for sublist in all_rows for item in sublist]
    return pd.DataFrame(flat)


# ----------------------- PyTorch model ---------------------------------------

class SimpleLSTM(nn.Module):
    def __init__(self, input_dim=3, hidden_dim=64, n_layers=1, dropout=0.1):
        super().__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers=n_layers, batch_first=True, dropout=dropout)
        self.fc = nn.Sequential(
            nn.Linear(hidden_dim, hidden_dim//2),
            nn.ReLU(),
            nn.Linear(hidden_dim//2, 1),
        )

    def forward(self, x):
        # x: batch, seq_len, features
        out, _ = self.lstm(x)
        # take last timestep
        last = out[:, -1, :]
        return self.fc(last).squeeze(-1)


class TimeSeriesDataset(Dataset):
    def __init__(self, df: pd.DataFrame, seq_len: int = 8):
        # sort by synthetic trajectory: we only have flattened rows, so group by runs by cell_id
        self.seq_len = seq_len
        self.features = []
        self.labels = []

        # Iterate through each cell's trajectory
        for cell_id, cell_df in df.groupby('cell_id'):
            arr = cell_df[['telomere_length', 'age', 'damage', 'stress_index', 'oxidative_events', 'activation_count', 'label']].values
            n = len(arr)
            # Create sliding windows for each cell's trajectory
            for i in range(0, n - seq_len - 1):
                window = arr[i:i+seq_len]
                next_label = arr[i+seq_len][6] # 'label' is the 7th column (index 6)
                # Features for LSTM: telomere_length, age, damage, stress_index, oxidative_events, activation_count
                X = window[:, :6].astype(np.float32) # Take first 6 columns as features
                y = float(next_label)
                self.features.append(X)
                self.labels.append(y)

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        return torch.from_numpy(self.features[idx]), torch.tensor(self.labels[idx], dtype=torch.float32)


def train_model(df: pd.DataFrame,
                seq_len: int = 8,
                lr: float = 1e-3,
                batch_size: int = 128,
                epochs: int = 8,
                device: str = 'cpu') -> Tuple[nn.Module, Dict[str, Any]]:
    train_df, val_df = train_test_split(df, test_size=0.2, random_state=42, shuffle=True)
    train_ds = TimeSeriesDataset(train_df, seq_len=seq_len)
    val_ds = TimeSeriesDataset(val_df, seq_len=seq_len)
    train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_ds, batch_size=batch_size, shuffle=False)

    # Adjust input_dim based on the new features in TimeSeriesDataset
    model = SimpleLSTM(input_dim=6, hidden_dim=64).to(device)
    opt = torch.optim.Adam(model.parameters(), lr=lr)
    criterion = nn.BCEWithLogitsLoss()

    history = {'train_loss': [], 'val_loss': [], 'val_auc': []}
    for ep in range(epochs):
        model.train()
        running_loss = 0.0
        for Xb, yb in train_loader:
            Xb = Xb.to(device)
            yb = yb.to(device)
            logits = model(Xb)
            loss = criterion(logits, yb)
            opt.zero_grad()
            loss.backward()
            opt.step()
            running_loss += loss.item() * Xb.size(0)
        avg_train_loss = running_loss / len(train_loader.dataset)

        # validation
        model.eval()
        val_loss = 0.0
        preds, trues = [], []
        with torch.no_grad():
            for Xv, yv in val_loader:
                Xv = Xv.to(device)
                yv = yv.to(device)
                logits = model(Xv)
                loss = criterion(logits, yv)
                val_loss += loss.item() * Xv.size(0)
                probs = torch.sigmoid(logits).cpu().numpy()
                preds.extend(probs.tolist())
                trues.extend(yv.cpu().numpy().tolist())
        avg_val_loss = val_loss / len(val_loader.dataset)
        try:
            auc = roc_auc_score(trues, preds)
        except Exception:
            auc = 0.5

        history['train_loss'].append(avg_train_loss)
        history['val_loss'].append(avg_val_loss)
        history['val_auc'].append(auc)

        print(f"Epoch {ep+1}/{epochs}  train_loss={avg_train_loss:.4f}  val_loss={avg_val_loss:.4f}  val_auc={auc:.4f}")

    return model, history


# ----------------------- Quick demonstration / main --------------------------

if __name__ == '__main__':
    # Quick smoke-run to produce a tiny dataset and train quickly. Adjust parameters for real experiments.
    sim = TelomereSimulator()
    print("Generating synthetic dataset (this may take a moment)...")
    # Updated generate_dataset call
    df = generate_dataset(sim, n_cells=600, max_cycles=60, policy='threshold', threshold_on=120.0, lineage_size=5)
    print("Dataset size:", len(df))

    # Inspect distribution
    print(df[['telomere_length','damage','label']].describe())

    # Train with a higher learning rate option
    lr = 5e-3  # "higher" learning rate - tune carefully
    model, history = train_model(df, seq_len=6, lr=lr, batch_size=64, epochs=6)

    # Plot training history
    plt.figure()
    plt.plot(history['train_loss'], label='train_loss')
    plt.plot(history['val_loss'], label='val_loss')
    plt.legend()
    plt.title('Training loss')
    plt.show()

    # Generate candidate sequences
    sg = SequenceGenerator()
    candidates = sg.generate_candidates(n=50)
    # sort by simulated stability
    candidates_sorted = sorted(candidates, key=lambda x: x[1], reverse=True)
    print("Top sequence stability scores (top 5):")
    for seq, score in candidates_sorted[:5]:
        print(score)

    # example inference on a single trajectory
    # updated sample to include all features used for training
    sample = df.sample(12)
    sample_sorted = sample.sort_values(['cycle']).head(8)[['telomere_length','age','damage','stress_index','oxidative_events','activation_count']].values.astype(np.float32)
    X = torch.from_numpy(sample_sorted[np.newaxis, :, :])
    model.eval()
    with torch.no_grad():
        logit = model(X)
        prob = torch.sigmoid(logit).item()
    print(f"Model predicts telomerase activation probability next cycle: {prob:.3f}")

    print("""Done.\nNotes: this framework is intentionally heuristic; replace or extend subcomponents with\n          domain-specific simulators or more elaborate policies as you progress.""")

# Synthetic Multimodal Imaging Module
"""python
import numpy as np
from PIL import Image, ImageDraw, ImageFilter

# Virtual Cell Imaging Generator
# (Synthetic microscopy-like textures for multimodal ML training)

def generate_cell_image(telomere_length: float,
                        damage: float,
                        activation: int,
                        size: int = 256) -> Image.Image:
    img = Image.new("RGB", (size, size), "black")
    draw = ImageDraw.Draw(img)
    r = int((telomere_length / 15000) * (size/3)) + 10
    cx, cy = size//2, size//2
    draw.ellipse((cx-r, cy-r, cx+r, cy+r), outline=(180, 180, 255), width=3)
    glow_strength = min(activation, 10) * 8
    if glow_strength > 0:
        glow = Image.new("RGB", (size, size), (50, 50, 120))
        img = Image.blend(img, glow, glow_strength/100)
    noise = np.random.randint(0, int(50 + damage*3), (size, size), dtype=np.uint8)
    noise_img = Image.fromarray(np.stack([noise]*3, axis=-1))
    img = Image.blend(img, noise_img, 0.08)
    img = img.filter(ImageFilter.GaussianBlur(radius=1.2))
    return img

# Multimodal Sample Generator

def generate_multimodal_sample(record: dict, idx: int, out_dir: str = "multimodal_data"):
    import os
    os.makedirs(out_dir, exist_ok=True)
    img = generate_cell_image(
        telomere_length=record.get("current_length", 8000),
        damage=record.get("damage", 0),
        activation=record.get("activation_count", 0)
    )
    img.save(f"{out_dir}/cell_{record.get('cell_id','x')}_{idx}.png")

# Multimodal Batch Exporter

def export_multimodal_from_dataset(df, out_dir="multimodal_data"):
    for idx, row in df.iterrows():
        generate_multimodal_sample(row, idx, out_dir)
"""


# ------------------------------------------------------
# Telomere Texture Encoder (Safe, Synthetic-only)
# ------------------------------------------------------
# PyTorch encoder to transform synthetic microscopy-like images into a compact embedding
# that captures visual telomere-related texture. Designed for multimodal ML (image + tabular).
# This encoder is explicitly synthetic-only and not intended for any wet-lab or diagnostic use.
# ------------------------------------------------------

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import os

class TelomereImageDataset(Dataset):
    """Loads images generated by the multimodal exporter and pairs them with tabular features.
    Expects images named like: cell_<cell_id>_<idx>.png and a dataframe `df` with matching rows.
    """
    def __init__(self, df, img_dir='multimodal_data', transform=None):
        self.df = df.reset_index(drop=True)
        self.img_dir = img_dir
        self.transform = transform or transforms.Compose([
            transforms.Resize((128,128)),
            transforms.ToTensor(),
        ])

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        # find image path (best-effort)
        pattern = f"cell_{row.get('cell_id')}_"
        files = [f for f in os.listdir(self.img_dir) if f.startswith(pattern)]
        img_path = os.path.join(self.img_dir, files[0]) if files else None
        if img_path and os.path.exists(img_path):
            img = Image.open(img_path).convert('RGB')
            img = self.transform(img)
        else:
            # fallback: zero image
            img = torch.zeros(3,128,128)

        tab = torch.tensor([row.get('telomere_length',0), row.get('damage',0), row.get('stress_index',0)], dtype=torch.float32)
        label = torch.tensor(row.get('label',0), dtype=torch.float32)
        return img, tab, label


class TelomereEncoderCNN(nn.Module):
    """Compact CNN encoder producing a fixed-size embedding from images."""
    def __init__(self, embedding_dim: int = 128):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((4,4)),
        )
        self.fc = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128*4*4, 512),
            nn.ReLU(),
            nn.Linear(512, embedding_dim)
        )

    def forward(self, x):
        x = self.conv(x)
        x = self.fc(x)
        x = F.normalize(x, dim=1)
        return x


class MultimodalClassifier(nn.Module):
    """Combine image embedding + tabular features for a supervised objective (label prediction).
    This can be used as a proxy task to train the encoder.
    """
    def __init__(self, encoder: TelomereEncoderCNN, tab_dim: int = 3, embedding_dim: int = 128):
        super().__init__()
        self.encoder = encoder
        self.head = nn.Sequential(
            nn.Linear(embedding_dim + tab_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 1)
        )

    def forward(self, img, tab):
        emb = self.encoder(img)
        x = torch.cat([emb, tab], dim=1)
        return self.head(x).squeeze(1), emb


def train_encoder(df, img_dir='multimodal_data', epochs=10, batch_size=64, lr=1e-3, device='cpu'):
    ds = TelomereImageDataset(df, img_dir=img_dir)
    loader = DataLoader(ds, batch_size=batch_size, shuffle=True)
    encoder = TelomereEncoderCNN(embedding_dim=128).to(device)
    model = MultimodalClassifier(encoder, tab_dim=3, embedding_dim=128).to(device)

    opt = torch.optim.Adam(model.parameters(), lr=lr)
    crit = nn.BCEWithLogitsLoss()

    for ep in range(epochs):
        model.train()
        total_loss = 0.0
        for img, tab, label in loader:
            img = img.to(device)
            tab = tab.to(device)
            label = label.to(device)
            logits, emb = model(img, tab)
            loss = crit(logits, label)
            opt.zero_grad()
            loss.backward()
            opt.step()
            total_loss += loss.item() * img.size(0)
        print(f"Epoch {ep+1}/{epochs} loss={(total_loss/len(loader.dataset)):.4f}")

    # Return trained encoder for downstream use
    return encoder


# Safety note:
# This encoder operates entirely on synthetic images created by earlier modules. It is
# explicitly designed for computational experiments and feature learning on procedurally
# generated data. It must NOT be used for any form of biological diagnosis, wet-lab
# guidance, or medical decision-making.



# ----------------------- Evolving Cell + Efficient Optimizers ------------------
"""
Evolving cell framework
- Evolutionary Strategy (ES) to evolve a small parameterized policy for telomerase activation
- Supervised / gradient-based optimizer (Adam) to train a neural policy on simulation data
- Fitness = longevity (cycles survived) - risk_penalty * risk_score

All code is synthetic-only and intended for in-silico experiments.
"""

import copy

# --- Policy representations ----------------------------------------------------

class ThresholdPolicy:
    """Simple policy: activate telomerase when telomere_length < threshold."""
    def __init__(self, threshold: float = 120.0):
        self.threshold = threshold

    def act(self, state: CellState) -> int:
        return 1 if state.telomere_length < self.threshold else 0

    def params(self):
        return np.array([self.threshold], dtype=np.float32)

    def set_params(self, p: np.ndarray):
        self.threshold = float(p[0])


class LinearPolicy:
    """Linear policy on features: wx + b -> sigmoid -> threshold 0.5"""
    def __init__(self, dim=3):
        self.w = np.random.randn(dim).astype(np.float32)
        self.b = np.random.randn(1).astype(np.float32)

    def act(self, state: CellState) -> int:
        x = np.array([state.telomere_length, state.damage, state.age], dtype=np.float32)
        s = 1.0 / (1.0 + np.exp(- (self.w.dot(x) + self.b)))
        return int(s > 0.5)

    def params(self):
        return np.concatenate([self.w, self.b])

    def set_params(self, p: np.ndarray):
        self.w = p[:len(self.w)]
        self.b = p[len(self.w):]


# --- Evolutionary Strategy (ES) -----------------------------------------------

def evaluate_policy(sim: TelomereSimulator, policy, max_cycles=200, early_stop=True):
    """Run one simulated cell under `policy`, return fitness and diagnostic info."""
    state = CellState(telomere_length=sim.init_length * random.uniform(0.9,1.1), age=0, damage=random.uniform(0,0.05))
    activation_count = 0
    for cycle in range(max_cycles):
        action = policy.act(state)
        state = sim.step(state, telomerase_active=bool(action), oxidative_burst=0.0)
        if action:
            activation_count += 1
        ev = sim.check_events(state, activation_count)
        if ev['senescent']:
            break
    # fitness: cycles lived (age) penalized by risk_score and activation_count
    risk = ev.get('risk_score', 0.0)
    fitness = state.age - 2.0 * risk - 0.01 * activation_count
    return fitness, {'age': state.age, 'risk': risk, 'activation_count': activation_count}


def es_optimize(sim: TelomereSimulator, policy_class, n_params: int, pop_size: int = 50, iters: int = 100, sigma: float = 1.0, lr: float = 0.1):
    """Simple OpenAI-ES style optimizer for policy parameter vectors.
    Returns best parameter vector and history.
    """
    # initialize mean
    mean = np.zeros(n_params, dtype=np.float32)
    history = []
    for it in range(iters):
        noises = np.random.randn(pop_size, n_params).astype(np.float32)
        params_pop = mean[None,:] + sigma * noises
        rewards = np.zeros(pop_size, dtype=np.float32)
        for i in range(pop_size):
            pol = policy_class()
            pol.set_params(params_pop[i])
            r, _ = evaluate_policy(sim, pol)
            rewards[i] = r
        # normalize rewards
        A = (rewards - rewards.mean()) / (rewards.std() + 1e-8)
        grad = np.dot(A, noises) / pop_size
        mean = mean + lr * grad
        best_idx = int(np.argmax(rewards))
        best_r = float(rewards[best_idx])
        history.append({'iter': it, 'best_reward': best_r, 'mean': mean.copy()})
        if it % max(1, iters//10) == 0:
            print(f"ES iter {it}/{iters} best_reward={best_r:.3f}")
    return mean, history


# --- Gradient-based policy training (supervised from simulated labels) --------

class PolicyNet(nn.Module):
    def __init__(self, input_dim=3, hidden=64):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden),
            nn.ReLU(),
            nn.Linear(hidden, hidden//2),
            nn.ReLU(),
            nn.Linear(hidden//2, 1)
        )

    def forward(self, x):
        return self.net(x).squeeze(1)


def build_supervised_dataset(df, seq_len=1):
    """Create a simple dataset for supervised learning: features -> label (activate now).
    We use current features and the label column from the simulation.
    """
    X = df[['telomere_length','damage','age']].astype(np.float32).values
    y = df['label'].astype(np.float32).values
    return X, y


def train_policy_supervised(df, epochs=20, lr=1e-3, batch_size=256, device='cpu'):
    X, y = build_supervised_dataset(df)
    ds = torch.utils.data.TensorDataset(torch.from_numpy(X), torch.from_numpy(y))
    loader = DataLoader(ds, batch_size=batch_size, shuffle=True)
    model = PolicyNet(input_dim=X.shape[1]).to(device)
    opt = torch.optim.Adam(model.parameters(), lr=lr)
    crit = nn.BCEWithLogitsLoss()
    for ep in range(epochs):
        total = 0.0
        for xb, yb in loader:
            xb = xb.to(device)
            yb = yb.to(device)
            logits = model(xb)
            loss = crit(logits, yb)
            opt.zero_grad()
            loss.backward()
            opt.step()
            total += loss.item() * xb.size(0)
        print(f"Supervised epoch {ep+1}/{epochs} loss={(total/len(ds)):.4f}")
    return model


# --- Quick demo: run ES, then supervised training --------------------------------
if __name__ == '__main__':
    sim = TelomereSimulator()
    print('\n=== Evolutionary Strategy optimizing threshold policy ===')
    # ES to find a good threshold
    mean_params, hist = es_optimize(sim, lambda: ThresholdPolicy(), n_params=1, pop_size=40, iters=40, sigma=10.0, lr=0.5)
    print('ES result threshold:', mean_params)

    # Generate a dataset using the policy found (to train a supervised network)
    best_policy = ThresholdPolicy(mean_params[0])
    df = generate_dataset(sim, n_cells=2000, max_cycles=80, policy='threshold', threshold_on=best_policy.threshold)
    print('Generated dataset size:', len(df))

    print('\n=== Supervised training of policy network on generated dataset ===')
    device = 'cpu'
    model = train_policy_supervised(df, epochs=6, lr=1e-3, batch_size=128, device=device)

    # Evaluate learned policy using the simulator
    def model_policy_act(state):
        x = torch.tensor([[state.telomere_length, state.damage, state.age]], dtype=torch.float32)
        with torch.no_grad():
            logit = model(x).item()
        return 1 if torch.sigmoid(torch.tensor(logit)) > 0.5 else 0

    class ModelPolicyWrapper:
        def act(self, state):
            return model_policy_act(state)

    fitness, diag = evaluate_policy(sim, ModelPolicyWrapper())
    print('Model policy fitness:', fitness, diag)

    print("""\nDone. Replace or extend the objective and constraints as you see fit.""")
```

```python
"""
virtual_cell_foundational_model.py

Foundational, purely in-silico framework to simulate telomere dynamics, a "virtual telomerase"
control policy, generate synthetic "resistant" repetitive DNA sequences, and train a
deep-learning model (PyTorch) to predict when to activate telomerase.

This is a simulation + ML sandbox - intended for computational experiments only.
Do NOT use for wet-lab protocols or biological manipulation.

Requirements:
- Python 3.8+
- numpy, pandas, matplotlib, torch, scikit-learn

High-level components:
1. TelomereSimulator: stochastic model of telomere length across divisions, oxidative stress,
   and telomerase action.
2. SequenceGenerator: create synthetic repetitive telomeric-like sequences and score them for
   in-silico "stability" (heuristic score based on GC content, repeat periodicity and entropy).
3. Dataset builder: run many simulated cells with varied policies to build train / val sets.
4. PyTorch predictor: small LSTM/Transformer that predicts whether telomerase should activate
   next cycle given recent state features. Training loop with adjustable learning rate.
5. Example run at bottom demonstrating quick training and evaluation.

Adapt and expand: replace heuristics with biologically-derived features or experimental
simulations (in-silico chemistry engines) if available. Keep experiments computational only.
"""

import math
import random
import time
from dataclasses import dataclass
from typing import List, Tuple, Dict, Any

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, accuracy_score

# ----------------------- Simulation core ---------------------------------

@dataclass
class CellState:
    telomere_length: float  # abstract units (repeat units)
    age: int                # cell divisions experienced
    damage: float           # accumulated damage (0..1)


class TelomereSimulator:
    """Simple stochastic telomere model.

    - Each division shortens telomere by `shorten_mean` +/- `shorten_var`.
    - Oxidative damage increases damage; damage increases shortening rate.
    - Virtual telomerase when active adds repeats (with efficieny factor).
    - If telomere_length < senescence_thresh -> senescence event.
    - If telomerase_active too often or telomere exceeds cancer_thresh -> risk_of_uncontrolled_growth.
    """

    def __init__(self,
                 init_length: float = 200.0,
                 shorten_mean: float = 5.0,
                 shorten_std: float = 1.5,
                 damage_rate: float = 0.01,
                 damage_var: float = 0.005,
                 telomerase_add_mean: float = 50.0,
                 telomerase_add_std: float = 8.0,
                 senescence_thresh: float = 50.0,
                 cancer_thresh: float = 1000.0):
        self.init_length = init_length
        self.shorten_mean = shorten_mean
        self.shorten_std = shorten_std
        self.damage_rate = damage_rate
        self.damage_var = damage_var
        self.telomerase_add_mean = telomerase_add_mean
        self.telomerase_add_std = telomerase_add_std
        self.senescence_thresh = senescence_thresh
        self.cancer_thresh = cancer_thresh

    def step(self, state: CellState, telomerase_active: bool, oxidative_burst: float = 0.0) -> CellState:
        """Advance one division/cycle and return new state."""
        # Natural shortening influenced by damage
        shorten = np.random.normal(self.shorten_mean * (1.0 + state.damage), self.shorten_std)
        shorten += oxidative_burst  # occasional extra shortening
        shorten = max(0.0, shorten)

        new_length = state.telomere_length - shorten

        # Telomerase effect
        if telomerase_active:
            add = np.random.normal(self.telomerase_add_mean, self.telomerase_add_std)
            add = max(0.0, add)
            new_length += add

        # Damage accumulation
        damage_inc = np.random.normal(self.damage_rate, self.damage_var)
        damage_inc = max(0.0, damage_inc) + oxidative_burst * 0.001
        new_damage = min(1.0, state.damage + damage_inc)

        new_age = state.age + 1

        return CellState(telomere_length=new_length, age=new_age, damage=new_damage)

    def check_events(self, state: CellState, telomerase_activation_count: int) -> Dict[str, Any]:
        """Return events: senescence, uncontrolled_growth_risk (simple heuristic)."""
        senescent = state.telomere_length <= self.senescence_thresh
        # If telomere grows above cancer threshold OR telomerase used too frequently -> risk
        risk_score = 0.0
        if state.telomere_length >= self.cancer_thresh:
            risk_score += 1.0
        # scale risk with activation frequency and low damage control
        if telomerase_activation_count > max(1, state.age * 0.3):
            risk_score += 0.5
        return {"senescent": senescent, "risk_score": risk_score}


# ----------------------- Sequence heuristics ---------------------------------

class SequenceGenerator:
    """Generate repetitive 'telomeric-like' sequences and a heuristic stability score.

    Heuristic considers:
    - GC content (higher GC can increase stability in some contexts)
    - Repeat periodicity (clean repeats better)
    - Shannon entropy (lower sometimes more stable for repeats)
    """

    bases = ['A', 'C', 'G', 'T']

    def make_repeat(self, motif: str, n_repeat: int) -> str:
        return motif * n_repeat

    def random_motif(self, length: int = 3) -> str:
        return ''.join(random.choices(self.bases, k=length))

    def stability_score(self, seq: str) -> float:
        # GC content
        gc = (seq.count('G') + seq.count('C')) / max(1, len(seq))
        # entropy
        freqs = [seq.count(b) / len(seq) for b in self.bases]
        entropy = -sum([p * math.log2(p) for p in freqs if p > 0])
        # periodicity: check best small motif that reconstructs the string
        best_period = 0
        for p in range(1, min(8, len(seq))):
            motif = seq[:p]
            if motif * (len(seq) // p) == seq[:p * (len(seq) // p)]:
                best_period = p
                break
        # heuristic combination: higher GC, lower entropy, clear periodic motif preferred
        score = (gc * 0.6) + ((2.0 - entropy) * 0.2) + ((1.0 if best_period > 0 else 0.0) * 0.2)
        # normalize roughly 0..1
        return max(0.0, min(1.0, score))

    def generate_candidates(self, n: int = 100, motif_lengths: List[int] = [2,3,4]) -> List[Tuple[str,float]]:
        results = []
        for _ in range(n):
            mlen = random.choice(motif_lengths)
            motif = self.random_motif(mlen)
            repeats = random.randint(10, 200)
            seq = self.make_repeat(motif, repeats)
            score = self.stability_score(seq)
            results.append((seq, score))
        return results


# ----------------------- Data generation for ML -------------------------------

def generate_dataset(sim: TelomereSimulator,
                     n_cells: int = 2000,
                     max_cycles: int = 100,
                     policy: str = 'threshold',
                     threshold_on: float = 120.0,
                     lineage_size: int = 5) -> pd.DataFrame:
    """Expanded dataset generator.

    Additions:
    - cell_id: unique identifier for each simulated cell
    - lineage_id: simulate small families of related cells
    - oxidative_events: count of oxidative bursts per cycle
    - telomerase_activation_count: running count of activations
    - extra engineered feature: "stress_index" = damage * (initial_length / current_length)
    """
    rows = []
    cell_id = 0
    for lineage in range(n_cells // lineage_size):
        for _ in range(lineage_size):
            cell_id += 1
            state = CellState(telomere_length=sim.init_length * random.uniform(0.8, 1.2),
                               age=0,
                               damage=random.uniform(0.0,0.1))
            activation_count = 0
            oxidative_events = 0
            for cycle in range(max_cycles):
                # determine label
                if policy == 'threshold':
                    label = 1 if state.telomere_length < threshold_on else 0
                elif policy == 'heuristic':
                    label = 1 if (state.damage > 0.2 and state.telomere_length < sim.init_length*0.9) else 0
                else:
                    label = 1 if random.random() < 0.05 else 0

                oxidative = 0.0
                if random.random() < 0.03:
                    oxidative = random.uniform(0.0, 10.0)
                    oxidative_events += 1

                # engineered feature
                stress_index = state.damage * (sim.init_length / max(1e-3, state.telomere_length))

                rows.append({
                    'cell_id': cell_id,
                    'lineage_id': lineage,
                    'telomere_length': state.telomere_length,
                    'age': state.age,
                    'damage': state.damage,
                    'cycle': cycle,
                    'oxidative_events': oxidative_events,
                    'activation_count': activation_count,
                    'stress_index': stress_index,
                    'label': label
                })

                state = sim.step(state, telomerase_active=bool(label), oxidative_burst=oxidative)
                if label:
                    activation_count += 1
                ev = sim.check_events(state, activation_count)
                if ev['senescent']:
                    break

    df = pd.DataFrame(rows)
    return df

# ----------------------- Multiprocessing dataset generation --------------------
from multiprocessing import Pool, cpu_count

def _simulate_lineage(args):
    """Worker function to simulate a subset of lineages in parallel."""
    sim, lineage_start, lineage_end, max_cycles, policy, threshold_on, lineage_size = args
    rows = []
    cell_id = lineage_start * lineage_size
    for lineage in range(lineage_start, lineage_end):
        for _ in range(lineage_size):
            cell_id += 1
            state = CellState(telomere_length=sim.init_length * random.uniform(0.8, 1.2),
                               age=0,
                               damage=random.uniform(0.0,0.1))
            activation_count = 0
            oxidative_events = 0
            for cycle in range(max_cycles):
                if policy == 'threshold':
                    label = 1 if state.telomere_length < threshold_on else 0
                elif policy == 'heuristic':
                    label = 1 if (state.damage > 0.2 and state.telomere_length < sim.init_length*0.9) else 0
                else:
                    label = 1 if random.random() < 0.05 else 0

                oxidative = 0.0
                if random.random() < 0.03:
                    oxidative = random.uniform(0.0, 10.0)
                    oxidative_events += 1

                stress_index = state.damage * (sim.init_length / max(1e-3, state.telomere_length))

                rows.append({
                    'cell_id': cell_id,
                    'lineage_id': lineage,
                    'telomere_length': state.telomere_length,
                    'age': state.age,
                    'damage': state.damage,
                    'cycle': cycle,
                    'oxidative_events': oxidative_events,
                    'activation_count': activation_count,
                    'stress_index': stress_index,
                    'label': label
                })

                state = sim.step(state, telomerase_active=bool(label), oxidative_burst=oxidative)
                if label:
                    activation_count += 1
                if sim.check_events(state, activation_count)['senescent']:
                    break

    return rows

def generate_dataset_mp(sim: TelomereSimulator,
                         n_cells: int = 20000,
                         max_cycles: int = 100,
                         policy: str = 'threshold',
                         threshold_on: float = 120.0,
                         lineage_size: int = 5,
                         n_workers: int = None) -> pd.DataFrame:
    """Generate large datasets using multiprocessing.

    Splits the total number of lineages across available CPU cores.
    Returns a combined pandas DataFrame.
    """
    if n_workers is None:
        n_workers = max(1, cpu_count() - 1)

    n_lineages = n_cells // lineage_size
    chunk = math.ceil(n_lineages / n_workers)

    tasks = []
    for w in range(n_workers):
        start = w * chunk
        end = min((w + 1) * chunk, n_lineages)
        if start >= end:
            continue
        tasks.append((sim, start, end, max_cycles, policy, threshold_on, lineage_size))

    with Pool(processes=n_workers) as pool:
        all_rows = pool.map(_simulate_lineage, tasks)

    # flatten results
    flat = [item for sublist in all_rows for item in sublist]
    return pd.DataFrame(flat)


# ----------------------- PyTorch model ---------------------------------------

class SimpleLSTM(nn.Module):
    def __init__(self, input_dim=3, hidden_dim=64, n_layers=1, dropout=0.1):
        super().__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers=n_layers, batch_first=True, dropout=dropout)
        self.fc = nn.Sequential(
            nn.Linear(hidden_dim, hidden_dim//2),
            nn.ReLU(),
            nn.Linear(hidden_dim//2, 1),
        )

    def forward(self, x):
        # x: batch, seq_len, features
        out, _ = self.lstm(x)
        # take last timestep
        last = out[:, -1, :]
        return self.fc(last).squeeze(-1)


class TimeSeriesDataset(Dataset):
    def __init__(self, df: pd.DataFrame, seq_len: int = 8):
        # sort by synthetic trajectory: we only have flattened rows, so group by runs by cell_id
        self.seq_len = seq_len
        self.features = []
        self.labels = []

        # Iterate through each cell's trajectory
        for cell_id, cell_df in df.groupby('cell_id'):
            arr = cell_df[['telomere_length', 'age', 'damage', 'stress_index', 'oxidative_events', 'activation_count', 'label']].values
            n = len(arr)
            # Create sliding windows for each cell's trajectory
            for i in range(0, n - seq_len - 1):
                window = arr[i:i+seq_len]
                next_label = arr[i+seq_len][6] # 'label' is the 7th column (index 6)
                # Features for LSTM: telomere_length, age, damage, stress_index, oxidative_events, activation_count
                X = window[:, :6].astype(np.float32) # Take first 6 columns as features
                y = float(next_label)
                self.features.append(X)
                self.labels.append(y)

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        return torch.from_numpy(self.features[idx]), torch.tensor(self.labels[idx], dtype=torch.float32)


def train_model(df: pd.DataFrame,
                seq_len: int = 8,
                lr: float = 1e-3,
                batch_size: int = 128,
                epochs: int = 8,
                device: str = 'cpu') -> Tuple[nn.Module, Dict[str, Any]]:
    train_df, val_df = train_test_split(df, test_size=0.2, random_state=42, shuffle=True)
    train_ds = TimeSeriesDataset(train_df, seq_len=seq_len)
    val_ds = TimeSeriesDataset(val_df, seq_len=seq_len)
    train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_ds, batch_size=batch_size, shuffle=False)

    # Adjust input_dim based on the new features in TimeSeriesDataset
    model = SimpleLSTM(input_dim=6, hidden_dim=64).to(device)
    opt = torch.optim.Adam(model.parameters(), lr=lr)
    criterion = nn.BCEWithLogitsLoss()

    history = {'train_loss': [], 'val_loss': [], 'val_auc': []}
    for ep in range(epochs):
        model.train()
        running_loss = 0.0
        for Xb, yb in train_loader:
            Xb = Xb.to(device)
            yb = yb.to(device)
            logits = model(Xb)
            loss = criterion(logits, yb)
            opt.zero_grad()
            loss.backward()
            opt.step()
            running_loss += loss.item() * Xb.size(0)
        avg_train_loss = running_loss / len(train_loader.dataset)

        # validation
        model.eval()
        val_loss = 0.0
        preds, trues = [], []
        with torch.no_grad():
            for Xv, yv in val_loader:
                Xv = Xv.to(device)
                yv = yv.to(device)
                logits = model(Xv)
                loss = criterion(logits, yv)
                val_loss += loss.item() * Xv.size(0)
                probs = torch.sigmoid(logits).cpu().numpy()
                preds.extend(probs.tolist())
                trues.extend(yv.cpu().numpy().tolist())
        avg_val_loss = val_loss / len(val_loader.dataset)
        try:
            auc = roc_auc_score(trues, preds)
        except Exception:
            auc = 0.5

        history['train_loss'].append(avg_train_loss)
        history['val_loss'].append(avg_val_loss)
        history['val_auc'].append(auc)

        print(f"Epoch {ep+1}/{epochs}  train_loss={avg_train_loss:.4f}  val_loss={avg_val_loss:.4f}  val_auc={auc:.4f}")

    return model, history


# ----------------------- Quick demonstration / main --------------------------

if __name__ == '__main__':
    # Quick smoke-run to produce a tiny dataset and train quickly. Adjust parameters for real experiments.
    sim = TelomereSimulator()
    print("Generating synthetic dataset (this may take a moment)...")
    # Updated generate_dataset call
    df = generate_dataset(sim, n_cells=600, max_cycles=60, policy='threshold', threshold_on=120.0, lineage_size=5)
    print("Dataset size:", len(df))

    # Inspect distribution
    print(df[['telomere_length','damage','label']].describe())

    # Train with a higher learning rate option
    lr = 5e-3  # "higher" learning rate - tune carefully
    model, history = train_model(df, seq_len=6, lr=lr, batch_size=64, epochs=6)

    # Plot training history
    plt.figure()
    plt.plot(history['train_loss'], label='train_loss')
    plt.plot(history['val_loss'], label='val_loss')
    plt.legend()
    plt.title('Training loss')
    plt.show()

    # Generate candidate sequences
    sg = SequenceGenerator()
    candidates = sg.generate_candidates(n=50)
    # sort by simulated stability
    candidates_sorted = sorted(candidates, key=lambda x: x[1], reverse=True)
    print("Top sequence stability scores (top 5):")
    for seq, score in candidates_sorted[:5]:
        print(score)

    # example inference on a single trajectory
    # updated sample to include all features used for training
    sample = df.sample(12)
    sample_sorted = sample.sort_values(['cycle']).head(8)[['telomere_length','age','damage','stress_index','oxidative_events','activation_count']].values.astype(np.float32)
    X = torch.from_numpy(sample_sorted[np.newaxis, :, :])
    model.eval()
    with torch.no_grad():
        logit = model(X)
        prob = torch.sigmoid(logit).item()
    print(f"Model predicts telomerase activation probability next cycle: {prob:.3f}")

    print("""Done.\nNotes: this framework is intentionally heuristic; replace or extend subcomponents with\n          domain-specific simulators or more elaborate policies as you progress.""")

# Synthetic Multimodal Imaging Module
import numpy as np
from PIL import Image, ImageDraw, ImageFilter

# Virtual Cell Imaging Generator
# (Synthetic microscopy-like textures for multimodal ML training)

def generate_cell_image(telomere_length: float,
                        damage: float,
                        activation: int,
                        size: int = 256) -> Image.Image:
    img = Image.new("RGB", (size, size), "black")
    draw = ImageDraw.Draw(img)
    r = int((telomere_length / 15000) * (size/3)) + 10
    cx, cy = size//2, size//2
    draw.ellipse((cx-r, cy-r, cx+r, cy+r), outline=(180, 180, 255), width=3)
    glow_strength = min(activation, 10) * 8
    if glow_strength > 0:
        glow = Image.new("RGB", (size, size), (50, 50, 120))
        img = Image.blend(img, glow, glow_strength/100)
    noise = np.random.randint(0, int(50 + damage*3), (size, size), dtype=np.uint8)
    noise_img = Image.fromarray(np.stack([noise]*3, axis=-1))
    img = Image.blend(img, noise_img, 0.08)
    img = img.filter(ImageFilter.GaussianBlur(radius=1.2))
    return img

# Multimodal Sample Generator

def generate_multimodal_sample(record: dict, idx: int, out_dir: str = "multimodal_data"):
    import os
    os.makedirs(out_dir, exist_ok=True)
    img = generate_cell_image(
        telomere_length=record.get("current_length", 8000),
        damage=record.get("damage", 0),
        activation=record.get("activation_count", 0)
    )
    img.save(f"{out_dir}/cell_{record.get('cell_id','x')}_{idx}.png")

# Multimodal Batch Exporter

def export_multimodal_from_dataset(df, out_dir="multimodal_data"):
    for idx, row in df.iterrows():
        generate_multimodal_sample(row, idx, out_dir)


# ------------------------------------------------------
# Telomere Texture Encoder (Safe, Synthetic-only)
# ------------------------------------------------------
# PyTorch encoder to transform synthetic microscopy-like images into a compact embedding
# that captures visual telomere-related texture. Designed for multimodal ML (image + tabular).
# This encoder is explicitly synthetic-only and not intended for any wet-lab or diagnostic use.
# ------------------------------------------------------

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import os

class TelomereImageDataset(Dataset):
    """Loads images generated by the multimodal exporter and pairs them with tabular features.
    Expects images named like: cell_<cell_id>_<idx>.png and a dataframe `df` with matching rows.
    """
    def __init__(self, df, img_dir='multimodal_data', transform=None):
        self.df = df.reset_index(drop=True)
        self.img_dir = img_dir
        self.transform = transform or transforms.Compose([
            transforms.Resize((128,128)),
            transforms.ToTensor(),
        ])

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        # find image path (best-effort)
        pattern = f"cell_{row.get('cell_id')}_"
        files = [f for f in os.listdir(self.img_dir) if f.startswith(pattern)]
        img_path = os.path.join(self.img_dir, files[0]) if files else None
        if img_path and os.path.exists(img_path):
            img = Image.open(img_path).convert('RGB')
            img = self.transform(img)
        else:
            # fallback: zero image
            img = torch.zeros(3,128,128)

        tab = torch.tensor([row.get('telomere_length',0), row.get('damage',0), row.get('stress_index',0)], dtype=torch.float32)
        label = torch.tensor(row.get('label',0), dtype=torch.float32)
        return img, tab, label


class TelomereEncoderCNN(nn.Module):
    """Compact CNN encoder producing a fixed-size embedding from images."""
    def __init__(self, embedding_dim: int = 128):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((4,4)),
        )
        self.fc = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128*4*4, 512),
            nn.ReLU(),
            nn.Linear(512, embedding_dim)
        )

    def forward(self, x):
        x = self.conv(x)
        x = self.fc(x)
        x = F.normalize(x, dim=1)
        return x


class MultimodalClassifier(nn.Module):
    """Combine image embedding + tabular features for a supervised objective (label prediction).
    This can be used as a proxy task to train the encoder.
    """
    def __init__(self, encoder: TelomereEncoderCNN, tab_dim: int = 3, embedding_dim: int = 128):
        super().__init__()
        self.encoder = encoder
        self.head = nn.Sequential(
            nn.Linear(embedding_dim + tab_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 1)
        )

    def forward(self, img, tab):
        emb = self.encoder(img)
        x = torch.cat([emb, tab], dim=1)
        return self.head(x).squeeze(1), emb


def train_encoder(df, img_dir='multimodal_data', epochs=10, batch_size=64, lr=1e-3, device='cpu'):
    ds = TelomereImageDataset(df, img_dir=img_dir)
    loader = DataLoader(ds, batch_size=batch_size, shuffle=True)
    encoder = TelomereEncoderCNN(embedding_dim=128).to(device)
    model = MultimodalClassifier(encoder, tab_dim=3, embedding_dim=128).to(device)

    opt = torch.optim.Adam(model.parameters(), lr=lr)
    crit = nn.BCEWithLogitsLoss()

    for ep in range(epochs):
        model.train()
        total_loss = 0.0
        for img, tab, label in loader:
            img = img.to(device)
            tab = tab.to(device)
            label = label.to(device)
            logits, emb = model(img, tab)
            loss = crit(logits, label)
            opt.zero_grad()
            loss.backward()
            opt.step()
            total_loss += loss.item() * img.size(0)
        print(f"Epoch {ep+1}/{epochs} loss={(total_loss/len(loader.dataset)):.4f}")

    # Return trained encoder for downstream use
    return encoder


# Safety note:
# This encoder operates entirely on synthetic images created by earlier modules. It is
# explicitly designed for computational experiments and feature learning on procedurally
# generated data. It must NOT be used for any form of biological diagnosis, wet-lab
# guidance, or medical decision-making.



# ----------------------- Evolving Cell + Efficient Optimizers ------------------
"""
Evolving cell framework
- Evolutionary Strategy (ES) to evolve a small parameterized policy for telomerase activation
- Supervised / gradient-based optimizer (Adam) to train a neural policy on simulation data
- Fitness = longevity (cycles survived) - risk_penalty * risk_score

All code is synthetic-only and intended for in-silico experiments.
"""

import copy

# --- Policy representations ----------------------------------------------------

class ThresholdPolicy:
    """Simple policy: activate telomerase when telomere_length < threshold."""
    def __init__(self, threshold: float = 120.0):
        self.threshold = threshold

    def act(self, state: CellState) -> int:
        return 1 if state.telomere_length < self.threshold else 0

    def params(self):
        return np.array([self.threshold], dtype=np.float32)

    def set_params(self, p: np.ndarray):
        self.threshold = float(p[0])


class LinearPolicy:
    """Linear policy on features: wx + b -> sigmoid -> threshold 0.5"""
    def __init__(self, dim=3):
        self.w = np.random.randn(dim).astype(np.float32)
        self.b = np.random.randn(1).astype(np.float32)

    def act(self, state: CellState) -> int:
        x = np.array([state.telomere_length, state.damage, state.age], dtype=np.float32)
        s = 1.0 / (1.0 + np.exp(- (self.w.dot(x) + self.b)))
        return int(s > 0.5)

    def params(self):
        return np.concatenate([self.w, self.b])

    def set_params(self, p: np.ndarray):
        self.w = p[:len(self.w)]
        self.b = p[len(self.w):]


# --- Evolutionary Strategy (ES) -----------------------------------------------

def evaluate_policy(sim: TelomereSimulator, policy, max_cycles=200, early_stop=True):
    """Run one simulated cell under `policy`, return fitness and diagnostic info."""
    state = CellState(telomere_length=sim.init_length * random.uniform(0.9,1.1), age=0, damage=random.uniform(0,0.05))
    activation_count = 0
    for cycle in range(max_cycles):
        action = policy.act(state)
        state = sim.step(state, telomerase_active=bool(action), oxidative_burst=0.0)
        if action:
            activation_count += 1
        ev = sim.check_events(state, activation_count)
        if ev['senescent']:
            break
    # fitness: cycles lived (age) penalized by risk_score and activation_count
    risk = ev.get('risk_score', 0.0)
    fitness = state.age - 2.0 * risk - 0.01 * activation_count
    return fitness, {'age': state.age, 'risk': risk, 'activation_count': activation_count}


def es_optimize(sim: TelomereSimulator, policy_class, n_params: int, pop_size: int = 50, iters: int = 100, sigma: float = 1.0, lr: float = 0.1):
    """Simple OpenAI-ES style optimizer for policy parameter vectors.
    Returns best parameter vector and history.
    """
    # initialize mean
    mean = np.zeros(n_params, dtype=np.float32)
    history = []
    for it in range(iters):
        noises = np.random.randn(pop_size, n_params).astype(np.float32)
        params_pop = mean[None,:] + sigma * noises
        rewards = np.zeros(pop_size, dtype=np.float32)
        for i in range(pop_size):
            pol = policy_class()
            pol.set_params(params_pop[i])
            r, _ = evaluate_policy(sim, pol)
            rewards[i] = r
        # normalize rewards
        A = (rewards - rewards.mean()) / (rewards.std() + 1e-8)
        grad = np.dot(A, noises) / pop_size
        mean = mean + lr * grad
        best_idx = int(np.argmax(rewards))
        best_r = float(rewards[best_idx])
        history.append({'iter': it, 'best_reward': best_r, 'mean': mean.copy()})
        if it % max(1, iters//10) == 0:
            print(f"ES iter {it}/{iters} best_reward={best_r:.3f}")
    return mean, history


# --- Gradient-based policy training (supervised from simulated labels) --------

class PolicyNet(nn.Module):
    def __init__(self, input_dim=3, hidden=64):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden),
            nn.ReLU(),
            nn.Linear(hidden, hidden//2),
            nn.ReLU(),
            nn.Linear(hidden//2, 1)
        )

    def forward(self, x):
        return self.net(x).squeeze(1)


def build_supervised_dataset(df, seq_len=1):
    """Create a simple dataset for supervised learning: features -> label (activate now).
    We use current features and the label column from the simulation.
    """
    X = df[['telomere_length','damage','age']].astype(np.float32).values
    y = df['label'].astype(np.float32).values
    return X, y


def train_policy_supervised(df, epochs=20, lr=1e-3, batch_size=256, device='cpu'):
    X, y = build_supervised_dataset(df)
    ds = torch.utils.data.TensorDataset(torch.from_numpy(X), torch.from_numpy(y))
    loader = DataLoader(ds, batch_size=batch_size, shuffle=True)
    model = PolicyNet(input_dim=X.shape[1]).to(device)
    opt = torch.optim.Adam(model.parameters(), lr=lr)
    crit = nn.BCEWithLogitsLoss()
    for ep in range(epochs):
        total = 0.0
        for xb, yb in loader:
            xb = xb.to(device)
            yb = yb.to(device)
            logits = model(xb)
            loss = crit(logits, yb)
            opt.zero_grad()
            loss.backward()
            opt.step()
            total += loss.item() * xb.size(0)
        print(f"Supervised epoch {ep+1}/{epochs} loss={(total/len(ds)):.4f}")
    return model


# --- Quick demo: run ES, then supervised training --------------------------------
if __name__ == '__main__':
    sim = TelomereSimulator()
    print('\n=== Evolutionary Strategy optimizing threshold policy ===')
    # ES to find a good threshold
    mean_params, hist = es_optimize(sim, lambda: ThresholdPolicy(), n_params=1, pop_size=40, iters=40, sigma=10.0, lr=0.5)
    print('ES result threshold:', mean_params)

    # Generate a dataset using the policy found (to train a supervised network)
    best_policy = ThresholdPolicy(mean_params[0])
    df = generate_dataset(sim, n_cells=2000, max_cycles=80, policy='threshold', threshold_on=best_policy.threshold)
    print('Generated dataset size:', len(df))

    print('\n=== Supervised training of policy network on generated dataset ===')
    device = 'cpu'
    model = train_policy_supervised(df, epochs=6, lr=1e-3, batch_size=128, device=device)

    # Evaluate learned policy using the simulator
    def model_policy_act(state):
        x = torch.tensor([[state.telomere_length, state.damage, state.age]], dtype=torch.float32)
        with torch.no_grad():
            logit = model(x).item()
        return 1 if torch.sigmoid(torch.tensor(logit)) > 0.5 else 0

    class ModelPolicyWrapper:
        def act(self, state):
            return model_policy_act(state)

    fitness, diag = evaluate_policy(sim, ModelPolicyWrapper())
    print('Model policy fitness:', fitness, diag)

    print("""\nDone. Replace or extend the objective and constraints as you see fit.""")
```

```python
# virtual_cell_pipeline.py
"""
Virtual Cell  in silico simulation + representation learning pipeline (safe, abstracted).
NOT biological instructions. All "sequences" and behaviors are synthetic proxies.
"""

import math
import random
import argparse
from typing import Tuple, List
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from tqdm import tqdm
from sklearn.metrics import roc_auc_score, accuracy_score
from sklearn.cluster import KMeans
import umap
import matplotlib.pyplot as plt
import os

# ----------------------------
# 1) Synthetic data generation
# ----------------------------
TOKEN_VOCAB = list("ACGTX")  # abstract tokens, not real DNA instruction
VOCAB_TO_IDX = {c: i+1 for i, c in enumerate(TOKEN_VOCAB)}  # reserve 0 for padding

def generate_abstract_sequence(length: int) -> str:
    # Create sequences where patterns correlate with "stability"
    # 'X' token acts as special repeat that increases simulated stability
    seq = []
    for i in range(length):
        p = random.random()
        if p < 0.05:
            seq.append('X')  # synthetic resistant motif
        else:
            seq.append(random.choice(['A','C','G','T']))
    return ''.join(seq)

def seq_to_indices(seq: str, maxlen: int):
    arr = [VOCAB_TO_IDX.get(c, 0) for c in seq][:maxlen]
    if len(arr) < maxlen:
        arr += [0] * (maxlen - len(arr))
    return arr

def simulate_telomere_dynamics(seq: str, rng_seed=None) -> Tuple[float, float, float, int]:
    # Abstract simulation returning:
    # - initial_length proxy
    # - oxidative_stress (0..1)
    # - repair_efficiency (0..1)
    # - label: 1 => healthy division / desirable; 0 => risk (uncontrolled or senescent)
    if rng_seed is not None:
        random.seed(rng_seed)
    # features derived from sequence:
    resistant_motif_count = seq.count('X')  # proxy for "resistant repeats"
    initial_length = 100 + resistant_motif_count * 15 + random.gauss(0, 5)
    oxidative_stress = min(max(random.betavariate(2,5) + 0.2*(1/(1+resistant_motif_count)), 0.0), 1.0)
    repair_eff = min(max(random.betavariate(5,2) + 0.1 * resistant_motif_count/10.0, 0.0), 1.0)
    # telomerase activation history (abstract)
    telomerase_activity_score = min(max(0.2 * resistant_motif_count + repair_eff - oxidative_stress + random.gauss(0,0.05), 0.0), 2.0)
    # label logic (abstract/rule-based + noise):
    score = 0.5* (initial_length/150.0) + 0.3*repair_eff - 0.4*oxidative_stress + 0.2*(telomerase_activity_score/2.0)
    prob_good = 1/(1+math.exp(-6*(score-0.4)))  # sigmoid with steeper slope
    label = 1 if random.random() < prob_good else 0
    return initial_length, oxidative_stress, repair_eff, label

class VirtualCellDataset(Dataset):
    def __init__(self, n_samples:int=5000, seq_len:int=60, maxlen:int=64, seed:int=42):
        self.samples = []
        random.seed(seed)
        np.random.seed(seed)
        for i in range(n_samples):
            s = generate_abstract_sequence(seq_len)
            initial_len, ox, repair, label = simulate_telomere_dynamics(s)
            numeric_feats = np.array([initial_len, ox, repair], dtype=np.float32)
            idxs = np.array(seq_to_indices(s, maxlen), dtype=np.int64)
            self.samples.append((idxs, numeric_feats, int(label)))
    def __len__(self): return len(self.samples)
    def __getitem__(self, i):
        return self.samples[i]

# ----------------------------
# 2) Model: little Transformer + numeric feature head
# ----------------------------
class SeqTransformer(nn.Module):
    def __init__(self, vocab_size:int, emb_dim:int=64, n_heads:int=4, num_layers:int=2, seq_maxlen:int=64, numeric_dim:int=3, hidden_dim:int=128, proj_dim:int=64):
        super().__init__()
        self.token_emb = nn.Embedding(vocab_size+1, emb_dim, padding_idx=0)  # +1 for padding idx=0
        self.pos_emb = nn.Embedding(seq_maxlen, emb_dim)
        encoder_layer = nn.TransformerEncoderLayer(d_model=emb_dim, nhead=n_heads, dim_feedforward=emb_dim*4, activation='gelu')
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        self.pool = nn.AdaptiveAvgPool1d(1)  # pool sequence to vector
        # numeric feature MLP
        self.num_mlp = nn.Sequential(
            nn.Linear(numeric_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, emb_dim)
        )
        # final projection
        self.project = nn.Sequential(
            nn.Linear(emb_dim*2, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, proj_dim)
        )
        # prediction head
        self.classifier = nn.Sequential(
            nn.Linear(proj_dim, hidden_dim//2),
            nn.ReLU(),
            nn.Linear(hidden_dim//2, 1)
        )

    def forward(self, token_idxs, numeric_feats):
        # token_idxs: [B, L]
        B, L = token_idxs.shape
        tok = self.token_emb(token_idxs)  # [B, L, E]
        pos = self.pos_emb(torch.arange(L, device=token_idxs.device).unsqueeze(0).expand(B,-1))
        x = tok + pos
        # transformer expects [L, B, E]
        x = x.permute(1,0,2)
        x = self.transformer(x)  # [L,B,E]
        x = x.permute(1,2,0)  # [B,E,L]
        x = self.pool(x).squeeze(-1)  # [B, E]
        num_v = self.num_mlp(numeric_feats)  # [B, E]
        joint = torch.cat([x, num_v], dim=1)  # [B, 2E]
        proj = self.project(joint)  # [B, proj_dim] -> embeddings
        logits = self.classifier(proj).squeeze(-1)
        return logits, proj  # return both for training and embedding extraction

# ----------------------------
# 3) Training loop & utilities
# ----------------------------
def train_one_epoch(model, loader, optim, device, criterion, scaler=None):
    model.train()
    losses = []
    for tokens, nums, labels in tqdm(loader, desc="train"):
        tokens = tokens.to(device)
        nums = nums.to(device)
        labels = labels.to(device).float()
        optim.zero_grad()
        logits, emb = model(tokens, nums)
        loss = criterion(logits, labels)
        loss.backward()
        optim.step()
        losses.append(loss.item())
    return float(np.mean(losses))

def eval_model(model, loader, device):
    model.eval()
    ys, preds, scores = [], [], []
    embs = []
    with torch.no_grad():
        for tokens, nums, labels in loader:
            tokens = tokens.to(device)
            nums = nums.to(device)
            labels = labels.numpy()
            logits, proj = model(tokens, nums)
            prob = torch.sigmoid(logits).cpu().numpy()
            preds.append((prob>0.5).astype(int))
            ys.append(labels)
            scores.append(prob)
            embs.append(proj.cpu().numpy())
    ys = np.concatenate(ys).ravel()
    preds = np.concatenate(preds).ravel()
    scores = np.concatenate(scores).ravel()
    embs = np.vstack(embs)
    auc = roc_auc_score(ys, scores)
    acc = accuracy_score(ys, preds)
    return {'auc': auc, 'acc': acc, 'embeddings': embs, 'labels': ys, 'scores': scores}

# ----------------------------
# 4) Main runnable pipeline
# ----------------------------
def main(args):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print("Device:", device)
    # create datasets
    ds = VirtualCellDataset(n_samples=args.n_samples, seq_len=args.seq_len, maxlen=args.maxlen, seed=args.seed)
    n = len(ds)
    idxs = list(range(n))
    random.shuffle(idxs)
    split1 = int(n*0.8)
    split2 = int(n*0.9)
    train_idx = idxs[:split1]
    val_idx = idxs[split1:split2]
    test_idx = idxs[split2:]
    def subset(ds, idxs):
        s = torch.utils.data.Subset(ds, idxs)
        return s
    train_dl = DataLoader(subset(ds, train_idx), batch_size=args.batch_size, shuffle=True, drop_last=True)
    val_dl = DataLoader(subset(ds, val_idx), batch_size=args.batch_size, shuffle=False)
    test_dl = DataLoader(subset(ds, test_idx), batch_size=args.batch_size, shuffle=False)

    # model
    model = SeqTransformer(vocab_size=len(TOKEN_VOCAB), emb_dim=args.emb_dim, n_heads=args.n_heads,
                           num_layers=args.num_layers, seq_maxlen=args.maxlen, numeric_dim=3,
                           hidden_dim=args.hidden_dim, proj_dim=args.proj_dim).to(device)

    # optimizer with high learning rate suggestion and scheduler
    optim = torch.optim.AdamW(model.parameters(), lr=args.learning_rate, weight_decay=1e-4)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optim, 'max', patience=3, factor=0.5)
    criterion = nn.BCEWithLogitsLoss()

    best_val_auc = -1.0
    ckpt_dir = args.ckpt_dir
    os.makedirs(ckpt_dir, exist_ok=True)

    for epoch in range(1, args.epochs+1):
        print(f"\n=== Epoch {epoch} ===")
        train_loss = train_one_epoch(model, train_dl, optim, device, criterion)
        val_res = eval_model(model, val_dl, device)
        print(f"Train loss {train_loss:.4f} | Val AUC {val_res['auc']:.4f} | Val ACC {val_res['acc']:.4f}")
        scheduler.step(val_res['auc'])
        if val_res['auc'] > best_val_auc:
            best_val_auc = val_res['auc']
            torch.save({'model_state': model.state_dict(), 'epoch': epoch}, os.path.join(ckpt_dir, 'best.pth'))
            print("Saved best checkpoint.")

    # final test eval
    test_res = eval_model(model, test_dl, device)
    print(f"\nTest AUC {test_res['auc']:.4f} | Test ACC {test_res['acc']:.4f}")

    # Step: dimensionality reduction + clustering for pattern discovery
    embeddings = test_res['embeddings']
    labels = test_res['labels']
    reducer = umap.UMAP(n_components=2, random_state=42)
    proj2 = reducer.fit_transform(embeddings)
    km = KMeans(n_clusters=3, random_state=42).fit(embeddings)
    cluster_ids = km.labels_

    # plot UMAP colored by cluster and by label
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    plt.scatter(proj2[:,0], proj2[:,1], c=cluster_ids, cmap='tab10', s=8)
    plt.title("UMAP of embeddings (clusters)")
    plt.subplot(1,2,2)
    plt.scatter(proj2[:,0], proj2[:,1], c=labels, cmap='coolwarm', s=8)
    plt.title("UMAP colored by true label (1=good)")
    plt.tight_layout()
    plt.savefig(os.path.join(ckpt_dir, 'umap_clusters.png'))
    print("Saved UMAP plot to", os.path.join(ckpt_dir, 'umap_clusters.png'))

    # simple nearest-neighbor lookup example for "drug discovery"
    # cluster centroids -> score, rank clusters by average predicted score
    cluster_scores = []
    for c in np.unique(cluster_ids):
        mask = cluster_ids == c
        cluster_scores.append((c, embeddings[mask].mean(axis=0).dot(np.ones(embeddings.shape[1]))))  # simple proxy
    print("Cluster scores (proxy):", cluster_scores)
    print("Pipeline complete. Check", ckpt_dir)

if __name__ == "__main__":
    parser = argparse.ArgumentParser()
    parser.add_argument("--n_samples", type=int, default=6000)
    parser.add_argument("--seq_len", type=int, default=60)
    parser.add_argument("--maxlen", type=int, default=64)
    parser.add_argument("--batch_size", type=int, default=64)
    parser.add_argument("--epochs", type=int, default=14)
    parser.add_argument("--learning_rate", type=float, default=2e-3)  # higher LR suggestion; tune with scheduler
    parser.add_argument("--emb_dim", type=int, default=64)
    parser.add_argument("--n_heads", type=int, default=4)
    parser.add_argument("--num_layers", type=int, default=2)
    parser.add_argument("--hidden_dim", type=int, default=128)
    parser.add_argument("--proj_dim", type=int, default=64)
    parser.add_argument("--seed", type=int, default=42)
    parser.add_argument("--ckpt_dir", type=str, default="checkpoints")
    args = parser.parse_args(args=[]) # Modified this line
    main(args)
```
