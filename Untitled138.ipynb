```python
import os
import math
import time
import random
from dataclasses import dataclass, field
from typing import List, Dict, Any, Tuple

import numpy as np
import pandas as pd

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from tqdm import trange, tqdm

# ----------------------------- Simulation primitives -----------------------------

@dataclass
class Telomere:
    """Simple telomere representation: length (int) and binding-protein status.
    length measured in arbitrary repeat units.
    """
    length: float
    shelterin_bound: bool = True

    def shorten(self, amount: float = 1.0):
        self.length = max(0.0, self.length - amount)

    def lengthen(self, amount: float = 1.0):
        self.length += amount

    def copy(self):
        return Telomere(self.length, self.shelterin_bound)


class VirtualCell:
    """A minimal virtual cell with telomere dynamics, TERT expression state,
    and a phagocytic activity counter.
    """
    def __init__(self, cell_id: int, init_tel_length: float = 100.0, is_stem: bool = False):
        self.id = cell_id
        self.telomere = Telomere(init_tel_length)
        self.is_stem = is_stem
        self.cell_cycle_phase = 'G0'  # G0, G1, S, G2, M
        self.alive = True
        self.age = 0
        self.dna_damage = 0.0  # simplified damage score
        self.tert_expression = 0.0  # regulator variable, not actual sequence
        self.phagocytosis_capacity = 1.0 if is_stem else 0.3  # stems are better at clean-up
        self.recent_divisions = 0
        # Logging history if needed
        self.history: List[Dict[str, Any]] = []

    def step(self, telomerase_pulse: float = 0.0, nutrient_level: float = 1.0, mech_stress: float = 0.0):
        """Advance the cell by one time unit.
        - Apply telomerase_pulse (adds to telomere length proportional to pulse)
        - Cells may divide if certain conditions met (simplified probabilistic rule)
        - Phagocytosis activity reduces damage
        """
        if not self.alive:
            return

        # telomerase effect: focused on stem cells but controller can pulse others
        telomerase_efficiency = 0.6 if self.is_stem else 0.2
        telomerase_amount = telomerase_pulse * telomerase_efficiency
        if telomerase_amount > 0:
            self.telomere.lengthen(telomerase_amount)

        # basal shortening per cell cycle
        shorten_rate = 0.05 + 0.01 * mech_stress
        self.telomere.shorten(shorten_rate * (1.0 - nutrient_level))

        # Damage accrual under stress
        self.dna_damage += mech_stress * 0.01

        # Phagocytosis reduces damage probabilistically
        cleanup = self.phagocytosis_capacity * (0.5 + 0.5 * nutrient_level)
        self.dna_damage = max(0.0, self.dna_damage - 0.01 * cleanup)

        # Division decision (very simplified): stem cells divide more readily on demand
        division_prob = 0.0
        if self.is_stem:
            division_prob = 0.01 + 0.1 * (1.0 - (self.telomere.length / 200.0))
            # respond to DNA damage or low neighbor density would be added here
        else:
            division_prob = 0.001

        if random.random() < division_prob and nutrient_level > 0.3 and self.telomere.length > 1.0:
            self.divide()

        # death by critical damage or telomere loss
        if self.dna_damage > 1.0 or self.telomere.length < 0.5:
            self.alive = False

        self.age += 1

        # record a compact history row
        self.history.append({
            'id': self.id,
            'age': self.age,
            'telomere_length': self.telomere.length,
            'dna_damage': self.dna_damage,
            'tert_expression': self.tert_expression,
            'alive': self.alive,
            'is_stem': self.is_stem,
        })

    def divide(self):
        # Division shortens telomeres and increases division counter
        self.telomere.shorten(1.0)
        self.recent_divisions += 1
        # After division, minor damage and TERT expression changes
        self.dna_damage += 0.01

    def apply_genetic_instability(self, severity: float = 0.1):
        # introduces stochastic damage that may alter outcomes (for cancer testing)
        self.dna_damage += severity

    def snapshot(self) -> Dict[str, Any]:
        return {
            'id': self.id,
            'age': self.age,
            'telomere_length': self.telomere.length,
            'dna_damage': self.dna_damage,
            'alive': self.alive,
            'is_stem': self.is_stem,
        }


class VirtualTissue:
    """A simple collection of VirtualCells that interact via shared nutrient and flow fields.
    The OOC will host a VirtualTissue instance and update nutrient/mech stress per-cell.
    """
    def __init__(self, n_cells: int = 100, stem_fraction: float = 0.05):
        self.cells: List[VirtualCell] = []
        for i in range(n_cells):
            is_stem = (random.random() < stem_fraction)
            init_len = random.gauss(120.0 if is_stem else 100.0, 5.0)
            self.cells.append(VirtualCell(i, init_tel_length=init_len, is_stem=is_stem))

    def step(self, telomerase_pulses: List[float], nutrient_field: np.ndarray, mech_field: np.ndarray):
        """Update each cell with corresponding environmental values. Arrays are expected len==n_cells.
        """
        for c, pulse, nutr, mech in zip(self.cells, telomerase_pulses, nutrient_field, mech_field):
            c.step(telomerase_pulse=pulse, nutrient_level=float(nutr), mech_stress=float(mech))

    def snapshot(self) -> pd.DataFrame:
        rows = [c.snapshot() for c in self.cells]
        return pd.DataFrame(rows)


# ----------------------------- Telomerase Controller (DL) -----------------------------

class TelomeraseDataset(Dataset):
    """Dataset that uses generated simulation snapshots to create training examples:
    Input: per-cell features (telomere length, damage, nutrient, mech, is_stem)
    Target: desired telomerase pulse (float)
    For demonstration, we use simple heuristics to generate targets for supervised training.
    """
    def __init__(self, records: pd.DataFrame):
        # records: expected columns id, telomere_length, dna_damage, nutrient, mech, is_stem
        self.X = records[['telomere_length', 'dna_damage', 'nutrient', 'mech', 'is_stem']].values.astype(np.float32)
        self.y = records['target_pulse'].values.astype(np.float32).reshape(-1, 1)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]


class TelomeraseControllerNet(nn.Module):
    def __init__(self, input_dim: int = 5, hidden: int = 64):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden),
            nn.ReLU(),
            nn.LayerNorm(hidden),
            nn.Linear(hidden, hidden),
            nn.ReLU(),
            nn.Linear(hidden, 1),
            nn.Sigmoid(),  # output pulse 0..1
        )

    def forward(self, x):
        return self.net(x)


class TelomeraseController:
    """Holds model, optimizer, and an inference wrapper. Training uses supervised 'oracle' targets
    produced by a rule-based generator below, but this class makes it easy to swap in RL later.
    """
    def __init__(self, device: str = 'cpu', lr: float = 1e-3, high_lr: bool = False):
        self.device = device
        self.model = TelomeraseControllerNet().to(device)
        # choose a higher learning rate if requested, but include safe defaults
        self.lr = 1e-2 if high_lr else lr
        self.optimizer = optim.AdamW(self.model.parameters(), lr=self.lr)
        self.criterion = nn.MSELoss()
        # scheduler to avoid runaway when using high lr
        self.scheduler = optim.lr_scheduler.OneCycleLR(self.optimizer, max_lr=self.lr, total_steps=1000) if high_lr else None

    def predict(self, features: np.ndarray) -> np.ndarray:
        self.model.eval()
        with torch.no_grad():
            x = torch.from_numpy(features.astype(np.float32)).to(self.device)
            out = self.model(x).cpu().numpy().reshape(-1)
        return out

    def train_one_epoch(self, dataloader: DataLoader, clip_value: float = 1.0):
        self.model.train()
        running_loss = 0.0
        for i, (X, y) in enumerate(dataloader):
            X = X.to(self.device)
            y = y.to(self.device)
            pred = self.model(X)
            loss = self.criterion(pred, y)
            self.optimizer.zero_grad()
            loss.backward()
            # gradient clipping to avoid divergence with high lr
            torch.nn.utils.clip_grad_norm_(self.model.parameters(), clip_value)
            self.optimizer.step()
            if self.scheduler:
                try:
                    self.scheduler.step()
                except Exception:
                    pass
            running_loss += loss.item()
        return running_loss / len(dataloader)


# ----------------------------- Heuristic "Oracle" target generator -----------------------------

def oracle_target(row: Dict[str, Any]) -> float:
    """Generates a supervised target pulse value (0..1) based on heuristic rules:
    - prioritize stem cells with short telomeres and elevated damage
    - avoid pulsing cells with high DNA damage (cancer prevention)
    - small pulses for somatic cells only in severe damage
    """
    tel = row['telomere_length']
    dmg = row['dna_damage']
    stem = row['is_stem']

    # base priority: shorter telomeres => higher need
    need = float(max(0.0, (150.0 - tel) / 150.0))
    # damage penalty: if damage too high, reduce pulse to mitigate cancer risk
    damage_penalty = float(max(0.0, 1.0 - dmg))
    pulse = need * damage_penalty
    # boost for true stem cells
    if stem:
        pulse *= 1.5
    pulse = min(1.0, max(0.0, pulse))
    return pulse


# ----------------------------- Environment + Data generator -----------------------------

class VirtualOOC:
    """Simple organ on chip environment that provides nutrient and mechanical stress
    fields and collects data for training the controller.
    """
    def __init__(self, tissue: VirtualTissue, flow_rate: float = 1.0):
        self.tissue = tissue
        self.n = len(tissue.cells)
        self.flow_rate = flow_rate
        # nutrient and mech fields per cell (start uniform)
        self.nutrient = np.ones(self.n) * 1.0
        self.mech = np.zeros(self.n)

    def step(self, controller: TelomeraseController = None) -> pd.DataFrame:
        # Build features for controller
        features = []
        for i, c in enumerate(self.tissue.cells):
            features.append({
                'id': c.id,
                'telomere_length': c.telomere.length,
                'dna_damage': c.dna_damage,
                'nutrient': float(self.nutrient[i]),
                'mech': float(self.mech[i]),
                'is_stem': 1.0 if c.is_stem else 0.0,
            })
        df = pd.DataFrame(features)

        # if controller present: infer pulses
        if controller is not None:
            X = df[['telomere_length', 'dna_damage', 'nutrient', 'mech', 'is_stem']].values.astype(np.float32)
            pulses = controller.predict(X)
        else:
            # no controller -> baseline zero pulses
            pulses = np.zeros(self.n)

        # Apply pulses and advance tissue
        self.tissue.step(pulses.tolist(), self.nutrient, self.mech)

        # Collect outputs (after step) and add the pulses used
        out = self.tissue.snapshot()
        out['nutrient'] = self.nutrient
        out['mech'] = self.mech
        out['applied_pulse'] = pulses
        # compute heuristic target for supervised training (oracle)
        out['target_pulse'] = out.apply(oracle_target, axis=1)
        return out

    def perturb(self, kind: str = 'injury', intensity: float = 0.5):
        """Introduce a perturbation: injury reduces nutrient locally and increases mech stress.
        """
        if kind == 'injury':
            # choose a patch
            idx = random.randint(0, self.n - 1)
            radius = max(1, int(self.n * 0.05))
            start = max(0, idx - radius)
            end = min(self.n, idx + radius)
            self.nutrient[start:end] *= (1.0 - intensity)
            self.mech[start:end] += intensity * 0.5
        elif kind == 'pathogen':
            # increase damage across tissue slightly
            for c in self.tissue.cells:
                c.dna_damage += 0.05 * intensity


# ----------------------------- Training / Experiment Scripts -----------------------------

def generate_training_data(env: VirtualOOC, steps: int = 200, perturb_every: int = 50) -> pd.DataFrame:
    records = []
    for t in range(steps):
        if perturb_every and (t % perturb_every == 0) and t > 0:
            env.perturb(kind='injury', intensity=min(0.6, 0.1 + 0.1 * (t // perturb_every)))
        out = env.step(controller=None)  # oracle uses target_pulse
        out['time'] = t
        records.append(out)
    df = pd.concat(records, ignore_index=True)
    return df


def train_controller_from_data(df: pd.DataFrame, device: str = 'cpu', high_lr: bool = False, epochs: int = 5):
    ds = TelomeraseDataset(df)
    loader = DataLoader(ds, batch_size=64, shuffle=True)
    controller = TelomeraseController(device=device, high_lr=high_lr)

    for e in range(epochs):
        loss = controller.train_one_epoch(loader, clip_value=1.0)
        print(f"Epoch {e+1}/{epochs}  loss={loss:.6f} lr={controller.lr}")

    return controller


def demo_run(output_dir: str = './outputs', seed: int = 42, high_lr: bool = True):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)

    os.makedirs(output_dir, exist_ok=True)

    tissue = VirtualTissue(n_cells=200, stem_fraction=0.08)
    env = VirtualOOC(tissue)

    # generate supervised data using oracle
    print("Generating training data (demo)...")
    df = generate_training_data(env, steps=300, perturb_every=60)
    df.to_csv(os.path.join(output_dir, 'sim_data.csv'), index=False)
    print("Data generated. Training controller...")

    # train controller (with optional high lr)
    controller = train_controller_from_data(df, device='cpu', high_lr=high_lr, epochs=8)

    # Evaluate controller in a fresh environment with an injury and pathogen challenge
    print("Evaluating trained controller in new environment...")
    tissue2 = VirtualTissue(n_cells=200, stem_fraction=0.08)
    env2 = VirtualOOC(tissue2)
    # run baseline (no controller) for reference
    baseline = []
    for t in range(50):
        if t == 15:
            env2.perturb('injury', intensity=0.5)
        baseline.append(env2.step(None))
    baseline_df = pd.concat(baseline, ignore_index=True)
    baseline_df.to_csv(os.path.join(output_dir, 'baseline_run.csv'), index=False)

    # reset and run with controller
    tissue3 = VirtualTissue(n_cells=200, stem_fraction=0.08)
    env3 = VirtualOOC(tissue3)
    controlled = []
    for t in range(50):
        if t == 15:
            env3.perturb('injury', intensity=0.5)
        out = env3.step(controller)
        controlled.append(out)
    controlled_df = pd.concat(controlled, ignore_index=True)
    controlled_df.to_csv(os.path.join(output_dir, 'controlled_run.csv'), index=False)

    print(f"Demo outputs written to {output_dir}")
    return controller, df, baseline_df, controlled_df


# ----------------------------- Command-line interface -----------------------------

if __name__ == '__main__':
    import argparse

    parser = argparse.ArgumentParser(description='Virtual OOC telomere simulation + controller demo')
    parser.add_argument('--mode', type=str, default='demo', choices=['demo', 'gen', 'train'], help='Mode')
    parser.add_argument('--out', type=str, default='./outputs', help='Output directory')
    parser.add_argument('--seed', type=int, default=42)
    parser.add_argument('--high_lr', action='store_true', help='Use higher learning rate for controller')
    args, unknown = parser.parse_known_args() # Modified line

    if args.mode == 'demo':
        demo_run(output_dir=args.out, seed=args.seed, high_lr=args.high_lr)
    elif args.mode == 'gen':
        tissue = VirtualTissue(n_cells=200, stem_fraction=0.08)
        env = VirtualOOC(tissue)
        df = generate_training_data(env, steps=500, perturb_every=80)
        os.makedirs(args.out, exist_ok=True)
        df.to_csv(os.path.join(args.out, 'sim_data.csv'), index=False)
        print('Data written')
    elif args.mode == 'train':
        df = pd.read_csv(os.path.join(args.out, 'sim_data.csv'))
        controller = train_controller_from_data(df, device='cpu', high_lr=args.high_lr, epochs=10)
        torch.save(controller.model.state_dict(), os.path.join(args.out, 'controller.pth'))
        print('Trained controller saved')

# End of file

# ---- Real Biological Data Integration (Safe, Public Datasets) ----
# Telomere length reference dataset (e.g., NHANES)  placeholder path
TEL_LENGTH_DATA = "data/telomere/nhanes_telomere_length.csv"

# Gene expression profiles (e.g., GTEx  safe, de-identified)
GTEX_EXPRESSION = "data/gtex/gtex_gene_expression.parquet"

# Drug reference fingerprints (ChEMBL / PubChem preprocessed)
CHEMBL_DB = "data/chembl/chembl_processed.csv"
PUBCHEM_DB = "data/pubchem/pubchem_fingerprints.feather"


def load_real_data():
    import pandas as pd
    data = {}
    try:
        data["telomere"] = pd.read_csv(TEL_LENGTH_DATA)
    except Exception:
        data["telomere"] = None
    try:
        data["gtex"] = pd.read_parquet(GTEX_EXPRESSION)
    except Exception:
        data["gtex"] = None
    try:
        data["chembl"] = pd.read_csv(CHEMBL_DB)
    except Exception:
        data["chembl"] = None
    try:
        import feather
        data["pubchem"] = feather.read_dataframe(PUBCHEM_DB)
    except Exception:
        data["pubchem"] = None
    return data

REAL_DATA = load_real_data()


# ---- GTEx Downloader (Public, Safe, De-identified Data) ----
# This module fetches GTEx v8 gene expression matrices from the official public repository.
# GTEx is fully de-identified and safe to use.

import os
import requests

GTEX_URL = "https://storage.googleapis.com/gtex_analysis_v8/rna_seq_data/GTEx_Analysis_2017-06-05_v8_RNASeQCv1.1.9_gene_tpm.gct.gz"
GTEX_LOCAL = "data/gtex/gtex_v8_tpm.gct.gz"


def download_gtex(overwrite=False):
    os.makedirs("data/gtex", exist_ok=True)
    if os.path.exists(GTEX_LOCAL) and not overwrite:
        print("GTEx already downloaded.")
        return GTEX_LOCAL

    print("Downloading GTEx v8 TPM matrix")
    resp = requests.get(GTEX_URL, stream=True)
    resp.raise_for_status()

    with open(GTEX_LOCAL, "wb") as f:
        for chunk in resp.iter_content(chunk_size=8192):
            if chunk:
                f.write(chunk)

    print("GTEx download complete ", GTEX_LOCAL)
    return GTEX_LOCAL


# ---- GTEx Loader ----
# Converts the GCT matrix into a pandas dataframe and caches it.

def load_gtex_tpm(path=GTEX_LOCAL):
    import pandas as pd

    if not os.path.exists(path):
        raise FileNotFoundError("GTEx TPM file not found. Run download_gtex().")

    print("Loading GTEx TPM gene expression")
    df = pd.read_csv(path, sep='    ', comment='#', skiprows=2)

    # Standard GCT format: Name, Description, Sample1...SampleN
    df = df.rename(columns={"Name": "gene_id", "Description": "gene_symbol"})
    df.set_index("gene_id", inplace=True)

    return df


# ---- Integrate GTEx into REAL_DATA ----
try:
    REAL_DATA["gtex_raw"] = load_gtex_tpm()
except Exception:
    REAL_DATA["gtex_raw"] = None



# ----------------------------- GTEx Preprocessing & Embeddings -----------------------------
# This block adds preprocessing (logTPM/scaling), gene filtering, tissue-level aggregation,
# a small PyTorch autoencoder for embeddings, dataset classes that combine GTex
# embeddings with simulated per-cell features, and contrastive-learning placeholders.

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Optional imports (make sure to pip install if you use them)
try:
    import umap
except Exception:
    umap = None


def preprocess_gtex(df: pd.DataFrame, min_tpm: float = 0.1, min_samples: int = 10) -> pd.DataFrame:
    """Preprocess raw GTEx TPM matrix (genes x samples).
    - Remove low-expression genes
    - Log1p transform
    - Return gene x sample dataframe (float32)
    """
    print("Preprocessing GTEx: filtering low-expression genes...")
    # df is genes x samples
    gene_mask = (df > min_tpm).sum(axis=1) >= min_samples
    df_filt = df[gene_mask]
    # log1p transform
    df_lp = np.log1p(df_filt.astype(np.float32))
    return df_lp


def aggregate_tissue_level(gtex_df: pd.DataFrame, sample_metadata: pd.DataFrame = None) -> pd.DataFrame:
    """Aggregate GTEx into tissue-level mean expression vectors.
    If sample_metadata provided (with columns sample_id, tissue_site_detail), it will use that mapping.
    Otherwise, attempt to infer tissue from sample column names (not guaranteed).
    """
    print("Aggregating GTEx to tissue-level expressions...")
    cols = gtex_df.columns.to_list()
    tissues = None
    if sample_metadata is not None:
        meta = sample_metadata.set_index('sample_id')
        tissues = [meta.loc[c, 'tissue_site_detail'] if c in meta.index else 'unknown' for c in cols]
    else:
        # fallback: try to parse by GTEx column naming patterns (may not work for all files)
        tissues = [c.split('-')[1] if '-' in c else 'unknown' for c in cols]
    tissue_map = {}
    for col, tissue in zip(cols, tissues):
        tissue_map.setdefault(tissue, []).append(col)

    agg = {}
    for tissue, samples in tissue_map.items():
        if len(samples) < 1:
            continue
        agg[tissue] = gtex_df[samples].mean(axis=1)
    agg_df = pd.DataFrame(agg)
    return agg_df


# Small PyTorch Autoencoder to compute embeddings for tissue expression vectors
class ExpressionAutoencoder(nn.Module):
    def __init__(self, input_dim: int, latent_dim: int = 64):
        super().__init__()
        self.enc = nn.Sequential(
            nn.Linear(input_dim, 256),
            nn.ReLU(),
            nn.Linear(256, latent_dim),
        )
        self.dec = nn.Sequential(
            nn.Linear(latent_dim, 256),
            nn.ReLU(),
            nn.Linear(256, input_dim),
        )

    def forward(self, x):
        z = self.enc(x)
        out = self.dec(z)
        return out, z


def compute_expression_embeddings(expr_df: pd.DataFrame, latent_dim: int = 64, epochs: int = 200, device: str = 'cpu') -> Tuple[pd.DataFrame, np.ndarray]:
    """Compute tissue embeddings using autoencoder + PCA fallback.
    Returns: (embeddings_df, z_matrix)
    """
    X = expr_df.T.fillna(0).values.astype(np.float32)  # samples x genes -> tissues x genes after aggregation
    scaler = StandardScaler()
    Xs = scaler.fit_transform(X)

    input_dim = Xs.shape[1]
    if input_dim < 16:
        # if too few features, fallback to PCA
        pca = PCA(n_components=min(latent_dim, max(2, input_dim - 1)))
        Z = pca.fit_transform(Xs)
        emb_df = pd.DataFrame(Z, index=expr_df.columns)
        return emb_df, Z

    ae = ExpressionAutoencoder(input_dim=input_dim, latent_dim=latent_dim).to(device)
    opt = optim.Adam(ae.parameters(), lr=1e-3)
    Xtorch = torch.from_numpy(Xs).to(device)

    for e in range(epochs):
        ae.train()
        opt.zero_grad()
        recon, z = ae(Xtorch)
        loss = ((recon - Xtorch) ** 2).mean()
        loss.backward()
        opt.step()
        if (e + 1) % 50 == 0:
            print(f"AE epoch {e+1}/{epochs} loss={loss.item():.6f}")

    ae.eval()
    with torch.no_grad():
        _, Z = ae(Xtorch)
    Z = Z.cpu().numpy()
    emb_df = pd.DataFrame(Z, index=expr_df.columns)
    return emb_df, Z


# ----------------------------- Integration: GTEx embeddings -> virtual cells -----------------------------

def map_tissue_embeddings_to_cells(sim_df: pd.DataFrame, tissue_embeddings: pd.DataFrame, tissue_col: str = 'tissue') -> pd.DataFrame:
    """Attach a tissue embedding to each simulated cell row.
    sim_df: per-cell dataframe from simulation (must have a column 'is_stem' and optionally 'tissue')
    tissue_embeddings: index=tissue_name -> embedding dims
    If sim_df lacks a tissue column, assign the most common tissue or random.
    """
    emb_cols = [f'emb_{i}' for i in range(tissue_embeddings.shape[1])]

    # ensure sim_df has a tissue column
    if tissue_col not in sim_df.columns:
        # assign random tissue labels
        tissues = tissue_embeddings.index.tolist()
        sim_df = sim_df.copy()
        sim_df[tissue_col] = np.random.choice(tissues, size=len(sim_df))

    # map
    def fetch_emb(t):
        if t in tissue_embeddings.index:
            return tissue_embeddings.loc[t].values
        else:
            return tissue_embeddings.iloc[0].values

    emb_matrix = np.vstack(sim_df[tissue_col].map(fetch_emb).values)
    emb_df = pd.DataFrame(emb_matrix, columns=emb_cols, index=sim_df.index)
    out = pd.concat([sim_df.reset_index(drop=True), emb_df.reset_index(drop=True)], axis=1)
    return out


# ----------------------------- Contrastive & Joint Models (placeholders) -----------------------------

def contrastive_placeholder():
    """Placeholder for contrastive training: pair telomere-stable cells with high-embedding similarity,
    and telomere-unstable ones as negatives. Implement with SimCLR or BYOL if desired.
    """
    print("Contrastive pipeline placeholder. Implement SimCLR/BYOL or use existing libs.")


# ----------------------------- Synthetic perturbation generator using GTEx priors -----------------------------

def synthesize_expression_perturbation(tissue_expr: pd.Series, fold_change_genes: List[str] = None, intensity: float = 0.5) -> pd.Series:
    """Create a perturbed tissue expression vector by up/down regulating a subset of genes.
    This is for in-silico drug or injury simulation.
    """
    v = tissue_expr.copy()
    n_genes = max(1, int(len(v) * 0.02))
    if fold_change_genes is None:
        idx = np.random.choice(v.index, size=n_genes, replace=False)
    else:
        idx = [g for g in fold_change_genes if g in v.index][:n_genes]
    for g in idx:
        v.loc[g] = v.loc[g] * (1.0 + (np.random.rand() - 0.5) * 2.0 * intensity)
    return v


# ----------------------------- High-level pipeline runner -----------------------------

def run_full_pipeline(download_gtex_flag: bool = False, recompute_embeddings: bool = True, device: str = 'cpu'):
    """Runs: download GTEx (optional) -> preprocess -> aggregate -> compute embeddings ->
    generate simulation data -> map embeddings to sim cells -> train controller on joint features.
    """
    if download_gtex_flag:
        try:
            download_gtex()
        except Exception as e:
            print("GTEx download failed:", e)

    # load/prepare GTEx
    gtex_raw = REAL_DATA.get('gtex_raw')
    if gtex_raw is None:
        print("GTEx raw not loaded into REAL_DATA. Please run download_gtex() and load_gtex_tpm().")
        return

    gtex_pp = preprocess_gtex(gtex_raw)
    tissue_expr = aggregate_tissue_level(gtex_pp)

    # embeddings
    emb_df, Z = compute_expression_embeddings(tissue_expr, latent_dim=64, epochs=150, device=device)

    # generate simulated data
    tissue = VirtualTissue(n_cells=300, stem_fraction=0.08)
    env = VirtualOOC(tissue)
    sim_df = generate_training_data(env, steps=400, perturb_every=80)

    # map embeddings
    joint = map_tissue_embeddings_to_cells(sim_df, emb_df)
    out_dir = 'outputs/full_pipeline'
    os.makedirs(out_dir, exist_ok=True)
    joint.to_parquet(os.path.join(out_dir, 'sim_with_gtex_embeddings.parquet'))
    print("Wrote joint dataset to", out_dir)

    # create dataset for controller training using extended features
    feature_cols = ['telomere_length', 'dna_damage', 'nutrient', 'mech', 'is_stem'] + [c for c in joint.columns if c.startswith('emb_')]
    # normalize embeddings
    X = joint[feature_cols].fillna(0).values.astype(np.float32)
    y = joint['target_pulse'].values.astype(np.float32).reshape(-1, 1)

    # simple retrain of controller using combined features (small net change)
    ds = torch.utils.data.TensorDataset(torch.from_numpy(X), torch.from_numpy(y))
    loader = DataLoader(ds, batch_size=128, shuffle=True)

    controller = TelomeraseController(device=device, high_lr=True)
    # replace model input_dim to match new features (recreate net)
    controller.model = TelomeraseControllerNet(input_dim=X.shape[1], hidden=128).to(device)
    controller.optimizer = optim.AdamW(controller.model.parameters(), lr=controller.lr)

    for e in range(6):
        loss = controller.train_one_epoch(loader)
        print(f"Joint training epoch {e+1} loss={loss:.6f}")

    torch.save(controller.model.state_dict(), os.path.join(out_dir, 'joint_controller.pth'))
    print("Saved trained joint controller to", out_dir)

    return joint, controller


# ----------------------------- End of additions -----------------------------
```

```python
"""
Telomere OOC Simulator - runnable single-file demo

Files included:
- This single script contains:
  - Telomere ODE model (telomere_sim)
  - Simple agent-based OOC grid (ooc_agent_grid)
  - Tiny PyTorch controller model and micro-training loop (control_model)
  - Small demo that runs the sim, produces synthetic dataset, trains a tiny model,
    and plots a couple of results.

Requirements (install with pip):
  pip install numpy scipy matplotlib pandas torch

Run:
  python telomere_ooc_simulator.py

Notes:
- This is computational-only code (no wet-lab/in-vivo instructions).
- The demo is intentionally small and CPU-friendly.
- Extendable: replace synthetic dataset creation with real public datasets (GTEx/TCGA/GEO/TelSeq outputs).

"""

import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import random
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd

# ---------------------------
# Telomere ODE (per-cell)
# ---------------------------
SHORTEN_PER_DIV = 50.0    # effective bp loss per division (toy units)
TELOMERASE_K = 1.0        # bp extension rate when fully active (toy units)
TRIM_THRESHOLD = 10000.0  # above this, trimming acts
TRIM_RATE = 0.01
DIV_RATE_BASE = 0.01


def telomere_ode(t, y, telomerase_activity, env_injury):
    L = y[0]
    # telomerase_activity is numeric 0..1 (we pass a scalar here since ODE solver calls it)
    dL_dt = -DIV_RATE_BASE * SHORTEN_PER_DIV + TELOMERASE_K * telomerase_activity - TRIM_RATE * max(0, L - TRIM_THRESHOLD)
    return [dL_dt]


# ---------------------------
# Agent-based OOC grid
# ---------------------------
class Cell:
    def __init__(self, cid, L=8000.0):
        self.id = cid
        self.L = L
        self.state = 'quiescent'  # 'proliferating','senescent'
        self.age = 0
        self.mutations = 0  # surrogate genomic instability

    def step(self, local_env, control_policy):
        # control_policy returns telomerase activity 0..1
        ta = control_policy(self, local_env)
        # simple discrete division logic
        if local_env.get('injury', 0) > 0 and self.state != 'senescent':
            # during injury, some cells proliferate
            self.state = 'proliferating'
            # apply division shortening
            self.L -= SHORTEN_PER_DIV
            # small chance to accumulate mutation during division
            if random.random() < 0.02:
                self.mutations += 1
        else:
            # baseline quiescent
            self.state = 'quiescent'
        # continuous telomerase extension
        self.L += TELOMERASE_K * ta * 0.5  # scaled
        # trimming
        if self.L > TRIM_THRESHOLD:
            self.L -= TRIM_RATE * (self.L - TRIM_THRESHOLD)
        # senescence threshold
        if self.L < 2000:
            self.state = 'senescent'
        self.age += 1


class OOCGrid:
    def __init__(self, nx=24, ny=8):
        self.nx = nx
        self.ny = ny
        self.cells = [Cell(i, L=np.random.normal(8000, 400)) for i in range(nx * ny)]

    def step(self, t, control_policy, injury_pattern=None):
        # injury_pattern: callable(cell, t) -> 0/1
        for c in self.cells:
            env = {'nutrient': 1.0, 'injury': injury_pattern(c, t) if injury_pattern is not None else 0}
            c.step(env, control_policy)

    def collect_telomere_snapshot(self):
        return np.array([c.L for c in self.cells])

    def collect_states(self):
        return [c.state for c in self.cells]

    def cancer_proxy_fraction(self):
        # crude proxy: cells with mutations>1 and state proliferating
        bad = sum(1 for c in self.cells if c.mutations > 1 and c.state == 'proliferating')
        return bad / len(self.cells)

# ---------------------------
# Control policy (baseline)
# ---------------------------

def simple_policy(cell, env):
    # activate telomerase when injury present and cell not senescent
    if env.get('injury', 0) > 0 and cell.state != 'senescent':
        return 1.0
    return 0.0

# ---------------------------
# Generate synthetic dataset for ML
# ---------------------------

def generate_synthetic_dataset(grid, steps=40):
    # We'll save short time-series per-cell: telomere history and env signals
    seqs = []  # each: (seq_len x features)
    envs = []  # env features aggregated
    targets = []

    # injury pattern: localized injury in first 3 timesteps for a band of cells
    def injury_pattern(cell, t):
        # injure first column cells for t in [1..3]
        if (cell.id % grid.nx) < 3 and t < 3:
            return 1
        return 0

    # run simulation and record per-step telomere lengths
    histories = {c.id: [] for c in grid.cells}
    for t in range(steps):
        # use simple_policy to generate data
        grid.step(t, simple_policy, injury_pattern)
        snap = grid.collect_telomere_snapshot()
        for i, val in enumerate(snap):
            histories[i].append(val)

    # synthesize per-cell sequences and targets (target = desired telomerase intensity during repair: 0 or 1)
    for cid, series in histories.items():
        # use windows of length 8 for model input
        if len(series) < 8:
            continue
        for start in range(0, len(series) - 8):
            window = np.array(series[start:start+8])
            # env feature: was there injury in window? (simple binary)
            injury_flag = 1.0 if any((start+i) < 3 and (cid % grid.nx) < 3 for i in range(8)) else 0.0
            # target: telomerase desired intensity for next step (if injury present and not senescent => 1)
            target = 1.0 if injury_flag and window[-1] > 2000 else 0.0
            # stack features: reshape window into T x 1
            seqs.append(window.reshape(8,1))
            envs.append([injury_flag])
            targets.append([target])

    seqs = np.array(seqs, dtype=np.float32)
    envs = np.array(envs, dtype=np.float32)
    targets = np.array(targets, dtype=np.float32)
    return seqs, envs, targets

# ---------------------------
# PyTorch controller model
# ---------------------------
class TinyController(nn.Module):
    def __init__(self, seq_in=1, env_in=1, hidden=64):
        super().__init__()
        self.lstm = nn.LSTM(input_size=seq_in, hidden_size=hidden, batch_first=True)
        self.env_fc = nn.Sequential(nn.Linear(env_in, hidden), nn.ReLU())
        self.fusion = nn.Sequential(nn.Linear(hidden*2, hidden), nn.ReLU(), nn.Linear(hidden,1), nn.Sigmoid())

    def forward(self, seq, env):
        # seq: batch x T x seq_in
        _, (h_n, _) = self.lstm(seq)
        h = h_n[-1]
        e = self.env_fc(env)
        x = torch.cat([h, e], dim=-1)
        out = self.fusion(x)
        return out

# ---------------------------
# Training helper
# ---------------------------

def train_controller(model, seqs, envs, targets, epochs=6, batch_size=64):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.to(device)
    opt = optim.AdamW(model.parameters(), lr=1e-3)
    loss_fn = nn.BCELoss()
    dataset = torch.utils.data.TensorDataset(torch.tensor(seqs), torch.tensor(envs), torch.tensor(targets))
    dl = torch.utils.data.DataLoader(dataset, batch_size=batch_size, shuffle=True)
    for ep in range(epochs):
        total_loss = 0.0
        for xb, xe, yb in dl:
            xb = xb.to(device); xe = xe.to(device); yb = yb.to(device)
            pred = model(xb, xe)
            loss = loss_fn(pred, yb)
            opt.zero_grad(); loss.backward(); opt.step()
            total_loss += loss.item() * xb.size(0)
        print(f"Epoch {ep+1}/{epochs} loss={total_loss/len(dataset):.4f}")

# ---------------------------
# Demo main
# ---------------------------

def demo_run():
    print("Starting demo... building grid and generating synthetic dataset")
    grid = OOCGrid(nx=24, ny=8)
    seqs, envs, targets = generate_synthetic_dataset(grid, steps=24)
    print(f"Generated {len(seqs)} training examples")

    # quick train
    model = TinyController(seq_in=1, env_in=1, hidden=64)
    train_controller(model, seqs, envs, targets, epochs=8, batch_size=128)

    # evaluate on a few examples
    with torch.no_grad():
        x = torch.tensor(seqs[:8])
        e = torch.tensor(envs[:8])
        preds = model(x, e).numpy().ravel()
    print("Sample preds:", preds)

    # run a validation sim using the trained policy
    def learned_policy(cell, env):
        # prepare sequence: last 8 steps we will approximate by repeating current L
        seq = np.array([cell.L]*8, dtype=np.float32).reshape(1,8,1)
        env_feat = np.array([1.0 if env.get('injury',0)>0 else 0.0], dtype=np.float32).reshape(1,1)
        with torch.no_grad():
            pred = model(torch.tensor(seq), torch.tensor(env_feat)).item()
        return float(pred)

    # new grid for eval
    eval_grid = OOCGrid(nx=24, ny=8)

    def injury_pattern(cell, t):
        if (cell.id % eval_grid.nx) < 3 and t < 4:
            return 1
        return 0

    times = list(range(0, 24))
    mean_telomeres = []
    cancer_proxies = []
    for t in times:
        eval_grid.step(t, learned_policy, injury_pattern)
        mean_telomeres.append(np.mean(eval_grid.collect_telomere_snapshot()))
        cancer_proxies.append(eval_grid.cancer_proxy_fraction())

    # plot results
    plt.figure(figsize=(8,3))
    plt.subplot(1,2,1)
    plt.plot(times, mean_telomeres)
    plt.title('Mean telomere length (eval)')
    plt.xlabel('t')
    plt.subplot(1,2,2)
    plt.plot(times, cancer_proxies)
    plt.title('Cancer-proxy fraction (eval)')
    plt.xlabel('t')
    plt.tight_layout()
    plt.show()

    # save a small CSV summary
    df = pd.DataFrame({'t': times, 'mean_telomere': mean_telomeres, 'cancer_proxy': cancer_proxies})
    df.to_csv('demo_eval_summary.csv', index=False)
    print("Demo finished. Summary saved to demo_eval_summary.csv")


if __name__ == '__main__':
    demo_run()
```
