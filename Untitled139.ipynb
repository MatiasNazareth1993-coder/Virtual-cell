```python
import math
import random
import argparse
from collections import deque
from dataclasses import dataclass
from typing import List, Dict, Any

import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim

# ----------------------------- Simulator core ---------------------------------

@dataclass
class Cell:
    id: int
    age: float
    telomere: float
    cell_type: str = "generic"
    division_count: int = 0
    alive: bool = True
    generation: int = 0

    def step_age(self, dt=1.0):
        self.age += dt


class OOCSimulator:
    """Abstract organ-on-a-chip simulator (agent-based).

    No biological details are provided  states are purely numeric.
    """

    def __init__(self,
                 initial_cells=100,
                 initial_telomere=100.0,
                 min_telomere_to_divide=10.0,
                 division_age=20.0,
                 telomere_loss_per_division=5.0,
                 telomerase_effect_per_activation=4.0,
                 phagocytosis_capacity_per_step=10,
                 max_steps=1000,
                 rng_seed=None):
        if rng_seed is not None:
            random.seed(rng_seed)
            np.random.seed(rng_seed)
            torch.manual_seed(rng_seed)

        self.cells: Dict[int, Cell] = {}
        self.next_id = 0
        self.step_count = 0
        self.history = {"population": [], "avg_telomere": [], "divisions": [], "deaths": []}

        self.min_telomere_to_divide = min_telomere_to_divide
        self.division_age = division_age
        self.telomere_loss_per_division = telomere_loss_per_division
        self.telomerase_effect_per_activation = telomerase_effect_per_activation
        self.phagocytosis_capacity_per_step = phagocytosis_capacity_per_step
        self.max_steps = max_steps

        # initialize
        for _ in range(initial_cells):
            self._add_cell(age=random.uniform(0, division_age), telomere=initial_telomere + random.uniform(-5, 5))

    def _add_cell(self, age=0.0, telomere=100.0, generation=0, cell_type="generic"):
        c = Cell(id=self.next_id, age=age, telomere=telomere, generation=generation, cell_type=cell_type)
        self.cells[self.next_id] = c
        self.next_id += 1
        return c

    def step(self, telomerase_activation: float, dt=1.0):
        """Advance simulation by one step.

        telomerase_activation: scalar in [0, 1] describing fraction of maximum activation.
        (The policy network outputs this.)
        """
        divisions = 0
        deaths = 0

        # Per-cell updates
        for c in list(self.cells.values()):
            if not c.alive:
                continue
            c.step_age(dt)

            # Telomerase effect is modeled as a fractional increase to telomere length
            # proportional to activation; deliberately abstract.
            c.telomere += telomerase_activation * self.telomerase_effect_per_activation * (dt)

            # Check division condition (age + sufficient telomere)
            if c.age >= self.division_age and c.telomere > self.min_telomere_to_divide:
                # division happens probabilistically  we model stochasticity abstractly
                if random.random() < 0.5:
                    # parent loses telomere length, new daughter receives reduced telomere
                    c.telomere -= self.telomere_loss_per_division
                    c.age = 0.0
                    c.generation += 1
                    # daughter inherits a fraction of parent's telomere
                    daughter_tel = max(0.0, c.telomere - random.uniform(0, self.telomere_loss_per_division/2))
                    self._add_cell(age=0.0, telomere=daughter_tel, generation=c.generation, cell_type=c.cell_type)
                    divisions += 1

            # Death condition: telomere too short or extreme age
            if c.telomere <= 0 or c.age > self.division_age * 10:
                c.alive = False
                deaths += 1

        # Phagocytosis: remove dead cells up to capacity
        dead_ids = [cid for cid, cell in self.cells.items() if not cell.alive]
        removed = 0
        for cid in dead_ids:
            if removed >= self.phagocytosis_capacity_per_step:
                break
            del self.cells[cid]
            removed += 1

        self.step_count += 1

        # record history
        pop = len(self.cells)
        avg_tel = np.mean([c.telomere for c in self.cells.values()]) if pop > 0 else 0.0
        self.history['population'].append(pop)
        self.history['avg_telomere'].append(avg_tel)
        self.history['divisions'].append(divisions)
        self.history['deaths'].append(deaths)

        return {"population": pop, "avg_telomere": avg_tel, "divisions": divisions, "deaths": deaths}

    def reset(self):
        # reinitialize to a default small population while preserving RNG
        self.cells = {}
        self.next_id = 0
        self.step_count = 0
        self.history = {"population": [], "avg_telomere": [], "divisions": [], "deaths": []}
        for _ in range(100):
            self._add_cell(age=random.uniform(0, self.division_age), telomere=100.0 + random.uniform(-5, 5))

    def get_observation(self):
        """Return a compact observation vector for the policy network.

        We'll return population size, average telomere, telomere std, recent division rate.
        """
        pop = len(self.cells)
        telomeres = np.array([c.telomere for c in self.cells.values()]) if pop > 0 else np.array([0.0])
        avg = float(np.mean(telomeres))
        std = float(np.std(telomeres))
        recent_divs = float(np.sum(self.history['divisions'][-10:])) if len(self.history['divisions'])>0 else 0.0
        return np.array([pop, avg, std, recent_divs], dtype=np.float32)


# ----------------------------- Policy network ---------------------------------

class PolicyNet(nn.Module):
    def __init__(self, obs_dim=4, hidden=64):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(obs_dim, hidden),
            nn.ReLU(),
            nn.Linear(hidden, hidden//2),
            nn.ReLU(),
            nn.Linear(hidden//2, 1),
            nn.Sigmoid()  # output between 0 and 1
        )

    def forward(self, x):
        return self.net(x).squeeze(-1)


# ----------------------------- Trainer / REINFORCE ----------------------------

class Trainer:
    """Simple REINFORCE trainer to train the policy to manage telomerase activation.

    Reward is designed to encourage:
     - average telomere close to a target
     - moderate division rate (penalize large spikes)
     - population survival (avoid extinction)
    """

    def __init__(self, sim: OOCSimulator, policy: PolicyNet, lr=1e-2, device='cpu'):
        self.sim = sim
        self.policy = policy.to(device)
        self.optimizer = optim.Adam(self.policy.parameters(), lr=lr)
        self.device = device

    def compute_reward(self, obs, info, target_telomere=100.0):
        # reward components
        avg_tel = info['avg_telomere']
        pop = info['population']
        divisions = info['divisions']

        r_tel = -abs(avg_tel - target_telomere) / (target_telomere + 1e-6)  # closeness
        r_div = -divisions * 0.1  # penalize too many divisions
        r_pop = math.log(pop + 1) if pop > 0 else -5.0  # encourage non-extinction

        return r_tel + r_div + 0.5 * r_pop

    def run_episode(self, max_steps=200):
        # Reset the simulator for each new episode
        self.sim.reset()

        rewards = []
        log_probs = []
        infos = []

        for t in range(max_steps):
            obs = self.sim.get_observation()
            obs_t = torch.tensor(obs, dtype=torch.float32, device=self.device)
            act_prob = self.policy(obs_t)
            # sample telomerase activation from Bernoulli-like continuous action
            # we add small Gaussian noise to encourage exploration
            noise = torch.randn_like(act_prob) * 0.05
            activation = float((act_prob + noise).clamp(0.0, 1.0).item())

            # store log prob for policy gradient
            # treat the action as a continuous distribution centered on act_prob
            # log_prob approximated by gaussian
            mean = act_prob
            std = torch.tensor(0.05, device=self.device)
            var = std**2
            log_prob = -((activation - mean)**2) / (2*var) - torch.log(std * torch.sqrt(torch.tensor(2*math.pi)))
            log_probs.append(log_prob)

            info = self.sim.step(activation)
            infos.append(info)

            r = self.compute_reward(obs, info)
            rewards.append(r)

            # early stop if population extinct or too large
            if info['population'] == 0 or info['population'] > 5000:
                break

        # compute discounted returns
        returns = []
        G = 0.0
        gamma = 0.99
        for r in reversed(rewards):
            G = r + gamma * G
            returns.insert(0, G)
        returns = torch.tensor(returns, dtype=torch.float32, device=self.device)
        # Only normalize if there's more than one element to avoid std() == 0
        if len(returns) > 1:
            returns = (returns - returns.mean()) / (returns.std() + 1e-8)
        else:
            returns = returns - returns.mean() # Just center it if only one element

        loss = 0.0
        for lp, G in zip(log_probs, returns):
            loss = loss - (lp * G)  # gradient ascent -> minimize negative
        loss = loss / (len(log_probs) + 1e-8)

        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()

        return {'loss': loss.item(), 'episode_length': len(rewards), 'final_info': infos[-1] if infos else None}


# ----------------------------- Utilities / run script -------------------------

def run_training(seed=42, episodes=100, episode_steps=200, lr=1e-2):
    sim = OOCSimulator(initial_cells=100, # Reduced initial cells
                       initial_telomere=100.0,
                       min_telomere_to_divide=15.0, # Increased min telomere to divide
                       division_age=20.0,
                       telomere_loss_per_division=5.0,
                       telomerase_effect_per_activation=2.0, # Reduced telomerase effect
                       phagocytosis_capacity_per_step=50, # Increased phagocytosis capacity
                       rng_seed=seed)

    policy = PolicyNet(obs_dim=4, hidden=128)
    trainer = Trainer(sim, policy, lr=lr)

    metrics = {'loss': [], 'pop': [], 'avg_tel': [], 'divs': []}

    for ep in range(episodes):
        res = trainer.run_episode(max_steps=episode_steps)
        metrics['loss'].append(res['loss'])
        fin = res['final_info']
        if fin is not None:
            metrics['pop'].append(fin['population'])
            metrics['avg_tel'].append(fin['avg_telomere'])
            metrics['divs'].append(fin['divisions'])
        print(f"Episode {ep+1}/{episodes}: loss={res['loss']:.4f}, steps={res['episode_length']}, pop={fin['population'] if fin else 'N/A'}")

    # Simple plotting
    fig, axs = plt.subplots(2, 2, figsize=(12, 8))
    axs = axs.flatten()
    axs[0].plot(metrics['loss'])
    axs[0].set_title('Policy loss per episode')
    if metrics['pop']:
        axs[1].plot(metrics['pop'])
        axs[1].set_title('Final population per episode')
    if metrics['avg_tel']:
        axs[2].plot(metrics['avg_tel'])
        axs[2].set_title('Final average telomere per episode')
    if metrics['divs']:
        axs[3].plot(metrics['divs'])
        axs[3].set_title('Final divisions per episode')
    plt.tight_layout()
    plt.show()

    # Save model
    torch.save(policy.state_dict(), 'policy_telomerase_control.pth')
    print('Saved policy to policy_telomerase_control.pth')


if __name__ == '__main__':
    parser = argparse.ArgumentParser()
    parser.add_argument('--seed', type=int, default=42)
    parser.add_argument('--episodes', type=int, default=50)
    parser.add_argument('--steps', type=int, default=200)
    parser.add_argument('--lr', type=float, default=1e-2)
    args, unknown = parser.parse_known_args()

    run_training(seed=args.seed, episodes=args.episodes, episode_steps=args.steps, lr=args.lr)
```

```python
import numpy as np
from dataclasses import dataclass

# ------------------------------------------------------------
# SOMATIC CELL PROFILE (defines what an adult cell looks like)
# ------------------------------------------------------------

@dataclass
class SomaticCellProfile:
    name: str = "adult_fibroblast"
    initial_telomere: float = 60.0         # shorter than stem cells
    baseline_telomerase: float = 0.001     # somatic cells  inactive
    min_telomere_to_divide: float = 12.0
    division_age: float = 24.0
    max_divisions: int = 40                # Hayflick-like limit
    telomere_loss_per_division: float = 5.5
    senescence_threshold: float = 5.0
    phagocytosis_priority: int = 1         # how soon this cell is removed


# ------------------------------------------------------------
# IN-SILICO SOMATIC CELL OBJECT
# ------------------------------------------------------------

class SomaticCell:
    def __init__(self, profile: SomaticCellProfile):
        self.profile = profile
        self.age = 0.0
        self.generation = 0
        self.telomere = profile.initial_telomere
        self.division_count = 0
        self.alive = True
        self.senescent = False

    # ------------------------------------
    # Update cell age every simulation step
    # ------------------------------------
    def step_age(self, dt=1.0):
        if not self.alive:
            return
        self.age += dt

    # --------------------------------------------------------
    # Apply telomerase from PPO + baseline somatic telomerase
    # --------------------------------------------------------
    def apply_telomerase(self, activation: float, effect_per_step=0.5):
        if not self.alive:
            return
        activation = np.clip(activation, 0.0, 1.0)
        baseline = self.profile.baseline_telomerase

        total = activation + baseline
        total = np.clip(total, 0.0, 1.0)

        self.telomere += total * effect_per_step

    # --------------------------------------------------------
    # Determine whether the cell can divide (somatic constraints)
    # --------------------------------------------------------
    def ready_to_divide(self):
        if self.senescent or not self.alive:
            return False

        p = self.profile

        return (
            self.age >= p.division_age and
            self.telomere > p.min_telomere_to_divide and
            self.division_count < p.max_divisions
        )

    # --------------------------------------------------------
    # Perform division  returns a daughter cell
    # --------------------------------------------------------
    def divide(self):
        if not self.ready_to_divide():
            return None

        p = self.profile

        self.division_count += 1
        self.age = 0.0

        # Telomere loss in parent
        self.telomere -= p.telomere_loss_per_division
        if self.telomere < 0:
            self.telomere = 0

        # Create daughter cell
        daughter_tel = max(
            0.0,
            self.telomere - np.random.uniform(0, p.telomere_loss_per_division * 0.5)
        )

        daughter = SomaticCell(profile=p)
        daughter.telomere = daughter_tel
        daughter.generation = self.generation + 1

        return daughter

    # --------------------------------------------------------
    # Determine senescence and death
    # --------------------------------------------------------
    def evaluate_viability(self):
        p = self.profile

        if self.telomere <= 0 or self.telomere < p.senescence_threshold:
            self.senescent = True

        if self.division_count >= p.max_divisions:
            self.senescent = True

        if self.senescent and np.random.random() < 0.1:
            self.alive = False

    # --------------------------------------------------------
    # Run complete single simulation step
    # --------------------------------------------------------
    def step(self, telomerase_signal):
        if not self.alive:
            return None

        self.step_age(1.0)
        self.apply_telomerase(telomerase_signal)
        daughter = None

        if self.ready_to_divide():
            daughter = self.divide()

        self.evaluate_viability()
        return daughter


# ------------------------------------------------------------
# EXAMPLE MINI-SIMULATION
# ------------------------------------------------------------

if __name__ == "__main__":
    profile = SomaticCellProfile()
    cell = SomaticCell(profile)

    population = [cell]

    for t in range(200):
        new_cells = []
        for c in population:
            d = c.step(telomerase_signal=0.05)  # low PPO signal
            if d is not None:
                new_cells.append(d)

        population.extend(new_cells)

        # Remove dead cells
        population = [c for c in population if c.alive]

        print(f"t={t}, population={len(population)}, avg_telomere={np.mean([c.telomere for c in population]):.2f}")
```

```python
import torch
import torch.nn as nn
from typing import Tuple

# Re-define ActorCritic class as it's needed to load the model
class ActorCritic(nn.Module):
    def __init__(self, obs_dim=5, hidden=128):
        super().__init__()
        # Simpler shared encoder
        self.shared_encoder = nn.Sequential(
            nn.Linear(obs_dim, hidden),
            nn.GELU(),
            nn.LayerNorm(hidden),
            nn.Linear(hidden, hidden),
            nn.GELU(),
            nn.LayerNorm(hidden)
        )

        # actor head
        self.policy_mean = nn.Sequential(
            nn.Linear(hidden, hidden // 2),
            nn.GELU(),
            nn.LayerNorm(hidden // 2),
            nn.Linear(hidden // 2, hidden // 4),
            nn.GELU(),
            nn.LayerNorm(hidden // 4),
            nn.Linear(hidden // 4, 1),
            nn.Sigmoid()
        )
        # learned log-std (clamped for stability)
        self.log_std = nn.Parameter(torch.tensor(-2.0)) # Initializing to -2.0 for slightly higher initial std
        self.min_log_std_val = -5.0 # min std = exp(-5) approx 0.0067
        self.max_log_std_val = 0.0 # max std = 1.0 (exp(0)=1)

        # critic head
        self.value_head = nn.Sequential(
            nn.Linear(hidden, hidden // 2),
            nn.GELU(),
            nn.LayerNorm(hidden // 2),
            nn.Linear(hidden // 2, hidden // 4),
            nn.GELU(),
            nn.LayerNorm(hidden // 4),
            nn.Linear(hidden // 4, 1)
        )

    def forward(self, x: torch.Tensor) -> Tuple[torch.Tensor, torch.Tensor, torch.Tensor]:
        h = self.shared_encoder(x)

        mean = self.policy_mean(h).squeeze(-1)
        # Clamp log_std to a reasonable range during forward pass
        log_std_clamped = torch.clamp(self.log_std, self.min_log_std_val, self.max_log_std_val)
        std = torch.exp(log_std_clamped)

        value = self.value_head(h).squeeze(-1)
        return mean, std, value

# Instantiate the model with the correct dimensions
obs_dim = 5 # population, avg_tel, std_tel, neuron_count, fibro_count
hidden_dim = 128
loaded_policy = ActorCritic(obs_dim=obs_dim, hidden=hidden_dim)

# Load the saved state dictionary
loaded_policy.load_state_dict(torch.load('ppo_multi_tissue.pth'))
loaded_policy.eval() # Set the model to evaluation mode

print("Policy loaded successfully and set to evaluation mode.")
```

```python
"""
Virtual Organ-on-a-Chip (in-silico)  Multi-cell-type tissue simulation + PPO

This file builds on the earlier simulator and PPO trainer and adds a multi-cell-type
`tissue` simulation that includes adult fibroblasts, hepatocytes, and neurons as
somatic cell presets. The implementation remains purely computational and safe  no
wet-lab protocols or biological instructions are included.

Features added in this version:
- CellTypeProfile and SomaticCell classes for type-specific behavior
- TissueSimulator: manages mixed populations, phagocytosis, and type-aware logic
- Integration with the ActorCritic PPO trainer (existing)  trainer interacts with the tissue
- CLI switches to run short demonstrations or full PPO training
- Visualization helpers for per-type telomere distributions

Run examples:
- `python virtual_ooc_telomere_simulation.py --demo`  # run short multi-type sim
- `python virtual_ooc_telomere_simulation.py --ppo`   # train PPO on multi-type tissue

"""

import math
import random
import argparse
from collections import deque, namedtuple
from dataclasses import dataclass
from typing import List, Dict, Any, Tuple

import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim

# ----------------------------- Cell Type Profiles ---------------------------------

@dataclass
class CellTypeProfile:
    name: str
    initial_telomere: float
    baseline_telomerase: float
    min_telomere_to_divide: float
    division_age: float
    max_divisions: int
    telomere_loss_per_division: float
    senescence_threshold: float
    phagocytosis_priority: int
    notes: str = ""

# Presets for adult somatic cell types
CELL_PROFILES = {
    "fibroblast": CellTypeProfile(
        name="fibroblast",
        initial_telomere=60.0,
        baseline_telomerase=0.001,
        min_telomere_to_divide=12.0,
        division_age=24.0,
        max_divisions=40,
        telomere_loss_per_division=5.5,
        senescence_threshold=5.0,
        phagocytosis_priority=1,
        notes="Adult fibroblast: moderate division, limited replicative capacity"
    ),
    "hepatocyte": CellTypeProfile(
        name="hepatocyte",
        initial_telomere=70.0,
        baseline_telomerase=0.01,
        min_telomere_to_divide=10.0,
        division_age=40.0,
        max_divisions=100,
        telomere_loss_per_division=4.0,
        senescence_threshold=4.0,
        phagocytosis_priority=2,
        notes="Hepatocyte: quiescent but regenerative"
    ),
    "neuron": CellTypeProfile(
        name="neuron",
        initial_telomere=65.0,
        baseline_telomerase=0.0,
        min_telomere_to_divide=1e6,  # effectively never divides
        division_age=1e6,
        max_divisions=0,
        telomere_loss_per_division=0.0,
        senescence_threshold=1.0,
        phagocytosis_priority=3,
        notes="Post-mitotic neuron: effectively non-dividing in adult model"
    )
}

# ----------------------------- Somatic Cell ---------------------------------

class SomaticCell:
    def __init__(self, profile: CellTypeProfile, cid: int):
        self.id = cid
        self.profile = profile
        self.age = random.uniform(0, profile.division_age)  # staggered initial ages
        self.generation = 0
        self.telomere = profile.initial_telomere + random.uniform(-3, 3)
        self.division_count = 0
        self.alive = True
        self.senescent = False

    def step_age(self, dt=1.0):
        if not self.alive:
            return
        self.age += dt

    def apply_telomerase(self, activation: float, effect_per_step=0.5):
        if not self.alive:
            return
        activation = float(np.clip(activation, 0.0, 1.0))
        baseline = self.profile.baseline_telomerase
        total = min(1.0, activation + baseline)
        self.telomere += total * effect_per_step

    def ready_to_divide(self):
        if self.senescent or not self.alive:
            return False
        p = self.profile
        return (
            self.age >= p.division_age and
            self.telomere > p.min_telomere_to_divide and
            (p.max_divisions == 0 and False or self.division_count < p.max_divisions)
        )

    def divide(self):
        if not self.ready_to_divide():
            return None
        p = self.profile
        self.division_count += 1
        self.age = 0.0
        self.generation += 1
        self.telomere -= p.telomere_loss_per_division
        if self.telomere < 0:
            self.telomere = 0.0
        daughter = SomaticCell(profile=p, cid=-1)
        daughter.telomere = max(0.0, self.telomere - random.uniform(0, p.telomere_loss_per_division * 0.5))
        daughter.generation = self.generation
        return daughter

    def evaluate_viability(self):
        p = self.profile
        if self.telomere <= 0 or self.telomere < p.senescence_threshold:
            self.senescent = True
        if p.max_divisions > 0 and self.division_count >= p.max_divisions:
            self.senescent = True
        # stochastic clearance
        if self.senescent and random.random() < 0.1:
            self.alive = False

    def step(self, telomerase_signal):
        if not self.alive:
            return None
        self.step_age(1.0)
        self.apply_telomerase(telomerase_signal)
        daughter = None
        if self.ready_to_divide():
            daughter = self.divide()
        self.evaluate_viability()
        return daughter


# ----------------------------- Tissue Simulator ---------------------------------

class TissueSimulator:
    def __init__(self,
                 initial_distribution: Dict[str, int] = None,
                 phagocytosis_capacity_per_step: int = 20,
                 telomerase_effect_per_activation: float = 0.5,
                 rng_seed: int = None):
        if rng_seed is not None:
            random.seed(rng_seed)
            np.random.seed(rng_seed)
            torch.manual_seed(rng_seed)

        self.cells: Dict[int, SomaticCell] = {}
        self.next_id = 0
        self.step_count = 0
        self.history = {"population": [], "avg_tel": [], "by_type": {k: [] for k in CELL_PROFILES.keys()}}
        self.phagocytosis_capacity_per_step = phagocytosis_capacity_per_step
        self.telomerase_effect_per_activation = telomerase_effect_per_activation

        if initial_distribution is None:
            initial_distribution = {"fibroblast": 80, "hepatocyte": 15, "neuron": 5}

        for tname, count in initial_distribution.items():
            for _ in range(count):
                self._add_cell(tname)

    def _add_cell(self, type_name: str):
        profile = CELL_PROFILES[type_name]
        c = SomaticCell(profile=profile, cid=self.next_id)
        self.cells[self.next_id] = c
        self.next_id += 1
        return c

    def step(self, telomerase_activation: float):
        divisions = 0
        deaths = 0
        # apply per-cell step
        new_cells = []
        for c in list(self.cells.values()):
            if not c.alive:
                continue
            # combine global activation with type-specific baseline internally
            c.apply_telomerase(telomerase_activation, effect_per_step=self.telomerase_effect_per_activation)
            c.step_age(1.0)
            if c.ready_to_divide():
                d = c.divide()
                if d is not None:
                    # give proper unique id
                    d.id = self.next_id
                    self.cells[self.next_id] = d
                    self.next_id += 1
                    divisions += 1
            c.evaluate_viability()

        # collect dead ids
        dead_ids = [cid for cid, cell in self.cells.items() if not cell.alive]
        # sort by phagocytosis priority (small -> removed first)
        dead_ids_sorted = sorted(dead_ids, key=lambda cid: CELL_PROFILES[self.cells[cid].profile.name].phagocytosis_priority)
        removed = 0
        for cid in dead_ids_sorted:
            if removed >= self.phagocytosis_capacity_per_step:
                break
            del self.cells[cid]
            removed += 1
            deaths += 1

        self.step_count += 1

        # record history
        pop = len(self.cells)
        avg_tel = np.mean([c.telomere for c in self.cells.values()]) if pop > 0 else 0.0
        self.history['population'].append(pop)
        self.history['avg_tel'].append(avg_tel)
        for tname in CELL_PROFILES.keys():
            self.history['by_type'][tname].append(len([c for c in self.cells.values() if c.profile.name == tname]))

        return {"population": pop, "avg_tel": avg_tel, "divisions": divisions, "deaths": deaths}

    def get_observation(self):
        pop = len(self.cells)
        telomeres = np.array([c.telomere for c in self.cells.values()]) if pop > 0 else np.array([0.0])
        avg = float(np.mean(telomeres))
        std = float(np.std(telomeres))
        # recent divisions: sum last 10
        recent_divs = 0
        if len(self.history['population']) > 0:
            recent_divs = int(sum([max(0, self.history['by_type'][t][-1] - self.history['by_type'][t][-2]) if len(self.history['by_type'][t])>1 else 0 for t in self.history['by_type']]))
        # observation vector: total pop, avg tel, std tel, neurons count (proxy), fibro count
        neuron_count = len([c for c in self.cells.values() if c.profile.name == 'neuron'])
        fibro_count = len([c for c in self.cells.values() if c.profile.name == 'fibroblast'])
        return np.array([pop, avg, std, neuron_count, fibro_count], dtype=np.float32)

    def reset(self, initial_distribution: Dict[str, int] = None):
        self.cells = {}
        self.next_id = 0
        self.step_count = 0
        self.history = {"population": [], "avg_tel": [], "by_type": {k: [] for k in CELL_PROFILES.keys()}}
        if initial_distribution is None:
            initial_distribution = {"fibroblast": 80, "hepatocyte": 15, "neuron": 5}
        for tname, count in initial_distribution.items():
            for _ in range(count):
                self._add_cell(tname)


# ----------------------------- Actor-Critic and PPO Trainer (integrated) ---------------------------

class ActorCritic(nn.Module):
    def __init__(self, obs_dim=5, hidden=128): # Enforce hidden=128
        super().__init__()
        # Simpler shared encoder
        self.shared_encoder = nn.Sequential(
            nn.Linear(obs_dim, hidden),
            nn.GELU(),
            nn.LayerNorm(hidden),
            nn.Linear(hidden, hidden),
            nn.GELU(),
            nn.LayerNorm(hidden)
        )

        # actor head
        self.policy_mean = nn.Sequential(
            nn.Linear(hidden, hidden // 2),
            nn.GELU(),
            nn.LayerNorm(hidden // 2),
            nn.Linear(hidden // 2, hidden // 4),
            nn.GELU(),
            nn.LayerNorm(hidden // 4),
            nn.Linear(hidden // 4, 1),
            nn.Sigmoid()
        )
        # learned log-std (clamped for stability)
        self.log_std = nn.Parameter(torch.tensor(-2.0)) # Initializing to -2.0 for slightly higher initial std
        self.min_log_std_val = -5.0 # min std = exp(-5) approx 0.0067
        self.max_log_std_val = 0.0 # max std = 1.0 (exp(0)=1)

        # critic head
        self.value_head = nn.Sequential(
            nn.Linear(hidden, hidden // 2),
            nn.GELU(),
            nn.LayerNorm(hidden // 2),
            nn.Linear(hidden // 2, hidden // 4),
            nn.GELU(),
            nn.LayerNorm(hidden // 4),
            nn.Linear(hidden // 4, 1)
        )

    def forward(self, x: torch.Tensor) -> Tuple[torch.Tensor, torch.Tensor, torch.Tensor]:
        h = self.shared_encoder(x)

        mean = self.policy_mean(h).squeeze(-1)
        # Clamp log_std to a reasonable range during forward pass
        log_std_clamped = torch.clamp(self.log_std, self.min_log_std_val, self.max_log_std_val)
        std = torch.exp(log_std_clamped)

        value = self.value_head(h).squeeze(-1)
        return mean, std, value

Transition = namedtuple('Transition', ['obs', 'action', 'logp', 'reward', 'value', 'done'])

class PPOTrainer:
    def __init__(self,
                 sim: TissueSimulator,
                 model: ActorCritic,
                 lr=3e-4,
                 gamma=0.99,
                 lam=0.95,
                 clip_eps=0.2,
                 epochs=8,
                 batch_size=64,
                 device='cpu'):
        self.sim = sim
        self.model = model.to(device)
        self.optimizer = optim.AdamW(self.model.parameters(), lr=lr, weight_decay=1e-4)
        self.gamma = gamma
        self.lam = lam
        self.clip_eps = clip_eps
        self.epochs = epochs
        self.batch_size = batch_size
        self.device = device

    def select_action(self, obs: np.ndarray):
        obs_t = torch.tensor(obs, dtype=torch.float32, device=self.device)
        mean, std, value = self.model(obs_t)
        dist = torch.distributions.Normal(mean, std)
        action = dist.sample()
        action_clamped = action.clamp(0.0, 1.0)
        logp = dist.log_prob(action).sum()
        return float(action_clamped.item()), float(logp.item()), float(value.item())

    def compute_gae(self, transitions: List[Transition], last_value: float):
        rewards = [t.reward for t in transitions]
        values = [t.value for t in transitions]
        dones = [t.done for t in transitions]
        gae = 0.0
        returns = []
        advantages = []
        values_plus = values + [last_value]
        for step in reversed(range(len(rewards))):
            delta = rewards[step] + self.gamma * values_plus[step+1] * (1 - dones[step]) - values_plus[step]
            gae = delta + self.gamma * self.lam * (1 - dones[step]) * gae
            advantages.insert(0, gae)
            returns.insert(0, gae + values[step])
        adv = np.array(advantages, dtype=np.float32)
        ret = np.array(returns, dtype=np.float32)
        adv = (adv - adv.mean()) / (adv.std() + 1e-8)
        return adv, ret

    def update(self, transitions: List[Transition], advantages: np.ndarray, returns: np.ndarray):
        obs = torch.tensor(np.vstack([t.obs for t in transitions]), dtype=torch.float32, device=self.device)
        actions = torch.tensor([t.action for t in transitions], dtype=torch.float32, device=self.device)
        old_logps = torch.tensor([t.logp for t in transitions], dtype=torch.float32, device=self.device)
        returns_t = torch.tensor(returns, dtype=torch.float32, device=self.device)
        advantages_t = torch.tensor(advantages, dtype=torch.float32, device=self.device)

        dataset_size = obs.shape[0]
        indices = np.arange(dataset_size)

        for epoch in range(self.epochs):
            np.random.shuffle(indices)
            for start in range(0, dataset_size, self.batch_size):
                batch_idx = indices[start:start+self.batch_size]
                obs_b = obs[batch_idx]
                actions_b = actions[batch_idx]
                old_logps_b = old_logps[batch_idx]
                returns_b = returns_t[batch_idx]
                adv_b = advantages_t[batch_idx]

                mean, std, values = self.model(obs_b)
                dist = torch.distributions.Normal(mean, std)
                new_logps = dist.log_prob(actions_b).sum(dim=-1)

                ratio = torch.exp(new_logps - old_logps_b)
                surr1 = ratio * adv_b
                surr2 = torch.clamp(ratio, 1.0 - self.clip_eps, 1.0 + self.clip_eps) * adv_b
                policy_loss = -torch.min(surr1, surr2).mean()

                value_loss = nn.functional.mse_loss(values, returns_b)
                entropy = dist.entropy().mean()

                loss = policy_loss + 0.5 * value_loss - 0.01 * entropy

                self.optimizer.zero_grad()
                loss.backward()
                nn.utils.clip_grad_norm_(self.model.parameters(), 0.5)
                self.optimizer.step()

    def train(self, total_timesteps=20000, rollout_length=128):
        transitions: List[Transition] = []
        timesteps = 0
        episode = 0
        metrics = {'avg_reward': [], 'pop': [], 'avg_tel': []}

        self.sim.reset()
        obs = self.sim.get_observation()

        while timesteps < total_timesteps:
            transitions.clear()
            for _ in range(rollout_length):
                action, logp, value = self.select_action(obs)
                info = self.sim.step(action)
                reward = self.compute_reward(obs, info)
                done = (info['population'] == 0) or (info['population'] > 5000)

                transitions.append(Transition(obs=obs, action=action, logp=logp, reward=reward, value=value, done=done))

                obs = self.sim.get_observation()
                timesteps += 1
                if done:
                    metrics['avg_reward'].append(sum([t.reward for t in transitions]))
                    metrics['pop'].append(info['population'])
                    metrics['avg_tel'].append(info['avg_tel'])
                    self.sim.reset()
                    obs = self.sim.get_observation()
                    episode += 1

                if timesteps >= total_timesteps:
                    break

            _, _, last_value = self.model(torch.tensor(obs, dtype=torch.float32, device=self.device))
            last_value = float(last_value.item())
            advantages, returns = self.compute_gae(transitions, last_value)
            self.update(transitions, advantages, returns)

            if len(metrics['avg_reward']) > 0:
                print(f"Episode {episode}, timesteps {timesteps}, recent avg reward {metrics['avg_reward'][-1]:.3f}")
            else:
                print(f"Timesteps {timesteps}, episode {episode}")

        torch.save(self.model.state_dict(), 'ppo_multi_tissue.pth')
        print('Saved PPO policy to ppo_multi_tissue.pth')

        if metrics['avg_reward']:
            plt.figure(figsize=(8,4))
            plt.plot(metrics['avg_reward'])
            plt.title('Episode returns')
            plt.show()

    def compute_reward(self, obs, info, target_telomere=65.0):
        avg_tel = info['avg_tel']
        pop = info['population']
        divisions = info['divisions']
        r_tel = -abs(avg_tel - target_telomere) / (target_telomere + 1e-6)
        r_div = -divisions * 0.05
        r_pop = math.log(pop + 1) if pop > 0 else -5.0
        # extra bonus for preserving neurons
        neuron_count = len([c for c in self.sim.cells.values() if c.profile.name == 'neuron'])
        r_neurons = 0.2 * (neuron_count / (neuron_count + 1))
        return r_tel + r_div + 0.5 * r_pop + r_neurons


# ----------------------------- Utilities / Demo ---------------------------------

def run_demo(seed=42, steps=200):
    sim = TissueSimulator(rng_seed=seed)
    # simple policy: low constant activation
    activation = 0.02
    pops = []
    for t in range(steps):
        info = sim.step(activation)
        pops.append(info['population'])
    # plot populations per cell type
    plt.figure(figsize=(8,5))
    for tname in sim.history['by_type']:
        plt.plot(sim.history['by_type'][tname], label=tname)
    plt.legend()
    plt.title('Per-type population over time (demo)')
    plt.xlabel('Steps')
    plt.ylabel('Count')
    plt.show()


def run_ppo_training(seed=42, timesteps=20000):
    sim = TissueSimulator(rng_seed=seed)
    model = ActorCritic(obs_dim=5, hidden=128) # Enforce hidden=128
    trainer = PPOTrainer(sim, model, lr=3e-4, gamma=0.99, lam=0.95, clip_eps=0.2, epochs=8, batch_size=64) # Ensure lr=3e-4
    trainer.train(total_timesteps=timesteps, rollout_length=128)


if __name__ == '__main__':
    parser = argparse.ArgumentParser()
    parser.add_argument('--demo', action='store_true', help='Run a short multi-type simulation demo')
    parser.add_argument('--ppo', action='store_true', help='Train PPO on multi-type tissue')
    parser.add_argument('--timesteps', type=int, default=20000)
    args, unknown = parser.parse_known_args() # Change from parse_args() to parse_known_args()

    # Programmatically set args.ppo to True to initiate training as requested
    args.ppo = True

    if args.demo:
        run_demo()
    elif args.ppo:
        run_ppo_training(timesteps=args.timesteps)
    else:
        print('Use --demo to see a short multi-type tissue simulation, or --ppo to train the PPO controller.')
```
