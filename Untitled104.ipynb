```python
# ===============================================
# Virtual Cell Simulation Core
# AI-Regulated Telomerase & Synthetic Phagocytosis
# ===============================================

import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor

# -------------------------------
# 1. Parameters and Initialization
# -------------------------------

class VirtualCell:
    def __init__(self, telomere_length=10000, damage_rate=0.01):
        self.telomere_length = telomere_length  # base pairs
        self.telomerase_activity = 0.0
        self.cell_age = 0
        self.damage_level = 0.0
        self.damage_rate = damage_rate
        self.divisions = 0
        self.alive = True
        self.debris = 0.0  # accumulated waste

    # --------------------------
    # 2. Telomerase Regulation
    # --------------------------
    def update_telomerase(self, ai_model):
        input_data = np.array([[self.telomere_length, self.damage_level, self.divisions]])
        self.telomerase_activity = ai_model.predict(input_data)[0]
        self.telomerase_activity = np.clip(self.telomerase_activity, 0, 1)

    # --------------------------
    # 3. Cell Division Control
    # --------------------------
    def divide(self):
        if self.telomerase_activity > 0.4 and self.damage_level < 0.3:
            self.telomere_length += int(100 * self.telomerase_activity)
            self.divisions += 1
        else:
            self.telomere_length -= int(50 * (1 - self.telomerase_activity))
        self.cell_age += 1
        if self.telomere_length <= 3000 or self.damage_level > 0.8:
            self.alive = False

    # --------------------------
    # 4. Synthetic Phagocytosis
    # --------------------------
    def phagocytosis(self):
        # Debris clearance proportional to efficiency
        efficiency = 1 - (self.damage_level + np.random.uniform(0, 0.1))
        cleared = self.debris * efficiency
        self.debris -= cleared
        self.damage_level = max(0.0, self.damage_level - cleared * 0.1)
        return cleared

    # --------------------------
    # 5. Environmental Stress
    # --------------------------
    def accumulate_damage(self):
        self.damage_level += np.random.uniform(0, self.damage_rate)
        self.debris += np.random.uniform(0, 0.05)


# ------------------------------
# 6. AI Model for Telomerase Control
# ------------------------------
def train_telomerase_ai():
    # Synthetic training data: telomere length, damage, divisions  telomerase activity
    X = np.random.rand(1000, 3)
    y = np.clip(1 - X[:,0]*0.3 - X[:,1]*0.5 + X[:,2]*0.1, 0, 1)
    model = MLPRegressor(hidden_layer_sizes=(32,16), activation='relu', max_iter=2000)
    model.fit(X, y)
    return model


# ------------------------------
# 7. Simulation Loop
# ------------------------------
def run_simulation(steps=200):
    ai_model = train_telomerase_ai()
    cell = VirtualCell()
    history = {"telomere": [], "damage": [], "activity": [], "divisions": []}

    for _ in range(steps):
        if not cell.alive:
            break
        cell.accumulate_damage()
        cell.update_telomerase(ai_model)
        cell.divide()
        cell.phagocytosis()

        history["telomere"].append(cell.telomere_length)
        history["damage"].append(cell.damage_level)
        history["activity"].append(cell.telomerase_activity)
        history["divisions"].append(cell.divisions)

    return history


# ------------------------------
# 8. Visualization
# ------------------------------
def plot_results(history):
    plt.figure(figsize=(10,6))
    plt.subplot(3,1,1)
    plt.plot(history["telomere"], label="Telomere Length")
    plt.ylabel("Length (bp)")
    plt.legend()

    plt.subplot(3,1,2)
    plt.plot(history["damage"], label="Damage Level", color='orange')
    plt.ylabel("Damage")
    plt.legend()

    plt.subplot(3,1,3)
    plt.plot(history["activity"], label="Telomerase Activity", color='green')
    plt.ylabel("Activity")
    plt.xlabel("Time Step")
    plt.legend()

    plt.tight_layout()
    plt.show()


# ------------------------------
# 9. Run the Simulation
# ------------------------------
if __name__ == "__main__":
    history = run_simulation(steps=300)
    plot_results(history)
```

```python
# Multi-cell Organ-on-a-Chip Simulation
# AI-shared controller for telomerase regulation, cell-cell signaling, waste recycling
# Colab-ready: requires numpy, matplotlib, scikit-learn
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from matplotlib import animation

# -----------------------------
# 1) Virtual Cell Definition
# -----------------------------
class VirtualCell:
    def __init__(self, x, y, telomere_length=8000, damage=0.0):
        self.x = x
        self.y = y
        self.telomere_length = float(telomere_length)  # base pairs
        self.damage = float(damage)                    # 0..1
        self.debris = 0.0                              # accumulated waste
        self.divisions = 0
        self.alive = True
        self.telomerase_activity = 0.0                 # 0..1

    def state_vector(self):
        # Vector used by the shared AI controller
        return np.array([self.telomere_length, self.damage, self.debris, self.divisions], dtype=float)

    def apply_telomerase(self):
        # Telomerase extends telomeres proportionally; absence leads to steady shortening
        activity = float(np.clip(self.telomerase_activity, 0.0, 1.0))
        # Extension and shortening scales
        extension = 50.0 * activity
        shortening = 20.0 * (1.0 - activity) * (1.0 + self.damage)
        self.telomere_length += extension - shortening

    def attempt_division(self):
        # Conditions to divide: adequate telomere length, low damage, and telomerase support
        if (self.telomere_length > 5000) and (self.damage < 0.4) and (self.telomerase_activity > 0.35):
            self.divisions += 1
            # small cost to damage and debris with division
            self.damage = min(1.0, self.damage + np.random.uniform(0.0, 0.02))
            self.debris += np.random.uniform(0.0, 0.05)
            # modest telomere shortening due to imperfect replication (balanced by telomerase)
            self.telomere_length -= np.random.uniform(0, 30)
            return True
        return False

    def accumulate_environmental_stress(self, stress_rate):
        # Random environmental stress increases damage and debris
        self.damage = min(1.0, self.damage + np.random.exponential(scale=stress_rate))
        self.debris += np.random.exponential(scale=0.02)

    def phagocytose(self, efficiency):
        # Clear debris; efficiency 0..1
        cleared = min(self.debris, self.debris * efficiency)
        self.debris -= cleared
        # clearing reduces damage slightly
        self.damage = max(0.0, self.damage - cleared * 0.05)

    def check_viability(self):
        # Death rules: critically short telomeres or very high damage
        if (self.telomere_length < 3000) or (self.damage >= 0.95):
            self.alive = False

# -----------------------------
# 2) Shared AI Controller
# -----------------------------
class SharedAIController:
    def __init__(self):
        # Pre-train on synthetic data to map local state + neighbor influence to telomerase activity
        self.model = self._train_initial_model()

    def _train_initial_model(self):
        # Generate synthetic training data
        # Inputs: [telomere_len_norm, damage, debris_norm, divisions_norm, neighbor_signal]
        n = 2000
        tel_norm = np.random.rand(n)
        damage = np.random.rand(n)
        debris_norm = np.random.rand(n)
        div_norm = np.random.rand(n)
        neighbor_signal = np.random.rand(n)

        X = np.vstack([tel_norm, damage, debris_norm, div_norm, neighbor_signal]).T
        # Target telomerase activity: favor moderate telomere lengths, low damage, positive neighbor signal
        # Synthetic rule (bounded 0..1):
        y = 0.6 * (1 - damage) + 0.3 * neighbor_signal + 0.1 * (1 - np.abs(tel_norm - 0.5))
        y = np.clip(y, 0.0, 1.0)

        model = MLPRegressor(hidden_layer_sizes=(64,32), activation='relu', max_iter=2000, random_state=42)
        model.fit(X, y)
        return model

    def predict_activity_grid(self, grid_cells, neighbor_signal_grid):
        # grid_cells: 2D array of VirtualCell or None; neighbor_signal_grid: same-shaped float array
        rows, cols = neighbor_signal_grid.shape
        activities = np.zeros((rows, cols))
        for i in range(rows):
            for j in range(cols):
                cell = grid_cells[i][j]
                if (cell is None) or (not cell.alive):
                    activities[i, j] = 0.0
                    continue
                # normalize telomere_length and debris/divisions for model input
                tel_norm = np.clip(cell.telomere_length / 10000.0, 0.0, 1.0)
                debris_norm = np.clip(cell.debris / 5.0, 0.0, 1.0)
                div_norm = np.tanh(cell.divisions / 10.0)  # saturating transform
                nbr = float(np.clip(neighbor_signal_grid[i, j], 0.0, 1.0))
                X = np.array([[tel_norm, cell.damage, debris_norm, div_norm, nbr]])
                activities[i, j] = float(np.clip(self.model.predict(X)[0], 0.0, 1.0))
        return activities

# -----------------------------
# 3) Organ-on-Chip Environment
# -----------------------------
class OrganOnChip:
    def __init__(self, rows=12, cols=12, stress_rate=0.02, flow_rate=0.2):
        self.rows = rows
        self.cols = cols
        self.grid = [[VirtualCell(i, j, telomere_length=np.random.normal(8000, 300), damage=np.random.rand()*0.1)
                      for j in range(cols)] for i in range(rows)]
        self.shared_ai = SharedAIController()
        # diffusing signals and waste fields
        self.signal = np.zeros((rows, cols), dtype=float)    # pro-regenerative signal
        self.waste = np.zeros((rows, cols), dtype=float)     # accumulated waste in environment
        self.flow_rate = flow_rate
        self.stress_rate = stress_rate
        self.time = 0

    def neighbor_indices(self, i, j):
        # 4-neighborhood wrapping off (non-periodic)
        inds = []
        for di, dj in [(-1,0),(1,0),(0,-1),(0,1)]:
            ni, nj = i+di, j+dj
            if 0 <= ni < self.rows and 0 <= nj < self.cols:
                inds.append((ni, nj))
        return inds

    def diffuse(self, field, diff_coeff=0.2):
        # Simple discrete diffusion: each cell exchanges with neighbors
        new = np.copy(field)
        for i in range(self.rows):
            for j in range(self.cols):
                nbrs = self.neighbor_indices(i,j)
                exchange = sum(field[ni, nj] for (ni, nj) in nbrs) - len(nbrs)*field[i, j]
                new[i, j] += diff_coeff * exchange / max(1, len(nbrs))
        return np.clip(new, 0.0, None)

    def step(self):
        # 1) Cells produce signals/waste based on damage and divisions
        for i in range(self.rows):
            for j in range(self.cols):
                cell = self.grid[i][j]
                if (cell is None) or (not cell.alive):
                    continue
                # more damaged cells produce more pro-regenerative signal (call for help)
                self.signal[i, j] += 0.05 * cell.damage
                # divisions and debris add to environmental waste
                self.waste[i, j] += 0.02 * cell.divisions + 0.1 * cell.debris

        # 2) Diffuse signal and waste across chip
        self.signal = self.diffuse(self.signal, diff_coeff=0.25)
        self.waste = self.diffuse(self.waste, diff_coeff=0.15)

        # 3) Flow: simulate microfluidic washout that reduces environmental waste
        washout = np.exp(-self.flow_rate)
        self.waste *= washout

        # 4) Compute neighbor signal (normalized) for AI input: combine local + averaged neighbors
        neighbor_signal_grid = np.zeros((self.rows, self.cols))
        for i in range(self.rows):
            for j in range(self.cols):
                nbrs = self.neighbor_indices(i,j)
                val = self.signal[i,j] + sum(self.signal[ni,nj] for (ni,nj) in nbrs)
                denom = 1 + len(nbrs)
                neighbor_signal_grid[i,j] = np.clip(val / denom, 0.0, 1.0)

        # 5) Shared AI decides telomerase activities (vectorized)
        activities = self.shared_ai.predict_activity_grid(self.grid, neighbor_signal_grid)

        # 6) Update each cell with activity, internal dynamics, phagocytosis, and viability checks
        for i in range(self.rows):
            for j in range(self.cols):
                cell = self.grid[i][j]
                if (cell is None) or (not cell.alive):
                    continue
                cell.telomerase_activity = activities[i,j]
                cell.accumulate_environmental_stress(self.stress_rate)
                cell.apply_telomerase()
                cell.attempt_division()
                # phagocytosis efficiency can be influenced by neighbor signal and low environmental waste
                efficiency = np.clip(0.5 + 0.4 * neighbor_signal_grid[i,j] - 0.2 * self.waste[i,j], 0.0, 1.0)
                cell.phagocytose(efficiency)
                cell.check_viability()

        # 7) Sink some signal gradually (decay)
        self.signal *= 0.95

        self.time += 1

    def grid_state_matrices(self):
        tel = np.zeros((self.rows, self.cols))
        dmg = np.zeros((self.rows, self.cols))
        act = np.zeros((self.rows, self.cols))
        alive = np.zeros((self.rows, self.cols))
        for i in range(self.rows):
            for j in range(self.cols):
                cell = self.grid[i][j]
                if (cell is None) or (not cell.alive):
                    tel[i,j] = 0.0
                    dmg[i,j] = 1.0
                    act[i,j] = 0.0
                    alive[i,j] = 0
                else:
                    tel[i,j] = cell.telomere_length
                    dmg[i,j] = cell.damage
                    act[i,j] = cell.telomerase_activity
                    alive[i,j] = 1
        return tel, dmg, act, alive

# -----------------------------
# 4) Visualization & Run
# -----------------------------
def run_experiment(steps=200, rows=12, cols=12, animate=True):
    env = OrganOnChip(rows=rows, cols=cols)
    history = {"avg_tel": [], "avg_damage": [], "avg_activity": [], "alive_fraction": []}
    frames = []

    for t in range(steps):
        env.step()
        tel, dmg, act, alive = env.grid_state_matrices()
        history["avg_tel"].append(np.nanmean(tel[tel>0]) if np.any(tel>0) else 0.0)
        history["avg_damage"].append(np.mean(dmg))
        history["avg_activity"].append(np.mean(act))
        history["alive_fraction"].append(np.mean(alive))
        if animate:
            frames.append((tel.copy(), dmg.copy(), act.copy(), env.waste.copy()))

    # Plot summary timeseries
    fig, axs = plt.subplots(2,2, figsize=(12,9))
    axs = axs.ravel()
    axs[0].plot(history["avg_tel"]); axs[0].set_title("Avg Telomere Length")
    axs[1].plot(history["avg_damage"]); axs[1].set_title("Avg Damage")
    axs[2].plot(history["avg_activity"]); axs[2].set_title("Avg Telomerase Activity")
    axs[3].plot(history["alive_fraction"]); axs[3].set_title("Alive Fraction")
    for a in axs:
        a.grid(True)
    plt.tight_layout()
    plt.show()

    # Animation (grid heatmaps)
    if animate:
        fig2, axarr = plt.subplots(1,4, figsize=(16,4))
        def update_frame(k):
            tel, dmg, act, waste = frames[k]
            axarr[0].clear(); axarr[1].clear(); axarr[2].clear(); axarr[3].clear()
            im0 = axarr[0].imshow(np.clip(tel, 0, 12000), vmin=3000, vmax=10000, aspect='auto')
            axarr[0].set_title("Telomere")
            im1 = axarr[1].imshow(dmg, vmin=0, vmax=1, aspect='auto')
            axarr[1].set_title("Damage")
            im2 = axarr[2].imshow(act, vmin=0, vmax=1, aspect='auto')
            axarr[2].set_title("Telomerase Activity")
            im3 = axarr[3].imshow(waste, aspect='auto')
            axarr[3].set_title("Environmental Waste")
            # colorbars
            fig2.suptitle(f"Time {k}")
            return im0, im1, im2, im3

        anim = animation.FuncAnimation(fig2, update_frame, frames=len(frames), interval=150, blit=False)
        plt.show()
    return env, history

# -----------------------------
# 5) Run the simulation
# -----------------------------
if __name__ == "__main__":
    env, history = run_experiment(steps=300, rows=14, cols=14, animate=True)
```

```python
# ============================================
# Organ-on-a-Chip Simulation with Genetic Drift & Mutation
# AI-regulated Telomerase and Cellular Evolution
# ============================================

import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from matplotlib import animation

# -----------------------------
# 1. Virtual Cell with Genome
# -----------------------------
class VirtualCell:
    def __init__(self, x, y, telomere_length=8000, damage=0.0):
        self.x, self.y = x, y
        self.telomere_length = float(telomere_length)
        self.damage = float(damage)
        self.debris = 0.0
        self.divisions = 0
        self.alive = True
        self.telomerase_activity = 0.0
        self.genetic_integrity = 1.0  # Initialize genetic_integrity
        self.epigenetic_state = 0.5 # Initialize epigenetic_state

        # Genomic traits (each 01)
        self.genes = {
            "TERT_reg": np.clip(np.random.normal(0.7, 0.1), 0.0, 1.0),
            "DNA_repair": np.clip(np.random.normal(0.7, 0.1), 0.0, 1.0),
            "Phago_eff": np.clip(np.random.normal(0.7, 0.1), 0.0, 1.0)
        }

    def state_vector(self):
        return np.array([self.telomere_length, self.damage, self.debris, self.divisions], dtype=float)

    # ----------------------------------
    # Genetic drift & mutation behavior
    # ----------------------------------
    def mutate(self, mutation_rate=0.001, drift_rate=0.01):
        for gene, val in self.genes.items():
            # Continuous drift
            self.genes[gene] = np.clip(val + np.random.normal(0, drift_rate), 0.0, 1.0)
            # Discrete mutation event
            if np.random.rand() < mutation_rate:
                change = np.random.uniform(-0.3, 0.3)
                self.genes[gene] = np.clip(val + change, 0.0, 1.0)
                # Small chance of malignancy (TERT_reg spikes uncontrollably)
                if gene == "TERT_reg" and np.random.rand() < 0.02:
                    self.genes["TERT_reg"] = 1.2  # cancer-like hyperactivation

    # -----------------------------
    # Telomerase regulation (AI + gene)
    # -----------------------------
    def apply_telomerase(self):
        activity = float(np.clip(self.telomerase_activity, 0.0, 1.0))
        response_factor = np.clip(self.genes["TERT_reg"], 0.0, 1.2)
        extension = 50.0 * activity * response_factor
        shortening = 20.0 * (1.0 - activity) * (1.0 + self.damage)
        self.telomere_length += extension - shortening

    # -----------------------------
    # Cell division & aging
    # -----------------------------
    def attempt_division(self):
        if (self.telomere_length > 5000) and (self.damage < 0.4):
            self.divisions += 1
            self.damage += np.random.uniform(0.0, 0.02)
            self.debris += np.random.uniform(0.0, 0.05)
            self.telomere_length -= np.random.uniform(0, 30)
            return True
        return False

    # -----------------------------
    # Environmental stress
    # -----------------------------
    def accumulate_stress(self, base_rate):
        repair_factor = 1.0 - self.genes["DNA_repair"]
        self.damage = min(1.0, self.damage + np.random.exponential(scale=base_rate * repair_factor))
        self.debris += np.random.exponential(scale=0.02)
        # Genetic drift: small integrity decrease
        self.genetic_integrity = max(0.0, self.genetic_integrity - abs(np.random.normal(0, 0.0005)))
        # Memory adaptation: damage increases epigenetic responsiveness
        if self.damage > 0.5:
            self.epigenetic_state = min(1.0, self.epigenetic_state + 0.01)

    # -----------------------------
    # Phagocytosis
    # -----------------------------
    def phagocytose(self, base_efficiency):
        genetic_boost = np.clip(self.genes["Phago_eff"], 0.0, 1.0)
        efficiency = np.clip(base_efficiency * (0.5 + 0.5 * genetic_boost), 0.0, 1.0)
        cleared = self.debris * efficiency
        self.debris -= cleared
        self.damage = max(0.0, self.damage - cleared * 0.05)

    def check_viability(self):
        if (self.telomere_length < 3000) or (self.damage >= 0.95):
            self.alive = False

# -----------------------------
# 2. Shared AI Controller
# -----------------------------
class SharedAIController:
    def __init__(self):
        self.model = self._train_model()

    def _train_model(self):
        n = 2000
        X = np.random.rand(n, 5)
        tel, dmg, debris, div, signal = X.T
        y = 0.5*(1 - dmg) + 0.3*signal + 0.2*(1 - np.abs(tel - 0.5))
        y = np.clip(y, 0.0, 1.0)
        model = MLPRegressor(hidden_layer_sizes=(64,32), activation='relu', max_iter=1500)
        model.fit(X, y)
        return model

    def predict_activity_grid(self, grid, neighbor_signal):
        r, c = neighbor_signal.shape
        act = np.zeros((r, c))
        for i in range(r):
            for j in range(c):
                cell = grid[i][j]
                if (cell is None) or (not cell.alive):
                    act[i, j] = 0.0
                    continue
                tel_norm = np.clip(cell.telomere_length / 10000.0, 0.0, 1.0)
                debris_norm = np.clip(cell.debris / 5.0, 0.0, 1.0)
                div_norm = np.tanh(cell.divisions / 10.0)
                X = np.array([[tel_norm, cell.damage, debris_norm, div_norm, neighbor_signal[i, j]]])
                raw = np.clip(self.model.predict(X)[0], 0.0, 1.0)
                # AI output modulated by the cells TERT_reg gene
                act[i, j] = np.clip(raw * cell.genes["TERT_reg"], 0.0, 1.0)
        return act

# -----------------------------
# 3. Organ-on-a-Chip Environment
# -----------------------------
class OrganOnChip:
    def __init__(self, rows=12, cols=12, stress_rate=0.02):
        self.rows, self.cols = rows, cols
        self.grid = [[VirtualCell(i, j) for j in range(cols)] for i in range(rows)]
        self.ai = SharedAIController()
        self.signal = np.zeros((rows, cols))
        self.waste = np.zeros((rows, cols))
        self.stress_rate = stress_rate
        self.time = 0

    def neighbors(self, i, j):
        dirs = [(-1,0),(1,0),(0,-1),(0,1)]
        return [(i+di,j+dj) for di,dj in dirs if 0<=i+di<self.rows and 0<=j+dj<self.cols]

    def diffuse(self, field, D=0.25):
        new = np.copy(field)
        for i in range(self.rows):
            for j in range(self.cols):
                vals = [field[x,y] for x,y in self.neighbors(i,j)]
                if vals:
                    avg = np.mean(vals)
                    new[i,j] += D*(avg - field[i,j])
        return np.clip(new, 0.0, None)

    def step(self):
        for i in range(self.rows):
            for j in range(self.cols):
                c = self.grid[i][j]
                if not c.alive: continue
                self.signal[i,j] += 0.05 * c.damage
                self.waste[i,j] += 0.05 * c.debris

        self.signal = self.diffuse(self.signal, D=0.3)
        self.waste = self.diffuse(self.waste, D=0.2)
        neighbor_signal = np.clip(self.signal, 0.0, 1.0)

        activities = self.ai.predict_activity_grid(self.grid, neighbor_signal)

        for i in range(self.rows):
            for j in range(self.cols):
                c = self.grid[i][j]
                if not c.alive: continue
                c.telomerase_activity = activities[i,j]
                c.accumulate_stress(self.stress_rate)
                c.apply_telomerase()
                c.attempt_division()
                base_eff = np.clip(0.5 + 0.4*neighbor_signal[i,j] - 0.3*self.waste[i,j], 0.0, 1.0)
                c.phagocytose(base_eff)
                c.mutate(mutation_rate=0.005, drift_rate=0.005)
                c.check_viability()

        self.signal *= 0.95
        self.time += 1

    def grid_matrices(self):
        tel = np.zeros((self.rows, self.cols))
        dmg = np.zeros((self.rows, self.cols))
        act = np.zeros((self.rows, self.cols))
        tert = np.zeros((self.rows, self.cols))
        alive = np.zeros((self.rows, self.cols))
        for i in range(self.rows):
            for j in range(self.cols):
                c = self.grid[i][j]
                if c.alive:
                    tel[i,j] = c.telomere_length
                    dmg[i,j] = c.damage
                    act[i,j] = c.telomerase_activity
                    tert[i,j] = c.genes["TERT_reg"]
                    alive[i,j] = 1
                else:
                    tel[i,j] = 0.0
                    dmg[i,j] = 1.0
                    act[i,j] = 0.0
                    tert[i,j] = 0.0
                    alive[i,j] = 0
        return tel, dmg, act, tert, alive

# -----------------------------
# 4. Simulation Runner
# -----------------------------
def run_simulation(steps=300, rows=12, cols=12, animate=True):
    env = OrganOnChip(rows, cols) # Re-initialize the environment
    history = {"tel":[], "damage":[], "activity":[], "alive":[]}
    frames=[]

    for _ in range(steps):
        env.step()
        tel,dmg,act,tert,alive = env.grid_matrices()
        history["tel"].append(np.mean(tel[tel>0]))
        history["damage"].append(np.mean(dmg))
        history["activity"].append(np.mean(act))
        history["alive"].append(np.mean(alive))
        if animate:
            frames.append((tel.copy(), dmg.copy(), act.copy(), tert.copy()))

    fig, ax = plt.subplots(2,2, figsize=(10,8))
    ax[0, 0].plot(history["tel"]); ax[0, 0].set_title("Mean Telomere Length")
    ax[0, 1].plot(history["damage"]); ax[0, 1].set_title("Mean Damage")
    ax[1, 0].plot(history["activity"]); ax[1, 0].set_title("Mean Telomerase Activity")
    ax[1, 1].plot(history["alive"]); ax[1, 1].set_title("Alive Fraction")
    plt.tight_layout(); plt.show()

    if animate:
        fig2, axs = plt.subplots(1,4, figsize=(16,4))
        def update(k):
            tel,dmg,act,tert = frames[k]
            axs[0].imshow(tel, vmin=3000, vmax=10000); axs[0].set_title("Telomere")
            axs[1].imshow(dmg, vmin=0, vmax=1); axs[1].set_title("Damage")
            axs[2].imshow(act, vmin=0, vmax=1); axs[2].set_title("Activity")
            axs[3].imshow(tert, vmin=0, vmax=1.2); axs[3].set_title("TERT_reg Gene")
        anim = animation.FuncAnimation(fig2, update, frames=len(frames), interval=150)
        plt.show()

    return env, history

# -----------------------------
# 5. Run
# -----------------------------
if __name__ == "__main__":
    env, hist = run_simulation(steps=300, rows=14, cols=14)
```

```python
# =========================================================
# Virtual Cell Simulation v3
# AI-Regulated Telomerase + Phagocytosis + Mutation + Epigenetic Memory
# =========================================================

import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor

# -------------------------------
# 1. Virtual Cell Class
# -------------------------------

class VirtualCell:
    def __init__(self, telomere_length=10000, damage_rate=0.01, mutation_rate=0.0001):
        self.telomere_length = telomere_length  # base pairs
        self.telomerase_activity = 0.0
        self.cell_age = 0
        self.damage_level = 0.0
        self.damage_rate = damage_rate
        self.divisions = 0
        self.alive = True
        self.debris = 0.0
        self.genetic_integrity = 1.0  # 1 = perfect genome, <1 = accumulated mutations
        self.mutation_rate = mutation_rate
        self.epigenetic_state = 0.5  # 0 = silent, 1 = fully active epigenetic memory
        self.memory_decay = 0.98  # how fast the cell forgets experiences

    # --------------------------
    # Telomerase Regulation (AI)
    # --------------------------
    def update_telomerase(self, ai_model):
        # Epigenetic modulation: past experience biases future telomerase activity
        input_data = np.array([[self.telomere_length, self.damage_level, self.divisions, self.epigenetic_state]])
        activity = ai_model.predict(input_data)[0]
        # Epigenetic influence: positive memory boosts repair; negative stress suppresses
        activity *= (0.8 + 0.4 * self.epigenetic_state)
        self.telomerase_activity = np.clip(activity, 0, 1)

    # --------------------------
    # Division and Aging
    # --------------------------
    def divide(self):
        if self.telomerase_activity > 0.4 and self.damage_level < 0.4 and self.genetic_integrity > 0.8:
            self.telomere_length += int(120 * self.telomerase_activity)
            self.divisions += 1
            # Memory reinforcement: successful divisions enhance epigenetic stability
            self.epigenetic_state = min(1.0, self.epigenetic_state + 0.02)
        else:
            self.telomere_length -= int(60 * (1 - self.telomerase_activity))
            self.epigenetic_state *= self.memory_decay

        self.cell_age += 1

        # If telomeres too short or damage too high  apoptosis
        if self.telomere_length <= 3000 or self.damage_level > 0.85 or self.genetic_integrity < 0.6:
            self.alive = False

    # --------------------------
    # Synthetic Phagocytosis
    # --------------------------
    def phagocytosis(self):
        efficiency = 1 - (self.damage_level + np.random.uniform(0, 0.1))
        cleared = self.debris * max(efficiency, 0)
        self.debris -= cleared
        self.damage_level = max(0.0, self.damage_level - cleared * 0.1)
        return cleared

    # --------------------------
    # Damage Accumulation
    # --------------------------
    def accumulate_damage(self):
        self.damage_level += np.random.uniform(0, self.damage_rate)
        self.debris += np.random.uniform(0, 0.05)
        self.genetic_drift()

    # --------------------------
    # Genetic Drift / Mutation
    # --------------------------
    def genetic_drift(self):
        mutation = np.random.normal(0, self.mutation_rate)
        self.genetic_integrity = max(0.0, min(1.0, self.genetic_integrity - abs(mutation)))
        # Memory adaptation: damage increases epigenetic responsiveness
        if self.damage_level > 0.5:
            self.epigenetic_state = min(1.0, self.epigenetic_state + 0.01)

# -------------------------------
# 2. AI Model for Telomerase Regulation
# -------------------------------
def train_telomerase_ai():
    X = np.random.rand(1500, 4)  # telomere, damage, divisions, epigenetic state
    y = np.clip(1 - X[:,0]*0.25 - X[:,1]*0.5 + X[:,2]*0.15 + X[:,3]*0.2, 0, 1)
    model = MLPRegressor(hidden_layer_sizes=(64,32), activation='relu', max_iter=3000)
    model.fit(X, y)
    return model

# -------------------------------
# 3. Simulation Loop
# -------------------------------
def run_simulation(steps=400):
    ai_model = train_telomerase_ai()
    cell = VirtualCell()
    history = {"telomere": [], "damage": [], "activity": [], "divisions": [],
               "integrity": [], "epigenetic": []}

    for _ in range(steps):
        if not cell.alive:
            break
        cell.accumulate_damage()
        cell.update_telomerase(ai_model)
        cell.divide()
        cell.phagocytosis()

        history["telomere"].append(cell.telomere_length)
        history["damage"].append(cell.damage_level)
        history["activity"].append(cell.telomerase_activity)
        history["divisions"].append(cell.divisions)
        history["integrity"].append(cell.genetic_integrity)
        history["epigenetic"].append(cell.epigenetic_state)

    return history

# -------------------------------
# 4. Visualization
# -------------------------------
def plot_results(history):
    plt.figure(figsize=(10,10))
    plt.subplot(4,1,1)
    plt.plot(history["telomere"], label="Telomere Length")
    plt.ylabel("Telomere (bp)")
    plt.legend()

    plt.subplot(4,1,2)
    plt.plot(history["damage"], label="Damage", color='orange')
    plt.plot(history["integrity"], label="Genetic Integrity", color='purple')
    plt.ylabel("Cell Health")
    plt.legend()

    plt.subplot(4,1,3)
    plt.plot(history["activity"], label="Telomerase Activity", color='green')
    plt.plot(history["epigenetic"], label="Epigenetic Memory", color='red', linestyle='--')
    plt.ylabel("Activity / Memory")
    plt.legend()

    plt.subplot(4,1,4)
    plt.plot(history["divisions"], label="Divisions", color='gray')
    plt.xlabel("Time Step")
    plt.ylabel("Divisions")
    plt.legend()

    plt.tight_layout()
    plt.show()

# -------------------------------
# 5. Run Simulation
# -------------------------------
if __name__ == "__main__":
    history = run_simulation(steps=500)
    plot_results(history)
```

```python
# Organ-on-Chip Simulation
# Multi-cell grid, AI-regulated telomerase, mutation/genetic drift, and EPIGENETIC MEMORY propagation
# Colab-ready: requires numpy, matplotlib, scikit-learn

import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from matplotlib import animation

# -----------------------------
# Virtual Cell with Genome & Epigenetics
# -----------------------------
class VirtualCell:
    def __init__(self, i, j, telomere_length=8000.0):
        self.i = i
        self.j = j
        self.telomere_length = float(telomere_length)
        self.damage = 0.0
        self.debris = 0.0
        self.divisions = 0
        self.alive = True
        self.telomerase_activity = 0.0

        # Genetic traits (0..1)
        self.genes = {
            'TERT_reg': np.clip(np.random.normal(0.7, 0.12), 0.0, 1.0),
            'DNA_repair': np.clip(np.random.normal(0.7, 0.12), 0.0, 1.0),
            'Phago_eff': np.clip(np.random.normal(0.7, 0.12), 0.0, 1.0)
        }

        # Epigenetic state: 0 (silenced) .. 1 (fully active). Represents chromatin marks
        self.epigenetic_state = np.clip(np.random.beta(2,5), 0.0, 1.0)
        # Epigenetic plasticity: how easily the epigenetic state changes in response to stimuli
        self.plasticity = np.clip(np.random.normal(0.3, 0.1), 0.05, 1.0)
        # Genetic integrity measure (1 = pristine)
        self.genetic_integrity = 1.0

    def copy_for_division(self):
        """Return a new VirtualCell representing a daughter cell (inheritance + epigenetic transmission)."""
        daughter = VirtualCell(self.i, self.j, telomere_length=self.telomere_length)
        daughter.damage = max(0.0, self.damage + np.random.normal(0, 0.01))
        daughter.debris = self.debris * 0.5
        daughter.divisions = self.divisions
        daughter.genetic_integrity = max(0.0, self.genetic_integrity - np.random.normal(0, 0.001))
        # Genes inherit with small chance of mutation
        daughter.genes = {k: np.clip(v + np.random.normal(0, 0.01), 0.0, 1.0) for k, v in self.genes.items()}
        # Epigenetic inheritance: partial transmission of parent's epigenetic marks
        inherit_fraction = np.clip(0.6 * (1 - daughter.plasticity) , 0.0, 1.0)
        daughter.epigenetic_state = np.clip(self.epigenetic_state * inherit_fraction + np.random.normal(0, 0.05), 0.0, 1.0)
        daughter.plasticity = self.plasticity
        return daughter

    def attempt_division(self):
        """Attempt cell division. Returns True if division occurred."""
        if not self.alive:
            return False
        # Conditions: adequate telomere, reasonable damage, genetic integrity
        if (self.telomere_length > 5000) and (self.damage < 0.45) and (self.genetic_integrity > 0.6):
            # Division cost
            self.divisions += 1
            self.telomere_length -= np.random.uniform(0, 40)
            self.damage = min(1.0, self.damage + np.random.uniform(0.0, 0.02))
            self.debris += np.random.uniform(0.0, 0.05)
            # Successful division reinforces epigenetic state a bit (stabilization)
            self.epigenetic_state = np.clip(self.epigenetic_state + 0.01 * (1 - self.epigenetic_state), 0.0, 1.0)
            return True
        else:
            # If division not allowed, epigenetic state relaxes slightly
            self.epigenetic_state *= 0.995
            return False

    def apply_telomerase(self):
        # Telomerase extension depends on both AI signal and intrinsic TERT_reg
        response_factor = np.clip(self.genes['TERT_reg'], 0.0, 1.2)
        activity = float(np.clip(self.telomerase_activity, 0.0, 1.0))
        extension = 50.0 * activity * response_factor
        shortening = 20.0 * (1.0 - activity) * (1.0 + self.damage)
        self.telomere_length += extension - shortening

    def accumulate_environmental_stress(self, base_rate=0.02):
        # DNA repair gene reduces effective damage accumulation
        repair_factor = 1.0 - self.genes['DNA_repair']
        # Ensure repair_factor is not negative
        repair_factor = np.clip(repair_factor, 0.0, None)
        self.damage = min(1.0, self.damage + np.random.exponential(scale=base_rate * repair_factor))
        self.debris += np.random.exponential(scale=0.02)
        # Genetic drift: small integrity decrease
        self.genetic_integrity = max(0.0, self.genetic_integrity - abs(np.random.normal(0, 0.0005)))
        # Damage-driven epigenetic adaptation (stress marks)
        if self.damage > 0.5:
            self.epigenetic_state = np.clip(self.epigenetic_state + 0.02 * self.plasticity, 0.0, 1.0)

    def phagocytose(self, base_efficiency):
        genetic_boost = np.clip(self.genes['Phago_eff'], 0.0, 1.0)
        efficiency = np.clip(base_efficiency * (0.5 + 0.5 * genetic_boost), 0.0, 1.0)
        cleared = self.debris * efficiency
        self.debris -= cleared
        self.damage = max(0.0, self.damage - cleared * 0.05)

    def mutate_drift(self, mutation_rate=0.005, drift_rate=0.002):
        # Continuous drift
        for g in self.genes:
            self.genes[g] = np.clip(self.genes[g] + np.random.normal(0, drift_rate), 0.0, 1.0)
        # Discrete mutation events
        if np.random.rand() < mutation_rate:
            gene = np.random.choice(list(self.genes.keys()))
            change = np.random.uniform(-0.25, 0.25)
            self.genes[gene] = np.clip(self.genes[gene] + change, 0.0, 1.2)
            # small chance of catastrophic TERT dysregulation
            if gene == 'TERT_reg' and np.random.rand() < 0.01:
                self.genes['TERT_reg'] = 1.5  # hyper-activation (cancer-like)
                # epigenetic shift towards active state
                self.epigenetic_state = np.clip(self.epigenetic_state + 0.2, 0.0, 1.0)

    def check_viability(self):
        if (self.telomere_length < 3000) or (self.damage >= 0.98) or (self.genetic_integrity < 0.2):
            self.alive = False


# -----------------------------
# Shared AI controller (tissue-level)
# -----------------------------
class SharedAIController:
    def __init__(self):
        self.model = self._train_initial_model()

    def _train_initial_model(self):
        # Synthetic training set: inputs - tel_norm, damage, debris_norm, div_norm, epigenetic, neighbor_signal
        n = 3000
        tel = np.random.rand(n)
        dmg = np.random.rand(n)
        debris = np.random.rand(n)
        div = np.random.rand(n)
        epi = np.random.rand(n)
        neigh = np.random.rand(n)
        X = np.vstack([tel, dmg, debris, div, epi, neigh]).T
        # synthetic target: encourage telomerase when damage low, epi supportive, neighbors signal;
        y = 0.5*(1 - dmg) + 0.25*neigh + 0.15*(1 - np.abs(tel - 0.5)) + 0.1*epi
        y = np.clip(y, 0.0, 1.0)
        model = MLPRegressor(hidden_layer_sizes=(128,64), activation='relu', max_iter=2000, random_state=1)
        model.fit(X, y)
        return model

    def predict_activity_grid(self, grid, neighbor_signal):
        rows, cols = neighbor_signal.shape
        activities = np.zeros((rows, cols))
        for i in range(rows):
            for j in range(cols):
                cell = grid[i][j]
                if (cell is None) or (not cell.alive):
                    activities[i, j] = 0.0
                    continue
                tel_norm = np.clip(cell.telomere_length / 12000.0, 0.0, 1.0)
                debris_norm = np.clip(cell.debris / 5.0, 0.0, 1.0)
                div_norm = np.tanh(cell.divisions / 10.0)
                epi = np.clip(cell.epigenetic_state, 0.0, 1.0)
                nbr = np.clip(neighbor_signal[i, j], 0.0, 1.0)
                X = np.array([[tel_norm, cell.damage, debris_norm, div_norm, epi, nbr]])
                raw = float(np.clip(self.model.predict(X)[0], 0.0, 1.0))
                # Modulate by cell's TERT_reg (genetic) and epigenetic_state
                mod = raw * np.clip(cell.genes['TERT_reg'] * (0.7 + 0.6*epi), 0.0, 1.8)
                activities[i, j] = np.clip(mod, 0.0, 1.0)
        return activities


# -----------------------------
# Organ-on-Chip Environment
# -----------------------------
class OrganOnChip:
    def __init__(self, rows=14, cols=14, stress_rate=0.02, flow_rate=0.2):
        self.rows = rows
        self.cols = cols
        self.grid = [[VirtualCell(i, j, telomere_length=np.random.normal(8000, 400))
                      for j in range(cols)] for i in range(rows)]
        self.ai = SharedAIController()
        self.signal = np.zeros((rows, cols))  # pro-regenerative/cytokine-like signal
        self.epi_signal = np.zeros((rows, cols))  # epigenetic-diffusible signal (represents signaling that biases epi-state)
        self.waste = np.zeros((rows, cols))
        self.flow_rate = flow_rate
        self.stress_rate = stress_rate
        self.time = 0

    def neighbors(self, i, j):
        dirs = [(-1,0),(1,0),(0,-1),(0,1)]
        return [(i+di, j+dj) for di, dj in dirs if 0 <= i+di < self.rows and 0 <= j+dj < self.cols]

    def diffuse(self, field, diff_coeff=0.2):
        new = np.copy(field)
        for i in range(self.rows):
            for j in range(self.cols):
                nbrs = self.neighbors(i, j)
                if not nbrs:
                    continue
                exchange = sum(field[ni, nj] for (ni, nj) in nbrs) - len(nbrs) * field[i, j]
                new[i, j] += diff_coeff * exchange / max(1, len(nbrs))
        return np.clip(new, 0.0, None)

    def step(self):
        # 1) Cells produce signals based on damage, epigenetic_state and divisions
        for i in range(self.rows):
            for j in range(self.cols):
                cell = self.grid[i][j]
                if (cell is None) or (not cell.alive):
                    continue
                # damaged cells emit pro-regenerative signals (calls for help)
                self.signal[i, j] += 0.05 * cell.damage
                # cells with active epigenetic marks release epi-biasing factors
                self.epi_signal[i, j] += 0.03 * cell.epigenetic_state
                # debris & divisions add to environmental waste
                self.waste[i, j] += 0.02 * cell.divisions + 0.08 * cell.debris

        # 2) Diffuse both signals and epi_signal
        self.signal = self.diffuse(self.signal, diff_coeff=0.25)
        self.epi_signal = self.diffuse(self.epi_signal, diff_coeff=0.2)
        self.waste = self.diffuse(self.waste, diff_coeff=0.15)

        # 3) Microfluidic washout reduces environmental waste
        self.waste *= np.exp(-self.flow_rate)

        # 4) Neighbor signal normalization for AI input
        neighbor_signal = np.zeros((self.rows, self.cols))
        for i in range(self.rows):
            for j in range(self.cols):
                nbrs = self.neighbors(i, j)
                val = self.signal[i, j] + sum(self.signal[ni, nj] for (ni, nj) in nbrs)
                denom = 1 + len(nbrs)
                neighbor_signal[i, j] = np.clip(val / denom, 0.0, 1.0)

        # 5) AI predicts telomerase activities
        activities = self.ai.predict_activity_grid(self.grid, neighbor_signal)

        # 6) Update cells
        for i in range(self.rows):
            for j in range(self.cols):
                cell = self.grid[i][j]
                if (cell is None) or (not cell.alive):
                    continue
                cell.telomerase_activity = activities[i, j]
                cell.accumulate_environmental_stress(self.stress_rate)
                cell.apply_telomerase()
                divided = cell.attempt_division()
                # phagocytosis efficiency influenced by neighbor signal & waste
                base_eff = np.clip(0.45 + 0.4 * neighbor_signal[i, j] - 0.25 * self.waste[i, j], 0.0, 1.0)
                cell.phagocytose(base_eff)
                # mutation & drift
                cell.mutate_drift(mutation_rate=0.006, drift_rate=0.003)
                # Epigenetic influence from local epi_signal (neighbor-driven epigenetic propagation)
                # Cells adjust epigenetic_state towards local epi_signal scaled by their plasticity
                local_epi = np.clip(self.epi_signal[i, j], 0.0, 1.0)
                # update towards local signal
                cell.epigenetic_state = np.clip(cell.epigenetic_state + cell.plasticity * (local_epi - cell.epigenetic_state) * 0.1, 0.0, 1.0)
                # small spontaneous epigenetic decay
                cell.epigenetic_state *= 0.997
                cell.check_viability()

        # 7) Epigenetic signal sink (decay)
        self.signal *= 0.94
        self.epi_signal *= 0.92
        self.time += 1

    def grid_state_matrices(self):
        tel = np.zeros((self.rows, self.cols))
        dmg = np.ones((self.rows, self.cols)) * np.nan
        act = np.zeros((self.rows, self.cols))
        tert = np.zeros((self.rows, self.cols))
        epi = np.zeros((self.rows, self.cols))
        alive = np.zeros((self.rows, self.cols))
        for i in range(self.rows):
            for j in range(self.cols):
                c = self.grid[i][j]
                if c.alive:
                    tel[i, j] = c.telomere_length
                    dmg[i, j] = c.damage
                    act[i, j] = c.telomerase_activity
                    tert[i, j] = c.genes['TERT_reg']
                    epi[i, j] = c.epigenetic_state
                    alive[i, j] = 1
                else:
                    tel[i, j] = 0.0
                    dmg[i, j] = 1.0
                    act[i, j] = 0.0
                    tert[i, j] = 0.0
                    epi[i, j] = 0.0
                    alive[i, j] = 0
        return tel, dmg, act, tert, epi, alive


# -----------------------------
# Visualization & Runner
# -----------------------------
def run_experiment(steps=300, rows=14, cols=14, animate=True):
    env = OrganOnChip(rows=rows, cols=cols)
    history = {'avg_tel': [], 'avg_damage': [], 'avg_activity': [], 'alive_fraction': [], 'avg_epi': []}
    frames = []

    for t in range(steps):
        env.step()
        tel, dmg, act, tert, epi, alive = env.grid_state_matrices()
        alive_mask = alive.astype(bool)
        history['avg_tel'].append(np.nanmean(tel[alive_mask]) if np.any(alive_mask) else 0.0)
        history['avg_damage'].append(np.nanmean(dmg))
        history['avg_activity'].append(np.nanmean(act))
        history['alive_fraction'].append(np.mean(alive))
        history['avg_epi'].append(np.nanmean(epi[alive_mask]) if np.any(alive_mask) else 0.0)
        if animate:
            frames.append((tel.copy(), dmg.copy(), act.copy(), epi.copy(), env.waste.copy(), env.signal.copy()))

    # Time-series plots
    fig, axs = plt.subplots(2, 3, figsize=(14,8))
    axs = axs.ravel()
    axs[0].plot(history['avg_tel']); axs[0].set_title('Avg Telomere Length')
    axs[1].plot(history['avg_damage']); axs[1].set_title('Avg Damage')
    axs[2].plot(history['avg_activity']); axs[2].set_title('Avg Telomerase Activity')
    axs[3].plot(history['alive_fraction']); axs[3].set_title('Alive Fraction')
    axs[4].plot(history['avg_epi']); axs[4].set_title('Avg Epigenetic State')
    axs[5].axis('off')
    for a in axs:
        a.grid(True)
    plt.tight_layout(); plt.show()

    # Animation
    if animate:
        fig2, axarr = plt.subplots(1, 6, figsize=(18, 3.5))

        def update_frame(k):
            tel, dmg, act, epi, waste, sig = frames[k]
            axarr[0].clear(); axarr[1].clear(); axarr[2].clear(); axarr[3].clear(); axarr[4].clear(); axarr[5].clear()
            im0 = axarr[0].imshow(np.clip(tel, 0, 12000), vmin=3000, vmax=10000, aspect='auto')
            axarr[0].set_title('Telomere')
            im1 = axarr[1].imshow(dmg, vmin=0, vmax=1, aspect='auto')
            axarr[1].set_title('Damage')
            im2 = axarr[2].imshow(act, vmin=0, vmax=1, aspect='auto')
            axarr[2].set_title('Telomerase')
            im3 = axarr[3].imshow(epi, vmin=0, vmax=1, aspect='auto')
            axarr[3].set_title('Epigenetic State')
            im4 = axarr[4].imshow(waste, aspect='auto')
            axarr[4].set_title('Waste')
            im5 = axarr[5].imshow(sig, aspect='auto')
            axarr[5].set_title('Signal')
            fig2.suptitle(f'Time {k}')
            return im0, im1, im2, im3, im4, im5

        anim = animation.FuncAnimation(fig2, update_frame, frames=len(frames), interval=150, blit=False)
        plt.show()

    return env, history


# -----------------------------
# If run as a script
# -----------------------------
if __name__ == '__main__':
    env, history = run_experiment(steps=300, rows=16, cols=16, animate=True)
    print('Done')
```

```python
# Organ-on-Chip Simulation
# Multi-cell grid, AI-regulated telomerase, mutation/genetic drift, and EPIGENETIC MEMORY propagation
# Colab-ready: requires numpy, matplotlib, scikit-learn

import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from matplotlib import animation

# -----------------------------
# Virtual Cell with Genome & Epigenetics
# -----------------------------
class VirtualCell:
    def __init__(self, i, j, telomere_length=8000.0):
        self.i = i
        self.j = j
        self.telomere_length = float(telomere_length)
        self.damage = 0.0
        self.debris = 0.0
        self.divisions = 0
        self.alive = True
        self.telomerase_activity = 0.0

        # Genetic traits (0..1)
        self.genes = {
            'TERT_reg': np.clip(np.random.normal(0.7, 0.12), 0.0, 1.0),
            'DNA_repair': np.clip(np.random.normal(0.7, 0.12), 0.0, 1.0),
            'Phago_eff': np.clip(np.random.normal(0.7, 0.12), 0.0, 1.0)
        }

        # Epigenetic state: 0 (silenced) .. 1 (fully active). Represents chromatin marks
        self.epigenetic_state = np.clip(np.random.beta(2,5), 0.0, 1.0)
        # Epigenetic plasticity: how easily the epigenetic state changes in response to stimuli
        self.plasticity = np.clip(np.random.normal(0.3, 0.1), 0.05, 1.0)
        # Genetic integrity measure (1 = pristine)
        self.genetic_integrity = 1.0

    def copy_for_division(self):
        """Return a new VirtualCell representing a daughter cell (inheritance + epigenetic transmission)."""
        daughter = VirtualCell(self.i, self.j, telomere_length=self.telomere_length)
        daughter.damage = max(0.0, self.damage + np.random.normal(0, 0.01))
        daughter.debris = self.debris * 0.5
        daughter.divisions = self.divisions
        daughter.genetic_integrity = max(0.0, self.genetic_integrity - np.random.normal(0, 0.001))
        # Genes inherit with small chance of mutation
        daughter.genes = {k: np.clip(v + np.random.normal(0, 0.01), 0.0, 1.0) for k, v in self.genes.items()}
        # Epigenetic inheritance: partial transmission of parent's epigenetic marks
        inherit_fraction = np.clip(0.6 * (1 - daughter.plasticity) , 0.0, 1.0)
        daughter.epigenetic_state = np.clip(self.epigenetic_state * inherit_fraction + np.random.normal(0, 0.05), 0.0, 1.0)
        daughter.plasticity = self.plasticity
        return daughter

    def attempt_division(self):
        """Attempt cell division. Returns True if division occurred."""
        if not self.alive:
            return False
        # Conditions: adequate telomere, reasonable damage, genetic integrity
        if (self.telomere_length > 5000) and (self.damage < 0.45) and (self.genetic_integrity > 0.6):
            # Division cost
            self.divisions += 1
            self.telomere_length -= np.random.uniform(0, 40)
            self.damage = min(1.0, self.damage + np.random.uniform(0.0, 0.02))
            self.debris += np.random.uniform(0.0, 0.05)
            # Successful division reinforces epigenetic state a bit (stabilization)
            self.epigenetic_state = np.clip(self.epigenetic_state + 0.01 * (1 - self.epigenetic_state), 0.0, 1.0)
            return True
        else:
            # If division not allowed, epigenetic state relaxes slightly
            self.epigenetic_state *= 0.995
            return False

    def apply_telomerase(self):
        # Telomerase extension depends on both AI signal and intrinsic TERT_reg
        response_factor = np.clip(self.genes['TERT_reg'], 0.0, 1.2)
        activity = float(np.clip(self.telomerase_activity, 0.0, 1.0))
        extension = 50.0 * activity * response_factor
        shortening = 20.0 * (1.0 - activity) * (1.0 + self.damage)
        self.telomere_length += extension - shortening

    def accumulate_environmental_stress(self, base_rate=0.02):
        # DNA repair gene reduces effective damage accumulation
        repair_factor = 1.0 - self.genes['DNA_repair']
        # Ensure repair_factor is not negative
        repair_factor = np.clip(repair_factor, 0.0, None)
        self.damage = min(1.0, self.damage + np.random.exponential(scale=base_rate * repair_factor))
        self.debris += np.random.exponential(scale=0.02)
        # Genetic drift: small integrity decrease
        self.genetic_integrity = max(0.0, self.genetic_integrity - abs(np.random.normal(0, 0.0005)))
        # Damage-driven epigenetic adaptation (stress marks)
        if self.damage > 0.5:
            self.epigenetic_state = np.clip(self.epigenetic_state + 0.02 * self.plasticity, 0.0, 1.0)

    def phagocytose(self, base_efficiency):
        genetic_boost = np.clip(self.genes['Phago_eff'], 0.0, 1.0)
        efficiency = np.clip(base_efficiency * (0.5 + 0.5 * genetic_boost), 0.0, 1.0)
        cleared = self.debris * efficiency
        self.debris -= cleared
        self.damage = max(0.0, self.damage - cleared * 0.05)

    def mutate_drift(self, mutation_rate=0.005, drift_rate=0.002):
        # Continuous drift
        for g in self.genes:
            self.genes[g] = np.clip(self.genes[g] + np.random.normal(0, drift_rate), 0.0, 1.0)
        # Discrete mutation events
        if np.random.rand() < mutation_rate:
            gene = np.random.choice(list(self.genes.keys()))
            change = np.random.uniform(-0.25, 0.25)
            self.genes[gene] = np.clip(self.genes[gene] + change, 0.0, 1.2)
            # small chance of catastrophic TERT dysregulation
            if gene == 'TERT_reg' and np.random.rand() < 0.01:
                self.genes['TERT_reg'] = 1.5  # hyper-activation (cancer-like)
                # epigenetic shift towards active state
                self.epigenetic_state = np.clip(self.epigenetic_state + 0.2, 0.0, 1.0)

    def check_viability(self):
        if (self.telomere_length < 3000) or (self.damage >= 0.98) or (self.genetic_integrity < 0.2):
            self.alive = False


# -----------------------------
# Shared AI controller (tissue-level)
# -----------------------------
class SharedAIController:
    def __init__(self):
        self.model = self._train_initial_model()

    def _train_initial_model(self):
        # Synthetic training set: inputs - tel_norm, damage, debris_norm, div_norm, epigenetic, neighbor_signal
        n = 3000
        tel = np.random.rand(n)
        dmg = np.random.rand(n)
        debris = np.random.rand(n)
        div = np.random.rand(n)
        epi = np.random.rand(n)
        neigh = np.random.rand(n)
        X = np.vstack([tel, dmg, debris, div, epi, neigh]).T
        # synthetic target: encourage telomerase when damage low, epi supportive, neighbors signal;
        y = 0.5*(1 - dmg) + 0.25*neigh + 0.15*(1 - np.abs(tel - 0.5)) + 0.1*epi
        y = np.clip(y, 0.0, 1.0)
        model = MLPRegressor(hidden_layer_sizes=(128,64), activation='relu', max_iter=2000, random_state=1)
        model.fit(X, y)
        return model

    def predict_activity_grid(self, grid, neighbor_signal):
        rows, cols = neighbor_signal.shape
        activities = np.zeros((rows, cols))
        for i in range(rows):
            for j in range(cols):
                cell = grid[i][j]
                if (cell is None) or (not cell.alive):
                    activities[i, j] = 0.0
                    continue
                tel_norm = np.clip(cell.telomere_length / 12000.0, 0.0, 1.0)
                debris_norm = np.clip(cell.debris / 5.0, 0.0, 1.0)
                div_norm = np.tanh(cell.divisions / 10.0)
                epi = np.clip(cell.epigenetic_state, 0.0, 1.0)
                nbr = np.clip(neighbor_signal[i, j], 0.0, 1.0)
                X = np.array([[tel_norm, cell.damage, debris_norm, div_norm, epi, nbr]])
                raw = float(np.clip(self.model.predict(X)[0], 0.0, 1.0))
                # Modulate by cell's TERT_reg (genetic) and epigenetic_state
                mod = raw * np.clip(cell.genes['TERT_reg'] * (0.7 + 0.6*epi), 0.0, 1.8)
                activities[i, j] = np.clip(mod, 0.0, 1.0)
        return activities


# -----------------------------
# Organ-on-Chip Environment
# -----------------------------
class OrganOnChip:
    def __init__(self, rows=14, cols=14, stress_rate=0.02, flow_rate=0.2):
        self.rows = rows
        self.cols = cols
        self.grid = [[VirtualCell(i, j, telomere_length=np.random.normal(8000, 400))
                      for j in range(cols)] for i in range(rows)]
        self.ai = SharedAIController()
        self.signal = np.zeros((rows, cols))  # pro-regenerative/cytokine-like signal
        self.epi_signal = np.zeros((rows, cols))  # epigenetic-diffusible signal (represents signaling that biases epi-state)
        self.waste = np.zeros((rows, cols))
        self.flow_rate = flow_rate
        self.stress_rate = stress_rate
        self.time = 0

    def neighbors(self, i, j):
        dirs = [(-1,0),(1,0),(0,-1),(0,1)]
        return [(i+di, j+dj) for di, dj in dirs if 0 <= i+di < self.rows and 0 <= j+dj < self.cols]

    def diffuse(self, field, diff_coeff=0.2):
        new = np.copy(field)
        for i in range(self.rows):
            for j in range(self.cols):
                nbrs = self.neighbors(i, j)
                if not nbrs:
                    continue
                exchange = sum(field[ni, nj] for (ni, nj) in nbrs) - len(nbrs) * field[i, j]
                new[i, j] += diff_coeff * exchange / max(1, len(nbrs))
        return np.clip(new, 0.0, None)

    def step(self):
        # 1) Cells produce signals based on damage, epigenetic_state and divisions
        for i in range(self.rows):
            for j in range(self.cols):
                cell = self.grid[i][j]
                if (cell is None) or (not cell.alive):
                    continue
                # damaged cells emit pro-regenerative signals (calls for help)
                self.signal[i, j] += 0.05 * cell.damage
                # cells with active epigenetic marks release epi-biasing factors
                self.epi_signal[i, j] += 0.03 * cell.epigenetic_state
                # debris & divisions add to environmental waste
                self.waste[i, j] += 0.02 * cell.divisions + 0.08 * cell.debris

        # 2) Diffuse both signals and epi_signal
        self.signal = self.diffuse(self.signal, diff_coeff=0.25)
        self.epi_signal = self.diffuse(self.epi_signal, diff_coeff=0.2)
        self.waste = self.diffuse(self.waste, diff_coeff=0.15)

        # 3) Microfluidic washout reduces environmental waste
        self.waste *= np.exp(-self.flow_rate)

        # 4) Neighbor signal normalization for AI input
        neighbor_signal = np.zeros((self.rows, self.cols))
        for i in range(self.rows):
            for j in range(self.cols):
                nbrs = self.neighbors(i, j)
                val = self.signal[i, j] + sum(self.signal[ni, nj] for (ni, nj) in nbrs)
                denom = 1 + len(nbrs)
                neighbor_signal[i, j] = np.clip(val / denom, 0.0, 1.0)

        # 5) AI predicts telomerase activities
        activities = self.ai.predict_activity_grid(self.grid, neighbor_signal)

        # 6) Update cells
        for i in range(self.rows):
            for j in range(self.cols):
                cell = self.grid[i][j]
                if (cell is None) or (not cell.alive):
                    continue
                cell.telomerase_activity = activities[i, j]
                cell.accumulate_environmental_stress(self.stress_rate)
                cell.apply_telomerase()
                divided = cell.attempt_division()
                # phagocytosis efficiency influenced by neighbor signal & waste
                base_eff = np.clip(0.45 + 0.4 * neighbor_signal[i, j] - 0.25 * self.waste[i, j], 0.0, 1.0)
                cell.phagocytose(base_eff)
                # mutation & drift
                cell.mutate_drift(mutation_rate=0.006, drift_rate=0.003)
                # Epigenetic influence from local epi_signal (neighbor-driven epigenetic propagation)
                # Cells adjust epigenetic_state towards local epi_signal scaled by their plasticity
                local_epi = np.clip(self.epi_signal[i, j], 0.0, 1.0)
                # update towards local signal
                cell.epigenetic_state = np.clip(cell.epigenetic_state + cell.plasticity * (local_epi - cell.epigenetic_state) * 0.1, 0.0, 1.0)
                # small spontaneous epigenetic decay
                cell.epigenetic_state *= 0.997
                cell.check_viability()

        # 7) Epigenetic signal sink (decay)
        self.signal *= 0.94
        self.epi_signal *= 0.92
        self.time += 1

    def grid_state_matrices(self):
        tel = np.zeros((self.rows, self.cols))
        dmg = np.ones((self.rows, self.cols)) * np.nan
        act = np.zeros((self.rows, self.cols))
        tert = np.zeros((self.rows, self.cols))
        epi = np.zeros((self.rows, self.cols))
        alive = np.zeros((self.rows, self.cols))
        for i in range(self.rows):
            for j in range(self.cols):
                c = self.grid[i][j]
                if c.alive:
                    tel[i, j] = c.telomere_length
                    dmg[i, j] = c.damage
                    act[i, j] = c.telomerase_activity
                    tert[i, j] = c.genes['TERT_reg']
                    epi[i, j] = c.epigenetic_state
                    alive[i, j] = 1
                else:
                    tel[i, j] = 0.0
                    dmg[i, j] = 1.0
                    act[i, j] = 0.0
                    tert[i, j] = 0.0
                    epi[i, j] = 0.0
                    alive[i, j] = 0
        return tel, dmg, act, tert, epi, alive


# -----------------------------
# Visualization & Runner
# -----------------------------
def run_experiment(steps=300, rows=14, cols=14, animate=True):
    env = OrganOnChip(rows=rows, cols=cols)
    history = {'avg_tel': [], 'avg_damage': [], 'avg_activity': [], 'alive_fraction': [], 'avg_epi': []}
    frames = []

    for t in range(steps):
        env.step()
        tel, dmg, act, tert, epi, alive = env.grid_state_matrices()
        alive_mask = alive.astype(bool)
        history['avg_tel'].append(np.nanmean(tel[alive_mask]) if np.any(alive_mask) else 0.0)
        history['avg_damage'].append(np.nanmean(dmg))
        history['avg_activity'].append(np.nanmean(act))
        history['alive_fraction'].append(np.mean(alive))
        history['avg_epi'].append(np.nanmean(epi[alive_mask]) if np.any(alive_mask) else 0.0)
        if animate:
            frames.append((tel.copy(), dmg.copy(), act.copy(), epi.copy(), env.waste.copy(), env.signal.copy()))

    # Time-series plots
    fig, axs = plt.subplots(2, 3, figsize=(14,8))
    axs = axs.ravel()
    axs[0].plot(history['avg_tel']); axs[0].set_title('Avg Telomere Length')
    axs[1].plot(history['avg_damage']); axs[1].set_title('Avg Damage')
    axs[2].plot(history['avg_activity']); axs[2].set_title('Avg Telomerase Activity')
    axs[3].plot(history['alive_fraction']); axs[3].set_title('Alive Fraction')
    axs[4].plot(history['avg_epi']); axs[4].set_title('Avg Epigenetic State')
    axs[5].axis('off')
    for a in axs:
        a.grid(True)
    plt.tight_layout(); plt.show()

    # Animation
    if animate:
        fig2, axarr = plt.subplots(1, 6, figsize=(18, 3.5))

        def update_frame(k):
            tel, dmg, act, epi, waste, sig = frames[k]
            for ax in axarr:
                ax.clear()
            im0 = axarr[0].imshow(np.clip(tel, 0, 12000), vmin=3000, vmax=10000, aspect='auto')
            axarr[0].set_title('Telomere')
            im1 = axarr[1].imshow(dmg, vmin=0, vmax=1, aspect='auto')
            axarr[1].set_title('Damage')
            im2 = axarr[2].imshow(act, vmin=0, vmax=1, aspect='auto')
            axarr[2].set_title('Telomerase')
            im3 = axarr[3].imshow(epi, vmin=0, vmax=1, aspect='auto')
            axarr[3].set_title('Epigenetic State')
            im4 = axarr[4].imshow(waste, aspect='auto')
            axarr[4].set_title('Waste')
            im5 = axarr[5].imshow(sig, aspect='auto')
            axarr[5].set_title('Signal')
            # overlay immune positions
            if immune_pos.size:
                ys = immune_pos[:,0]
                xs = immune_pos[:,1]
                for ax in axarr:
                    ax.scatter(xs, ys, marker='x', s=40)
            fig2.suptitle(f'Time {k}')
            return im0, im1, im2, im3, im4, im5

        anim = animation.FuncAnimation(fig2, update_frame, frames=len(frames), interval=150, blit=False)
        plt.show()

    return env, history


# -----------------------------
# If run as a script
# -----------------------------
if __name__ == '__main__':
    env, history = run_experiment(steps=300, rows=16, cols=16, animate=True)
    print('Done')

# -----------------------------
# Immune Surveillance Module
# -----------------------------
class ImmuneCell:
    """Simple immune patrolling agent that moves, senses, and removes malignant or highly damaged cells."""
    def __init__(self, i, j, sensing_radius=1, kill_prob=0.9, patrol_speed=1):
        self.i = i
        self.j = j
        self.sensing_radius = sensing_radius
        self.kill_prob = kill_prob
        self.patrol_speed = patrol_speed
        self.alive = True

    def move_towards(self, target_i, target_j, rows, cols):
        # move one step towards target (bounded by grid)
        di = np.sign(target_i - self.i)
        dj = np.sign(target_j - self.j)
        self.i = int(np.clip(self.i + di, 0, rows-1))
        self.j = int(np.clip(self.j + dj, 0, cols-1))

    def random_patrol(self, rows, cols):
        # random walk limited by patrol_speed
        self.i = int(np.clip(self.i + np.random.randint(-1, 2), 0, rows-1))
        self.j = int(np.clip(self.j + np.random.randint(-1, 2), 0, cols-1))

    def attempt_kill(self, cell):
        # Decide whether to remove a threatening cell
        if not cell.alive:
            return False
        # Threat criteria: hyper-TERT, low genetic integrity, or extreme damage
        threat_score = 0.0
        if cell.genes.get('TERT_reg', 0) > 1.2:
            threat_score += 0.7
        if cell.genetic_integrity < 0.4:
            threat_score += 0.6
        if cell.damage > 0.9:
            threat_score += 0.4
        # Epigenetic hyperactivity may also be suspicious
        if cell.epigenetic_state > 0.95:
            threat_score += 0.2
        # Probability to act scales with threat score
        if threat_score <= 0:
            return False
        p = 1 - np.exp(-2.5 * threat_score)
        # combined with immune intrinsic kill probability
        if np.random.rand() < min(1.0, p * self.kill_prob):
            # kill: mark cell dead and leave debris
            cell.alive = False
            cell.debris += np.random.uniform(0.1, 0.5)
            return True
        return False

# Integrate immune agents into OrganOnChip
# We'll add an immune list and simple recruitment based on tissue distress signals

def _organ_on_chip_add_immune_support():
    # patch class to add immune support attributes and methods
    def init_with_immune(self, rows=14, cols=14, stress_rate=0.02, flow_rate=0.2):
        # call original __init__ by reusing current attributes already set in file; this function
        # will only be used if user re-instantiates OrganOnChip manually. For backward compatibility
        # we update the instance attributes below when called from run_experiment.
        pass

    # Add immune list attribute to existing OrganOnChip class via monkey patch
    setattr(OrganOnChip, 'immune_agents', None)

    def spawn_immune_agents(self, n=6):
        # initialize immune agents spread around the grid
        agents = []
        for k in range(n):
            i = np.random.randint(0, self.rows)
            j = np.random.randint(0, self.cols)
            agents.append(ImmuneCell(i, j))
        self.immune_agents = agents

    def immune_step(self):
        # Simple recruitment: areas with high epi_signal or high waste attract immune cells
        if self.immune_agents is None:
            self.spawn_immune_agents(n=max(4, (self.rows*self.cols)//50))
        # For each immune agent: sense local threats, move and act
        for agent in list(self.immune_agents):
            if not agent.alive:
                continue
            # find highest local distress cell within a radius
            best = None
            best_score = 0.0
            rmin = max(0, agent.i - 3)
            rmax = min(self.rows-1, agent.i + 3)
            cmin = max(0, agent.j - 3)
            cmax = min(self.cols-1, agent.j + 3)
            for ii in range(rmin, rmax+1):
                for jj in range(cmin, cmax+1):
                    cell = self.grid[ii][jj]
                    if not cell.alive:
                        continue
                    # threat heuristics similar to ImmuneCell.attempt_kill
                    score = 0.0
                    if cell.genes.get('TERT_reg', 0) > 1.2:
                        score += 3.0
                    score += max(0, 1.0 - cell.genetic_integrity) * 2.0
                    score += cell.damage * 1.5
                    # environmental waste and epi_signal increase local score
                    score += 1.0 * self.waste[ii, jj]
                    score += 0.6 * self.epi_signal[ii, jj]
                    if score > best_score:
                        best_score = score
                        best = (ii, jj)
            # Move towards best target if found, otherwise random patrol
            if best and best_score > 0.5:
                agent.move_towards(best[0], best[1], self.rows, self.cols)
                target_cell = self.grid[agent.i][agent.j]
                # attempt kill
                agent.attempt_kill(target_cell)
            else:
                agent.random_patrol(self.rows, self.cols)
            # small chance immune agent is exhausted/removed
            if np.random.rand() < 0.001:
                agent.alive = False
        # clean dead agents occasionally
        self.immune_agents = [a for a in self.immune_agents if a.alive]

    # Attach methods to OrganOnChip
    setattr(OrganOnChip, 'spawn_immune_agents', spawn_immune_agents)
    setattr(OrganOnChip, 'immune_step', immune_step)

# apply monkey patch
_organ_on_chip_add_immune_support()

# Modify OrganOnChip.step to call immune functions: we'll provide a small wrapper that the user
# can call instead of the original step() if they want immune surveillance active.

def step_with_immune(self):
    # Run immune actions first (they can pre-emptively remove dangerous cells)
    if getattr(self, 'immune_agents', None) is None:
        self.spawn_immune_agents(n=max(4, (self.rows*self.cols)//50))
    # recruitment increases when tissue distress high
    # spawn extra agents if many high-waste spots
    hotspots = np.sum(self.waste > 0.5)
    if hotspots > (self.rows * self.cols) * 0.02 and len(self.immune_agents) < (self.rows*self.cols)//10:
        self.spawn_immune_agents(n=2)
    # immune patrol & act
    self.immune_step()
    # then run the original physiological step
    # reuse existing step logic from class (we call it directly)
    OrganOnChip.step(self)
    # after step, immune cells can scavenge debris (reduce waste locally where they are present)
    for ag in getattr(self, 'immune_agents', [])[:]:
        ii, jj = ag.i, ag.j
        # immune scavenging reduces waste and increases local phagocytosis indirectly
        self.waste[ii, jj] = max(0.0, self.waste[ii, jj] - np.random.uniform(0.01, 0.1))

# Attach wrapper
setattr(OrganOnChip, 'step_with_immune', step_with_immune)

# Update run_experiment to optionally use immune surveillance (non-destructive change)
import inspect
_old_run_experiment = run_experiment

def run_experiment_with_immune(steps=300, rows=14, cols=14, animate=True, immune=True):
    env = OrganOnChip(rows=rows, cols=cols)
    # start immune agents if requested
    if immune:
        env.spawn_immune_agents(n=max(6, (rows*cols)//40))
    history = {'avg_tel': [], 'avg_damage': [], 'avg_activity': [], 'alive_fraction': [], 'avg_epi': [], 'immune_count': []}
    frames = []

    for t in range(steps):
        if immune:
            env.step_with_immune()
        else:
            env.step()
        tel, dmg, act, tert, epi, alive = env.grid_state_matrices()
        alive_mask = alive.astype(bool)
        history['avg_tel'].append(np.nanmean(tel[alive_mask]) if np.any(alive_mask) else 0.0)
        history['avg_damage'].append(np.nanmean(dmg))
        history['avg_activity'].append(np.nanmean(act))
        history['alive_fraction'].append(np.mean(alive))
        history['avg_epi'].append(np.nanmean(epi[alive_mask]) if np.any(alive_mask) else 0.0)
        history['immune_count'].append(len(env.immune_agents) if getattr(env, 'immune_agents', None) is not None else 0)
        if animate:
            frames.append((tel.copy(), dmg.copy(), act.copy(), epi.copy(), env.waste.copy(), env.signal.copy(), np.array([ (a.i, a.j) for a in (env.immune_agents or []) ])))

    # Time-series plots (include immune count)
    fig, axs = plt.subplots(2, 3, figsize=(14,8))
    axs = axs.ravel()
    axs[0].plot(history['avg_tel']); axs[0].set_title('Avg Telomere Length')
    axs[1].plot(history['avg_damage']); axs[1].set_title('Avg Damage')
    axs[2].plot(history['avg_activity']); axs[2].set_title('Avg Telomerase Activity')
    axs[3].plot(history['alive_fraction']); axs[3].set_title('Alive Fraction')
    axs[4].plot(history['avg_epi']); axs[4].set_title('Avg Epigenetic State')
    axs[5].plot(history['immune_count']); axs[5].set_title('Immune Agent Count')
    for a in axs:
        a.grid(True)
    plt.tight_layout(); plt.show()

    # Animation with immune overlay
    if animate:
        fig2, axarr = plt.subplots(1, 6, figsize=(18, 3.5))

        def update_frame(k):
            tel, dmg, act, epi, waste, sig, immune_pos = frames[k]
            for ax in axarr:
                ax.clear()
            im0 = axarr[0].imshow(np.clip(tel, 0, 12000), vmin=3000, vmax=10000, aspect='auto')
            axarr[0].set_title('Telomere')
            im1 = axarr[1].imshow(dmg, vmin=0, vmax=1, aspect='auto')
            axarr[1].set_title('Damage')
            im2 = axarr[2].imshow(act, vmin=0, vmax=1, aspect='auto')
            axarr[2].set_title('Telomerase')
            im3 = axarr[3].imshow(epi, vmin=0, vmax=1, aspect='auto')
            axarr[3].set_title('Epigenetic State')
            im4 = axarr[4].imshow(waste, aspect='auto')
            axarr[4].set_title('Waste')
            im5 = axarr[5].imshow(sig, aspect='auto')
            axarr[5].set_title('Signal')
            # overlay immune positions
            if immune_pos.size:
                ys = immune_pos[:,0]
                xs = immune_pos[:,1]
                for ax in axarr:
                    ax.scatter(xs, ys, marker='x', s=40)
            fig2.suptitle(f'Time {k}')
            return im0, im1, im2, im3, im4, im5

        anim = animation.FuncAnimation(fig2, update_frame, frames=len(frames), interval=150, blit=False)
        plt.show()

    return env, history

# expose the new runner
# setattr(__import__(__name__).__class__, 'run_experiment_with_immune', run_experiment_with_immune) # Removed setattr call

# End of immune surveillance extension

# -----------------------------
# Reinforcement-Learned Immune Controller (Centralized Policy)
# -----------------------------
# This module replaces the heuristic patrol with a centralized policy network trained
# online using a simple policy-gradient (REINFORCE) algorithm implemented in NumPy.
# The policy controls all immune agents (shared parameters) and issues actions per agent.
# Actions per agent: 0=stay,1=up,2=down,3=left,4=right,5=attack,6=recruit

class RLImmuneController:
    def __init__(self, env, hidden_sizes=(64,32), lr=1e-3, gamma=0.99):
        self.env = env
        # policy parameters (MLP): input -> hidden -> hidden -> output
        # Input features: [agent_i, agent_j, local_damage, local_tert, local_epi, local_waste, alive_frac]
        self.input_dim = 7
        self.hidden_sizes = hidden_sizes
        self.output_dim = 7  # actions
        # initialize weights
        self.params = self._init_params()
        self.lr = lr
        self.gamma = gamma
        # episode memory for policy gradient
        self.episode = []
        self.baseline = 0.0
        self.baseline_alpha = 0.01

    def _init_params(self):
        rng = np.random.RandomState(1)
        params = {}
        dims = [self.input_dim] + list(self.hidden_sizes) + [self.output_dim]
        for i in range(len(dims)-1):
            params[f'W{i}'] = rng.normal(0, 0.1, size=(dims[i], dims[i+1]))
            params[f'b{i}'] = np.zeros((dims[i+1],))
        return params

    def _forward(self, x):
        h = x.astype(float)
        L = len(self.hidden_sizes)
        for i in range(L):
            h = np.tanh(h.dot(self.params[f'W{i}']) + self.params[f'b{i}'])
        logits = h.dot(self.params[f'W{L}']) + self.params[f'b{L}']
        # softmax
        exp = np.exp(logits - np.max(logits))
        probs = exp / np.sum(exp)
        return probs

    def sample_action(self, features):
        probs = self._forward(features)
        action = np.random.choice(len(probs), p=probs)
        return action, probs

    def store_transition(self, features, action, prob, reward):
        self.episode.append((features, action, prob, reward))

    def finish_episode(self):
        # compute discounted returns
        returns = []
        R = 0.0
        for _, _, _, r in reversed(self.episode):
            R = r + self.gamma * R
            returns.insert(0, R)
        returns = np.array(returns)
        # normalize
        returns = (returns - returns.mean()) / (returns.std() + 1e-8)
        # policy gradient updates (simple REINFORCE)
        grads = {k: np.zeros_like(v) for k, v in self.params.items()}
        for (features, action, prob, _), G in zip(self.episode, returns):
            # compute gradient via finite-difference style for softmax cross-entropy
            probs = self._forward(features)
            # dlogpi/dz = one-hot(action) - probs
            dlog = -probs
            dlog[action] += 1.0
            # backprop through MLP (compute gradients w.r.t params)
            # simple backprop with tanh activations
            # forward pass stores intermediates
            acts = [features.astype(float)]
            h = features.astype(float)
            L = len(self.hidden_sizes)
            for i in range(L):
                z = h.dot(self.params[f'W{i}']) + self.params[f'b{i}']
                h = np.tanh(z)
                acts.append(h)
            # logits layer input
            logits_in = acts[-1]
            # gradient at logits
            g_logits = dlog * G
            # grads for last layer
            grads[f'W{L}'] += np.outer(acts[-1], g_logits)
            grads[f'b{L}'] += g_logits
            # propagate to hidden layers
            g = g_logits.dot(self.params[f'W{L}'].T)
            for i in reversed(range(L)):
                # derivative of tanh
                dz = g * (1 - acts[i+1]**2)
                grads[f'W{i}'] += np.outer(acts[i], dz)
                grads[f'b{i}'] += dz
                g = dz.dot(self.params[f'W{i}'].T)
        # gradient step
        for k in self.params:
            self.params[k] += self.lr * grads[k]
        # update baseline
        avg_return = np.mean(returns) if returns.size>0 else 0.0
        self.baseline = (1 - self.baseline_alpha) * self.baseline + self.baseline_alpha * avg_return
        self.episode = []

    # High-level policy: decide actions for each immune agent given environment
    def decide_and_act(self):
        env = self.env
        alive_frac = np.mean(env.grid_state_matrices()[-1])
        # for each immune agent, form features and sample action
        for ag in getattr(env, 'immune_agents', []):
            features = self._agent_features(ag, alive_frac)
            action, probs = self.sample_action(features)
            # execute action
            reward = self._execute_action(ag, action)
            # simple immediate reward: change in alive fraction (we'll approximate with current alive frac)
            current_alive = np.mean(env.grid_state_matrices()[-1])
            r = current_alive - alive_frac
            self.store_transition(features, action, probs[action], r)
        # after actions in this time step, occasionally update policy
        if len(self.episode) > max(200, len(getattr(env, 'immune_agents', []))*10):
            self.finish_episode()

    def _agent_features(self, ag, alive_frac):
        env = self.env
        ii, jj = ag.i, ag.j
        # local cell
        cell = env.grid[ii][jj]
        local_damage = cell.damage if cell.alive else 1.0
        local_tert = cell.genes.get('TERT_reg', 0.0) if cell.alive else 0.0
        local_epi = cell.epigenetic_state if cell.alive else 0.0
        local_waste = env.waste[ii, jj]
        feat = np.array([ii / max(1, env.rows-1), jj / max(1, env.cols-1), local_damage, local_tert/1.5, local_epi, local_waste, alive_frac])
        return feat

    def _execute_action(self, ag, action):
        env = self.env
        if action == 0:
            # stay
            pass
        elif action == 1:
            ag.i = int(np.clip(ag.i - 1, 0, env.rows-1))
        elif action == 2:
            ag.i = int(np.clip(ag.i + 1, 0, env.rows-1))
        elif action == 3:
            ag.j = int(np.clip(ag.j - 1, 0, env.cols-1))
        elif action == 4:
            ag.j = int(np.clip(ag.j + 1, 0, env.cols-1))
        elif action == 5:
            # attack local cell
            target = env.grid[ag.i][ag.j]
            if target.alive:
                killed = ag.attempt_kill(target)
                # reward positive if target was high-threat, negative if false-positive
                threat = 0.0
                if target.genes.get('TERT_reg',0) > 1.2:
                    threat += 1.0
                threat += max(0, 1.0 - target.genetic_integrity)
                threat += target.damage
                if killed:
                    return 1.0 * threat - 0.5 * (1 - min(1.0, threat))
                else:
                    return -0.2
        elif action == 6:
            # recruit: spawn a new immune agent near current location if capacity allows
            if len(env.immune_agents) < (env.rows * env.cols) // 6:
                new_i = int(np.clip(ag.i + np.random.randint(-1,2), 0, env.rows-1))
                new_j = int(np.clip(ag.j + np.random.randint(-1,2), 0, env.cols-1))
                env.immune_agents.append(ImmuneCell(new_i, new_j))
                return -0.05  # small recruitment cost
        return 0.0

# Monkey patch: replace heuristic immune_step with RL-based driver

def enable_rl_immune(env):
    # create controller and attach to env
    controller = RLImmuneController(env)
    env.rl_immune = controller
    # override spawn_immune_agents to ensure at least one agent
    if getattr(env, 'immune_agents', None) is None:
        env.spawn_immune_agents(n=max(4, (env.rows*env.cols)//50))
    return controller

# New runner that uses RL immune controller
def run_experiment_with_rl(steps=400, rows=14, cols=14, animate=True, init_immune=6):
    env = OrganOnChip(rows=rows, cols=cols)
    env.spawn_immune_agents(n=init_immune)
    controller = enable_rl_immune(env)

    history = {'avg_tel': [], 'avg_damage': [], 'avg_activity': [], 'alive_fraction': [], 'avg_epi': [], 'immune_count': []}
    frames = []

    for t in range(steps):
        # RL controller decides actions for immune agents
        controller.decide_and_act()
        # then environment physics step
        env.step()
        # occasional immune scavenging
        for ag in getattr(env, 'immune_agents', []):
            if np.random.rand() < 0.05:
                env.waste[ag.i, ag.j] = max(0.0, env.waste[ag.i, ag.j] - np.random.uniform(0.0, 0.05))
        tel, dmg, act, tert, epi, alive = env.grid_state_matrices()
        alive_mask = alive.astype(bool)
        history['avg_tel'].append(np.nanmean(tel[alive_mask]) if np.any(alive_mask) else 0.0)
        history['avg_damage'].append(np.nanmean(dmg))
        history['avg_activity'].append(np.nanmean(act))
        history['alive_fraction'].append(np.mean(alive))
        history['avg_epi'].append(np.nanmean(epi[alive_mask]) if np.any(alive_mask) else 0.0)
        history['immune_count'].append(len(env.immune_agents))
        if animate:
            frames.append((tel.copy(), dmg.copy(), act.copy(), epi.copy(), env.waste.copy(), env.signal.copy(), np.array([(a.i,a.j) for a in env.immune_agents])))

    # plotting
    fig, axs = plt.subplots(2, 3, figsize=(14,8))
    axs = axs.ravel()
    axs[0].plot(history['avg_tel']); axs[0].set_title('Avg Telomere Length')
    axs[1].plot(history['avg_damage']); axs[1].set_title('Avg Damage')
    axs[2].plot(history['avg_activity']); axs[2].set_title('Avg Telomerase Activity')
    axs[3].plot(history['alive_fraction']); axs[3].set_title('Alive Fraction')
    axs[4].plot(history['avg_epi']); axs[4].set_title('Avg Epigenetic State')
    axs[5].plot(history['immune_count']); axs[5].set_title('Immune Agent Count')
    for a in axs:
        a.grid(True)
    plt.tight_layout(); plt.show()

    # animation with immune overlay
    if animate:
        fig2, axarr = plt.subplots(1, 6, figsize=(18, 3.5))
        def update_frame(k):
            tel, dmg, act, epi, waste, sig, immune_pos = frames[k]
            for ax in axarr:
                ax.clear()
            im0 = axarr[0].imshow(np.clip(tel, 0, 12000), vmin=3000, vmax=10000, aspect='auto'); axarr[0].set_title('Telomere')
            im1 = axarr[1].imshow(dmg, vmin=0, vmax=1, aspect='auto'); axarr[1].set_title('Damage')
            im2 = axarr[2].imshow(act, vmin=0, vmax=1, aspect='auto'); axarr[2].set_title('Telomerase')
            im3 = axarr[3].imshow(epi, vmin=0, vmax=1, aspect='auto'); axarr[3].set_title('Epigenetic')
            im4 = axarr[4].imshow(waste, aspect='auto'); axarr[4].set_title('Waste')
            im5 = axarr[5].imshow(sig, aspect='auto'); axarr[5].set_title('Signal')
            if immune_pos.size:
                ys = immune_pos[:,0]; xs = immune_pos[:,1]
                for ax in axarr:
                    ax.scatter(xs, ys, marker='x', s=40)
            fig2.suptitle(f'Time {k}')
            return im0, im1, im2, im3, im4, im5
        anim = animation.FuncAnimation(fig2, update_frame, frames=len(frames), interval=150, blit=False)
        plt.show()

    return env, history

# Attach run_experiment_with_rl into module namespace for easy access
# setattr(__import__(__name__).__class__, 'run_experiment_with_rl', run_experiment_with_rl) # Removed setattr call

# End of RL immune controller
```

```python
# -----------------------------
# PyTorch Actor-Critic Immune Controller
# -----------------------------
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

# detect GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class TorchA2CImmuneController(nn.Module):
    def __init__(self, env, hidden_sizes=(128, 64), lr=3e-4, gamma=0.99):
        super().__init__()
        self.env = env
        self.input_dim = 7  # features per immune agent
        self.hidden_sizes = hidden_sizes
        self.output_dim = 7  # actions: stay, up, down, left, right, attack, recruit
        self.gamma = gamma

        # Actor network
        self.actor = nn.Sequential(
            nn.Linear(self.input_dim, hidden_sizes[0]),
            nn.ReLU(),
            nn.Linear(hidden_sizes[0], hidden_sizes[1]),
            nn.ReLU(),
            nn.Linear(hidden_sizes[1], self.output_dim)
        )

        # Critic network
        self.critic = nn.Sequential(
            nn.Linear(self.input_dim, hidden_sizes[0]),
            nn.ReLU(),
            nn.Linear(hidden_sizes[0], hidden_sizes[1]),
            nn.ReLU(),
            nn.Linear(hidden_sizes[1], 1)
        )

        self.optimizer = optim.Adam(self.parameters(), lr=lr)
        self.to(device)
        self.memory = []

    def forward_actor(self, x):
        logits = self.actor(x)
        probs = F.softmax(logits, dim=-1)
        return probs

    def forward_critic(self, x):
        value = self.critic(x)
        return value

    def select_action(self, features):
        features = torch.tensor(features, dtype=torch.float32, device=device)
        probs = self.forward_actor(features)
        m = torch.distributions.Categorical(probs)
        action = m.sample()
        return action.item(), m.log_prob(action), probs.detach().cpu().numpy()

    def store_transition(self, log_prob, value, reward):
        self.memory.append((log_prob, value, reward))

    def train_step(self):
        if not self.memory:
            return
        R = 0
        returns = []
        for _, _, reward in reversed(self.memory):
            R = reward + self.gamma * R
            returns.insert(0, R)
        returns = torch.tensor(returns, dtype=torch.float32, device=device)
        policy_losses = []
        value_losses = []
        for (log_prob, value, _), R in zip(self.memory, returns):
            advantage = R - value.item()
            policy_losses.append(-log_prob * advantage)
            value_losses.append(F.mse_loss(value, torch.tensor([R], device=device)))
        self.optimizer.zero_grad()
        loss = torch.stack(policy_losses).sum() + torch.stack(value_losses).sum()
        loss.backward()
        self.optimizer.step()
        self.memory = []

    def agent_features(self, ag, alive_frac):
        env = self.env
        ii, jj = ag.i, ag.j
        cell = env.grid[ii][jj]
        local_damage = cell.damage if cell.alive else 1.0
        local_tert = cell.genes.get('TERT_reg', 0.0) if cell.alive else 0.0
        local_epi = cell.epigenetic_state if cell.alive else 0.0
        local_waste = env.waste[ii, jj]
        feat = np.array([ii / max(1, env.rows-1), jj / max(1, env.cols-1), local_damage, local_tert/1.5, local_epi, local_waste, alive_frac])
        return feat

    def decide_and_act(self):
        env = self.env
        alive_frac = np.mean(env.grid_state_matrices()[-1])
        for ag in getattr(env, 'immune_agents', []):
            features = self.agent_features(ag, alive_frac)
            action, log_prob, _ = self.select_action(features)
            value = self.forward_critic(torch.tensor(features, dtype=torch.float32, device=device))
            reward = self._execute_action(ag, action)
            self.store_transition(log_prob, value, reward)
        # train after each step
        self.train_step()

    def _execute_action(self, ag, action):
        env = self.env
        if action == 0: pass
        elif action == 1: ag.i = int(np.clip(ag.i - 1, 0, env.rows-1))
        elif action == 2: ag.i = int(np.clip(ag.i + 1, 0, env.rows-1))
        elif action == 3: ag.j = int(np.clip(ag.j - 1, 0, env.cols-1))
        elif action == 4: ag.j = int(np.clip(ag.j + 1, 0, env.cols-1))
        elif action == 5:
            target = env.grid[ag.i][ag.j]
            if target.alive:
                killed = ag.attempt_kill(target)
                threat = 0.0
                if target.genes.get('TERT_reg',0) > 1.2: threat += 1.0
                threat += max(0, 1.0 - target.genetic_integrity)
                threat += target.damage
                if killed: return 1.0*threat - 0.5*(1-min(1.0, threat))
                else: return -0.2
        elif action == 6:
            if len(env.immune_agents) < (env.rows*env.cols)//6:
                new_i = int(np.clip(ag.i + np.random.randint(-1,2), 0, env.rows-1))
                new_j = int(np.clip(ag.j + np.random.randint(-1,2), 0, env.cols-1))
                env.immune_agents.append(ImmuneCell(new_i, new_j))
                return -0.05
        return 0.0

# Runner integrating PyTorch RL immune controller

def run_experiment_with_rl_torch(steps=400, rows=14, cols=14, animate=True, init_immune=6):
    env = OrganOnChip(rows=rows, cols=cols)
    env.spawn_immune_agents(n=init_immune)
    controller = TorchA2CImmuneController(env)

    history = {'avg_tel': [], 'avg_damage': [], 'avg_activity': [], 'alive_fraction': [], 'avg_epi': [], 'immune_count': []}
    frames = []

    for t in range(steps):
        controller.decide_and_act()
        env.step()
        for ag in getattr(env, 'immune_agents', []):
            if np.random.rand() < 0.05:
                env.waste[ag.i, ag.j] = max(0.0, env.waste[ag.i, ag.j] - np.random.uniform(0.0, 0.05))
        tel, dmg, act, tert, epi, alive = env.grid_state_matrices()
        alive_mask = alive.astype(bool)
        history['avg_tel'].append(np.nanmean(tel[alive_mask]) if np.any(alive_mask) else 0.0)
        history['avg_damage'].append(np.nanmean(dmg))
        history['avg_activity'].append(np.nanmean(act))
        history['alive_fraction'].append(np.mean(alive))
        history['avg_epi'].append(np.nanmean(epi[alive_mask]) if np.any(alive_mask) else 0.0)
        history['immune_count'].append(len(env.immune_agents))
        if animate:
            frames.append((tel.copy(), dmg.copy(), act.copy(), epi.copy(), env.waste.copy(), env.signal.copy(), np.array([(a.i,a.j) for a in env.immune_agents])))

    # plotting (same as previous RL runner)
    fig, axs = plt.subplots(2, 3, figsize=(14,8))
    axs = axs.ravel()
    axs[0].plot(history['avg_tel']); axs[0].set_title('Avg Telomere Length')
    axs[1].plot(history['avg_damage']); axs[1].set_title('Avg Damage')
    axs[2].plot(history['avg_activity']); axs[2].set_title('Avg Telomerase Activity')
    axs[3].plot(history['alive_fraction']); axs[3].set_title('Alive Fraction')
    axs[4].plot(history['avg_epi']); axs[4].set_title('Avg Epigenetic State')
    axs[5].plot(history['immune_count']); axs[5].set_title('Immune Agent Count')
    for a in axs: a.grid(True)
    plt.tight_layout(); plt.show()

    # animation overlay (immune positions)
    if animate:
        fig2, axarr = plt.subplots(1, 6, figsize=(18, 3.5))
        def update_frame(k):
            tel, dmg, act, epi, waste, sig, immune_pos = frames[k]
            for ax in axarr: ax.clear()
            axarr[0].imshow(np.clip(tel, 0, 12000), vmin=3000, vmax=10000, aspect='auto'); axarr[0].set_title('Telomere')
            axarr[1].imshow(dmg, vmin=0, vmax=1, aspect='auto'); axarr[1].set_title('Damage')
            axarr[2].imshow(act, vmin=0, vmax=1, aspect='auto'); axarr[2].set_title('Telomerase')
            axarr[3].imshow(epi, vmin=0, vmax=1, aspect='auto'); axarr[3].set_title('Epigenetic')
            axarr[4].imshow(waste, aspect='auto'); axarr[4].set_title('Waste')
            axarr[5].imshow(sig, aspect='auto'); axarr[5].set_title('Signal')
            if immune_pos.size:
                ys = immune_pos[:,0]; xs = immune_pos[:,1]
                for ax in axarr: ax.scatter(xs, ys, marker='x', s=40)
            fig2.suptitle(f'Time {k}')
            return axarr
        anim = animation.FuncAnimation(fig2, update_frame, frames=len(frames), interval=150, blit=False)
        plt.show()

    return env, history

# attach to module namespace
# setattr(__import__(__name__).__class__, 'run_experiment_with_rl_torch', run_experiment_with_rl_torch) # Removed setattr call

# End of PyTorch RL immune controller
```

```python
# -----------------------------
# Immune Surveillance Module
# -----------------------------
class ImmuneCell:
    """Simple immune patrolling agent that moves, senses, and removes malignant or highly damaged cells."""
    def __init__(self, i, j, sensing_radius=1, kill_prob=0.9, patrol_speed=1):
        self.i = i
        self.j = j
        self.sensing_radius = sensing_radius
        self.kill_prob = kill_prob
        self.patrol_speed = patrol_speed
        self.alive = True

    def move_towards(self, target_i, target_j, rows, cols):
        # move one step towards target (bounded by grid)
        di = np.sign(target_i - self.i)
        dj = np.sign(target_j - self.j)
        self.i = int(np.clip(self.i + di, 0, rows-1))
        self.j = int(np.clip(self.j + dj, 0, cols-1))

    def random_patrol(self, rows, cols):
        # random walk limited by patrol_speed
        self.i = int(np.clip(self.i + np.random.randint(-1, 2), 0, rows-1))
        self.j = int(np.clip(self.j + np.random.randint(-1, 2), 0, cols-1))

    def attempt_kill(self, cell):
        # Decide whether to remove a threatening cell
        if not cell.alive:
            return False
        # Threat criteria: hyper-TERT, low genetic integrity, or extreme damage
        threat_score = 0.0
        if cell.genes.get('TERT_reg', 0) > 1.2:
            threat_score += 0.7
        if cell.genetic_integrity < 0.4:
            threat_score += 0.6
        if cell.damage > 0.9:
            threat_score += 0.4
        # Epigenetic hyperactivity may also be suspicious
        if cell.epigenetic_state > 0.95:
            threat_score += 0.2
        # Probability to act scales with threat score
        if threat_score <= 0:
            return False
        p = 1 - np.exp(-2.5 * threat_score)
        # combined with immune intrinsic kill probability
        if np.random.rand() < min(1.0, p * self.kill_prob):
            # kill: mark cell dead and leave debris
            cell.alive = False
            cell.debris += np.random.uniform(0.1, 0.5)
            return True
        return False

# Integrate immune agents into OrganOnChip
# We'll add an immune list and simple recruitment based on tissue distress signals

def _organ_on_chip_add_immune_support():
    # patch class to add immune support attributes and methods
    def init_with_immune(self, rows=14, cols=14, stress_rate=0.02, flow_rate=0.2):
        # call original __init__ by reusing current attributes already set in file; this function
        # will only be used if user re-instantiates OrganOnChip manually. For backward compatibility
        # we update the instance attributes below when called from run_experiment.
        pass

    # Add immune list attribute to existing OrganOnChip class via monkey patch
    setattr(OrganOnChip, 'immune_agents', None)

    def spawn_immune_agents(self, n=6):
        # initialize immune agents spread around the grid
        agents = []
        for k in range(n):
            i = np.random.randint(0, self.rows)
            j = np.random.randint(0, self.cols)
            agents.append(ImmuneCell(i, j))
        self.immune_agents = agents

    def immune_step(self):
        # Simple recruitment: areas with high epi_signal or high waste attract immune cells
        if self.immune_agents is None:
            self.spawn_immune_agents(n=max(4, (self.rows*self.cols)//50))
        # For each immune agent: sense local threats, move and act
        for agent in list(self.immune_agents):
            if not agent.alive:
                continue
            # find highest local distress cell within a radius
            best = None
            best_score = 0.0
            rmin = max(0, agent.i - 3)
            rmax = min(self.rows-1, agent.i + 3)
            cmin = max(0, agent.j - 3)
            cmax = min(self.cols-1, agent.j + 3)
            for ii in range(rmin, rmax+1):
                for jj in range(cmin, cmax+1):
                    cell = self.grid[ii][jj]
                    if not cell.alive:
                        continue
                    # threat heuristics similar to ImmuneCell.attempt_kill
                    score = 0.0
                    if cell.genes.get('TERT_reg', 0) > 1.2:
                        score += 3.0
                    score += max(0, 1.0 - cell.genetic_integrity) * 2.0
                    score += cell.damage * 1.5
                    # environmental waste and epi_signal increase local score
                    score += 1.0 * self.waste[ii, jj]
                    score += 0.6 * self.epi_signal[ii, jj]
                    if score > best_score:
                        best_score = score
                        best = (ii, jj)
            # Move towards best target if found, otherwise random patrol
            if best and best_score > 0.5:
                agent.move_towards(best[0], best[1], self.rows, self.cols)
                target_cell = self.grid[agent.i][agent.j]
                # attempt kill
                agent.attempt_kill(target_cell)
            else:
                agent.random_patrol(self.rows, self.cols)
            # small chance immune agent is exhausted/removed
            if np.random.rand() < 0.001:
                agent.alive = False
        # clean dead agents occasionally
        self.immune_agents = [a for a in self.immune_agents if a.alive]

    # Attach methods to OrganOnChip
    setattr(OrganOnChip, 'spawn_immune_agents', spawn_immune_agents)
    setattr(OrganOnChip, 'immune_step', immune_step)

# apply monkey patch
_organ_on_chip_add_immune_support()

# Modify OrganOnChip.step to call immune functions: we'll provide a small wrapper that the user
# can call instead of the original step() if they want immune surveillance active.

def step_with_immune(self):
    # Run immune actions first (they can pre-emptively remove dangerous cells)
    if getattr(self, 'immune_agents', None) is None:
        self.spawn_immune_agents(n=max(4, (self.rows*self.cols)//50))
    # recruitment increases when tissue distress high
    # spawn extra agents if many high-waste spots
    hotspots = np.sum(self.waste > 0.5)
    if hotspots > (self.rows * self.cols) * 0.02 and len(self.immune_agents) < (self.rows*self.cols)//10:
        self.spawn_immune_agents(n=2)
    # immune patrol & act
    self.immune_step()
    # then run the original physiological step
    # reuse existing step logic from class (we call it directly)
    OrganOnChip.step(self)
    # after step, immune cells can scavenge debris (reduce waste locally where they are present)
    for ag in getattr(self, 'immune_agents', [])[:]:
        ii, jj = ag.i, ag.j
        # immune scavenging reduces waste and increases local phagocytosis indirectly
        self.waste[ii, jj] = max(0.0, self.waste[ii, jj] - np.random.uniform(0.01, 0.1))

# Attach wrapper
setattr(OrganOnChip, 'step_with_immune', step_with_immune)
```

```python
# -----------------------------
# PyTorch Actor-Critic Immune Controller
# -----------------------------
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

# detect GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class TorchA2CImmuneController(nn.Module):
    def __init__(self, env, hidden_sizes=(128, 64), lr=3e-4, gamma=0.99):
        super().__init__()
        self.env = env
        self.input_dim = 7  # features per immune agent
        self.hidden_sizes = hidden_sizes
        self.output_dim = 7  # actions: stay, up, down, left, right, attack, recruit
        self.gamma = gamma

        # Actor network
        self.actor = nn.Sequential(
            nn.Linear(self.input_dim, hidden_sizes[0]),
            nn.ReLU(),
            nn.Linear(hidden_sizes[0], hidden_sizes[1]),
            nn.ReLU(),
            nn.Linear(hidden_sizes[1], self.output_dim)
        )

        # Critic network
        self.critic = nn.Sequential(
            nn.Linear(self.input_dim, hidden_sizes[0]),
            nn.ReLU(),
            nn.Linear(hidden_sizes[0], hidden_sizes[1]),
            nn.ReLU(),
            nn.Linear(hidden_sizes[1], 1)
        )

        self.optimizer = optim.Adam(self.parameters(), lr=lr)
        self.to(device)
        self.memory = []

    def forward_actor(self, x):
        logits = self.actor(x)
        probs = F.softmax(logits, dim=-1)
        return probs

    def forward_critic(self, x):
        value = self.critic(x)
        return value

    def select_action(self, features):
        features = torch.tensor(features, dtype=torch.float32, device=device)
        probs = self.forward_actor(features)
        m = torch.distributions.Categorical(probs)
        action = m.sample()
        return action.item(), m.log_prob(action), probs.detach().cpu().numpy()

    def store_transition(self, log_prob, value, reward):
        self.memory.append((log_prob, value, reward))

    def train_step(self):
        if not self.memory:
            return
        R = 0
        returns = []
        for _, _, reward in reversed(self.memory):
            R = reward + self.gamma * R
            returns.insert(0, R)
        returns = torch.tensor(returns, dtype=torch.float32, device=device)
        policy_losses = []
        value_losses = []
        for (log_prob, value, _), R in zip(self.memory, returns):
            advantage = R - value.item()
            policy_losses.append(-log_prob * advantage)
            value_losses.append(F.mse_loss(value, torch.tensor([R], device=device)))
        self.optimizer.zero_grad()
        loss = torch.stack(policy_losses).sum() + torch.stack(value_losses).sum()
        loss.backward()
        self.optimizer.step()
        self.memory = []

    def agent_features(self, ag, alive_frac):
        env = self.env
        ii, jj = ag.i, ag.j
        cell = env.grid[ii][jj]
        local_damage = cell.damage if cell.alive else 1.0
        local_tert = cell.genes.get('TERT_reg', 0.0) if cell.alive else 0.0
        local_epi = cell.epigenetic_state if cell.alive else 0.0
        local_waste = env.waste[ii, jj]
        feat = np.array([ii / max(1, env.rows-1), jj / max(1, env.cols-1), local_damage, local_tert/1.5, local_epi, local_waste, alive_frac])
        return feat

    def decide_and_act(self):
        env = self.env
        alive_frac = np.mean(env.grid_state_matrices()[-1])
        for ag in getattr(env, 'immune_agents', []):
            features = self.agent_features(ag, alive_frac)
            action, log_prob, _ = self.select_action(features)
            value = self.forward_critic(torch.tensor(features, dtype=torch.float32, device=device))
            reward = self._execute_action(ag, action)
            self.store_transition(log_prob, value, reward)
        # train after each step
        self.train_step()

    def _execute_action(self, ag, action):
        env = self.env
        if action == 0: pass
        elif action == 1: ag.i = int(np.clip(ag.i - 1, 0, env.rows-1))
        elif action == 2: ag.i = int(np.clip(ag.i + 1, 0, env.rows-1))
        elif action == 3: ag.j = int(np.clip(ag.j - 1, 0, env.cols-1))
        elif action == 4: ag.j = int(np.clip(ag.j + 1, 0, env.cols-1))
        elif action == 5:
            target = env.grid[ag.i][ag.j]
            if target.alive:
                killed = ag.attempt_kill(target)
                threat = 0.0
                if target.genes.get('TERT_reg',0) > 1.2: threat += 1.0
                threat += max(0, 1.0 - target.genetic_integrity)
                threat += target.damage
                if killed: return 1.0*threat - 0.5*(1-min(1.0, threat))
                else: return -0.2
        elif action == 6:
            if len(env.immune_agents) < (env.rows*env.cols)//6:
                new_i = int(np.clip(ag.i + np.random.randint(-1,2), 0, env.rows-1))
                new_j = int(np.clip(ag.j + np.random.randint(-1,2), 0, env.cols-1))
                env.immune_agents.append(ImmuneCell(new_i, new_j))
                return -0.05
        return 0.0

# Runner integrating PyTorch RL immune controller

def run_experiment_with_rl_torch(steps=400, rows=14, cols=14, animate=True, init_immune=6):
    env = OrganOnChip(rows=rows, cols=cols)
    env.spawn_immune_agents(n=init_immune)
    controller = TorchA2CImmuneController(env)

    history = {'avg_tel': [], 'avg_damage': [], 'avg_activity': [], 'alive_fraction': [], 'avg_epi': [], 'immune_count': []}
    frames = []

    for t in range(steps):
        controller.decide_and_act()
        env.step()
        for ag in getattr(env, 'immune_agents', []):
            if np.random.rand() < 0.05:
                env.waste[ag.i, ag.j] = max(0.0, env.waste[ag.i, ag.j] - np.random.uniform(0.0, 0.05))
        tel, dmg, act, tert, epi, alive = env.grid_state_matrices()
        alive_mask = alive.astype(bool)
        history['avg_tel'].append(np.nanmean(tel[alive_mask]) if np.any(alive_mask) else 0.0)
        history['avg_damage'].append(np.nanmean(dmg))
        history['avg_activity'].append(np.nanmean(act))
        history['alive_fraction'].append(np.mean(alive))
        history['avg_epi'].append(np.nanmean(epi[alive_mask]) if np.any(alive_mask) else 0.0)
        history['immune_count'].append(len(env.immune_agents))
        if animate:
            frames.append((tel.copy(), dmg.copy(), act.copy(), epi.copy(), env.waste.copy(), env.signal.copy(), np.array([(a.i,a.j) for a in env.immune_agents])))

    # plotting (same as previous RL runner)
    fig, axs = plt.subplots(2, 3, figsize=(14,8))
    axs = axs.ravel()
    axs[0].plot(history['avg_tel']); axs[0].set_title('Avg Telomere Length')
    axs[1].plot(history['avg_damage']); axs[1].set_title('Avg Damage')
    axs[2].plot(history['avg_activity']); axs[2].set_title('Avg Telomerase Activity')
    axs[3].plot(history['alive_fraction']); axs[3].set_title('Alive Fraction')
    axs[4].plot(history['avg_epi']); axs[4].set_title('Avg Epigenetic State')
    axs[5].plot(history['immune_count']); axs[5].set_title('Immune Agent Count')
    for a in axs: a.grid(True)
    plt.tight_layout(); plt.show()

    # animation overlay (immune positions)
    if animate:
        fig2, axarr = plt.subplots(1, 6, figsize=(18, 3.5))
        def update_frame(k):
            tel, dmg, act, epi, waste, sig, immune_pos = frames[k]
            for ax in axarr: ax.clear()
            axarr[0].imshow(np.clip(tel, 0, 12000), vmin=3000, vmax=10000, aspect='auto'); axarr[0].set_title('Telomere')
            axarr[1].imshow(dmg, vmin=0, vmax=1, aspect='auto'); axarr[1].set_title('Damage')
            axarr[2].imshow(act, vmin=0, vmax=1, aspect='auto'); axarr[2].set_title('Telomerase')
            axarr[3].imshow(epi, vmin=0, vmax=1, aspect='auto'); axarr[3].set_title('Epigenetic')
            axarr[4].imshow(waste, aspect='auto'); axarr[4].set_title('Waste')
            axarr[5].imshow(sig, aspect='auto'); axarr[5].set_title('Signal')
            if immune_pos.size:
                ys = immune_pos[:,0]; xs = immune_pos[:,1]
                for ax in axarr: ax.scatter(xs, ys, marker='x', s=40)
            fig2.suptitle(f'Time {k}')
            return axarr
        anim = animation.FuncAnimation(fig2, update_frame, frames=len(frames), interval=150, blit=False)
        plt.show()

    return env, history

# attach to module namespace
# setattr(__import__(__name__).__class__, 'run_experiment_with_rl_torch', run_experiment_with_rl_torch) # Removed setattr call

# End of PyTorch RL immune controller
```
