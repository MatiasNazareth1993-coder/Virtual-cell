```python
"""
transformer_telomerase_regulator.py

Lightweight Transformer-based controller for telomerase regulation in a virtual cell.
This is a modeling skeleton ‚Äî not a wet-lab protocol.

Requirements:
- Python 3.9+
- PyTorch
"""

import math
import torch
import torch.nn as nn
import torch.nn.functional as F
from typing import Dict, Tuple, Optional

# ----------------------------
# Hyperparameters / Settings
# ----------------------------
EMBED_DIM = 128
NUM_LAYERS = 3
NUM_HEADS = 4
FF_DIM = 256
SEQ_LEN = 16              # number of timesteps fed to transformer
INPUT_FEATURES = 12       # telomere length + ROS + damage + etc
ACTION_DIM = 2            # [binary_on_prob, continuous_rate]
DROPOUT = 0.1
SAFE_TELOMERE_MAX = 2.0   # normalized units (example)


# ----------------------------
# Positional encoding
# ----------------------------
class PositionalEncoding(nn.Module):
    def __init__(self, d_model: int, max_len: int = 5000):
        super().__init__()
        pe = torch.zeros(max_len, d_model)  # (max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0)  # (1, max_len, d_model)
        self.register_buffer("pe", pe)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        # x: (batch, seq_len, d_model)
        return x + self.pe[:, : x.size(1), :]


# ----------------------------
# Transformer controller
# ----------------------------
class TelomeraseTransformerController(nn.Module):
    def __init__(self,
                 input_features: int = INPUT_FEATURES,
                 embed_dim: int = EMBED_DIM,
                 num_layers: int = NUM_LAYERS,
                 num_heads: int = NUM_HEADS,
                 ff_dim: int = FF_DIM,
                 action_dim: int = ACTION_DIM,
                 dropout: float = DROPOUT):
        super().__init__()
        self.input_proj = nn.Linear(input_features, embed_dim)
        self.pos_enc = PositionalEncoding(embed_dim, max_len=SEQ_LEN)
        encoder_layer = nn.TransformerEncoderLayer(d_model=embed_dim,
                                                   nhead=num_heads,
                                                   dim_feedforward=ff_dim,
                                                   dropout=dropout,
                                                   activation="gelu")
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        self.control_head = nn.Sequential(
            nn.LayerNorm(embed_dim),
            nn.Linear(embed_dim, embed_dim // 2),
            nn.ReLU(),
            nn.Linear(embed_dim // 2, action_dim)
        )

        # for interpretability: map attention weights via an auxiliary projection
        self.attention_map_proj = nn.Linear(embed_dim, 1)

    def forward(self, x: torch.Tensor, src_key_padding_mask: Optional[torch.Tensor] = None
                ) -> Dict[str, torch.Tensor]:
        """
        x: (batch, seq_len, input_features)
        returns dict with:
          - "action_logits": raw outputs (batch, action_dim)
          - "action": mapped action probabilities / rates
          - "seq_embedding": (batch, seq_len, embed_dim)
        """
        # Project inputs
        h = self.input_proj(x)            # (B, L, D)
        h = self.pos_enc(h)               # add positional encodings
        # Transformer expects shape (L, B, D)
        h_t = h.transpose(0, 1)           # (L, B, D)
        out = self.transformer(h_t, src_key_padding_mask=src_key_padding_mask)
        out = out.transpose(0, 1)         # (B, L, D)
        # Pooling: use last timestep embedding (could also use attentive pooling)
        pooled = out[:, -1, :]            # (B, D)
        logits = self.control_head(pooled)  # (B, action_dim)
        # Map logits -> actual actions
        # action[0]: sigmoid -> prob of activation; action[1]: sigmoid->continuous rate 0..1
        action_prob = torch.sigmoid(logits[:, 0:1])
        action_rate = torch.sigmoid(logits[:, 1:2])
        action = torch.cat([action_prob, action_rate], dim=-1)
        return {
            "action_logits": logits,
            "action": action,
            "seq_embedding": out
        }

# ----------------------------
# Environment wrapper (interface)
# ----------------------------
class VirtualCellEnvInterface:
    """
    Minimal interface expected by controller:
      - step(action): applies action and returns next_state, reward, done, info
      - reset(): returns initial observation sequence
    Observations are time-series windows (seq_len x feature_dim)
    """

    def __init__(self, seq_len: int = SEQ_LEN):
        self.seq_len = seq_len
        self.reset()

    def reset(self):
        # state holds recent history as list of feature vectors
        # Example features: [tel_len, ROS, dna_damage, cell_cycle, atp, mut_count, phag_queue, ...]
        self.history = [self._random_obs() for _ in range(self.seq_len)]
        self.step_count = 0
        self.done = False
        return self._get_obs()

    def _random_obs(self):
        # synthetic initial observation (normalize features to 0..1)
        obs = torch.rand(INPUT_FEATURES)
        # ensure telomere length is reasonable
        obs[0] = 0.5  # telomere normalized
        return obs

    def _get_obs(self):
        return torch.stack(self.history, dim=0)  # (seq_len, INPUT_FEATURES)

    def step(self, action: torch.Tensor) -> Tuple[torch.Tensor, float, bool, Dict]:
        """
        action: tensor shape (action_dim,) or (1, action_dim)
        Apply action to environment; returns next_obs, reward, done, info
        """
        if action.dim() == 2:
            action = action.squeeze(0)
        # Safety wrapper: clip rate & optionally override if telomere > safe max
        tel_len = float(self.history[-1][0].item())
        activation_prob = float(action[0].item())
        rate = float(action[1].item())
        # If telomere already too long, force suppression
        if tel_len > SAFE_TELOMERE_MAX:
            activation_prob = 0.0
            rate = 0.0

        # Simulate effect (very simplified)
        # telomere grows proportional to activation_prob * rate, shrinks with damage/replication
        delta_tel = activation_prob * (rate * 0.05) - 0.01 * self.history[-1][2].item()  # damage reduces telomere
        # mutate environment state
        new_obs = self.history[-1] + 0.02 * torch.randn_like(self.history[-1])  # small noise
        new_obs = new_obs.clone()
        new_obs[0] = max(0.0, min(2.5, tel_len + delta_tel))  # telomere in [0,2.5]
        # update mutational burden: increase slightly if activation is high
        mut_count_idx = 5 if INPUT_FEATURES > 5 else -1
        if mut_count_idx >= 0:
            new_obs[mut_count_idx] = min(1.0, new_obs[mut_count_idx] + 0.001 * activation_prob * rate)
        # shift history
        self.history.pop(0)
        self.history.append(new_obs)
        self.step_count += 1

        # reward: favor telomere within target window (0.6..1.2), penalize high mut_count & damage
        tel = new_obs[0].item()
        mut = new_obs[mut_count_idx].item() if mut_count_idx >= 0 else 0.0
        damage = new_obs[2].item() if INPUT_FEATURES > 2 else 0.0
        tel_target = 1.0
        reward = -abs(tel - tel_target) - 5.0 * mut - 2.0 * damage
        done = self.step_count >= 500
        info = {"telomere": tel, "mut": mut}
        return self._get_obs(), float(reward), done, info


# ----------------------------
# Losses and constraints
# ----------------------------
def safety_penalty(info: Dict, action: torch.Tensor) -> torch.Tensor:
    """
    Compute penalty if action risks unsafe behavior (e.g., excessive telomere extension).
    info: environment info dict
    action: (B, action_dim)
    """
    tel = info.get("telomere", 0.0)
    # if tel above a threshold and action tries to activate, penalize
    unsafe_mask = (tel > SAFE_TELOMERE_MAX * 0.9)
    penalty = 0.0
    if unsafe_mask:
        # encourage suppression
        penalty = 10.0 * action[:, 0].mean()
    return torch.tensor(penalty, dtype=torch.float)


# ----------------------------
# Example training loop (supervised + RL hybrid)
# ----------------------------
def train_example():
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = TelomeraseTransformerController().to(device)
    env = VirtualCellEnvInterface(seq_len=SEQ_LEN)
    opt = torch.optim.AdamW(model.parameters(), lr=3e-4, weight_decay=1e-4)

    # synthetic demonstration data for supervised pretraining (optional)
    # Here: teach the model to maintain telomere ~ 1.0 by simple heuristic
    def generate_demo_batch(batch_size=32):
        obs_batch = []
        act_batch = []
        for _ in range(batch_size):
            obs = torch.stack([env._random_obs() for _ in range(SEQ_LEN)], dim=0)  # (L, F)
            tel = obs[-1][0].item()
            # simple heuristic: if tel < 0.9 -> activate with moderate rate, if >1.1 suppress
            if tel < 0.9:
                a = torch.tensor([1.0, 0.6])
            elif tel > 1.1:
                a = torch.tensor([0.0, 0.0])
            else:
                a = torch.tensor([0.3, 0.2])
            obs_batch.append(obs)
            act_batch.append(a)
        obs_batch = torch.stack(obs_batch, dim=0)  # (B, L, F)
        act_batch = torch.stack(act_batch, dim=0)  # (B, action_dim)
        return obs_batch.to(device), act_batch.to(device)

    # supervised pretrain
    for epoch in range(50):
        model.train()
        obs_b, act_b = generate_demo_batch(batch_size=64)
        out = model(obs_b)
        pred = out["action"]
        loss = F.mse_loss(pred, act_b)
        opt.zero_grad()
        loss.backward()
        opt.step()
        if epoch % 10 == 0:
            print(f"[pretrain] epoch {epoch} loss {loss.item():.4f}")

    # RL fine-tuning using simple policy gradient-ish objective (high-level)
    # NOTE: this is placeholder ‚Äî integrate a proper RL library for production.
    for episode in range(200):
        obs_seq = env.reset().unsqueeze(0).to(device)  # (1, L, F)
        done = False
        ep_reward = 0.0
        traj = []
        while not done:
            model.eval()
            with torch.no_grad():
                out = model(obs_seq)  # batch=1
                action = out["action"]  # (1, action_dim)
            next_obs, r, done, info = env.step(action.cpu())
            ep_reward += r
            # record transition for simple policy gradient
            traj.append((obs_seq.cpu(), action.cpu(), r, info))
            obs_seq = next_obs.unsqueeze(0).to(device)
        # compute simple policy gradient surrogate (toy)
        # For real training use PPO or other stable algorithms.
        # Simple: encourage actions taken in high reward episodes
        if ep_reward > -100:  # tiny heuristic threshold
            # backprop through last few transitions as supervised targets (toy)
            model.train()
            for obs_seq_b, action_b, r, info in traj[-20:]:
                obs_seq_b = obs_seq_b.to(device)
                action_b = action_b.to(device)
                pred = model(obs_seq_b)["action"]
                # pseudo-target: amplify action towards positive reward
                target = action_b * (1.0 + 0.01 * r)
                loss = F.mse_loss(pred, target)
                # safety penalty
                loss = loss + 0.01 * safety_penalty(info, pred)
                opt.zero_grad()
                loss.backward()
                opt.step()
        if episode % 20 == 0:
            print(f"[rl] episode {episode} ep_reward={ep_reward:.2f}")

    return model


if __name__ == "__main__":
    # Run training example (this may be slow; for testing reduce episodes)
    model = train_example()
    torch.save(model.state_dict(), "telomerase_transformer.pth")
    print("Model trained & saved (example).")
```

```python
# ============================================================
# üß¨ Transformer-Based Telomerase Regulator + PPO/SAC (Colab)
# ============================================================

!pip install torch gymnasium stable-baselines3 matplotlib --quiet

import gymnasium as gym
import numpy as np
import torch
import torch.nn as nn
import math, random, matplotlib.pyplot as plt
from stable_baselines3 import PPO, SAC
from stable_baselines3.common.torch_layers import BaseFeaturesExtractor

# ============================================================
# Environment Definition (Gym-compatible)
# ============================================================
class VirtualCellEnv(gym.Env):
    """
    Virtual cell environment with telomerase regulation.
    Compatible with Stable-Baselines3 (continuous action space).
    """
    metadata = {"render.modes": ["human"]}
    def __init__(self, seq_len=16, n_features=12, safe_max=2.0):
        super().__init__()
        self.seq_len = seq_len
        self.n_features = n_features
        self.safe_max = safe_max
        self.action_space = gym.spaces.Box(low=0.0, high=1.0, shape=(2,), dtype=np.float32)
        self.observation_space = gym.spaces.Box(low=0.0, high=1.0, shape=(n_features,), dtype=np.float32)
        self.reset()

    def _random_obs(self):
        obs = np.random.rand(self.n_features).astype(np.float32)
        obs[0] = 0.5  # telomere
        return obs

    def reset(self, *, seed=None, options=None):
        super().reset(seed=seed)
        self.state = self._random_obs()
        self.mut = 0.0
        self.steps = 0
        return self.state, {}

    def step(self, action):
        tel, damage = self.state[0], self.state[2] if self.n_features>2 else 0.0
        act, rate = float(action[0]), float(action[1])
        if tel > self.safe_max: act, rate = 0.0, 0.0
        delta_tel = act * (rate*0.05) - 0.01*damage
        new_tel = np.clip(tel + delta_tel, 0.0, 2.5)
        self.mut = np.clip(self.mut + 0.001*act*rate, 0.0, 1.0)
        self.state = self.state + 0.02*np.random.randn(self.n_features).astype(np.float32)
        self.state[0] = new_tel
        self.state[5 if self.n_features>5 else -1] = self.mut
        reward = -abs(new_tel - 1.0) - 5*self.mut - 2*damage
        self.steps += 1
        done = self.steps >= 500
        return self.state, reward, done, False, {}

    def render(self):
        print(f"Telomere: {self.state[0]:.2f}, Mut: {self.mut:.3f}")

# ============================================================
# Custom Transformer Feature Extractor
# ============================================================
class TransformerExtractor(BaseFeaturesExtractor):
    """
    Transformer encoder as feature extractor for PPO/SAC.
    """
    def __init__(self, observation_space: gym.spaces.Box, embed_dim=64, n_heads=2, n_layers=2):
        super().__init__(observation_space, features_dim=embed_dim)
        obs_dim = observation_space.shape[0]
        self.proj = nn.Linear(obs_dim, embed_dim)
        layer = nn.TransformerEncoderLayer(d_model=embed_dim, nhead=n_heads, dim_feedforward=embed_dim*2)
        self.transformer = nn.TransformerEncoder(layer, num_layers=n_layers)
        self.pool = nn.AdaptiveAvgPool1d(1)
    def forward(self, obs: torch.Tensor) -> torch.Tensor:
        # obs shape: (batch, features)
        x = self.proj(obs).unsqueeze(1)  # (B, 1, D)
        x_t = x.transpose(0,1)           # (1, B, D)
        out = self.transformer(x_t).transpose(0,1) # (B, 1, D)
        return out.squeeze(1)            # (B, D)

# ============================================================
# Training configuration
# ============================================================
USE_PPO = True    # set False to use SAC instead
TOTAL_TIMESTEPS = 50_000

env = VirtualCellEnv()
policy_kwargs = dict(
    features_extractor_class=TransformerExtractor,
    features_extractor_kwargs=dict(embed_dim=64, n_heads=2, n_layers=2),
)

algo = PPO if USE_PPO else SAC
model = algo(
    "MlpPolicy",  # base policy (will override feature extractor)
    env,
    policy_kwargs=policy_kwargs,
    verbose=1,
    tensorboard_log="./telomerase_rl/",
    learning_rate=3e-4,
    n_steps=1024,
    batch_size=64,
)

# ============================================================
# Train the model
# ============================================================
print(f"Training {'PPO' if USE_PPO else 'SAC'} agent...")
model.learn(total_timesteps=TOTAL_TIMESTEPS)
model.save("telomerase_rl_agent")

# ============================================================
# Evaluation and Visualization
# ============================================================
rewards, tel_trajectories = [], []
obs, _ = env.reset()
for ep in range(10):
    done, total_r = False, 0
    tel_series = []
    obs, _ = env.reset()
    while not done:
        action, _ = model.predict(obs, deterministic=True)
        obs, r, done, _, _ = env.step(action)
        tel_series.append(obs[0])
        total_r += r
    rewards.append(total_r)
    tel_trajectories.append(tel_series)

print(f"Average reward (10 eval episodes): {np.mean(rewards):.2f}")

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(rewards, marker='o')
plt.title("Episode Reward (Evaluation)")
plt.xlabel("Episode"); plt.ylabel("Total Reward")

plt.subplot(1,2,2)
for t in tel_trajectories:
    plt.plot(t, alpha=0.7)
plt.axhline(1.0, ls="--", c="k", label="Target telomere length")
plt.title("Telomere Length Trajectories")
plt.xlabel("Step"); plt.ylabel("Telomere (normalized)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# ============================================================
# üß¨ Transformer-Based Telomerase Regulator (Colab / Jupyter)
# ------------------------------------------------------------
# Virtual cell controller with telomere regulation & safety
# ============================================================

!pip install torch matplotlib tqdm --quiet

import math, torch, torch.nn as nn, torch.nn.functional as F
import matplotlib.pyplot as plt
from tqdm import trange

# ============================================================
# Config
# ============================================================
EMBED_DIM = 128
NUM_LAYERS = 3
NUM_HEADS = 4
FF_DIM = 256
SEQ_LEN = 16
INPUT_FEATURES = 12
ACTION_DIM = 2
SAFE_TELOMERE_MAX = 2.0
LR = 3e-4

# ============================================================
# Positional Encoding
# ============================================================
class PositionalEncoding(nn.Module):
    def __init__(self, d_model, max_len=5000):
        super().__init__()
        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0)
        self.register_buffer("pe", pe)
    def forward(self, x):
        return x + self.pe[:, :x.size(1), :]

# ============================================================
# Transformer Controller
# ============================================================
class TelomeraseTransformerController(nn.Module):
    def __init__(self):
        super().__init__()
        self.input_proj = nn.Linear(INPUT_FEATURES, EMBED_DIM)
        self.pos_enc = PositionalEncoding(EMBED_DIM, max_len=SEQ_LEN)
        layer = nn.TransformerEncoderLayer(
            d_model=EMBED_DIM, nhead=NUM_HEADS, dim_feedforward=FF_DIM,
            dropout=0.1, activation="gelu"
        )
        self.transformer = nn.TransformerEncoder(layer, num_layers=NUM_LAYERS)
        self.head = nn.Sequential(
            nn.LayerNorm(EMBED_DIM),
            nn.Linear(EMBED_DIM, EMBED_DIM//2),
            nn.ReLU(),
            nn.Linear(EMBED_DIM//2, ACTION_DIM)
        )
    def forward(self, x):
        # x: (B, L, F)
        x = self.input_proj(x)
        x = self.pos_enc(x)
        h = self.transformer(x.transpose(0,1)).transpose(0,1)
        pooled = h[:,-1,:]
        logits = self.head(pooled)
        prob = torch.sigmoid(logits[:,0:1])
        rate = torch.sigmoid(logits[:,1:2])
        return torch.cat([prob, rate], dim=-1)

# ============================================================
# Virtual Cell Environment
# ============================================================
class VirtualCellEnv:
    def __init__(self, seq_len=SEQ_LEN):
        self.seq_len = seq_len
        self.reset()
    def _random_obs(self):
        x = torch.rand(INPUT_FEATURES)
        x[0] = 0.5  # telomere
        return x
    def reset(self):
        self.history = [self._random_obs() for _ in range(self.seq_len)]
        self.step_count = 0
        return torch.stack(self.history, 0)
    def step(self, action):
        tel = float(self.history[-1][0])
        act, rate = float(action[0]), float(action[1])
        if tel > SAFE_TELOMERE_MAX: act, rate = 0.0, 0.0
        damage = float(self.history[-1][2])
        delta_tel = act*(rate*0.05) - 0.01*damage
        new = self.history[-1] + 0.02*torch.randn_like(self.history[-1])
        new[0] = max(0.0, min(2.5, tel + delta_tel))
        if INPUT_FEATURES > 5:
            new[5] = min(1.0, new[5] + 0.001*act*rate)
        self.history.pop(0)
        self.history.append(new)
        tel = float(new[0]); mut = float(new[5]) if INPUT_FEATURES>5 else 0.0
        reward = -abs(tel-1.0) - 5*mut - 2*damage
        self.step_count += 1
        done = self.step_count>=500
        return torch.stack(self.history,0), reward, done, {"tel":tel,"mut":mut}

# ============================================================
# Pretraining data (heuristic)
# ============================================================
def make_batch(env, batch=64):
    X, Y = [], []
    for _ in range(batch):
        obs = torch.stack([env._random_obs() for _ in range(SEQ_LEN)],0)
        tel = obs[-1][0]
        if tel<0.9:  a = torch.tensor([1.0,0.6])
        elif tel>1.1:a = torch.tensor([0.0,0.0])
        else:        a = torch.tensor([0.3,0.2])
        X.append(obs); Y.append(a)
    return torch.stack(X), torch.stack(Y)

# ============================================================
# Training
# ============================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = TelomeraseTransformerController().to(device)
opt = torch.optim.AdamW(model.parameters(), lr=LR)
env = VirtualCellEnv()

print("Pretraining (behavior cloning)...")
for epoch in trange(50):
    xb, yb = make_batch(env)
    xb, yb = xb.to(device), yb.to(device)
    pred = model(xb)
    loss = F.mse_loss(pred, yb)
    opt.zero_grad(); loss.backward(); opt.step()
print("Pretrain done.")

# RL fine-tuning
rewards_hist, tel_hist = [], []
for ep in trange(200, desc="RL"):
    obs = env.reset().unsqueeze(0).to(device)
    done, tot_r = False, 0
    tel_series = []
    while not done:
        with torch.no_grad():
            action = model(obs)
        next_obs, r, done, info = env.step(action.cpu().squeeze(0))
        obs = next_obs.unsqueeze(0).to(device)
        tot_r += r; tel_series.append(info["tel"])
        # Simple reward shaping: supervised update on-the-fly
        target = action.clone()
        target[0,0] = 1.0 if info["tel"]<0.9 else 0.0
        target[0,1] = 0.6 if info["tel"]<0.9 else 0.0
        loss = F.mse_loss(model(obs), target.to(device))
        opt.zero_grad(); loss.backward(); opt.step()
    rewards_hist.append(tot_r); tel_hist.append(tel_series)

print("Training complete.")

# ============================================================
# Visualization
# ============================================================
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(rewards_hist)
plt.title("Episode Reward")
plt.xlabel("Episode"); plt.ylabel("Total Reward")

plt.subplot(1,2,2)
for t in tel_hist[-5:]:
    plt.plot(t, alpha=0.7)
plt.axhline(1.0, ls="--", c="k", label="Target length")
plt.title("Telomere length trajectories (last 5 episodes)")
plt.xlabel("Step"); plt.ylabel("Normalized telomere length")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# ============================================================
# üß¨ Transformer-Based Telomerase Regulator + PPO/SAC (Colab)
# ============================================================

!pip install torch gymnasium stable-baselines3 matplotlib --quiet

import gymnasium as gym
import numpy as np
import torch
import torch.nn as nn
import math, random, matplotlib.pyplot as plt
from stable_baselines3 import PPO, SAC
from stable_baselines3.common.torch_layers import BaseFeaturesExtractor

# ============================================================
# Environment Definition (Gym-compatible)
# ============================================================
class VirtualCellEnv(gym.Env):
    """
    Virtual cell environment with telomerase regulation.
    Compatible with Stable-Baselines3 (continuous action space).
    """
    metadata = {"render.modes": ["human"]}
    def __init__(self, seq_len=16, n_features=12, safe_max=2.0):
        super().__init__()
        self.seq_len = seq_len
        self.n_features = n_features
        self.safe_max = safe_max
        self.action_space = gym.spaces.Box(low=0.0, high=1.0, shape=(2,), dtype=np.float32)
        self.observation_space = gym.spaces.Box(low=0.0, high=1.0, shape=(n_features,), dtype=np.float32)
        self.reset()

    def _random_obs(self):
        obs = np.random.rand(self.n_features).astype(np.float32)
        obs[0] = 0.5  # telomere
        return obs

    def reset(self, *, seed=None, options=None):
        super().reset(seed=seed)
        self.state = self._random_obs()
        self.mut = 0.0
        self.steps = 0
        return self.state, {}

    def step(self, action):
        tel, damage = self.state[0], self.state[2] if self.n_features>2 else 0.0
        act, rate = float(action[0]), float(action[1])
        if tel > self.safe_max: act, rate = 0.0, 0.0
        delta_tel = act * (rate*0.05) - 0.01*damage
        new_tel = np.clip(tel + delta_tel, 0.0, 2.5)
        self.mut = np.clip(self.mut + 0.001*act*rate, 0.0, 1.0)
        self.state = self.state + 0.02*np.random.randn(self.n_features).astype(np.float32)
        self.state[0] = new_tel
        self.state[5 if self.n_features>5 else -1] = self.mut
        reward = -abs(new_tel - 1.0) - 5*self.mut - 2*damage
        self.steps += 1
        done = self.steps >= 500
        return self.state, reward, done, False, {}

    def render(self):
        print(f"Telomere: {self.state[0]:.2f}, Mut: {self.mut:.3f}")

# ============================================================
# Custom Transformer Feature Extractor
# ============================================================
class TransformerExtractor(BaseFeaturesExtractor):
    """
    Transformer encoder as feature extractor for PPO/SAC.
    """
    def __init__(self, observation_space: gym.spaces.Box, embed_dim=64, n_heads=2, n_layers=2):
        super().__init__(observation_space, features_dim=embed_dim)
        obs_dim = observation_space.shape[0]
        self.proj = nn.Linear(obs_dim, embed_dim)
        layer = nn.TransformerEncoderLayer(d_model=embed_dim, nhead=n_heads, dim_feedforward=embed_dim*2)
        self.transformer = nn.TransformerEncoder(layer, num_layers=n_layers)
        self.pool = nn.AdaptiveAvgPool1d(1)
    def forward(self, obs: torch.Tensor) -> torch.Tensor:
        # obs shape: (batch, features)
        x = self.proj(obs).unsqueeze(1)  # (B, 1, D)
        x_t = x.transpose(0,1)           # (1, B, D)
        out = self.transformer(x_t).transpose(0,1) # (B, 1, D)
        return out.squeeze(1)            # (B, D)

# ============================================================
# Training configuration
# ============================================================
USE_PPO = True    # set False to use SAC instead
TOTAL_TIMESTEPS = 50_000

env = VirtualCellEnv()
policy_kwargs = dict(
    features_extractor_class=TransformerExtractor,
    features_extractor_kwargs=dict(embed_dim=64, n_heads=2, n_layers=2),
)

algo = PPO if USE_PPO else SAC
model = algo(
    "MlpPolicy",  # base policy (will override feature extractor)
    env,
    policy_kwargs=policy_kwargs,
    verbose=1,
    tensorboard_log="./telomerase_rl/",
    learning_rate=3e-4,
    n_steps=1024,
    batch_size=64,
)

# ============================================================
# Train the model
# ============================================================
print(f"Training {'PPO' if USE_PPO else 'SAC'} agent...")
model.learn(total_timesteps=TOTAL_TIMESTEPS)
model.save("telomerase_rl_agent")

# ============================================================
# Evaluation and Visualization
# ============================================================
rewards, tel_trajectories = [], []
obs, _ = env.reset()
for ep in range(10):
    done, total_r = False, 0
    tel_series = []
    obs, _ = env.reset()
    while not done:
        action, _ = model.predict(obs, deterministic=True)
        obs, r, done, _, _ = env.step(action)
        tel_series.append(obs[0])
        total_r += r
    rewards.append(total_r)
    tel_trajectories.append(tel_series)

print(f"Average reward (10 eval episodes): {np.mean(rewards):.2f}")

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(rewards, marker='o')
plt.title("Episode Reward (Evaluation)")
plt.xlabel("Episode"); plt.ylabel("Total Reward")

plt.subplot(1,2,2)
for t in tel_trajectories:
    plt.plot(t, alpha=0.7)
plt.axhline(1.0, ls="--", c="k", label="Target telomere length")
plt.title("Telomere Length Trajectories")
plt.xlabel("Step"); plt.ylabel("Telomere (normalized)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# ============================================================
# üß¨ Multi-Cell Organ-on-Chip Simulation + RL (Colab / Jupyter)
# Single shared policy controlling N cells simultaneously.
# ============================================================

!pip install -q torch gymnasium stable-baselines3 matplotlib numpy

import math, random, numpy as np, torch
import torch.nn as nn
import matplotlib.pyplot as plt
from stable_baselines3 import PPO
from stable_baselines3.common.torch_layers import BaseFeaturesExtractor
import gymnasium as gym

# -------------------------
# Simulation hyperparameters
# -------------------------
NUM_CELLS = 8             # number of cells in the organ-on-chip
CELL_FEATURES = 8         # per-cell observation features (telomere, ROS, damage, phag_queue, mut, ATP, signals...)
SEQ_LEN = 1               # not using time-window here; single-step observations
ACTION_PER_CELL = 2       # [activation_prob, rate]
SAFE_TELOMERE_MAX = 2.0
TOTAL_TIMESTEPS = 80_000  # adjust for longer training in Colab

# -------------------------
# Multi-cell Environment
# -------------------------
class MultiCellEnv(gym.Env):
    """
    Gym-compatible multi-cell organ-on-chip environment.
    Observation: flattened array shape = (num_cells * cell_features,)
    Action: flattened array shape = (num_cells * action_per_cell,) with values in [0,1]
    Reward: sum of per-cell rewards (can be changed to cooperative or per-cell rewards).
    """
    metadata = {"render.modes": ["human"]}
    def __init__(self, num_cells=NUM_CELLS, cell_features=CELL_FEATURES, safe_max=SAFE_TELOMERE_MAX):
        super().__init__()
        self.num_cells = num_cells
        self.cell_features = cell_features
        self.safe_max = safe_max

        obs_low = np.zeros(self.num_cells * self.cell_features, dtype=np.float32)
        obs_high = np.ones(self.num_cells * self.cell_features, dtype=np.float32)
        self.observation_space = gym.spaces.Box(low=obs_low, high=obs_high, dtype=np.float32)

        act_low = np.zeros(self.num_cells * ACTION_PER_CELL, dtype=np.float32)
        act_high = np.ones(self.num_cells * ACTION_PER_CELL, dtype=np.float32)
        self.action_space = gym.spaces.Box(low=act_low, high=act_high, dtype=np.float32)

        self.reset()

    def _init_cells(self):
        # per-cell state layout (normalized 0..1):
        # idx0: telomere (0..~2.5 normalized to 0..1 by dividing by 2.5)
        # idx1: ROS (0..1)
        # idx2: damage (0..1)
        # idx3: phag_queue (0..1)
        # idx4: mutational_burden (0..1)
        # idx5: ATP/nutrient (0..1)
        # idx6: local_signal (0..1)  (from neighbors)
        # idx7: cell_cycle_phase (0..1)
        cells = np.zeros((self.num_cells, self.cell_features), dtype=np.float32)
        for i in range(self.num_cells):
            cells[i, 0] = 1.0/2.5 * 1.0  # target ~1 -> normalized 1/2.5 ~0.4 (we will use raw normalized values later)
            cells[i, 1] = np.random.rand() * 0.2  # low ROS
            cells[i, 2] = np.random.rand() * 0.1  # low damage
            cells[i, 3] = np.random.rand() * 0.05 # phag queue small
            cells[i, 4] = np.random.rand() * 0.02 # mut small
            cells[i, 5] = 0.8 + np.random.rand()*0.2 # ATP/nutrient
            cells[i, 6] = 0.0
            cells[i, 7] = np.random.rand()
        return cells

    def reset(self, *, seed=None, options=None):
        super().reset(seed=seed)
        self.cells = self._init_cells()
        # global nutrient field (scalar) normalized 0..1
        self.nutrient = 1.0
        self.steps = 0
        return self._get_obs(), {}

    def _get_obs(self):
        # flatten cells and clip 0..1
        return np.clip(self.cells.flatten().astype(np.float32), 0.0, 1.0)

    def step(self, action):
        """
        action: flattened actions for all cells, each action pair in [0,1]
        """
        self.steps += 1
        actions = np.clip(action.reshape(self.num_cells, ACTION_PER_CELL), 0.0, 1.0)
        rewards = np.zeros(self.num_cells, dtype=np.float32)

        # Simulate nutrient diffusion / consumption: nutrient decays + replenishment
        self.nutrient = np.clip(self.nutrient * 0.995 + 0.001, 0.0, 1.0)

        # local neighbor signal = average telomere of immediate neighbors (1D ring)
        telomeres_raw = self.cells[:, 0] * 2.5  # de-normalize to raw scale [0..2.5]
        neighbor_signal = np.zeros_like(telomeres_raw)
        for i in range(self.num_cells):
            left = (i - 1) % self.num_cells
            right = (i + 1) % self.num_cells
            neighbor_signal[i] = (telomeres_raw[left] + telomeres_raw[right]) / 2.0 / 2.5  # re-normalize

        # update each cell
        for i in range(self.num_cells):
            tel_norm = float(self.cells[i,0])
            tel = tel_norm * 2.5
            ros = float(self.cells[i,1])
            damage = float(self.cells[i,2])
            phag = float(self.cells[i,3])
            mut = float(self.cells[i,4])
            atp = float(self.cells[i,5])
            local_sig = neighbor_signal[i]

            act = float(actions[i,0])
            rate = float(actions[i,1])

            # Safety: prevent activation if telomere > safe max
            if tel > self.safe_max:
                act, rate = 0.0, 0.0

            # Telomere change: activation increases telomere, damage and replication shorten it
            delta_tel = act * (rate * 0.06) * (0.5 + atp*0.5) - 0.012 * damage - 0.005 * (1.0 - atp)
            tel += delta_tel
            tel = max(0.0, min(2.5, tel))

            # ROS dynamics (activity and nutrient scarcity may increase ROS)
            ros += 0.01 * (1.0 - atp) + 0.02 * act * rate - 0.02 * (phag)
            ros = np.clip(ros, 0.0, 1.0)

            # damage increases with ROS, decreases with phagocytosis
            damage += 0.02 * ros - 0.05 * phag
            damage = np.clip(damage, 0.0, 1.0)

            # phag queue: grows with damage and decreases if cell has energy
            phag += 0.05 * damage - 0.03 * (atp)
            phag = np.clip(phag, 0.0, 1.0)

            # mutation burden increases slowly with telomerase overactivation
            mut += 0.002 * act * rate + 0.001 * ros
            mut = np.clip(mut, 0.0, 1.0)

            # ATP consumption/production: nutrient field and activity
            atp += 0.01 * self.nutrient - 0.02 * act * rate - 0.01 * damage
            atp = np.clip(atp, 0.0, 1.0)

            # cell cycle toggles randomly influenced by ATP
            cell_cycle = float(self.cells[i,7])
            if random.random() < 0.05:
                cell_cycle = random.random() * (0.5 + 0.5*atp)

            # write back (normalized tel)
            self.cells[i,0] = tel / 2.5
            self.cells[i,1] = ros
            self.cells[i,2] = damage
            self.cells[i,3] = phag
            self.cells[i,4] = mut
            self.cells[i,5] = atp
            self.cells[i,6] = local_sig
            self.cells[i,7] = cell_cycle

            # reward: keep telomere close to 1.0 and penalize mut/damage
            target_tel = 1.0
            tel_score = -abs(tel - target_tel)
            penalty = -4.0 * mut - 2.0 * damage - 0.5 * phag
            rewards[i] = tel_score + penalty

        # global cooperative reward is sum of per-cell rewards
        total_reward = float(np.sum(rewards))

        # small global events: if too many cells have high damage, reduce nutrient
        avg_damage = float(np.mean(self.cells[:,2]))
        if avg_damage > 0.3:
            self.nutrient = max(0.0, self.nutrient - 0.01)

        done = (self.steps >= 1000)
        info = {
            "per_cell_rewards": rewards,
            "avg_telomere": float(np.mean(self.cells[:,0]) * 2.5),
            "avg_mut": float(np.mean(self.cells[:,4])),
            "nutrient": float(self.nutrient),
            "avg_damage": avg_damage,
        }
        return self._get_obs(), total_reward, done, False, info

    def render(self, mode='human'):
        tel_raw = self.cells[:,0] * 2.5
        print("Telomeres:", np.round(tel_raw, 3))
        print("Mutations:", np.round(self.cells[:,4], 3))
        print("Nutrient:", self.nutrient)

# -------------------------
# Custom Features Extractor (per-cell small transformer then flatten)
# -------------------------
class PerCellTransformerExtractor(BaseFeaturesExtractor):
    """
    Expects observation shape (num_cells * cell_features,)
    Produces flattened embedding of size num_cells * embed_dim for policy net
    """
    def __init__(self, observation_space: gym.spaces.Box, embed_dim=32, n_heads=2, n_layers=2):
        obs_shape = observation_space.shape
        num_inputs = obs_shape[0]
        super().__init__(observation_space, features_dim=embed_dim * NUM_CELLS)
        assert num_inputs == NUM_CELLS * CELL_FEATURES, "Observation shape mismatch"
        self.embed_dim = embed_dim

        # per-cell projection
        self.proj = nn.Linear(CELL_FEATURES, embed_dim)
        encoder_layer = nn.TransformerEncoderLayer(d_model=embed_dim, nhead=n_heads, dim_feedforward=embed_dim*2)
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=n_layers)
        # final linear to combine per-cell embeddings if needed
        self.out_proj = nn.Linear(embed_dim * NUM_CELLS, embed_dim * NUM_CELLS)

    def forward(self, observations: torch.Tensor) -> torch.Tensor:
        # observations: (batch, num_cells * cell_features)
        B = observations.shape[0]
        x = observations.view(B, NUM_CELLS, CELL_FEATURES)           # (B, N, F)
        # project and add a dummy time dimension 1 -> transformer expects (S,B,D) where S=1
        x_proj = self.proj(x)                                        # (B, N, D)
        # transpose to (N, B, D) so transformer treats each cell as a "sequence token"
        x_t = x_proj.transpose(0,1)                                  # (N, B, D)
        out = self.transformer(x_t)                                  # (N, B, D)
        out = out.transpose(0,1).contiguous()                        # (B, N, D)
        out_flat = out.view(B, -1)                                   # (B, N*D)
        return self.out_proj(out_flat)                               # (B, features_dim)

# -------------------------
# Create env and model
# -------------------------
env = MultiCellEnv()
policy_kwargs = dict(
    features_extractor_class=PerCellTransformerExtractor,
    features_extractor_kwargs=dict(embed_dim=32, n_heads=2, n_layers=2),
)

model = PPO(
    "MlpPolicy",
    env,
    verbose=1,
    policy_kwargs=policy_kwargs,
    tensorboard_log="./multi_cell_telomerase/",
    learning_rate=3e-4,
    n_steps=2048,
    batch_size=64,
)

# -------------------------
# Train
# -------------------------
print("Starting training (this may take several minutes)...")
model.learn(total_timesteps=TOTAL_TIMESTEPS)
model.save("multi_cell_telomerase_policy")
print("Training finished & model saved.")

# -------------------------
# Evaluation
# -------------------------
def evaluate_model(model, env, episodes=10):
    avg_rewards = []
    trajectories = []
    for ep in range(episodes):
        obs, _ = env.reset()
        done = False
        tot_r = 0.0
        tel_traj = []
        while not done:
            action, _ = model.predict(obs, deterministic=True)
            obs, r, done, _, info = env.step(action)
            tot_r += r
            tel_traj.append(info["avg_telomere"])
        avg_rewards.append(tot_r)
        trajectories.append(tel_traj)
    return avg_rewards, trajectories

rewards, trajs = evaluate_model(model, env, episodes=8)
print("Avg reward (eval):", np.mean(rewards))

# Plotting results
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(rewards, marker='o')
plt.title("Total Episode Reward (Evaluation)")
plt.xlabel("Episode"); plt.ylabel("Total Reward")

plt.subplot(1,2,2)
for t in trajs:
    plt.plot(t, alpha=0.7)
plt.axhline(1.0, linestyle='--', label='Target telomere')
plt.title("Average Telomere Across Cells (per step)")
plt.xlabel("Step"); plt.ylabel("Avg telomere (raw)")
plt.legend()
plt.tight_layout()
plt.show()

# Render one simulation with prints
obs, _ = env.reset()
for _ in range(50):
    action, _ = model.predict(obs, deterministic=True)
    obs, r, done, _, info = env.step(action)
env.render()
```

```python
# ============================================================
# üß¨ Multi-Cell Organ-on-Chip Simulation + RL (Colab / Jupyter)
# Single shared policy controlling N cells simultaneously.
# ============================================================

!pip install -q torch gymnasium stable-baselines3 matplotlib numpy

import math, random, numpy as np, torch
import torch.nn as nn
import matplotlib.pyplot as plt
from stable_baselines3 import PPO
from stable_baselines3.common.torch_layers import BaseFeaturesExtractor
import gymnasium as gym

# -------------------------
# Simulation hyperparameters
# -------------------------
NUM_CELLS = 8             # number of cells in the organ-on-chip
CELL_FEATURES = 8         # per-cell observation features (telomere, ROS, damage, phag_queue, mut, ATP, signals...)
SEQ_LEN = 1               # not using time-window here; single-step observations
ACTION_PER_CELL = 2       # [activation_prob, rate]
SAFE_TELOMERE_MAX = 2.0
TOTAL_TIMESTEPS = 80_000  # adjust for longer training in Colab

# -------------------------
# Multi-cell Environment
# -------------------------
class MultiCellEnv(gym.Env):
    """
    Gym-compatible multi-cell organ-on-chip environment.
    Observation: flattened array shape = (num_cells * cell_features,)
    Action: flattened array shape = (num_cells * action_per_cell,) with values in [0,1]
    Reward: sum of per-cell rewards (can be changed to cooperative or per-cell rewards).
    """
    metadata = {"render.modes": ["human"]}
    def __init__(self, num_cells=NUM_CELLS, cell_features=CELL_FEATURES, safe_max=SAFE_TELOMERE_MAX):
        super().__init__()
        self.num_cells = num_cells
        self.cell_features = cell_features
        self.safe_max = safe_max

        obs_low = np.zeros(self.num_cells * self.cell_features, dtype=np.float32)
        obs_high = np.ones(self.num_cells * self.cell_features, dtype=np.float32)
        self.observation_space = gym.spaces.Box(low=obs_low, high=obs_high, dtype=np.float32)

        act_low = np.zeros(self.num_cells * ACTION_PER_CELL, dtype=np.float32)
        act_high = np.ones(self.num_cells * ACTION_PER_CELL, dtype=np.float32)
        self.action_space = gym.spaces.Box(low=act_low, high=act_high, dtype=np.float32)

        self.reset()

    def _init_cells(self):
        # per-cell state layout (normalized 0..1):
        # idx0: telomere (0..~2.5 normalized to 0..1 by dividing by 2.5)
        # idx1: ROS (0..1)
        # idx2: damage (0..1)
        # idx3: phag_queue (0..1)
        # idx4: mutational_burden (0..1)
        # idx5: ATP/nutrient (0..1)
        # idx6: local_signal (0..1)  (from neighbors)
        # idx7: cell_cycle_phase (0..1)
        cells = np.zeros((self.num_cells, self.cell_features), dtype=np.float32)
        for i in range(self.num_cells):
            cells[i, 0] = 1.0/2.5 * 1.0  # target ~1 -> normalized 1/2.5 ~0.4 (we will use raw normalized values later)
            cells[i, 1] = np.random.rand() * 0.2  # low ROS
            cells[i, 2] = np.random.rand() * 0.1  # low damage
            cells[i, 3] = np.random.rand() * 0.05 # phag queue small
            cells[i, 4] = np.random.rand() * 0.02 # mut small
            cells[i, 5] = 0.8 + np.random.rand()*0.2 # ATP/nutrient
            cells[i, 6] = 0.0
            cells[i, 7] = np.random.rand()
        return cells

    def reset(self, *, seed=None, options=None):
        super().reset(seed=seed)
        self.cells = self._init_cells()
        # global nutrient field (scalar) normalized 0..1
        self.nutrient = 1.0
        self.steps = 0
        return self._get_obs(), {}

    def _get_obs(self):
        # flatten cells and clip 0..1
        return np.clip(self.cells.flatten().astype(np.float32), 0.0, 1.0)

    def step(self, action):
        """
        action: flattened actions for all cells, each action pair in [0,1]
        """
        self.steps += 1
        actions = np.clip(action.reshape(self.num_cells, ACTION_PER_CELL), 0.0, 1.0)
        rewards = np.zeros(self.num_cells, dtype=np.float32)

        # Simulate nutrient diffusion / consumption: nutrient decays + replenishment
        self.nutrient = np.clip(self.nutrient * 0.995 + 0.001, 0.0, 1.0)

        # local neighbor signal = average telomere of immediate neighbors (1D ring)
        telomeres_raw = self.cells[:, 0] * 2.5  # de-normalize to raw scale [0..2.5]
        neighbor_signal = np.zeros_like(telomeres_raw)
        for i in range(self.num_cells):
            left = (i - 1) % self.num_cells
            right = (i + 1) % self.num_cells
            neighbor_signal[i] = (telomeres_raw[left] + telomeres_raw[right]) / 2.0 / 2.5  # re-normalize

        # update each cell
        for i in range(self.num_cells):
            tel_norm = float(self.cells[i,0])
            tel = tel_norm * 2.5
            ros = float(self.cells[i,1])
            damage = float(self.cells[i,2])
            phag = float(self.cells[i,3])
            mut = float(self.cells[i,4])
            atp = float(self.cells[i,5])
            local_sig = neighbor_signal[i]

            act = float(actions[i,0])
            rate = float(actions[i,1])

            # Safety: prevent activation if telomere > safe max
            if tel > self.safe_max:
                act, rate = 0.0, 0.0

            # Telomere change: activation increases telomere, damage and replication shorten it
            delta_tel = act * (rate * 0.06) * (0.5 + atp*0.5) - 0.012 * damage - 0.005 * (1.0 - atp)
            tel += delta_tel
            tel = max(0.0, min(2.5, tel))

            # ROS dynamics (activity and nutrient scarcity may increase ROS)
            ros += 0.01 * (1.0 - atp) + 0.02 * act * rate - 0.02 * (phag)
            ros = np.clip(ros, 0.0, 1.0)

            # damage increases with ROS, decreases with phagocytosis
            damage += 0.02 * ros - 0.05 * phag
            damage = np.clip(damage, 0.0, 1.0)

            # phag queue: grows with damage and decreases if cell has energy
            phag += 0.05 * damage - 0.03 * (atp)
            phag = np.clip(phag, 0.0, 1.0)

            # mutation burden increases slowly with telomerase overactivation
            mut += 0.002 * act * rate + 0.001 * ros
            mut = np.clip(mut, 0.0, 1.0)

            # ATP consumption/production: nutrient field and activity
            atp += 0.01 * self.nutrient - 0.02 * act * rate - 0.01 * damage
            atp = np.clip(atp, 0.0, 1.0)

            # cell cycle toggles randomly influenced by ATP
            cell_cycle = float(self.cells[i,7])
            if random.random() < 0.05:
                cell_cycle = random.random() * (0.5 + 0.5*atp)

            # write back (normalized tel)
            self.cells[i,0] = tel / 2.5
            self.cells[i,1] = ros
            self.cells[i,2] = damage
            self.cells[i,3] = phag
            self.cells[i,4] = mut
            self.cells[i,5] = atp
            self.cells[i,6] = local_sig
            self.cells[i,7] = cell_cycle

            # reward: keep telomere close to 1.0 and penalize mut/damage
            target_tel = 1.0
            tel_score = -abs(tel - target_tel)
            penalty = -4.0 * mut - 2.0 * damage - 0.5 * phag
            rewards[i] = tel_score + penalty

        # global cooperative reward is sum of per-cell rewards
        total_reward = float(np.sum(rewards))

        # small global events: if too many cells have high damage, reduce nutrient
        avg_damage = float(np.mean(self.cells[:,2]))
        if avg_damage > 0.3:
            self.nutrient = max(0.0, self.nutrient - 0.01)

        done = (self.steps >= 1000)
        info = {
            "per_cell_rewards": rewards,
            "avg_telomere": float(np.mean(self.cells[:,0]) * 2.5),
            "avg_mut": float(np.mean(self.cells[:,4])),
            "nutrient": float(self.nutrient),
            "avg_damage": avg_damage,
        }
        return self._get_obs(), total_reward, done, False, info

    def render(self, mode='human'):
        tel_raw = self.cells[:,0] * 2.5
        print("Telomeres:", np.round(tel_raw, 3))
        print("Mutations:", np.round(self.cells[:,4], 3))
        print("Nutrient:", self.nutrient)

# -------------------------
# Custom Features Extractor (per-cell small transformer then flatten)
# Modified to capture attention weights
# -------------------------
class PerCellTransformerExtractor(BaseFeaturesExtractor):
    """
    Expects observation shape (num_cells * cell_features,)
    Produces flattened embedding of size num_cells * embed_dim for policy net
    Includes attention weight capture.
    """
    def __init__(self, observation_space: gym.spaces.Box, embed_dim=32, n_heads=2, n_layers=2):
        obs_shape = observation_space.shape
        num_inputs = obs_shape[0]
        super().__init__(observation_space, features_dim=embed_dim * NUM_CELLS)
        assert num_inputs == NUM_CELLS * CELL_FEATURES, "Observation shape mismatch"
        self.embed_dim = embed_dim
        self.n_cells = NUM_CELLS # Store number of cells

        # per-cell projection
        self.proj = nn.Linear(CELL_FEATURES, embed_dim)
        # Transformer encoder layers that can return attention weights
        encoder_layer = nn.TransformerEncoderLayer(d_model=embed_dim, nhead=n_heads, dim_feedforward=embed_dim*2, batch_first=True)
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=n_layers)

        # final linear to combine per-cell embeddings if needed
        self.out_proj = nn.Linear(embed_dim * NUM_CELLS, embed_dim * NUM_CELLS)

        # Attribute to store the last computed attention weights
        self.last_attention_weights = None

        # Register hook to capture attention weights from the first transformer layer
        # Find the first MultiheadAttention module in the first layer
        mha_module = None
        if hasattr(self.transformer.layers[0], 'self_attn'):
            mha_module = self.transformer.layers[0].self_attn

        if mha_module is not None:
            # Patch the MHA forward to capture weights if it doesn't natively return them
            # Note: This patching might be fragile depending on PyTorch version/implementation details.
            # A more robust way is to modify the TransformerEncoderLayer or MultiheadAttention classes directly.
            # We will try a hook that expects `attn_output_weights` to be set by the internal MHA forward
            # or try to get it from the output tuple if need_weights=True is handled by the layer.
            try:
                # Try to register a hook that checks for returned attention weights
                def hook_fn(module, input, output):
                    # output is typically (output_tensor, attn_weights_tensor) when need_weights=True
                    if isinstance(output, tuple) and len(output) == 2 and isinstance(output[1], torch.Tensor):
                         self.last_attention_weights = output[1].mean(dim=0).detach() # Average over heads, detach
                    elif hasattr(module, 'attn_output_weights') and module.attn_output_weights is not None:
                        self.last_attention_weights = module.attn_output_weights.mean(dim=0).detach() # Average over heads, detach


                # Temporarily modify the layer to request weights
                original_forward = self.transformer.layers[0].forward
                def patched_layer_forward(src, src_mask=None, src_key_padding_mask=None, is_causal=False):
                     # Call original forward with need_weights=True if the method signature supports it
                     # This is tricky without inspecting signature fully. Assume standard TransformerEncoderLayer
                     return original_forward(src, src_mask=src_mask, src_key_padding_mask=src_key_padding_mask, is_causal=is_causal) # need_weights is handled internally by MHA if batch_first=True

                self.transformer.layers[0].forward = patched_layer_forward

                # Now register the hook on the MHA sub-module
                self._hook_handle = mha_module.register_forward_hook(hook_fn)
                print("Registered attention capture hook.")

            except Exception as e:
                print(f"Warning: Could not register attention capture hook: {e}")
                self._hook_handle = None
        else:
             print("Warning: Could not find MultiheadAttention module for hook.")
             self._hook_handle = None


    def forward(self, observations: torch.Tensor) -> torch.Tensor:
        # observations: (batch, num_cells * cell_features)
        B = observations.shape[0]
        x = observations.view(B, NUM_CELLS, CELL_FEATURES)           # (B, N, F)
        # project
        x_proj = self.proj(x)                                        # (B, N, D)

        # Transformer expects shape (B, L, D) with batch_first=True
        out = self.transformer(x_proj)                               # (B, N, D)

        out_flat = out.view(B, -1)                                   # (B, N*D)
        return self.out_proj(out_flat)                               # (B, features_dim)

    def get_attention(self):
        """Retrieve latest attention weights (N_cells x N_cells, averaged over heads)"""
        return self.last_attention_weights

# -------------------------
# Create env and model
# -------------------------
env = MultiCellEnv()
policy_kwargs = dict(
    features_extractor_class=PerCellTransformerExtractor,
    features_extractor_kwargs=dict(embed_dim=32, n_heads=2, n_layers=2),
)

model = PPO(
    "MlpPolicy",
    env,
    verbose=1,
    policy_kwargs=policy_kwargs,
    tensorboard_log="./multi_cell_telomerase/",
    learning_rate=3e-4,
    n_steps=2048,
    batch_size=64,
)

# -------------------------
# Train
# -------------------------
print("Starting training (this may take several minutes)...")
model.learn(total_timesteps=TOTAL_TIMESTEPS)
model.save("multi_cell_telomerase_policy")
print("Training finished & model saved.")

# -------------------------
# Evaluation
# -------------------------
def evaluate_model(model, env, episodes=10):
    avg_rewards = []
    trajectories = []
    for ep in range(episodes):
        obs, _ = env.reset()
        done = False
        tot_r = 0.0
        tel_traj = []
        while not done:
            action, _ = model.predict(obs, deterministic=True)
            obs, r, done, _, info = env.step(action)
            tot_r += r
            tel_traj.append(info["avg_telomere"])
        avg_rewards.append(tot_r)
        trajectories.append(tel_traj)
    return avg_rewards, trajectories

rewards, trajs = evaluate_model(model, env, episodes=8)
print("Avg reward (eval):", np.mean(rewards))

# Plotting results
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(rewards, marker='o')
plt.title("Total Episode Reward (Evaluation)")
plt.xlabel("Episode"); plt.ylabel("Total Reward")

plt.subplot(1,2,2)
for t in trajs:
    plt.plot(t, alpha=0.7)
plt.axhline(1.0, linestyle='--', label='Target telomere')
plt.title("Average Telomere Across Cells (per step)")
plt.xlabel("Step"); plt.ylabel("Avg telomere (raw)")
plt.legend()
plt.tight_layout()
plt.show()

# Render one simulation with prints
obs, _ = env.reset()
for _ in range(50):
    action, _ = model.predict(obs, deterministic=True)
    obs, r, done, _, info = env.step(action)
env.render()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class AttentionReportingTransformerExtractor(nn.Module):
    """
    Feature extractor for per-cell embeddings with attention weight reporting.
    Returns both features and attention matrices for visualization.
    """

    def __init__(self, observation_space, embedding_dim=64, n_cells=8, n_heads=4, n_layers=2):
        super().__init__()
        self.obs_dim = observation_space.shape[0]
        self.n_cells = n_cells
        self.embedding_dim = embedding_dim

        # Each "cell" gets its own embedding vector
        self.embed = nn.Linear(self.obs_dim, embedding_dim)

        # Transformer Encoder Layers that report attention weights
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=embedding_dim,
            nhead=n_heads,
            dim_feedforward=embedding_dim * 2,
            batch_first=True,
        )
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=n_layers)

        # Store last attention weights for inspection
        self.last_attention = None

        # Patch the forward of each MHA submodule to capture attention weights
        def mha_forward_with_return(self_mha, query, key, value, **kwargs):
            out, attn_weights = self_mha.original_forward(query, key, value, need_weights=True)
            self_mha.attn_output_weights = attn_weights.detach()  # save for external use
            return out, attn_weights

        # Patch each MultiheadAttention to record weights
        for layer in self.transformer.layers:
            mha = layer.self_attn
            mha.original_forward = mha.forward
            mha.forward = lambda q, k, v, **kw: mha_forward_with_return(mha, q, k, v, **kw)

        self.output_proj = nn.Linear(embedding_dim, embedding_dim)

    def forward(self, obs, return_attention=False):
        """
        obs: (batch, obs_dim)
        Returns: features, attention_weights (if requested)
        """
        x = self.embed(obs).unsqueeze(1).repeat(1, self.n_cells, 1)  # simulate per-cell tokens
        x = self.transformer(x)

        # Extract attention weights from the last transformer layer
        last_mha = self.transformer.layers[-1].self_attn
        attn_weights = getattr(last_mha, "attn_output_weights", None)
        if attn_weights is not None:
            attn_weights = attn_weights.mean(0).detach()  # average across heads
            self.last_attention = attn_weights  # store for external access

        features = self.output_proj(x.mean(dim=1))

        if return_attention:
            return features, attn_weights
        return features
```

```python
from stable_baselines3 import PPO
from stable_baselines3.common.torch_layers import BaseFeaturesExtractor
from gymnasium import spaces

class CustomExtractor(BaseFeaturesExtractor):
    def __init__(self, observation_space: spaces.Box, features_dim: int = 64):
        super().__init__(observation_space, features_dim)
        self.extractor = AttentionReportingTransformerExtractor(observation_space)
        self._features_dim = features_dim

    def forward(self, observations):
        feats = self.extractor(observations)
        return feats

    def get_attention(self):
        """Retrieve latest attention weights (N x N)"""
        return getattr(self.extractor, "last_attention", None)
```

```python
# ============================================================
# üß¨ Multi-Cell Organ-on-Chip Simulation + RL (Colab / Jupyter)
# Single shared policy controlling N cells simultaneously.
# ============================================================

!pip install -q torch gymnasium stable-baselines3 matplotlib numpy

import math, random, numpy as np, torch
import torch.nn as nn
import matplotlib.pyplot as plt
from stable_baselines3 import PPO
from stable_baselines3.common.torch_layers import BaseFeaturesExtractor
import gymnasium as gym

# -------------------------
# Simulation hyperparameters
# -------------------------
NUM_CELLS = 8             # number of cells in the organ-on-chip
CELL_FEATURES = 8         # per-cell observation features (telomere, ROS, damage, phag_queue, mut, ATP, signals...)
SEQ_LEN = 1               # not using time-window here; single-step observations
ACTION_PER_CELL = 2       # [activation_prob, rate]
SAFE_TELOMERE_MAX = 2.0
TOTAL_TIMESTEPS = 80_000  # adjust for longer training in Colab

# -------------------------
# Multi-cell Environment
# -------------------------
class MultiCellEnv(gym.Env):
    """
    Gym-compatible multi-cell organ-on-chip environment.
    Observation: flattened array shape = (num_cells * cell_features,)
    Action: flattened array shape = (num_cells * action_per_cell,) with values in [0,1]
    Reward: sum of per-cell rewards (can be changed to cooperative or per-cell rewards).
    """
    metadata = {"render.modes": ["human"]}
    def __init__(self, num_cells=NUM_CELLS, cell_features=CELL_FEATURES, safe_max=SAFE_TELOMERE_MAX):
        super().__init__()
        self.num_cells = num_cells
        self.cell_features = cell_features
        self.safe_max = safe_max

        obs_low = np.zeros(self.num_cells * self.cell_features, dtype=np.float32)
        obs_high = np.ones(self.num_cells * self.cell_features, dtype=np.float32)
        self.observation_space = gym.spaces.Box(low=obs_low, high=obs_high, dtype=np.float32)

        act_low = np.zeros(self.num_cells * ACTION_PER_CELL, dtype=np.float32)
        act_high = np.ones(self.num_cells * ACTION_PER_CELL, dtype=np.float32)
        self.action_space = gym.spaces.Box(low=act_low, high=act_high, dtype=np.float32)

        self.reset()

    def _init_cells(self):
        # per-cell state layout (normalized 0..1):
        # idx0: telomere (0..~2.5 normalized to 0..1 by dividing by 2.5)
        # idx1: ROS (0..1)
        # idx2: damage (0..1)
        # idx3: phag_queue (0..1)
        # idx4: mutational_burden (0..1)
        # idx5: ATP/nutrient (0..1)
        # idx6: local_signal (0..1)  (from neighbors)
        # idx7: cell_cycle_phase (0..1)
        cells = np.zeros((self.num_cells, self.cell_features), dtype=np.float32)
        for i in range(self.num_cells):
            cells[i, 0] = 1.0/2.5 * 1.0  # target ~1 -> normalized 1/2.5 ~0.4 (we will use raw normalized values later)
            cells[i, 1] = np.random.rand() * 0.2  # low ROS
            cells[i, 2] = np.random.rand() * 0.1  # low damage
            cells[i, 3] = np.random.rand() * 0.05 # phag queue small
            cells[i, 4] = np.random.rand() * 0.02 # mut small
            cells[i, 5] = 0.8 + np.random.rand()*0.2 # ATP/nutrient
            cells[i, 6] = 0.0
            cells[i, 7] = np.random.rand()
        return cells

    def reset(self, *, seed=None, options=None):
        super().reset(seed=seed)
        self.cells = self._init_cells()
        # global nutrient field (scalar) normalized 0..1
        self.nutrient = 1.0
        self.steps = 0
        return self._get_obs(), {}

    def _get_obs(self):
        # flatten cells and clip 0..1
        return np.clip(self.cells.flatten().astype(np.float32), 0.0, 1.0)

    def step(self, action):
        """
        action: flattened actions for all cells, each action pair in [0,1]
        """
        self.steps += 1
        actions = np.clip(action.reshape(self.num_cells, ACTION_PER_CELL), 0.0, 1.0)
        rewards = np.zeros(self.num_cells, dtype=np.float32)

        # Simulate nutrient diffusion / consumption: nutrient decays + replenishment
        self.nutrient = np.clip(self.nutrient * 0.995 + 0.001, 0.0, 1.0)

        # local neighbor signal = average telomere of immediate neighbors (1D ring)
        telomeres_raw = self.cells[:, 0] * 2.5  # de-normalize to raw scale [0..2.5]
        neighbor_signal = np.zeros_like(telomeres_raw)
        for i in range(self.num_cells):
            left = (i - 1) % self.num_cells
            right = (i + 1) % self.num_cells
            neighbor_signal[i] = (telomeres_raw[left] + telomeres_raw[right]) / 2.0 / 2.5  # re-normalize

        # update each cell
        for i in range(self.num_cells):
            tel_norm = float(self.cells[i,0])
            tel = tel_norm * 2.5
            ros = float(self.cells[i,1])
            damage = float(self.cells[i,2])
            phag = float(self.cells[i,3])
            mut = float(self.cells[i,4])
            atp = float(self.cells[i,5])
            local_sig = neighbor_signal[i]

            act = float(actions[i,0])
            rate = float(actions[i,1])

            # Safety: prevent activation if telomere > safe max
            if tel > self.safe_max:
                act, rate = 0.0, 0.0

            # Telomere change: activation increases telomere, damage and replication shorten it
            delta_tel = act * (rate * 0.06) * (0.5 + atp*0.5) - 0.012 * damage - 0.005 * (1.0 - atp)
            tel += delta_tel
            tel = max(0.0, min(2.5, tel))

            # ROS dynamics (activity and nutrient scarcity may increase ROS)
            ros += 0.01 * (1.0 - atp) + 0.02 * act * rate - 0.02 * (phag)
            ros = np.clip(ros, 0.0, 1.0)

            # damage increases with ROS, decreases with phagocytosis
            damage += 0.02 * ros - 0.05 * phag
            damage = np.clip(damage, 0.0, 1.0)

            # phag queue: grows with damage and decreases if cell has energy
            phag += 0.05 * damage - 0.03 * (atp)
            phag = np.clip(phag, 0.0, 1.0)

            # mutation burden increases slowly with telomerase overactivation
            mut += 0.002 * act * rate + 0.001 * ros
            mut = np.clip(mut, 0.0, 1.0)

            # ATP consumption/production: nutrient field and activity
            atp += 0.01 * self.nutrient - 0.02 * act * rate - 0.01 * damage
            atp = np.clip(atp, 0.0, 1.0)

            # cell cycle toggles randomly influenced by ATP
            cell_cycle = float(self.cells[i,7])
            if random.random() < 0.05:
                cell_cycle = random.random() * (0.5 + 0.5*atp)

            # write back (normalized tel)
            self.cells[i,0] = tel / 2.5
            self.cells[i,1] = ros
            self.cells[i,2] = damage
            self.cells[i,3] = phag
            self.cells[i,4] = mut
            self.cells[i,5] = atp
            self.cells[i,6] = local_sig
            self.cells[i,7] = cell_cycle

            # reward: keep telomere close to 1.0 and penalize mut/damage
            target_tel = 1.0
            tel_score = -abs(tel - target_tel)
            penalty = -4.0 * mut - 2.0 * damage - 0.5 * phag
            rewards[i] = tel_score + penalty

        # global cooperative reward is sum of per-cell rewards
        total_reward = float(np.sum(rewards))

        # small global events: if too many cells have high damage, reduce nutrient
        avg_damage = float(np.mean(self.cells[:,2]))
        if avg_damage > 0.3:
            self.nutrient = max(0.0, self.nutrient - 0.01)

        done = (self.steps >= 1000)
        info = {
            "per_cell_rewards": rewards,
            "avg_telomere": float(np.mean(self.cells[:,0]) * 2.5),
            "avg_mut": float(np.mean(self.cells[:,4])),
            "nutrient": float(self.nutrient),
            "avg_damage": avg_damage,
        }
        return self._get_obs(), total_reward, done, False, info

    def render(self, mode='human'):
        tel_raw = self.cells[:,0] * 2.5
        print("Telomeres:", np.round(tel_raw, 3))
        print("Mutations:", np.round(self.cells[:,4], 3))
        print("Nutrient:", self.nutrient)

# -------------------------
# Custom Features Extractor (per-cell small transformer then flatten)
# Modified to capture attention weights
# -------------------------
class PerCellTransformerExtractor(BaseFeaturesExtractor):
    """
    Expects observation shape (num_cells * cell_features,)
    Produces flattened embedding of size num_cells * embed_dim for policy net
    Includes attention weight capture.
    """
    def __init__(self, observation_space: gym.spaces.Box, embed_dim=32, n_heads=2, n_layers=2):
        obs_shape = observation_space.shape
        num_inputs = obs_shape[0]
        super().__init__(observation_space, features_dim=embed_dim * NUM_CELLS)
        assert num_inputs == NUM_CELLS * CELL_FEATURES, "Observation shape mismatch"
        self.embed_dim = embed_dim
        self.n_cells = NUM_CELLS # Store number of cells

        # per-cell projection
        self.proj = nn.Linear(CELL_FEATURES, embed_dim)
        # Transformer encoder layers that can return attention weights
        encoder_layer = nn.TransformerEncoderLayer(d_model=embed_dim, nhead=n_heads, dim_feedforward=embed_dim*2, batch_first=True)
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=n_layers)

        # final linear to combine per-cell embeddings if needed
        self.out_proj = nn.Linear(embed_dim * NUM_CELLS, embed_dim * NUM_CELLS)

        # Attribute to store the last computed attention weights
        self.last_attention_weights = None

        # Register hook to capture attention weights from the first transformer layer
        # Find the first MultiheadAttention module in the first layer
        mha_module = None
        if hasattr(self.transformer.layers[0], 'self_attn'):
            mha_module = self.transformer.layers[0].self_attn

        if mha_module is not None:
            # Patch the MHA forward to capture weights if it doesn't natively return them
            # Note: This patching might be fragile depending on PyTorch version/implementation details.
            # A more robust way is to modify the TransformerEncoderLayer or MultiheadAttention classes directly.
            # We will try a hook that expects `attn_output_weights` to be set by the internal MHA forward
            # or try to get it from the output tuple if need_weights=True is handled by the layer.
            try:
                # Try to register a hook that checks for returned attention weights
                def hook_fn(module, input, output):
                    # output is typically (output_tensor, attn_weights_tensor) when need_weights=True
                    if isinstance(output, tuple) and len(output) == 2 and isinstance(output[1], torch.Tensor):
                         self.last_attention_weights = output[1].mean(dim=0).detach() # Average over heads, detach
                    elif hasattr(module, 'attn_output_weights') and module.attn_output_weights is not None:
                        self.last_attention_weights = module.attn_output_weights.mean(dim=0).detach() # Average over heads, detach


                # Temporarily modify the layer to request weights
                original_forward = self.transformer.layers[0].forward
                def patched_layer_forward(src, src_mask=None, src_key_padding_mask=None, is_causal=False):
                     # Call original forward with need_weights=True if the method signature supports it
                     # This is tricky without inspecting signature fully. Assume standard TransformerEncoderLayer
                     return original_forward(src, src_mask=src_mask, src_key_padding_mask=src_key_padding_mask, is_causal=is_causal) # need_weights is handled internally by MHA if batch_first=True

                self.transformer.layers[0].forward = patched_layer_forward

                # Now register the hook on the MHA sub-module
                self._hook_handle = mha_module.register_forward_hook(hook_fn)
                print("Registered attention capture hook.")

            except Exception as e:
                print(f"Warning: Could not register attention capture hook: {e}")
                self._hook_handle = None
        else:
             print("Warning: Could not find MultiheadAttention module for hook.")
             self._hook_handle = None


    def forward(self, observations: torch.Tensor) -> torch.Tensor:
        # observations: (batch, num_cells * cell_features)
        B = observations.shape[0]
        x = observations.view(B, NUM_CELLS, CELL_FEATURES)           # (B, N, F)
        # project
        x_proj = self.proj(x)                                        # (B, N, D)

        # Transformer expects shape (B, L, D) with batch_first=True
        out = self.transformer(x_proj)                               # (B, N, D)

        out_flat = out.view(B, -1)                                   # (B, N*D)
        return self.out_proj(out_flat)                               # (B, features_dim)

    def get_attention(self):
        """Retrieve latest attention weights (N_cells x N_cells, averaged over heads)"""
        return self.last_attention_weights

# -------------------------
# Create env and model
# -------------------------
env = MultiCellEnv()
policy_kwargs = dict(
    features_extractor_class=PerCellTransformerExtractor,
    features_extractor_kwargs=dict(embed_dim=32, n_heads=2, n_layers=2),
)

model = PPO(
    "MlpPolicy",
    env,
    verbose=1,
    policy_kwargs=policy_kwargs,
    tensorboard_log="./multi_cell_telomerase/",
    learning_rate=3e-4,
    n_steps=2048,
    batch_size=64,
)

# -------------------------
# Train
# -------------------------
print("Starting training (this may take several minutes)...")
model.learn(total_timesteps=TOTAL_TIMESTEPS)
model.save("multi_cell_telomerase_policy")
print("Training finished & model saved.")

# -------------------------
# Evaluation
# -------------------------
def evaluate_model(model, env, episodes=10):
    avg_rewards = []
    trajectories = []
    for ep in range(episodes):
        obs, _ = env.reset()
        done = False
        tot_r = 0.0
        tel_traj = []
        while not done:
            action, _ = model.predict(obs, deterministic=True)
            obs, r, done, _, info = env.step(action)
            tot_r += r
            tel_traj.append(info["avg_telomere"])
        avg_rewards.append(tot_r)
        trajectories.append(tel_traj)
    return avg_rewards, trajectories

rewards, trajs = evaluate_model(model, env, episodes=8)
print("Avg reward (eval):", np.mean(rewards))

# Plotting results
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(rewards, marker='o')
plt.title("Total Episode Reward (Evaluation)")
plt.xlabel("Episode"); plt.ylabel("Total Reward")

plt.subplot(1,2,2)
for t in trajs:
    plt.plot(t, alpha=0.7)
plt.axhline(1.0, linestyle='--', label='Target telomere')
plt.title("Average Telomere Across Cells (per step)")
plt.xlabel("Step"); plt.ylabel("Avg telomere (raw)")
plt.legend()
plt.tight_layout()
plt.show()

# Render one simulation with prints
obs, _ = env.reset()
for _ in range(50):
    action, _ = model.predict(obs, deterministic=True)
    obs, r, done, _, info = env.step(action)
env.render()
```

```python
# ============================================================
# üß¨ Multi-Cell Organ-on-Chip Simulation + RL (Colab / Jupyter)
# Single shared policy controlling N cells simultaneously.
# ============================================================

!pip install -q torch gymnasium stable_baselines3 matplotlib numpy

import math, random, numpy as np, torch
import torch.nn as nn
import matplotlib.pyplot as plt
from stable_baselines3 import PPO
from stable_baselines3.common.torch_layers import BaseFeaturesExtractor
import gymnasium as gym

# -------------------------
# Simulation hyperparameters
# -------------------------
NUM_CELLS = 8             # number of cells in the organ-on-chip
CELL_FEATURES = 8         # per-cell observation features (telomere, ROS, damage, phag_queue, mut, ATP, signals...)
SEQ_LEN = 1               # not using time-window here; single-step observations
ACTION_PER_CELL = 2       # [activation_prob, rate]
SAFE_TELOMERE_MAX = 2.0
TOTAL_TIMESTEPS = 80_000  # adjust for longer training in Colab

# -------------------------
# Multi-cell Environment
# -------------------------
class MultiCellEnv(gym.Env):
    """
    Gym-compatible multi-cell organ-on-chip environment.
    Observation: flattened array shape = (num_cells * cell_features,)
    Action: flattened array shape = (num_cells * action_per_cell,) with values in [0,1]
    Reward: sum of per-cell rewards (can be changed to cooperative or per-cell rewards).
    """
    metadata = {"render.modes": ["human"]}
    def __init__(self, num_cells=NUM_CELLS, cell_features=CELL_FEATURES, safe_max=SAFE_TELOMERE_MAX):
        super().__init__()
        self.num_cells = num_cells
        self.cell_features = cell_features
        self.safe_max = safe_max

        obs_low = np.zeros(self.num_cells * self.cell_features, dtype=np.float32)
        obs_high = np.ones(self.num_cells * self.cell_features, dtype=np.float32)
        self.observation_space = gym.spaces.Box(low=obs_low, high=obs_high, dtype=np.float32)

        act_low = np.zeros(self.num_cells * ACTION_PER_CELL, dtype=np.float32)
        act_high = np.ones(self.num_cells * ACTION_PER_CELL, dtype=np.float32)
        self.action_space = gym.spaces.Box(low=act_low, high=act_high, dtype=np.float32)

        self.reset()

    def _init_cells(self):
        # per-cell state layout (normalized 0..1):
        # idx0: telomere (0..~2.5 normalized to 0..1 by dividing by 2.5)
        # idx1: ROS (0..1)
        # idx2: damage (0..1)
        # idx3: phag_queue (0..1)
        # idx4: mutational_burden (0..1)
        # idx5: ATP/nutrient (0..1)
        # idx6: local_signal (0..1)  (from neighbors)
        # idx7: cell_cycle_phase (0..1)
        cells = np.zeros((self.num_cells, self.cell_features), dtype=np.float32)
        for i in range(self.num_cells):
            cells[i, 0] = 1.0/2.5 * 1.0  # target ~1 -> normalized 1/2.5 ~0.4 (we will use raw normalized values later)
            cells[i, 1] = np.random.rand() * 0.2  # low ROS
            cells[i, 2] = np.random.rand() * 0.1  # low damage
            cells[i, 3] = np.random.rand() * 0.05 # phag queue small
            cells[i, 4] = np.random.rand() * 0.02 # mut small
            cells[i, 5] = 0.8 + np.random.rand()*0.2 # ATP/nutrient
            cells[i, 6] = 0.0
            cells[i, 7] = np.random.rand()
        return cells

    def reset(self, *, seed=None, options=None):
        super().reset(seed=seed)
        self.cells = self._init_cells()
        # global nutrient field (scalar) normalized 0..1
        self.nutrient = 1.0
        self.steps = 0
        return self._get_obs(), {}

    def _get_obs(self):
        # flatten cells and clip 0..1
        return np.clip(self.cells.flatten().astype(np.float32), 0.0, 1.0)

    def step(self, action):
        """
        action: flattened actions for all cells, each action pair in [0,1]
        """
        self.steps += 1
        actions = np.clip(action.reshape(self.num_cells, ACTION_PER_CELL), 0.0, 1.0)
        rewards = np.zeros(self.num_cells, dtype=np.float32)

        # Simulate nutrient diffusion / consumption: nutrient decays + replenishment
        self.nutrient = np.clip(self.nutrient * 0.995 + 0.001, 0.0, 1.0)

        # local neighbor signal = average telomere of immediate neighbors (1D ring)
        telomeres_raw = self.cells[:, 0] * 2.5  # de-normalize to raw scale [0..2.5]
        neighbor_signal = np.zeros_like(telomeres_raw)
        for i in range(self.num_cells):
            left = (i - 1) % self.num_cells
            right = (i + 1) % self.num_cells
            neighbor_signal[i] = (telomeres_raw[left] + telomeres_raw[right]) / 2.0 / 2.5  # re-normalize

        # update each cell
        for i in range(self.num_cells):
            tel_norm = float(self.cells[i,0])
            tel = tel_norm * 2.5
            ros = float(self.cells[i,1])
            damage = float(self.cells[i,2])
            phag = float(self.cells[i,3])
            mut = float(self.cells[i,4])
            atp = float(self.cells[i,5])
            local_sig = neighbor_signal[i]

            act = float(actions[i,0])
            rate = float(actions[i,1])

            # Safety: prevent activation if telomere > safe max
            if tel > self.safe_max:
                act, rate = 0.0, 0.0

            # Telomere change: activation increases telomere, damage and replication shorten it
            delta_tel = act * (rate * 0.06) * (0.5 + atp*0.5) - 0.012 * damage - 0.005 * (1.0 - atp)
            tel += delta_tel
            tel = max(0.0, min(2.5, tel))

            # ROS dynamics (activity and nutrient scarcity may increase ROS)
            ros += 0.01 * (1.0 - atp) + 0.02 * act * rate - 0.02 * (phag)
            ros = np.clip(ros, 0.0, 1.0)

            # damage increases with ROS, decreases with phagocytosis
            damage += 0.02 * ros - 0.05 * phag
            damage = np.clip(damage, 0.0, 1.0)

            # phag queue: grows with damage and decreases if cell has energy
            phag += 0.05 * damage - 0.03 * (atp)
            phag = np.clip(phag, 0.0, 1.0)

            # mutation burden increases slowly with telomerase overactivation
            mut += 0.002 * act * rate + 0.001 * ros
            mut = np.clip(mut, 0.0, 1.0)

            # ATP consumption/production: nutrient field and activity
            atp += 0.01 * self.nutrient - 0.02 * act * rate - 0.01 * damage
            atp = np.clip(atp, 0.0, 1.0)

            # cell cycle toggles randomly influenced by ATP
            cell_cycle = float(self.cells[i,7])
            if random.random() < 0.05:
                cell_cycle = random.random() * (0.5 + 0.5*atp)

            # write back (normalized tel)
            self.cells[i,0] = tel / 2.5
            self.cells[i,1] = ros
            self.cells[i,2] = damage
            self.cells[i,3] = phag
            self.cells[i,4] = mut
            self.cells[i,5] = atp
            self.cells[i,6] = local_sig
            self.cells[i,7] = cell_cycle

            # reward: keep telomere close to 1.0 and penalize mut/damage
            target_tel = 1.0
            tel_score = -abs(tel - target_tel)
            penalty = -4.0 * mut - 2.0 * damage - 0.5 * phag
            rewards[i] = tel_score + penalty

        # global cooperative reward is sum of per-cell rewards
        total_reward = float(np.sum(rewards))

        # small global events: if too many cells have high damage, reduce nutrient
        avg_damage = float(np.mean(self.cells[:,2]))
        if avg_damage > 0.3:
            self.nutrient = max(0.0, self.nutrient - 0.01)

        done = (self.steps >= 1000)
        info = {
            "per_cell_rewards": rewards,
            "avg_telomere": float(np.mean(self.cells[:,0]) * 2.5),
            "avg_mut": float(np.mean(self.cells[:,4])),
            "nutrient": float(self.nutrient),
            "avg_damage": avg_damage,
        }
        return self._get_obs(), total_reward, done, False, info

    def render(self, mode='human'):
        tel_raw = self.cells[:,0] * 2.5
        print("Telomeres:", np.round(tel_raw, 3))
        print("Mutations:", np.round(self.cells[:,4], 3))
        print("Nutrient:", self.nutrient)

# -------------------------
# Custom Features Extractor (per-cell small transformer then flatten)
# Modified to capture attention weights
# -------------------------
class PerCellTransformerExtractor(BaseFeaturesExtractor):
    """
    Expects observation shape (num_cells * cell_features,)
    Produces flattened embedding of size num_cells * embed_dim for policy net
    Includes attention weight capture.
    """
    def __init__(self, observation_space: gym.spaces.Box, embed_dim=32, n_heads=2, n_layers=2):
        obs_shape = observation_space.shape
        num_inputs = obs_shape[0]
        super().__init__(observation_space, features_dim=embed_dim * NUM_CELLS)
        assert num_inputs == NUM_CELLS * CELL_FEATURES, "Observation shape mismatch"
        self.embed_dim = embed_dim
        self.n_cells = NUM_CELLS # Store number of cells

        # per-cell projection
        self.proj = nn.Linear(CELL_FEATURES, embed_dim)
        # Transformer encoder layers. Need to access attention.
        # Modify the layer to return attention or use a hook if needed.
        # A common way is to use batch_first=True and need_weights=True in MultiheadAttention
        # However, TransformerEncoderLayer doesn't directly expose need_weights.
        # We'll try to modify the forward pass to get attention.

        self.transformer_layers = nn.ModuleList([
            nn.TransformerEncoderLayer(d_model=embed_dim, nhead=n_heads, dim_feedforward=embed_dim*2, batch_first=True)
            for _ in range(n_layers)
        ])

        # final linear to combine per-cell embeddings if needed
        self.out_proj = nn.Linear(embed_dim * NUM_CELLS, embed_dim * NUM_CELLS)

        # Attribute to store the last computed attention weights
        self.last_attention_weights = None


    def forward(self, observations: torch.Tensor) -> torch.Tensor:
        # observations: (batch, num_cells * cell_features)
        B = observations.shape[0]
        x = observations.view(B, NUM_CELLS, CELL_FEATURES)           # (B, N, F)
        # project
        x_proj = self.proj(x)                                        # (B, N, D)

        # Pass through transformer layers and capture attention from the last layer
        out = x_proj
        for i, layer in enumerate(self.transformer_layers):
            # Call the layer's forward, attempting to get attention weights
            # This depends on the exact implementation of TransformerEncoderLayer/MultiheadAttention
            # If batch_first=True, MultiheadAttention's forward signature is usually (query, key, value, key_padding_mask, need_weights, ...)
            # TransformerEncoderLayer's forward doesn't directly expose need_weights,
            # but some versions/usages might pass it through or have internal flags.

            # Attempting to get attention from the MultiheadAttention submodule
            attn_output, attn_weights = layer.self_attn(
                 out, out, out, need_weights=True, average_attn_weights=True # average_attn_weights=True returns (B, N, N)
             )
            out = out + layer.dropout1(attn_output)
            out = layer.norm1(out)
            out = out + layer.dropout2(layer.linear2(layer.dropout(layer.activation(layer.linear1(out)))))
            out = layer.norm2(out)

            if i == len(self.transformer_layers) - 1: # Capture attention from the last layer
                self.last_attention_weights = attn_weights.detach() # Shape (B, N, N) if average_attn_weights=True, or (B, num_heads, N, N)

        # Average attention weights over the batch if batch_first was True
        if self.last_attention_weights is not None and self.last_attention_weights.ndim == 3: # Assuming (B, N, N)
             self.last_attention_weights = self.last_attention_weights.mean(dim=0) # Average over batch: (N, N)


        out_flat = out.view(B, -1)                                   # (B, N*D)
        return self.out_proj(out_flat)                               # (B, features_dim)

    def get_attention(self):
        """Retrieve latest attention weights (N_cells x N_cells, averaged over heads and batch)"""
        return self.last_attention_weights

# -------------------------
# Create env and model
# -------------------------
env = MultiCellEnv()
policy_kwargs = dict(
    features_extractor_class=PerCellTransformerExtractor,
    features_extractor_kwargs=dict(embed_dim=32, n_heads=2, n_layers=2),
)

model = PPO(
    "MlpPolicy",
    env,
    verbose=1,
    policy_kwargs=policy_kwargs,
    tensorboard_log="./multi_cell_telomerase/",
    learning_rate=3e-4,
    n_steps=2048,
    batch_size=64,
)

# -------------------------
# Train
# -------------------------
print("Starting training (this may take several minutes)...")
model.learn(total_timesteps=TOTAL_TIMESTEPS)
model.save("multi_cell_telomerase_policy")
print("Training finished & model saved.")

# -------------------------
# Evaluation
# -------------------------
def evaluate_model(model, env, episodes=10):
    avg_rewards = []
    trajectories = []
    for ep in range(episodes):
        obs, _ = env.reset()
        done = False
        tot_r = 0.0
        tel_traj = []
        while not done:
            action, _ = model.predict(obs, deterministic=True)
            obs, r, done, _, info = env.step(action)
            tot_r += r
            tel_traj.append(info["avg_telomere"])
        avg_rewards.append(tot_r)
        trajectories.append(tel_traj)
    return avg_rewards, trajectories

rewards, trajs = evaluate_model(model, env, episodes=8)
print("Avg reward (eval):", np.mean(rewards))

# Plotting results
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(rewards, marker='o')
plt.title("Total Episode Reward (Evaluation)")
plt.xlabel("Episode"); plt.ylabel("Total Reward")

plt.subplot(1,2,2)
for t in trajs:
    plt.plot(t, alpha=0.7)
plt.axhline(1.0, linestyle='--', label='Target telomere')
plt.title("Average Telomere Across Cells (per step)")
plt.xlabel("Step"); plt.ylabel("Avg telomere (raw)")
plt.legend()
plt.tight_layout()
plt.show()

# Render one simulation with prints
obs, _ = env.reset()
for _ in range(50):
    action, _ = model.predict(obs, deterministic=True)
    obs, r, done, _, info = env.step(action)
env.render()
```

```python
# ============================================================
# üß¨ Multi-Cell Organ-on-Chip Simulation + RL (Colab / Jupyter)
# Single shared policy controlling N cells simultaneously.
# ============================================================

!pip install -q torch gymnasium stable_baselines3 matplotlib numpy

import math, random, numpy as np, torch
import torch.nn as nn
import matplotlib.pyplot as plt
from stable_baselines3 import PPO
from stable_baselines3.common.torch_layers import BaseFeaturesExtractor
import gymnasium as gym

# -------------------------
# Simulation hyperparameters
# -------------------------
NUM_CELLS = 8             # number of cells in the organ-on-chip
CELL_FEATURES = 8         # per-cell observation features (telomere, ROS, damage, phag_queue, mut, ATP, signals...)
SEQ_LEN = 1               # not using time-window here; single-step observations
ACTION_PER_CELL = 2       # [activation_prob, rate]
SAFE_TELOMERE_MAX = 2.0
TOTAL_TIMESTEPS = 80_000  # adjust for longer training in Colab

# -------------------------
# Multi-cell Environment
# -------------------------
class MultiCellEnv(gym.Env):
    """
    Gym-compatible multi-cell organ-on-chip environment.
    Observation: flattened array shape = (num_cells * cell_features,)
    Action: flattened array shape = (num_cells * action_per_cell,) with values in [0,1]
    Reward: sum of per-cell rewards (can be changed to cooperative or per-cell rewards).
    """
    metadata = {"render.modes": ["human"]}
    def __init__(self, num_cells=NUM_CELLS, cell_features=CELL_FEATURES, safe_max=SAFE_TELOMERE_MAX):
        super().__init__()
        self.num_cells = num_cells
        self.cell_features = cell_features
        self.safe_max = safe_max

        obs_low = np.zeros(self.num_cells * self.cell_features, dtype=np.float32)
        obs_high = np.ones(self.num_cells * self.cell_features, dtype=np.float32)
        self.observation_space = gym.spaces.Box(low=obs_low, high=obs_high, dtype=np.float32)

        act_low = np.zeros(self.num_cells * ACTION_PER_CELL, dtype=np.float32)
        act_high = np.ones(self.num_cells * ACTION_PER_CELL, dtype=np.float32)
        self.action_space = gym.spaces.Box(low=act_low, high=act_high, dtype=np.float32)

        self.reset()

    def _init_cells(self):
        # per-cell state layout (normalized 0..1):
        # idx0: telomere (0..~2.5 normalized to 0..1 by dividing by 2.5)
        # idx1: ROS (0..1)
        # idx2: damage (0..1)
        # idx3: phag_queue (0..1)
        # idx4: mutational_burden (0..1)
        # idx5: ATP/nutrient (0..1)
        # idx6: local_signal (0..1)  (from neighbors)
        # idx7: cell_cycle_phase (0..1)
        cells = np.zeros((self.num_cells, self.cell_features), dtype=np.float32)
        for i in range(self.num_cells):
            cells[i, 0] = 1.0/2.5 * 1.0  # target ~1 -> normalized 1/2.5 ~0.4 (we will use raw normalized values later)
            cells[i, 1] = np.random.rand() * 0.2  # low ROS
            cells[i, 2] = np.random.rand() * 0.1  # low damage
            cells[i, 3] = np.random.rand() * 0.05 # phag queue small
            cells[i, 4] = np.random.rand() * 0.02 # mut small
            cells[i, 5] = 0.8 + np.random.rand()*0.2 # ATP/nutrient
            cells[i, 6] = 0.0
            cells[i, 7] = np.random.rand()
        return cells

    def reset(self, *, seed=None, options=None):
        super().reset(seed=seed)
        self.cells = self._init_cells()
        # global nutrient field (scalar) normalized 0..1
        self.nutrient = 1.0
        self.steps = 0
        return self._get_obs(), {}

    def _get_obs(self):
        # flatten cells and clip 0..1
        return np.clip(self.cells.flatten().astype(np.float32), 0.0, 1.0)

    def step(self, action):
        """
        action: flattened actions for all cells, each action pair in [0,1]
        """
        self.steps += 1
        actions = np.clip(action.reshape(self.num_cells, ACTION_PER_CELL), 0.0, 1.0)
        rewards = np.zeros(self.num_cells, dtype=np.float32)

        # Simulate nutrient diffusion / consumption: nutrient decays + replenishment
        self.nutrient = np.clip(self.nutrient * 0.995 + 0.001, 0.0, 1.0)

        # local neighbor signal = average telomere of immediate neighbors (1D ring)
        telomeres_raw = self.cells[:, 0] * 2.5  # de-normalize to raw scale [0..2.5]
        neighbor_signal = np.zeros_like(telomeres_raw)
        for i in range(self.num_cells):
            left = (i - 1) % self.num_cells
            right = (i + 1) % self.num_cells
            neighbor_signal[i] = (telomeres_raw[left] + telomeres_raw[right]) / 2.0 / 2.5  # re-normalize

        # update each cell
        for i in range(self.num_cells):
            tel_norm = float(self.cells[i,0])
            tel = tel_norm * 2.5
            ros = float(self.cells[i,1])
            damage = float(self.cells[i,2])
            phag = float(self.cells[i,3])
            mut = float(self.cells[i,4])
            atp = float(self.cells[i,5])
            local_sig = neighbor_signal[i]

            act = float(actions[i,0])
            rate = float(actions[i,1])

            # Safety: prevent activation if telomere > safe max
            if tel > self.safe_max:
                act, rate = 0.0, 0.0

            # Telomere change: activation increases telomere, damage and replication shorten it
            delta_tel = act * (rate * 0.06) * (0.5 + atp*0.5) - 0.012 * damage - 0.005 * (1.0 - atp)
            tel += delta_tel
            tel = max(0.0, min(2.5, tel))

            # ROS dynamics (activity and nutrient scarcity may increase ROS)
            ros += 0.01 * (1.0 - atp) + 0.02 * act * rate - 0.02 * (phag)
            ros = np.clip(ros, 0.0, 1.0)

            # damage increases with ROS, decreases with phagocytosis
            damage += 0.02 * ros - 0.05 * phag
            damage = np.clip(damage, 0.0, 1.0)

            # phag queue: grows with damage and decreases if cell has energy
            phag += 0.05 * damage - 0.03 * (atp)
            phag = np.clip(phag, 0.0, 1.0)

            # mutation burden increases slowly with telomerase overactivation
            mut += 0.002 * act * rate + 0.001 * ros
            mut = np.clip(mut, 0.0, 1.0)

            # ATP consumption/production: nutrient field and activity
            atp += 0.01 * self.nutrient - 0.02 * act * rate - 0.01 * damage
            atp = np.clip(atp, 0.0, 1.0)

            # cell cycle toggles randomly influenced by ATP
            cell_cycle = float(self.cells[i,7])
            if random.random() < 0.05:
                cell_cycle = random.random() * (0.5 + 0.5*atp)

            # write back (normalized tel)
            self.cells[i,0] = tel / 2.5
            self.cells[i,1] = ros
            self.cells[i,2] = damage
            self.cells[i,3] = phag
            self.cells[i,4] = mut
            self.cells[i,5] = atp
            self.cells[i,6] = local_sig
            self.cells[i,7] = cell_cycle

            # reward: keep telomere close to 1.0 and penalize mut/damage
            target_tel = 1.0
            tel_score = -abs(tel - target_tel)
            penalty = -4.0 * mut - 2.0 * damage - 0.5 * phag
            rewards[i] = tel_score + penalty

        # global cooperative reward is sum of per-cell rewards
        total_reward = float(np.sum(rewards))

        # small global events: if too many cells have high damage, reduce nutrient
        avg_damage = float(np.mean(self.cells[:,2]))
        if avg_damage > 0.3:
            self.nutrient = max(0.0, self.nutrient - 0.01)

        done = (self.steps >= 1000)
        info = {
            "per_cell_rewards": rewards,
            "avg_telomere": float(np.mean(self.cells[:,0]) * 2.5),
            "avg_mut": float(np.mean(self.cells[:,4])),
            "nutrient": float(self.nutrient),
            "avg_damage": avg_damage,
        }
        return self._get_obs(), total_reward, done, False, info

    def render(self, mode='human'):
        tel_raw = self.cells[:,0] * 2.5
        print("Telomeres:", np.round(tel_raw, 3))
        print("Mutations:", np.round(self.cells[:,4], 3))
        print("Nutrient:", self.nutrient)

# -------------------------
# Custom Features Extractor (per-cell small transformer then flatten)
# Modified to capture attention weights
# -------------------------
class PerCellTransformerExtractor(BaseFeaturesExtractor):
    """
    Expects observation shape (num_cells * cell_features,)
    Produces flattened embedding of size num_cells * embed_dim for policy net
    Includes attention weight capture.
    """
    def __init__(self, observation_space: gym.spaces.Box, embed_dim=32, n_heads=2, n_layers=2):
        obs_shape = observation_space.shape
        num_inputs = obs_shape[0]
        super().__init__(observation_space, features_dim=embed_dim * NUM_CELLS)
        assert num_inputs == NUM_CELLS * CELL_FEATURES, "Observation shape mismatch"
        self.embed_dim = embed_dim
        self.n_cells = NUM_CELLS # Store number of cells

        # per-cell projection
        self.proj = nn.Linear(CELL_FEATURES, embed_dim)
        # Transformer encoder layers. Need to access attention.
        # Modify the layer to return attention or use a hook if needed.
        # A common way is to use batch_first=True and need_weights=True in MultiheadAttention
        # However, TransformerEncoderLayer doesn't directly expose need_weights.
        # We'll try to modify the forward pass to get attention.

        self.transformer_layers = nn.ModuleList([
            nn.TransformerEncoderLayer(d_model=embed_dim, nhead=n_heads, dim_feedforward=embed_dim*2, batch_first=True)
            for _ in range(n_layers)
        ])

        # final linear to combine per-cell embeddings if needed
        self.out_proj = nn.Linear(embed_dim * NUM_CELLS, embed_dim * NUM_CELLS)

        # Attribute to store the last computed attention weights
        self.last_attention_weights = None


    def forward(self, observations: torch.Tensor) -> torch.Tensor:
        # observations: (batch, num_cells * cell_features)
        B = observations.shape[0]
        x = observations.view(B, NUM_CELLS, CELL_FEATURES)           # (B, N, F)
        # project
        x_proj = self.proj(x)                                        # (B, N, D)

        # Pass through transformer layers and capture attention from the last layer
        out = x_proj
        for i, layer in enumerate(self.transformer_layers):
            # Call the layer's forward, attempting to get attention weights
            # This depends on the exact implementation of TransformerEncoderLayer/MultiheadAttention
            # If batch_first=True, MultiheadAttention's forward signature is usually (query, key, value, key_padding_mask, need_weights, ...)
            # TransformerEncoderLayer's forward doesn't directly expose need_weights,
            # but some versions/usages might pass it through or have internal flags.

            # Attempting to get attention from the MultiheadAttention submodule
            attn_output, attn_weights = layer.self_attn(
                 out, out, out, need_weights=True, average_attn_weights=True # average_attn_weights=True returns (B, N, N)
             )
            out = out + layer.dropout1(attn_output)
            out = layer.norm1(out)
            out = out + layer.dropout2(layer.linear2(layer.dropout(layer.activation(layer.linear1(out)))))
            out = layer.norm2(out)

            if i == len(self.transformer_layers) - 1: # Capture attention from the last layer
                self.last_attention_weights = attn_weights.detach() # Shape (B, N, N) if average_attn_weights=True, or (B, num_heads, N, N)

        # Average attention weights over the batch if batch_first was True
        if self.last_attention_weights is not None and self.last_attention_weights.ndim == 3: # Assuming (B, N, N)
             self.last_attention_weights = self.last_attention_weights.mean(dim=0) # Average over batch: (N, N)


        out_flat = out.view(B, -1)                                   # (B, N*D)
        return self.out_proj(out_flat)                               # (B, features_dim)

    def get_attention(self):
        """Retrieve latest attention weights (N_cells x N_cells, averaged over heads and batch)"""
        return self.last_attention_weights

# -------------------------
# Create env and model
# -------------------------
env = MultiCellEnv()
policy_kwargs = dict(
    features_extractor_class=PerCellTransformerExtractor,
    features_extractor_kwargs=dict(embed_dim=32, n_heads=2, n_layers=2),
)

model = PPO(
    "MlpPolicy",
    env,
    verbose=1,
    policy_kwargs=policy_kwargs,
    tensorboard_log="./multi_cell_telomerase/",
    learning_rate=3e-4,
    n_steps=2048,
    batch_size=64,
)

# -------------------------
# Train
# -------------------------
print("Starting training (this may take several minutes)...")
model.learn(total_timesteps=TOTAL_TIMESTEPS)
model.save("multi_cell_telomerase_policy")
print("Training finished & model saved.")

# -------------------------
# Evaluation
# -------------------------
def evaluate_model(model, env, episodes=10):
    avg_rewards = []
    trajectories = []
    for ep in range(episodes):
        obs, _ = env.reset()
        done = False
        tot_r = 0.0
        tel_traj = []
        while not done:
            action, _ = model.predict(obs, deterministic=True)
            obs, r, done, _, info = env.step(action)
            tot_r += r
            tel_traj.append(info["avg_telomere"])
        avg_rewards.append(tot_r)
        trajectories.append(tel_traj)
    return avg_rewards, trajectories

rewards, trajs = evaluate_model(model, env, episodes=8)
print("Avg reward (eval):", np.mean(rewards))

# Plotting results
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(rewards, marker='o')
plt.title("Total Episode Reward (Evaluation)")
plt.xlabel("Episode"); plt.ylabel("Total Reward")

plt.subplot(1,2,2)
for t in trajs:
    plt.plot(t, alpha=0.7)
plt.axhline(1.0, linestyle='--', label='Target telomere')
plt.title("Average Telomere Across Cells (per step)")
plt.xlabel("Step"); plt.ylabel("Avg telomere (raw)")
plt.legend()
plt.tight_layout()
plt.show()

# Render one simulation with prints
obs, _ = env.reset()
for _ in range(50):
    action, _ = model.predict(obs, deterministic=True)
    obs, r, done, _, info = env.step(action)
env.render()
```
