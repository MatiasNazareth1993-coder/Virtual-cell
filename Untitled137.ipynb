```python
# telomere_sim.py
import numpy as np
from scipy.integrate import solve_ivp

# model parameters (tunable)
SHORTEN_PER_DIV = 50.0    # bp per division (example numeric scale)
TELOMERASE_K = 1.0        # max bp extension rate when active
TRIM_RATE = 0.01          # fractional trimming of excess length per unit time
DIV_RATE_BASE = 0.01      # baseline division rate

def telomere_ode(t, y, telomerase_pulse, env_signal):
    L = y[0]  # telomere length (bp)
    # telomerase_pulse: function of t or external control input (0..1)
    telomerase_activity = telomerase_pulse(t, env_signal)
    # shortening via replication approximated as continuous effective loss
    dL_dt = -DIV_RATE_BASE * SHORTEN_PER_DIV + TELOMERASE_K * telomerase_activity - TRIM_RATE * max(0, L - 10000)
    return [dL_dt]

# Example telomerase pulse function (policy)
def example_pulse(t, env):
    # env can be dictionary with signals; simple rule: pulse during injury (env['injury']>0)
    return 1.0 if env.get('injury',0) > 0 else 0.0

# simulate
y0 = [8000.0]  # starting telomere length
env = {'injury':1}  # example
sol = solve_ivp(lambda tt,y: telomere_ode(tt,y, example_pulse, env), [0,100], y0, max_step=0.5)
print(sol.t[-1], sol.y[0,-1])
```

```python
# ooc_agent_grid.py
import numpy as np

class Cell:
    def __init__(self, id, L=8000.0):
        self.id = id
        self.L = L
        self.state = 'quiescent'  # or 'proliferating','senescent','apoptotic'
        self.age = 0

    def step(self, local_env, control_policy):
        # control_policy: callable -> telomerase activity
        ta = control_policy(self, local_env)
        # update telomere length (discrete approx)
        if self.state == 'proliferating':
            self.L -= 50  # shortening per division
        self.L += 1.0 * ta  # small continuous extension
        # simple state transitions
        if self.L < 2000:
            self.state = 'senescent'
        self.age += 1

# grid simulation
def run_grid(steps=100, nx=32, ny=32):
    grid = {}
    for i in range(nx*ny):
        grid[i] = Cell(i)
    for t in range(steps):
        # produce environment field (nutrient, injury, shear) — simple placeholders
        for cell in grid.values():
            local_env = {'nutrient':1.0, 'injury': 1.0 if t<5 and cell.id%100==0 else 0.0}
            cell.step(local_env, lambda c,e: 1.0 if e['injury']>0 else 0.0)
    return grid

if __name__ == "__main__":
    final = run_grid()
    print("Sample cell L:", final[0].L, final[0].state)
```

```python
# control_model.py
import torch
import torch.nn as nn
import torch.optim as optim

class TelomereController(nn.Module):
    def __init__(self, seq_in=128, env_in=16, hidden=256):
        super().__init__()
        # time-series encoder (1D conv or LSTM)
        self.lstm = nn.LSTM(input_size=seq_in, hidden_size=hidden, batch_first=True)
        # env MLP
        self.env_fc = nn.Sequential(nn.Linear(env_in, hidden), nn.ReLU())
        # cross-attention / fusion
        self.fusion = nn.Sequential(nn.Linear(hidden*2, hidden), nn.ReLU(), nn.Linear(hidden,1), nn.Sigmoid())
    def forward(self, seq, env):
        # seq: batch x T x seq_in ; env: batch x env_in
        _, (h_n, _) = self.lstm(seq)
        h = h_n[-1]  # batch x hidden
        e = self.env_fc(env)
        x = torch.cat([h,e], dim=-1)
        out = self.fusion(x)
        return out  # telomerase activation probability / intensity in (0,1)

# training loop sketch
def train(model, dataloader, epochs=10):
    opt = optim.Adam(model.parameters(), lr=1e-4)
    loss_fn = nn.MSELoss()  # or custom loss combining repair metric and cancer proxy penalty
    for epoch in range(epochs):
        for seq, env, target in dataloader:
            pred = model(seq, env)
            loss = loss_fn(pred, target)
            opt.zero_grad(); loss.backward(); opt.step()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np

# -------------------------
# Synthetic DNA encoding
# -------------------------
BASE_TO_ID = {"T":0, "A":1, "G":2, "C":3, "X":4}

def encode_sequence(seq):
    return torch.tensor([BASE_TO_ID[b] for b in seq], dtype=torch.long)


# -------------------------
# Transformer for Telomere State
# -------------------------
class TelomereTransformer(nn.Module):
    def __init__(self, vocab=5, d_model=64, nhead=4, nlayers=3):
        super().__init__()
        self.embed = nn.Embedding(vocab, d_model)
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=d_model,
            nhead=nhead,
            dim_feedforward=256
        )
        self.encoder = nn.TransformerEncoder(encoder_layer, nlayers)
        self.fc = nn.Linear(d_model, 1)  # output = predicted telomere stability

    def forward(self, seq):
        x = self.embed(seq).unsqueeze(1)
        x = self.encoder(x)
        x = x.mean(dim=0)
        return self.fc(x)


# -------------------------
# Reinforcement Learning Telomerase Controller
# -------------------------
class TelomeraseRL(nn.Module):
    def __init__(self, input_dim=4, hidden=128):
        super().__init__()
        self.policy = nn.Sequential(
            nn.Linear(input_dim, hidden),
            nn.ReLU(),
            nn.Linear(hidden, 2),  # 0 = no activation, 1 = activate telomerase
            nn.Softmax(dim=-1)
        )

    def forward(self, state):
        return self.policy(state)


# -------------------------
# Virtual Organ-on-a-Chip Cell Simulation
# -------------------------
def simulate_step(cell):
    # Virtual dynamics (safe)
    cell["damage_index"] += np.random.normal(0.01, 0.005)
    cell["metabolic_load"] += 0.02 * (1 - cell["phagocytosis_efficiency"])
    cell["division_rate"] = max(0, 1 - cell["damage_index"])

    # Telomere shrinkage
    cell["telomere_length"] -= 0.5 * (1 - cell["telomerase_activity"])

    # Phagocytosis repair
    cell["damage_index"] *= (1 - cell["phagocytosis_efficiency"] * 0.3)

    return cell


# -------------------------
# Training Loop Example
# -------------------------
model = TelomereTransformer()
optimizer = optim.Adam(model.parameters(), lr=3e-3)  # high learning rate allowed for in silico

sequence = encode_sequence("TTAGGX" * 200)

for epoch in range(200):
    optimizer.zero_grad()
    pred = model(sequence)
    target = torch.tensor([0.1])  # desired stable-telomere score (virtual)
    loss = (pred - target).pow(2).mean()
    loss.backward()
    optimizer.step()

    if epoch % 20 == 0:
        print(f"Epoch {epoch}, Loss {loss.item():.4f}")
```

```python
"""
virtual_chip_sim.py

Safe, in-silico multi-cell organ-on-chip simulator with synthetic telomere tokens,
telomerase controller, phagocytosis, nutrient flow, and visualization.

NOT FOR BIOLOGICAL USE. All DNA tokens are synthetic placeholders and the
model is computational-only.
"""

import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
import networkx as nx
from tqdm import trange

# -------------------------
# 1) Synthetic telomere token (purely symbolic)
# -------------------------
TEL_UNIT = "TTAGGX"  # synthetic placeholder token (non-biological)
SYN_TEL_SHORT = 1000.0  # initial 'length' units for differentiated cells (virtual)
SYN_TEL_LONG = 10000.0  # for stem-like virtual cells

# -------------------------
# 2) Cell class (virtual cell state)
# -------------------------
class Cell:
    def __init__(self, pos, stemness=0.0, phago_eff=0.5):
        self.pos = pos
        self.telomere_length = SYN_TEL_SHORT if stemness < 0.5 else SYN_TEL_LONG
        self.telomerase_activity = 0.0  # 0..1
        self.division_rate = 0.0
        self.damage_index = np.random.uniform(0.0, 0.05)  # low initial damage
        self.metabolic_load = np.random.uniform(0.0, 0.1)
        self.phagocytosis_efficiency = phago_eff
        self.stemness = stemness  # 0..1
        self.alive = True

    def state_vector(self):
        # Feature vector exposed to controller
        return np.array([
            self.telomere_length / SYN_TEL_LONG,  # normalized
            self.telomerase_activity,
            self.damage_index,
            self.metabolic_load,
            self.phagocytosis_efficiency,
            self.stemness
        ], dtype=np.float32)

# -------------------------
# 3) Telomerase controller (small policy network)
# -------------------------
class TelomeraseController(nn.Module):
    def __init__(self, input_dim=6, hidden=64):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden),
            nn.ReLU(),
            nn.Linear(hidden, hidden//2),
            nn.ReLU(),
            nn.Linear(hidden//2, 1),
            nn.Sigmoid()  # output in [0,1] representing telomerase activation level
        )

    def forward(self, x):
        return self.net(x)

# -------------------------
# 4) Chip / Tissue environment
# -------------------------
class VirtualChip:
    def __init__(self, width=16, height=12, seed=42):
        np.random.seed(seed)
        self.width = width
        self.height = height
        self.cells = {}
        self.graph = nx.grid_2d_graph(width, height)
        self._populate_cells()
        # Nutrient field (flowing along x axis for simplicity)
        self.nutrient = np.ones((height, width)) * 1.0
        # Tissue demand signal (external; high values indicate need for divisions/repair)
        self.tissue_demand = np.zeros((height, width))
        self.time = 0

    def _populate_cells(self):
        for y in range(self.height):
            for x in range(self.width):
                # assign a few stem-like cells randomly near center
                dist_center = np.hypot(x - self.width/2, y - self.height/2)
                stem_prob = 0.05 if dist_center > max(self.width, self.height)/4 else 0.25
                stemness = np.random.choice([1.0, 0.0], p=[stem_prob, 1-stem_prob])
                phago_eff = np.clip(np.random.normal(0.5, 0.12), 0.05, 0.95)
                self.cells[(x,y)] = Cell(pos=(x,y), stemness=stemness, phago_eff=phago_eff)

    def neighborhood(self, pos):
        return [nbr for nbr in self.graph.neighbors(pos)]

    def step_nutrient_flow(self):
        # simple advection + decay
        decay = 0.002
        advect = 0.02
        self.nutrient = np.roll(self.nutrient, shift=1, axis=1) * (1 - advect)
        self.nutrient -= decay * self.nutrient
        self.nutrient = np.clip(self.nutrient, 0.0, 2.0)

    def update_tissue_demand(self):
        # demand if average damage in patch is high or nutrient low
        for (x,y), cell in self.cells.items():
            local_neighbors = self.neighborhood((x,y))
            avg_damage = np.mean([self.cells[n].damage_index for n in local_neighbors] + [cell.damage_index])
            local_nutr = self.nutrient[y,x]
            # scale demand 0..1
            demand = np.clip(0.5*(avg_damage/0.5) + 0.5*(1 - local_nutr/1.0), 0.0, 1.0)
            self.tissue_demand[y,x] = demand

# -------------------------
# 5) Dynamics: how a cell evolves per step
# -------------------------
def simulate_cell_dynamics(cell: Cell, local_nutrient, telomerase_cmd):
    # telomerase_cmd is in [0,1] applied by controller
    # Update telomerase activity (smooth)
    alpha = 0.3
    cell.telomerase_activity = (1 - alpha) * cell.telomerase_activity + alpha * telomerase_cmd

    # Damage accumulates with metabolic load and low phagocytosis
    damage_growth = 0.01 * (1 + cell.metabolic_load) * (1 - cell.phagocytosis_efficiency)
    cell.damage_index += np.random.normal(damage_growth, 0.005)
    cell.damage_index = np.clip(cell.damage_index, 0.0, 1.0)

    # Telomere change: shrinkage unless telomerase activity high
    basal_shrink = 1.0 * (1 + cell.metabolic_load*0.5)
    elongation = 2.0 * cell.telomerase_activity * (0.5 + cell.stemness)
    cell.telomere_length += -basal_shrink + elongation
    cell.telomere_length = max(1.0, cell.telomere_length)  # keep positive

    # Phagocytosis reduces damage
    phago_clear = cell.phagocytosis_efficiency * 0.05
    cell.damage_index = max(0.0, cell.damage_index - phago_clear)

    # Metabolic load responds to nutrient availability
    cell.metabolic_load = np.clip(cell.metabolic_load * (0.95 + 0.05*(1-local_nutrient)) + 0.01, 0, 1)

    # Decision for division: requires low damage, sufficient telomere length, tissue demand, and stemness
    can_divide = (cell.damage_index < 0.25 and cell.telomere_length > SYN_TEL_SHORT*0.5)
    return can_divide

# -------------------------
# 6) Division behavior (virtual)
# -------------------------
def perform_division(chip: VirtualChip, pos):
    cell = chip.cells[pos]
    if not cell.alive:
        return
    # division produces daughter cells that inherit telomeres (half distributed) and slight changes
    daughter1 = Cell(pos=pos, stemness=cell.stemness, phago_eff=cell.phagocytosis_efficiency)
    daughter2 = Cell(pos=pos, stemness=cell.stemness, phago_eff=cell.phagocytosis_efficiency)
    # split telomere length in-silico (virtual)
    split_len = cell.telomere_length * 0.5
    daughter1.telomere_length = split_len * (1.0 + 0.02*np.random.randn())
    daughter2.telomere_length = split_len * (1.0 + 0.02*np.random.randn())
    # damage carried over partially
    daughter1.damage_index = cell.damage_index * 0.5
    daughter2.damage_index = cell.damage_index * 0.5
    chip.cells[pos] = daughter1  # place one daughter at same pos
    # try to place second daughter in one neighbor if empty-ish (we'll simply overwrite a low-stem neighbor)
    neighbors = chip.neighborhood(pos)
    # Choose neighbor with highest damage to encourage replacement (virtual homeostasis)
    neighbors_sorted = sorted(neighbors, key=lambda n: chip.cells[n].damage_index, reverse=True)
    if neighbors_sorted:
        chosen = neighbors_sorted[0]
        chip.cells[chosen] = daughter2

# -------------------------
# 7) Training / controller usage
#    We'll use a simple supervised-like objective: maximize population telomere stability
#    while minimizing damage and excessive telomerase activation (parsimony).
# -------------------------
def run_simulation(seed=1, steps=400, train_controller=True, device='cpu'):
    np.random.seed(seed)
    torch.manual_seed(seed)

    chip = VirtualChip(width=20, height=14, seed=seed)
    controller = TelomeraseController(input_dim=6, hidden=80).to(device)
    opt = optim.Adam(controller.parameters(), lr=5e-3)  # relatively high learning rate for in-silico
    losses = []
    stats = {
        "avg_telomere": [],
        "avg_damage": [],
        "avg_phago": [],
        "divisions": []
    }

    for t in trange(steps, desc="Simulating"):
        chip.time = t
        chip.step_nutrient_flow()
        chip.update_tissue_demand()

        # Prepare batch for controller (all cells)
        states = []
        positions = []
        local_nutrients = []
        demand_vals = []

        for (x,y), cell in chip.cells.items():
            states.append(cell.state_vector())
            positions.append((x,y))
            local_nutrients.append(chip.nutrient[y,x])
            demand_vals.append(chip.tissue_demand[y,x])

        states = torch.tensor(np.stack(states), dtype=torch.float32).to(device)
        with torch.set_grad_enabled(train_controller):
            outputs = controller(states).squeeze(-1)  # telomerase activation level [0,1]
            tel_actions = outputs.detach().cpu().numpy()

            # Apply actions and simulate local dynamics
            total_divisions = 0
            loss_tensor = 0.0
            for i, pos in enumerate(positions):
                cell = chip.cells[pos]
                lnutr = local_nutrients[i]
                tel_cmd = float(outputs[i].item())
                can_divide = simulate_cell_dynamics(cell, lnutr, tel_cmd)

                # if tissue demand high at cell location and can_divide, attempt division probabilistically
                demand = demand_vals[i]
                if can_divide and (demand > 0.4) and (np.random.rand() < 0.15 + 0.6*demand):
                    perform_division(chip, pos)
                    total_divisions += 1

                # Build a surrogate loss: we want high telomere length, low damage, and low mean activation
                # Scale features into surrogate objectives:
                tel_norm = torch.clamp(torch.tensor(cell.telomere_length / SYN_TEL_LONG), 0.0, 2.0)
                dmg = torch.tensor(cell.damage_index)
                act = outputs[i]
                # encourage tel_norm ~ 0.8 and dmg ~ 0 and penalize large act
                loss_cell = (tel_norm - 0.8).pow(2) + 2.0*dmg + 0.5*act.pow(2)
                loss_tensor = loss_tensor + loss_cell

            # Backpropagate a mean loss across cells
            if train_controller:
                opt.zero_grad()
                loss_tensor = loss_tensor / len(positions)
                loss_tensor.backward()
                opt.step()
                losses.append(float(loss_tensor.detach().cpu().numpy()))

        # Record stats
        all_tels = np.array([c.telomere_length for c in chip.cells.values()])
        all_dmg = np.array([c.damage_index for c in chip.cells.values()])
        all_phago = np.array([c.phagocytosis_efficiency for c in chip.cells.values()])
        stats["avg_telomere"].append(all_tels.mean())
        stats["avg_damage"].append(all_dmg.mean())
        stats["avg_phago"].append(all_phago.mean())
        stats["divisions"].append(total_divisions)

    return chip, controller, stats, losses

# -------------------------
# 8) Visualization helpers
# -------------------------
def plot_stats(stats, losses, fname_prefix=None):
    steps = len(stats["avg_telomere"])
    t = np.arange(steps)
    fig, axes = plt.subplots(2,2, figsize=(12,8))
    axes = axes.flatten()
    axes[0].plot(t, stats["avg_telomere"])
    axes[0].set_title("Avg Telomere Length (virtual units)")
    axes[1].plot(t, stats["avg_damage"])
    axes[1].set_title("Avg Damage Index")
    axes[2].plot(t, stats["divisions"])
    axes[2].set_title("Divisions per Step")
    axes[3].plot(t, losses[:len(t)])
    axes[3].set_title("Controller Loss (surrogate)")
    plt.tight_layout()
    if fname_prefix:
        plt.savefig(f"{fname_prefix}_timeseries.png", dpi=150)
    plt.show()

def plot_chip_heatmap(chip: VirtualChip, metric="telomere"):
    grid = np.zeros((chip.height, chip.width))
    for (x,y), cell in chip.cells.items():
        if metric == "telomere":
            grid[y,x] = cell.telomere_length
        elif metric == "damage":
            grid[y,x] = cell.damage_index
        elif metric == "phago":
            grid[y,x] = cell.phagocytosis_efficiency
        elif metric == "stem":
            grid[y,x] = cell.stemness
    plt.figure(figsize=(6,5))
    plt.imshow(grid, origin='lower', cmap='viridis')
    plt.colorbar(label=metric)
    plt.title(f"Chip heatmap - {metric}")
    plt.show()

# -------------------------
# 9) Run example
# -------------------------
if __name__ == "__main__":
    chip, controller, stats, losses = run_simulation(seed=2, steps=350, train_controller=True)
    plot_stats(stats, losses, fname_prefix="virtual_chip")
    plot_chip_heatmap(chip, metric="telomere")
    plot_chip_heatmap(chip, metric="damage")
    plot_chip_heatmap(chip, metric="phago")
    print("Simulation complete. Final avg telomere:", stats["avg_telomere"][-1])
    print("Final avg damage:", stats["avg_damage"][-1])
```

```python
"""
virtual_multi_organ_chip.py

Safe, in-silico multi-organ organ-on-chip simulator (computational-only).

This extends the single-chip simulation to multiple organ compartments (e.g., liver, heart, lung)
connected by a shared microfluidic loop. Each organ has its own micro-environment, cell
populations, telomere dynamics, telomerase controllers, and inter-organ transport of
nutrients/waste/signals.

NOT FOR BIOLOGICAL USE: All DNA/telomere tokens are synthetic placeholders. No wet-lab
instructions or real sequence-level data are included.

How to run:
1. Save as virtual_multi_organ_chip.py
2. pip install numpy matplotlib torch networkx tqdm
3. python virtual_multi_organ_chip.py

Outputs: time-series plots of organ-level metrics, per-organ heatmaps, and simple cross-organ
transfer visualizations.
"""

import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
import networkx as nx
from tqdm import trange
from collections import defaultdict

# -------------------------
# Global synthetic telomere tokens (safe placeholders)
# -------------------------
TEL_UNIT = "TTAGGX"  # synthetic token only
SYN_TEL_SHORT = 1000.0
SYN_TEL_LONG = 10000.0

# -------------------------
# Generic virtual cell definition
# -------------------------
class Cell:
    def __init__(self, pos, stemness=0.0, phago_eff=0.5, organ_name='organ'):
        self.pos = pos
        self.organ = organ_name
        self.telomere_length = SYN_TEL_SHORT if stemness < 0.5 else SYN_TEL_LONG
        self.telomerase_activity = 0.0
        self.division_rate = 0.0
        self.damage_index = np.random.uniform(0.0, 0.05)
        self.metabolic_load = np.random.uniform(0.0, 0.1)
        self.phagocytosis_efficiency = phago_eff
        self.stemness = stemness
        self.alive = True

    def state_vector(self):
        return np.array([
            self.telomere_length / SYN_TEL_LONG,
            self.telomerase_activity,
            self.damage_index,
            self.metabolic_load,
            self.phagocytosis_efficiency,
            self.stemness
        ], dtype=np.float32)

# -------------------------
# Telomerase controller (same architecture reused per organ)
# -------------------------
class TelomeraseController(nn.Module):
    def __init__(self, input_dim=6, hidden=64):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden),
            nn.ReLU(),
            nn.Linear(hidden, hidden//2),
            nn.ReLU(),
            nn.Linear(hidden//2, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.net(x)

# -------------------------
# Organ compartment: holds a grid of cells and local environment
# -------------------------
class Organ:
    def __init__(self, name, width, height, stem_seed_prob=0.1, nutrient_base=1.0, seed=0):
        self.name = name
        self.width = width
        self.height = height
        self.cells = {}
        self.graph = nx.grid_2d_graph(width, height)
        self.nutrient = np.ones((height, width)) * nutrient_base
        self.waste = np.zeros((height, width))
        self.tissue_demand = np.zeros((height, width))
        self._populate_cells(stem_seed_prob, seed)

    def _populate_cells(self, stem_seed_prob, seed):
        np.random.seed(seed)
        for y in range(self.height):
            for x in range(self.width):
                stemness = 1.0 if np.random.rand() < stem_seed_prob else 0.0
                phago_eff = np.clip(np.random.normal(0.55, 0.12), 0.05, 0.95)
                self.cells[(x,y)] = Cell(pos=(x,y), stemness=stemness, phago_eff=phago_eff, organ_name=self.name)

    def neighborhood(self, pos):
        return [nbr for nbr in self.graph.neighbors(pos)]

    def step_local_diffusion(self):
        # simple diffusion for nutrient and waste (5-point stencil approximation)
        new_n = self.nutrient.copy()
        new_w = self.waste.copy()
        for y in range(self.height):
            for x in range(self.width):
                nbors = [(x+1,y),(x-1,y),(x,y+1),(x,y-1)]
                vals_n = [self.nutrient[y,x]]
                vals_w = [self.waste[y,x]]
                for nxp,nyp in nbors:
                    if 0 <= nxp < self.width and 0 <= nyp < self.height:
                        vals_n.append(self.nutrient[nyp,nxp])
                        vals_w.append(self.waste[nyp,nxp])
                new_n[y,x] = np.mean(vals_n)
                new_w[y,x] = np.mean(vals_w)
        # small decay / baseline consumption
        new_n = np.clip(new_n - 0.001, 0.0, 2.0)
        new_w = np.clip(new_w * 0.995, 0.0, 10.0)
        self.nutrient = new_n
        self.waste = new_w

    def update_tissue_demand(self):
        for (x,y), cell in self.cells.items():
            local_neighbors = self.neighborhood((x,y))
            avg_damage = np.mean([self.cells[n].damage_index for n in local_neighbors] + [cell.damage_index])
            local_nutr = self.nutrient[y,x]
            demand = np.clip(0.6*(avg_damage/0.5) + 0.4*(1 - local_nutr/1.0), 0.0, 1.0)
            self.tissue_demand[y,x] = demand

# -------------------------
# System that connects multiple organs through a shared circulation loop
# -------------------------
class MultiOrganSystem:
    def __init__(self, organ_specs, transfer_matrix=None, seed=0):
        """
        organ_specs: list of dicts with keys: name, width, height, stem_seed_prob, nutrient_base
        transfer_matrix: NxN matrix where entry (i,j) fraction flows from organ i -> organ j each circulation step
        """
        self.organs = []
        self.name_to_idx = {}
        for i, spec in enumerate(organ_specs):
            org = Organ(name=spec.get('name', f'org{i}'),
                        width=spec.get('width', 16),
                        height=spec.get('height', 12),
                        stem_seed_prob=spec.get('stem_seed_prob', 0.08),
                        nutrient_base=spec.get('nutrient_base', 1.0),
                        seed=seed+i)
            self.name_to_idx[org.name] = i
            self.organs.append(org)

        n = len(self.organs)
        if transfer_matrix is None:
            # default: circular loop with small cross transfer
            T = np.zeros((n,n))
            for i in range(n):
                T[i, (i+1)%n] = 0.95  # most fluid goes to next organ
                T[i, i] = 0.05         # some recirculates/stays local
            self.transfer_matrix = T
        else:
            self.transfer_matrix = np.array(transfer_matrix)

        # controllers per organ
        self.controllers = [TelomeraseController(input_dim=6, hidden=80) for _ in range(n)]
        self.optimizers = [optim.Adam(c.parameters(), lr=5e-3) for c in self.controllers]
        self.time = 0

    def circulate_and_transfer(self):
        # Aggregate organ-level bulk nutrient and waste, then redistribute based on transfer_matrix
        n = len(self.organs)
        bulk_n = np.zeros(n)
        bulk_w = np.zeros(n)
        # compute organ means
        for i, org in enumerate(self.organs):
            bulk_n[i] = org.nutrient.mean()
            bulk_w[i] = org.waste.mean()

        # apply transfer matrix: new_bulk_j = sum_i T[i,j] * bulk_i
        new_bulk_n = self.transfer_matrix.T.dot(bulk_n)
        new_bulk_w = self.transfer_matrix.T.dot(bulk_w)

        # spread redistributed bulk back to organ grids (simple uniform fill)
        for j, org in enumerate(self.organs):
            org.nutrient = np.ones_like(org.nutrient) * float(new_bulk_n[j])
            org.waste = np.ones_like(org.waste) * float(new_bulk_w[j])

    def step(self, train_controllers=True):
        self.time += 1
        # 1) local organ steps
        for org_idx, org in enumerate(self.organs):
            org.step_local_diffusion()
            org.update_tissue_demand()

            # prepare controller input batch
            states = []
            positions = []
            local_nuts = []
            demands = []
            for (x,y), cell in org.cells.items():
                states.append(cell.state_vector())
                positions.append((x,y))
                local_nuts.append(org.nutrient[y,x])
                demands.append(org.tissue_demand[y,x])
            if not states:
                continue
            states_t = torch.tensor(np.stack(states), dtype=torch.float32)
            controller = self.controllers[org_idx]

            with torch.set_grad_enabled(train_controllers):
                outputs = controller(states_t).squeeze(-1)
                # simulate each cell with its action
                loss_tensor = 0.0
                total_divisions = 0
                for i, pos in enumerate(positions):
                    cell = org.cells[pos]
                    tel_cmd = float(outputs[i].item())
                    can_divide = simulate_cell_dynamics(cell, local_nuts[i], tel_cmd)

                    # division probability tuned by organ-specific demand
                    demand = demands[i]
                    if can_divide and (demand > 0.4) and (np.random.rand() < 0.12 + 0.6*demand):
                        perform_division(org, pos)
                        total_divisions += 1

                    # surrogate loss
                    tel_norm = torch.clamp(torch.tensor(cell.telomere_length / SYN_TEL_LONG), 0.0, 2.0)
                    dmg = torch.tensor(cell.damage_index)
                    act = outputs[i]
                    loss_cell = (tel_norm - 0.8).pow(2) + 2.0*dmg + 0.5*act.pow(2)
                    loss_tensor = loss_tensor + loss_cell

                if train_controllers:
                    opt = self.optimizers[org_idx]
                    opt.zero_grad()
                    loss_tensor = loss_tensor / len(positions)
                    loss_tensor.backward()
                    opt.step()

        # 2) circulation step - exchange bulk fluid among organs
        self.circulate_and_transfer()

# -------------------------
# Reused dynamics & division functions
# -------------------------

def simulate_cell_dynamics(cell: Cell, local_nutrient, telomerase_cmd):
    alpha = 0.3
    cell.telomerase_activity = (1 - alpha) * cell.telomerase_activity + alpha * telomerase_cmd

    damage_growth = 0.01 * (1 + cell.metabolic_load) * (1 - cell.phagocytosis_efficiency)
    cell.damage_index += np.random.normal(damage_growth, 0.005)
    cell.damage_index = np.clip(cell.damage_index, 0.0, 1.0)

    basal_shrink = 1.0 * (1 + cell.metabolic_load*0.5)
    elongation = 2.0 * cell.telomerase_activity * (0.5 + cell.stemness)
    cell.telomere_length += -basal_shrink + elongation
    cell.telomere_length = max(1.0, cell.telomere_length)

    phago_clear = cell.phagocytosis_efficiency * 0.05
    cell.damage_index = max(0.0, cell.damage_index - phago_clear)

    cell.metabolic_load = np.clip(cell.metabolic_load * (0.95 + 0.05*(1-local_nutrient)) + 0.01, 0, 1)

    can_divide = (cell.damage_index < 0.25 and cell.telomere_length > SYN_TEL_SHORT*0.5)
    return can_divide


def perform_division(organ: Organ, pos):
    cell = organ.cells[pos]
    if not cell.alive:
        return
    daughter1 = Cell(pos=pos, stemness=cell.stemness, phago_eff=cell.phagocytosis_efficiency, organ_name=organ.name)
    daughter2 = Cell(pos=pos, stemness=cell.stemness, phago_eff=cell.phagocytosis_efficiency, organ_name=organ.name)
    split_len = cell.telomere_length * 0.5
    daughter1.telomere_length = split_len * (1.0 + 0.02*np.random.randn())
    daughter2.telomere_length = split_len * (1.0 + 0.02*np.random.randn())
    daughter1.damage_index = cell.damage_index * 0.5
    daughter2.damage_index = cell.damage_index * 0.5
    organ.cells[pos] = daughter1
    neighbors = organ.neighborhood(pos)
    neighbors_sorted = sorted(neighbors, key=lambda n: organ.cells[n].damage_index, reverse=True)
    if neighbors_sorted:
        chosen = neighbors_sorted[0]
        organ.cells[chosen] = daughter2

# -------------------------
# Monitoring & visualization
# -------------------------

def gather_system_stats(system: MultiOrganSystem):
    stats = {}
    for i, org in enumerate(system.organs):
        all_tels = np.array([c.telomere_length for c in org.cells.values()])
        all_dmg = np.array([c.damage_index for c in org.cells.values()])
        all_phago = np.array([c.phagocytosis_efficiency for c in org.cells.values()])
        stats[org.name] = {
            'avg_telomere': float(all_tels.mean()),
            'median_telomere': float(np.median(all_tels)),
            'avg_damage': float(all_dmg.mean()),
            'avg_phago': float(all_phago.mean()),
            'divisions': None
        }
    return stats


def plot_system_timeseries(history, fname_prefix=None):
    # history: list of dict snapshots (per time) as returned by gather_system_stats
    n_steps = len(history)
    organ_names = list(history[0].keys())
    fig, axes = plt.subplots(len(organ_names), 2, figsize=(12, 3*len(organ_names)))
    if len(organ_names) == 1:
        axes = axes.reshape(1,2)
    for i, oname in enumerate(organ_names):
        tel = [history[t][oname]['avg_telomere'] for t in range(n_steps)]
        dmg = [history[t][oname]['avg_damage'] for t in range(n_steps)]
        axes[i,0].plot(tel)
        axes[i,0].set_title(f"{oname} avg telomere")
        axes[i,1].plot(dmg)
        axes[i,1].set_title(f"{oname} avg damage")
    plt.tight_layout()
    if fname_prefix:
        plt.savefig(f"{fname_prefix}_system_timeseries.png", dpi=150)
    plt.show()


def plot_organ_heatmaps(system: MultiOrganSystem, metric='telomere'):
    n = len(system.organs)
    cols = min(3, n)
    rows = int(np.ceil(n/cols))
    fig, axs = plt.subplots(rows, cols, figsize=(4*cols, 3*rows))
    axs = np.array(axs).reshape(-1)
    for i, org in enumerate(system.organs):
        grid = np.zeros((org.height, org.width))
        for (x,y), cell in org.cells.items():
            if metric == 'telomere':
                grid[y,x] = cell.telomere_length
            elif metric == 'damage':
                grid[y,x] = cell.damage_index
            elif metric == 'phago':
                grid[y,x] = cell.phagocytosis_efficiency
            elif metric == 'nutrient':
                grid[y,x] = org.nutrient[y,x]
        ax = axs[i]
        im = ax.imshow(grid, origin='lower', cmap='viridis')
        ax.set_title(f"{org.name} - {metric}")
        plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    # hide unused axes
    for j in range(i+1, len(axs)):
        axs[j].axis('off')
    plt.tight_layout()
    plt.show()

# -------------------------
# Example experiment runner
# -------------------------

def run_multi_organ_experiment(seed=3, steps=300, train_controllers=True):
    # Define organs: liver-like (detox, higher stem seeds), heart-like (low stem), lung-like (exchange)
    organ_specs = [
        {'name': 'Liver', 'width': 20, 'height': 12, 'stem_seed_prob': 0.18, 'nutrient_base': 1.1},
        {'name': 'Heart', 'width': 14, 'height': 10, 'stem_seed_prob': 0.04, 'nutrient_base': 0.9},
        {'name': 'Lung', 'width': 18, 'height': 12, 'stem_seed_prob': 0.08, 'nutrient_base': 1.2},
    ]

    # Transfer matrix: Liver -> Heart -> Lung -> Liver loop with small leakage to self
    T = np.array([
        [0.05, 0.9, 0.05],  # Liver outputs mostly to Heart
        [0.05, 0.05, 0.9],  # Heart -> Lung
        [0.9, 0.05, 0.05],  # Lung -> Liver
    ])

    system = MultiOrganSystem(organ_specs, transfer_matrix=T, seed=seed)

    history = []
    for t in trange(steps, desc='Multi-organ sim'):
        system.step(train_controllers=train_controllers)
        if t % 5 == 0:
            stats = gather_system_stats(system)
            history.append(stats)

    return system, history

# -------------------------
# Main run
# -------------------------
if __name__ == '__main__':
    system, history = run_multi_organ_experiment(seed=7, steps=240, train_controllers=True)
    plot_system_timeseries(history, fname_prefix='multi_organ')
    plot_organ_heatmaps(system, metric='telomere')
    plot_organ_heatmaps(system, metric='damage')
    plot_organ_heatmaps(system, metric='nutrient')
    print('Multi-organ simulation finished.')
```

```python
"""
virtual_multi_organ_chip.py

Safe, in-silico multi-organ organ-on-chip simulator (computational-only).

This extends the single-chip simulation to multiple organ compartments (e.g., liver, heart, lung)
connected by a shared microfluidic loop. Each organ has its own micro-environment, cell
populations, telomere dynamics, telomerase controllers, and inter-organ transport of
nutrients/waste/signals.

NOT FOR BIOLOGICAL USE: All DNA/telomere tokens are synthetic placeholders. No wet-lab
instructions or real sequence-level data are included.

How to run:
1. Save as virtual_multi_organ_chip.py
2. pip install numpy matplotlib torch networkx tqdm
3. python virtual_multi_organ_chip.py

Outputs: time-series plots of organ-level metrics, per-organ heatmaps, and simple cross-organ
transfer visualizations.
"""

import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
import networkx as nx
from tqdm import trange
from collections import defaultdict

# -------------------------
# Global synthetic telomere tokens (safe placeholders)
# -------------------------
TEL_UNIT = "TTAGGX"  # synthetic token only
SYN_TEL_SHORT = 1000.0
SYN_TEL_LONG = 10000.0

# -------------------------
# Generic virtual cell definition
# -------------------------
class Cell:
    def __init__(self, pos, stemness=0.0, phago_eff=0.5, organ_name='organ'):
        self.pos = pos
        self.organ = organ_name
        self.telomere_length = SYN_TEL_SHORT if stemness < 0.5 else SYN_TEL_LONG
        self.telomerase_activity = 0.0
        self.division_rate = 0.0
        self.damage_index = np.random.uniform(0.0, 0.05)
        self.metabolic_load = np.random.uniform(0.0, 0.1)
        self.phagocytosis_efficiency = phago_eff
        self.stemness = stemness
        self.alive = True

    def state_vector(self):
        return np.array([
            self.telomere_length / SYN_TEL_LONG,
            self.telomerase_activity,
            self.damage_index,
            self.metabolic_load,
            self.phagocytosis_efficiency,
            self.stemness
        ], dtype=np.float32)

# -------------------------
# Telomerase controller (same architecture reused per organ)
# -------------------------
class TelomeraseController(nn.Module):
    def __init__(self, input_dim=6, hidden=64):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden),
            nn.ReLU(),
            nn.Linear(hidden, hidden//2),
            nn.ReLU(),
            nn.Linear(hidden//2, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.net(x)

# -------------------------
# Organ compartment: holds a grid of cells and local environment
# -------------------------
class Organ:
    def __init__(self, name, width, height, stem_seed_prob=0.1, nutrient_base=1.0, seed=0):
        self.name = name
        self.width = width
        self.height = height
        self.cells = {}
        self.graph = nx.grid_2d_graph(width, height)
        self.nutrient = np.ones((height, width)) * nutrient_base
        self.waste = np.zeros((height, width))
        self.tissue_demand = np.zeros((height, width))
        self._populate_cells(stem_seed_prob, seed)

    def _populate_cells(self, stem_seed_prob, seed):
        np.random.seed(seed)
        for y in range(self.height):
            for x in range(self.width):
                stemness = 1.0 if np.random.rand() < stem_seed_prob else 0.0
                phago_eff = np.clip(np.random.normal(0.55, 0.12), 0.05, 0.95)
                self.cells[(x,y)] = Cell(pos=(x,y), stemness=stemness, phago_eff=phago_eff, organ_name=self.name)

    def neighborhood(self, pos):
        return [nbr for nbr in self.graph.neighbors(pos)]

    def step_local_diffusion(self):
        # simple diffusion for nutrient and waste (5-point stencil approximation)
        new_n = self.nutrient.copy()
        new_w = self.waste.copy()
        for y in range(self.height):
            for x in range(self.width):
                nbors = [(x+1,y),(x-1,y),(x,y+1),(x,y-1)]
                vals_n = [self.nutrient[y,x]]
                vals_w = [self.waste[y,x]]
                for nxp,nyp in nbors:
                    if 0 <= nxp < self.width and 0 <= nyp < self.height:
                        vals_n.append(self.nutrient[nyp,nxp])
                        vals_w.append(self.waste[nyp,nxp])
                new_n[y,x] = np.mean(vals_n)
                new_w[y,x] = np.mean(vals_w)
        # small decay / baseline consumption
        new_n = np.clip(new_n - 0.001, 0.0, 2.0)
        new_w = np.clip(new_w * 0.995, 0.0, 10.0)
        self.nutrient = new_n
        self.waste = new_w

    def update_tissue_demand(self):
        for (x,y), cell in self.cells.items():
            local_neighbors = self.neighborhood((x,y))
            avg_damage = np.mean([self.cells[n].damage_index for n in local_neighbors] + [cell.damage_index])
            local_nutr = self.nutrient[y,x]
            demand = np.clip(0.6*(avg_damage/0.5) + 0.4*(1 - local_nutr/1.0), 0.0, 1.0)
            self.tissue_demand[y,x] = demand

# -------------------------
# System that connects multiple organs through a shared circulation loop
# -------------------------
class MultiOrganSystem:
    def __init__(self, organ_specs, transfer_matrix=None, seed=0):
        """
        organ_specs: list of dicts with keys: name, width, height, stem_seed_prob, nutrient_base
        transfer_matrix: NxN matrix where entry (i,j) fraction flows from organ i -> organ j each circulation step
        """
        self.organs = []
        self.name_to_idx = {}
        for i, spec in enumerate(organ_specs):
            org = Organ(name=spec.get('name', f'org{i}'),
                        width=spec.get('width', 16),
                        height=spec.get('height', 12),
                        stem_seed_prob=spec.get('stem_seed_prob', 0.08),
                        nutrient_base=spec.get('nutrient_base', 1.0),
                        seed=seed+i)
            self.name_to_idx[org.name] = i
            self.organs.append(org)

        n = len(self.organs)
        if transfer_matrix is None:
            # default: circular loop with small cross transfer
            T = np.zeros((n,n))
            for i in range(n):
                T[i, (i+1)%n] = 0.95  # most fluid goes to next organ
                T[i, i] = 0.05         # some recirculates/stays local
            self.transfer_matrix = T
        else:
            self.transfer_matrix = np.array(transfer_matrix)

        # controllers per organ
        self.controllers = [TelomeraseController(input_dim=6, hidden=80) for _ in range(n)]
        self.optimizers = [optim.Adam(c.parameters(), lr=5e-3) for c in self.controllers]
        self.time = 0

    def circulate_and_transfer(self):
        # Aggregate organ-level bulk nutrient and waste, then redistribute based on transfer_matrix
        n = len(self.organs)
        bulk_n = np.zeros(n)
        bulk_w = np.zeros(n)
        # compute organ means
        for i, org in enumerate(self.organs):
            bulk_n[i] = org.nutrient.mean()
            bulk_w[i] = org.waste.mean()

        # apply transfer matrix: new_bulk_j = sum_i T[i,j] * bulk_i
        new_bulk_n = self.transfer_matrix.T.dot(bulk_n)
        new_bulk_w = self.transfer_matrix.T.dot(bulk_w)

        # spread redistributed bulk back to organ grids (simple uniform fill)
        for j, org in enumerate(self.organs):
            org.nutrient = np.ones_like(org.nutrient) * float(new_bulk_n[j])
            org.waste = np.ones_like(org.waste) * float(new_bulk_w[j])

    def step(self, train_controllers=True):
        self.time += 1
        # 1) local organ steps
        for org_idx, org in enumerate(self.organs):
            org.step_local_diffusion()
            org.update_tissue_demand()

            # prepare controller input batch
            states = []
            positions = []
            local_nuts = []
            demands = []
            for (x,y), cell in org.cells.items():
                states.append(cell.state_vector())
                positions.append((x,y))
                local_nuts.append(org.nutrient[y,x])
                demands.append(org.tissue_demand[y,x])
            if not states:
                continue
            states_t = torch.tensor(np.stack(states), dtype=torch.float32)
            controller = self.controllers[org_idx]

            with torch.set_grad_enabled(train_controllers):
                outputs = controller(states_t).squeeze(-1)
                # simulate each cell with its action
                loss_tensor = 0.0
                total_divisions = 0
                for i, pos in enumerate(positions):
                    cell = org.cells[pos]
                    tel_cmd = float(outputs[i].item())
                    can_divide = simulate_cell_dynamics(cell, local_nuts[i], tel_cmd)

                    # division probability tuned by organ-specific demand
                    demand = demands[i]
                    if can_divide and (demand > 0.4) and (np.random.rand() < 0.12 + 0.6*demand):
                        perform_division(org, pos)
                        total_divisions += 1

                    # surrogate loss
                    tel_norm = torch.clamp(torch.tensor(cell.telomere_length / SYN_TEL_LONG), 0.0, 2.0)
                    dmg = torch.tensor(cell.damage_index)
                    act = outputs[i]
                    loss_cell = (tel_norm - 0.8).pow(2) + 2.0*dmg + 0.5*act.pow(2)
                    loss_tensor = loss_tensor + loss_cell

                if train_controllers:
                    opt = self.optimizers[org_idx]
                    opt.zero_grad()
                    loss_tensor = loss_tensor / len(positions)
                    loss_tensor.backward()
                    opt.step()

        # 2) circulation step - exchange bulk fluid among organs
        self.circulate_and_transfer()

# -------------------------
# Reused dynamics & division functions
# -------------------------

def simulate_cell_dynamics(cell: Cell, local_nutrient, telomerase_cmd):
    alpha = 0.3
    cell.telomerase_activity = (1 - alpha) * cell.telomerase_activity + alpha * telomerase_cmd

    damage_growth = 0.01 * (1 + cell.metabolic_load) * (1 - cell.phagocytosis_efficiency)
    cell.damage_index += np.random.normal(damage_growth, 0.005)
    cell.damage_index = np.clip(cell.damage_index, 0.0, 1.0)

    basal_shrink = 1.0 * (1 + cell.metabolic_load*0.5)
    elongation = 2.0 * cell.telomerase_activity * (0.5 + cell.stemness)
    cell.telomere_length += -basal_shrink + elongation
    cell.telomere_length = max(1.0, cell.telomere_length)

    phago_clear = cell.phagocytosis_efficiency * 0.05
    cell.damage_index = max(0.0, cell.damage_index - phago_clear)

    cell.metabolic_load = np.clip(cell.metabolic_load * (0.95 + 0.05*(1-local_nutrient)) + 0.01, 0, 1)

    can_divide = (cell.damage_index < 0.25 and cell.telomere_length > SYN_TEL_SHORT*0.5)
    return can_divide


def perform_division(organ: Organ, pos):
    cell = organ.cells[pos]
    if not cell.alive:
        return
    daughter1 = Cell(pos=pos, stemness=cell.stemness, phago_eff=cell.phagocytosis_efficiency, organ_name=organ.name)
    daughter2 = Cell(pos=pos, stemness=cell.stemness, phago_eff=cell.phagocytosis_efficiency, organ_name=organ.name)
    split_len = cell.telomere_length * 0.5
    daughter1.telomere_length = split_len * (1.0 + 0.02*np.random.randn())
    daughter2.telomere_length = split_len * (1.0 + 0.02*np.random.randn())
    daughter1.damage_index = cell.damage_index * 0.5
    daughter2.damage_index = cell.damage_index * 0.5
    organ.cells[pos] = daughter1
    neighbors = organ.neighborhood(pos)
    neighbors_sorted = sorted(neighbors, key=lambda n: organ.cells[n].damage_index, reverse=True)
    if neighbors_sorted:
        chosen = neighbors_sorted[0]
        organ.cells[chosen] = daughter2

# -------------------------
# Monitoring & visualization
# -------------------------

def gather_system_stats(system: MultiOrganSystem):
    stats = {}
    for i, org in enumerate(system.organs):
        all_tels = np.array([c.telomere_length for c in org.cells.values()])
        all_dmg = np.array([c.damage_index for c in org.cells.values()])
        all_phago = np.array([c.phagocytosis_efficiency for c in org.cells.values()])
        stats[org.name] = {
            'avg_telomere': float(all_tels.mean()),
            'median_telomere': float(np.median(all_tels)),
            'avg_damage': float(all_dmg.mean()),
            'avg_phago': float(all_phago.mean()),
            'divisions': None
        }
    return stats


def plot_system_timeseries(history, fname_prefix=None):
    # history: list of dict snapshots (per time) as returned by gather_system_stats
    n_steps = len(history)
    organ_names = list(history[0].keys())
    fig, axes = plt.subplots(len(organ_names), 2, figsize=(12, 3*len(organ_names)))
    if len(organ_names) == 1:
        axes = axes.reshape(1,2)
    for i, oname in enumerate(organ_names):
        tel = [history[t][oname]['avg_telomere'] for t in range(n_steps)]
        dmg = [history[t][oname]['avg_damage'] for t in range(n_steps)]
        axes[i,0].plot(tel)
        axes[i,0].set_title(f"{oname} avg telomere")
        axes[i,1].plot(dmg)
        axes[i,1].set_title(f"{oname} avg damage")
    plt.tight_layout()
    if fname_prefix:
        plt.savefig(f"{fname_prefix}_system_timeseries.png", dpi=150)
    plt.show()


def plot_organ_heatmaps(system: MultiOrganSystem, metric='telomere'):
    n = len(system.organs)
    cols = min(3, n)
    rows = int(np.ceil(n/cols))
    fig, axs = plt.subplots(rows, cols, figsize=(4*cols, 3*rows))
    axs = np.array(axs).reshape(-1)
    for i, org in enumerate(system.organs):
        grid = np.zeros((org.height, org.width))
        for (x,y), cell in org.cells.items():
            if metric == 'telomere':
                grid[y,x] = cell.telomere_length
            elif metric == 'damage':
                grid[y,x] = cell.damage_index
            elif metric == 'phago':
                grid[y,x] = cell.phagocytosis_efficiency
            elif metric == 'nutrient':
                grid[y,x] = org.nutrient[y,x]
        ax = axs[i]
        im = ax.imshow(grid, origin='lower', cmap='viridis')
        ax.set_title(f"{org.name} - {metric}")
        plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    # hide unused axes
    for j in range(i+1, len(axs)):
        axs[j].axis('off')
    plt.tight_layout()
    plt.show()

# -------------------------
# Example experiment runner
# -------------------------

def run_multi_organ_experiment(seed=3, steps=300, train_controllers=True):
    # Define organs: liver-like (detox, higher stem seeds), heart-like (low stem), lung-like (exchange)
    organ_specs = [
        {'name': 'Liver', 'width': 20, 'height': 12, 'stem_seed_prob': 0.18, 'nutrient_base': 1.1},
        {'name': 'Heart', 'width': 14, 'height': 10, 'stem_seed_prob': 0.04, 'nutrient_base': 0.9},
        {'name': 'Lung', 'width': 18, 'height': 12, 'stem_seed_prob': 0.08, 'nutrient_base': 1.2},
    ]

    # Transfer matrix: Liver -> Heart -> Lung -> Liver loop with small leakage to self
    T = np.array([
        [0.05, 0.9, 0.05],  # Liver outputs mostly to Heart
        [0.05, 0.05, 0.9],  # Heart -> Lung
        [0.9, 0.05, 0.05],  # Lung -> Liver
    ])

    system = MultiOrganSystem(organ_specs, transfer_matrix=T, seed=seed)

    history = []
    for t in trange(steps, desc='Multi-organ sim'):
        system.step(train_controllers=train_controllers)
        if t % 5 == 0:
            stats = gather_system_stats(system)
            history.append(stats)

    return system, history

# -------------------------
# Main run
# -------------------------
if __name__ == '__main__':
    system, history = run_multi_organ_experiment(seed=7, steps=240, train_controllers=True)
    plot_system_timeseries(history, fname_prefix='multi_organ')
    plot_organ_heatmaps(system, metric='telomere')
    plot_organ_heatmaps(system, metric='damage')
    plot_organ_heatmaps(system, metric='nutrient')
    print('Multi-organ simulation finished.')

# --- Streamlit Dashboard ---
# Add this section to enable a simple dashboard for visualization
# Run with: streamlit run virtual_multi_organ_chip.py

import streamlit as st
import pandas as pd
import numpy as np

st.set_page_config(layout="wide", page_title="Virtual Multi-Organ Chip")
st.title("Virtual Multi-Organ Chip Dashboard")

# Sidebar controls
org_names = [org.name for org in system.organs]
selected_org = st.sidebar.selectbox("Select Organ", org_names)
max_steps = st.sidebar.number_input("Simulation steps to run per 'Advance'", min_value=1, max_value=1000, value=5)
run_auto = st.sidebar.checkbox("Auto-run (continuous)", value=False)
step_button = st.sidebar.button("Advance")
reset_button = st.sidebar.button("Reset Simulation")

# Initialize session state for history and system
if 'sim_initialized' not in st.session_state:
    st.session_state.sim_initialized = True
    st.session_state.step_counter = 0
    st.session_state.history = []
    st.session_state.system = system  # use the in-memory system from this script

# Reset handler
if reset_button:
    st.session_state.system, _ = run_multi_organ_experiment(seed=7, steps=0, train_controllers=True)
    st.session_state.history = []
    st.session_state.step_counter = 0
    st.success("Simulation reset")

# Advance simulation
def advance_sim(steps=1):
    for _ in range(steps):
        st.session_state.system.step(train_controllers=True)
        st.session_state.step_counter += 1
        if st.session_state.step_counter % 5 == 0:
            st.session_state.history.append(gather_system_stats(st.session_state.system))

if step_button:
    advance_sim(int(max_steps))

# Auto-run (simple loop) - note: Streamlit will rerun; using experimental single-threaded loop
if run_auto:
    # run a chunk each rerun to avoid blocking UI
    advance_sim(3)

# Prepare time-series data for plotting
history = st.session_state.history
if not history:
    st.info("No history yet — press 'Advance' to run the sim and populate plots.")

# Construct per-organ time series DataFrames
if history:
    timesteps = len(history)
    # Build DataFrames
    organ_series = {name: { 'avg_telomere': [], 'avg_damage': [] } for name in org_names}
    for h in history:
        for name in org_names:
            organ_series[name]['avg_telomere'].append(h[name]['avg_telomere'])
            organ_series[name]['avg_damage'].append(h[name]['avg_damage'])

    # Display interactive charts
    cols = st.columns([2,1])
    with cols[0]:
        st.subheader("Avg Telomere Length per Organ")
        df_tel = pd.DataFrame({name: organ_series[name]['avg_telomere'] for name in org_names})
        st.line_chart(df_tel)

        st.subheader("Avg Damage Index per Organ")
        df_dmg = pd.DataFrame({name: organ_series[name]['avg_damage'] for name in org_names})
        st.line_chart(df_dmg)

    with cols[1]:
        st.subheader(f"Selected Organ: {selected_org}")
        last_snapshot = history[-1][selected_org]
        st.metric("Avg Telomere (latest)", f"{last_snapshot['avg_telomere']:.1f}")
        st.metric("Avg Damage (latest)", f"{last_snapshot['avg_damage']:.3f}")
        st.metric("Avg Phagocytosis", f"{last_snapshot['avg_phago']:.3f}")

    # Heatmap for selected organ
    st.subheader(f"Heatmaps for {selected_org}")
    org_obj = next(o for o in st.session_state.system.organs if o.name == selected_org)
    heat_metric = st.selectbox("Heatmap metric", ["telomere","damage","phago","nutrient"], index=0)
    # build grid
    grid = np.zeros((org_obj.height, org_obj.width))
    for (x,y), cell in org_obj.cells.items():
        if heat_metric == 'telomere':
            grid[y,x] = cell.telomere_length
        elif heat_metric == 'damage':
            grid[y,x] = cell.damage_index
        elif heat_metric == 'phago':
            grid[y,x] = cell.phagocytosis_efficiency
        elif heat_metric == 'nutrient':
            grid[y,x] = org_obj.nutrient[y,x]

    fig, ax = plt.subplots(figsize=(6,4))
    im = ax.imshow(grid, origin='lower')
    ax.set_title(f"{selected_org} - {heat_metric}")
    plt.colorbar(im, ax=ax)
    st.pyplot(fig)

# Footer
st.sidebar.markdown("---")
st.sidebar.write("Steps simulated:", int(st.session_state.step_counter))
st.sidebar.write("History length:", len(st.session_state.history))

st.caption("Note: This dashboard runs a safe in-silico simulation. No real biological sequences or wet-lab instructions are included.")
```
