```python
# virtual_organ_chip.py

import numpy as np
import matplotlib.pyplot as plt
from collections import deque

# -------------------------------
# Parameters
# -------------------------------
NUM_CELLS = 50
MAX_TELOMERE = 1000  # desired "immortal" length
MIN_TELOMERE = 200   # minimal length before senescence
DIVISION_THRESHOLD = 800
PHAGOCYTOSIS_RATE = 0.1
TIME_STEPS = 100

# -------------------------------
# Cell Class
# -------------------------------
class Cell:
    def __init__(self, cell_id, telomere_length=None):
        self.id = cell_id
        self.telomere = telomere_length if telomere_length else np.random.randint(500, 700)
        self.alive = True
        self.age = 0
        self.division_history = []

    def telomerase_extend(self, ai_factor=1.0):
        """
        Telomerase extension controlled by AI factor (0-1)
        """
        if self.alive:
            extension = np.random.randint(5, 15) * ai_factor
            self.telomere = min(self.telomere + extension, MAX_TELOMERE)

    def attempt_division(self):
        """
        Determines if the cell divides based on telomere length.
        Returns a new Cell if division occurs, None otherwise.
        """
        if self.alive and self.telomere >= DIVISION_THRESHOLD:
            self.telomere /= 2  # each daughter gets half
            self.division_history.append(self.age)
            return Cell(cell_id=f"{self.id}_d{len(self.division_history)}", telomere_length=self.telomere)
        return None

    def age_cell(self):
        """
        Ages the cell and checks for senescence.
        """
        self.age += 1
        if self.telomere < MIN_TELOMERE:
            self.alive = False

# -------------------------------
# Phagocytosis Module
# -------------------------------
def phagocytosis(cells):
    """
    Removes dead cells at a given rate
    """
    alive_cells = []
    for cell in cells:
        if cell.alive or np.random.rand() > PHAGOCYTOSIS_RATE:
            alive_cells.append(cell)
    return alive_cells

# -------------------------------
# Simple AI Controller
# -------------------------------
def simple_ai_controller(cell):
    """
    Returns a telomerase activation factor (0-1)
    Can be replaced with RL/Transformer later
    """
    if cell.telomere < DIVISION_THRESHOLD:
        return 1.0  # activate telomerase
    return 0.2  # low basal activity

# -------------------------------
# Simulation Loop
# -------------------------------
def run_simulation():
    # Initialize cells
    cells = [Cell(cell_id=f"C{i}") for i in range(NUM_CELLS)]
    telomere_history = []

    for t in range(TIME_STEPS):
        new_cells = []
        for cell in cells:
            if cell.alive:
                ai_factor = simple_ai_controller(cell)
                cell.telomerase_extend(ai_factor)
                daughter = cell.attempt_division()
                if daughter:
                    new_cells.append(daughter)
                cell.age_cell()
        cells.extend(new_cells)
        cells = phagocytosis(cells)

        # Record telomere lengths
        telomere_history.append([cell.telomere for cell in cells if cell.alive])

    return cells, telomere_history

# -------------------------------
# Visualization
# -------------------------------
def plot_telomere_history(telomere_history):
    plt.figure(figsize=(12,6))
    for i, cell_telomeres in enumerate(telomere_history):
        if cell_telomeres:
            plt.plot([i]*len(cell_telomeres), cell_telomeres, 'b.', alpha=0.5)
    plt.xlabel("Time Step")
    plt.ylabel("Telomere Length")
    plt.title("Telomere Dynamics in Virtual Organ-on-a-Chip")
    plt.show()

# -------------------------------
# Main Execution
# -------------------------------
if __name__ == "__main__":
    cells, telomere_history = run_simulation()
    plot_telomere_history(telomere_history)
    print(f"Final living cells: {len([c for c in cells if c.alive])}")
```

```python
# Multi-tissue Organ-on-a-Chip RL Simulation
!pip install stable-baselines3[extra] gymnasium plotly

import numpy as np
import gymnasium as gym
from gymnasium import spaces
from stable_baselines3 import PPO
import plotly.graph_objects as go
from IPython.display import clear_output
import time

# -------------------------------
# Parameters
# -------------------------------
NUM_TISSUES = 3
CELLS_PER_TISSUE = 10
NUM_CELLS = NUM_TISSUES * CELLS_PER_TISSUE

MAX_TELOMERE = 1000
MIN_TELOMERE = 200
DIVISION_THRESHOLD = 800
PHAGOCYTOSIS_RATE = 0.1
TIME_STEPS = 50

# Environmental stress signals
# Each tissue has a stress value (0-1)
STRESS_DECAY = 0.95
STRESS_NOISE = 0.05

# -------------------------------
# Cell Class
# -------------------------------
class Cell:
    def __init__(self, cell_id, tissue_id, telomere_length=None):
        self.id = cell_id
        self.tissue_id = tissue_id
        self.telomere = telomere_length if telomere_length else np.random.randint(500, 700)
        self.alive = True
        self.age = 0

    def apply_telomerase(self, factor):
        """Telomerase activity controlled by RL agent (0-1)"""
        if self.alive:
            extension = np.random.randint(5, 15) * factor
            self.telomere = min(self.telomere + extension, MAX_TELOMERE)

    def attempt_division(self):
        if self.alive and self.telomere >= DIVISION_THRESHOLD:
            self.telomere /= 2
            return Cell(cell_id=f"{self.id}_d", tissue_id=self.tissue_id)
        return None

    def age_cell(self, stress_factor=0):
        self.age += 1
        # Stress accelerates telomere shortening
        self.telomere -= stress_factor * 10
        if self.telomere < MIN_TELOMERE:
            self.alive = False

# -------------------------------
# Gym Environment
# -------------------------------
class OrganoChipEnv(gym.Env):
    metadata = {"render.modes": ["human"]}

    def __init__(self):
        super(OrganoChipEnv, self).__init__()
        self.cells = [Cell(f"C{i}", tissue_id=i//CELLS_PER_TISSUE) for i in range(NUM_CELLS)]
        self.stress_signals = np.random.rand(NUM_TISSUES) * 0.2
        self.observation_space = spaces.Box(low=0, high=MAX_TELOMERE,
                                            shape=(NUM_CELLS + NUM_TISSUES,), dtype=np.float32)
        self.action_space = spaces.Box(low=0, high=1, shape=(NUM_CELLS,), dtype=np.float32)
        self.time_step = 0

    def step(self, action):
        new_cells = []

        # Update environmental stress (stochastic decay)
        self.stress_signals = self.stress_signals * STRESS_DECAY + np.random.randn(NUM_TISSUES)*STRESS_NOISE
        self.stress_signals = np.clip(self.stress_signals, 0, 1)

        for i, cell in enumerate(self.cells):
            tissue_stress = self.stress_signals[cell.tissue_id]
            if cell.alive:
                cell.apply_telomerase(action[i])
                daughter = cell.attempt_division()
                if daughter:
                    new_cells.append(daughter)
                cell.age_cell(stress_factor=tissue_stress)

        self.cells.extend(new_cells)
        # Phagocytosis
        self.cells = [c for c in self.cells if c.alive or np.random.rand() > PHAGOCYTOSIS_RATE]

        # Observation: telomeres + stress per tissue
        obs_telomeres = [c.telomere for c in self.cells[:NUM_CELLS]] + [0]*(NUM_CELLS - len(self.cells))
        obs = np.concatenate([obs_telomeres, self.stress_signals])
        reward = sum([min(c.telomere/MAX_TELOMERE, 1.0) for c in self.cells if c.alive])
        self.time_step += 1
        done = self.time_step >= TIME_STEPS
        return obs.astype(np.float32), reward, done, False, {}

    def reset(self, seed=None, options=None):
        self.cells = [Cell(f"C{i}", tissue_id=i//CELLS_PER_TISSUE) for i in range(NUM_CELLS)]
        self.stress_signals = np.random.rand(NUM_TISSUES) * 0.2
        self.time_step = 0
        obs_telomeres = [c.telomere for c in self.cells]
        obs = np.concatenate([obs_telomeres, self.stress_signals])
        return obs.astype(np.float32), {}

    def render(self, attention=None):
        fig = go.Figure()
        for i, cell in enumerate(self.cells):
            if cell.alive:
                fig.add_trace(go.Scatter3d(
                    x=[self.time_step],
                    y=[i],
                    z=[cell.telomere],
                    mode='markers',
                    marker=dict(
                        size=5,
                        color=attention[i] if attention is not None else 0.5,
                        colorscale='Viridis',
                        cmin=0,
                        cmax=1
                    ),
                    name=f"Tissue {cell.tissue_id} Cell {i}"
                ))
        fig.update_layout(
            scene=dict(
                xaxis_title='Time Step',
                yaxis_title='Cell ID',
                zaxis_title='Telomere Length',
            ),
            title=f"Time Step {self.time_step}",
            width=900,
            height=700
        )
        clear_output(wait=True)
        fig.show()

# -------------------------------
# Train RL Agent
# -------------------------------
env = OrganoChipEnv()
agent = PPO("MlpPolicy", env, verbose=1)
agent.learn(total_timesteps=20000)

# -------------------------------
# Test & Real-Time Visualization
# -------------------------------
obs, _ = env.reset()
telomere_history = []
attention_history = []

for t in range(TIME_STEPS):
    action, _ = agent.predict(obs, deterministic=True)
    obs, reward, done, _, _ = env.step(action)
    telomere_history.append(obs[:NUM_CELLS].copy())
    attention_history.append(action.copy())
    env.render(attention=action)
    time.sleep(0.5)
    if done:
        break

# -------------------------------
# Optional: 3D Static Plot
# -------------------------------
fig = go.Figure()
for cell_id in range(NUM_CELLS):
    fig.add_trace(go.Scatter3d(
        x=np.arange(TIME_STEPS),
        y=[cell_id]*TIME_STEPS,
        z=[telomere_history[t][cell_id] for t in range(TIME_STEPS)],
        mode='lines+markers',
        marker=dict(
            size=4,
            color=[attention_history[t][cell_id] for t in range(TIME_STEPS)],
            colorscale='Viridis',
            cmin=0, cmax=1
        ),
        line=dict(width=2),
        name=f"Cell {cell_id}"
    ))
fig.update_layout(
    scene=dict(
        xaxis_title='Time Step',
        yaxis_title='Cell ID',
        zaxis_title='Telomere Length',
    ),
    title='3D Telomere Dynamics Across Multi-Tissue Organoids',
    width=900,
    height=700
)
fig.show()
```

```python
# Advanced Multi-Tissue Organ-on-a-Chip RL Simulation
!pip install stable-baselines3[extra] gymnasium plotly

import numpy as np
import gymnasium as gym
from gymnasium import spaces
from stable_baselines3 import PPO
import plotly.graph_objects as go
from IPython.display import clear_output
import time

# -------------------------------
# Simulation Parameters
# -------------------------------
NUM_TISSUES = 3
CELLS_PER_TISSUE = 10
NUM_CELLS = NUM_TISSUES * CELLS_PER_TISSUE

MAX_TELOMERE = 1000
MIN_TELOMERE = 200
DIVISION_THRESHOLD = 800

TIME_STEPS = 50

# Environmental signals
STRESS_DECAY = 0.95
STRESS_NOISE = 0.05
NUTRIENT_DECAY = 0.98
NUTRIENT_NOISE = 0.05

# Phagocytosis efficiency per cell (0-1)
PHAGOCYTOSIS_MEAN = 0.1
PHAGOCYTOSIS_STD = 0.05

# DNA damage per cell (0-1)
DNA_DAMAGE_NOISE = 0.02

# -------------------------------
# Cell Class
# -------------------------------
class Cell:
    def __init__(self, cell_id, tissue_id, telomere_length=None):
        self.id = cell_id
        self.tissue_id = tissue_id
        self.telomere = telomere_length if telomere_length else np.random.randint(500,700)
        self.alive = True
        self.age = 0
        self.dna_damage = np.random.rand() * 0.1
        self.phagocytosis_efficiency = np.clip(np.random.randn()*PHAGOCYTOSIS_STD + PHAGOCYTOSIS_MEAN, 0, 1)

    def apply_telomerase(self, factor):
        if self.alive:
            extension = np.random.randint(5,15) * factor
            self.telomere = min(self.telomere + extension, MAX_TELOMERE)

    def attempt_division(self):
        if self.alive and self.telomere >= DIVISION_THRESHOLD:
            self.telomere /= 2
            return Cell(cell_id=f"{self.id}_d", tissue_id=self.tissue_id)
        return None

    def age_cell(self, stress_factor=0, nutrient_factor=1.0):
        self.age += 1
        # Telomere shortening accelerated by stress and DNA damage, mitigated by nutrients
        damage_effect = self.dna_damage * 10
        self.telomere -= stress_factor*10 + damage_effect * (1-nutrient_factor)
        # DNA damage stochastic update
        self.dna_damage = np.clip(self.dna_damage + np.random.randn()*DNA_DAMAGE_NOISE, 0, 1)
        if self.telomere < MIN_TELOMERE:
            self.alive = False

# -------------------------------
# Gym Environment
# -------------------------------
class OrganoChipEnv(gym.Env):
    metadata = {"render.modes": ["human"]}

    def __init__(self):
        super().__init__()
        self.cells = [Cell(f"C{i}", tissue_id=i//CELLS_PER_TISSUE) for i in range(NUM_CELLS)]
        self.stress_signals = np.random.rand(NUM_TISSUES) * 0.2
        self.nutrient_signals = np.ones(NUM_TISSUES)
        obs_len = NUM_CELLS + NUM_TISSUES*2  # telomeres + stress + nutrient
        self.observation_space = spaces.Box(low=0, high=MAX_TELOMERE, shape=(obs_len,), dtype=np.float32)
        self.action_space = spaces.Box(low=0, high=1, shape=(NUM_CELLS,), dtype=np.float32)
        self.time_step = 0

    def step(self, action):
        new_cells = []

        # Update environmental signals
        self.stress_signals = np.clip(self.stress_signals*STRESS_DECAY + np.random.randn(NUM_TISSUES)*STRESS_NOISE, 0, 1)
        self.nutrient_signals = np.clip(self.nutrient_signals*NUTRIENT_DECAY + np.random.randn(NUM_TISSUES)*NUTRIENT_NOISE, 0, 1)

        for i, cell in enumerate(self.cells):
            tissue_stress = self.stress_signals[cell.tissue_id]
            tissue_nutrient = self.nutrient_signals[cell.tissue_id]
            if cell.alive:
                cell.apply_telomerase(action[i])
                daughter = cell.attempt_division()
                if daughter:
                    new_cells.append(daughter)
                cell.age_cell(stress_factor=tissue_stress, nutrient_factor=tissue_nutrient)

        self.cells.extend(new_cells)
        # Phagocytosis: remove dead cells probabilistically based on efficiency
        self.cells = [c for c in self.cells if c.alive or np.random.rand() > c.phagocytosis_efficiency]

        # Observation: telomeres + stress + nutrient signals
        obs_telomeres = [c.telomere for c in self.cells[:NUM_CELLS]] + [0]*(NUM_CELLS-len(self.cells))
        obs = np.concatenate([obs_telomeres, self.stress_signals, self.nutrient_signals])
        reward = sum([min(c.telomere/MAX_TELOMERE,1.0) for c in self.cells if c.alive])
        self.time_step += 1
        done = self.time_step >= TIME_STEPS
        return obs.astype(np.float32), reward, done, False, {}

    def reset(self, seed=None, options=None):
        self.cells = [Cell(f"C{i}", tissue_id=i//CELLS_PER_TISSUE) for i in range(NUM_CELLS)]
        self.stress_signals = np.random.rand(NUM_TISSUES) * 0.2
        self.nutrient_signals = np.ones(NUM_TISSUES)
        self.time_step = 0
        obs_telomeres = [c.telomere for c in self.cells]
        obs = np.concatenate([obs_telomeres, self.stress_signals, self.nutrient_signals])
        return obs.astype(np.float32), {}

    def render(self, attention=None):
        fig = go.Figure()
        for i, cell in enumerate(self.cells):
            if cell.alive:
                fig.add_trace(go.Scatter3d(
                    x=[self.time_step],
                    y=[i],
                    z=[cell.telomere],
                    mode='markers',
                    marker=dict(
                        size=5,
                        color=attention[i] if attention is not None else 0.5,
                        colorscale='Viridis',
                        cmin=0,
                        cmax=1
                    ),
                    name=f"Tissue {cell.tissue_id} Cell {i}"
                ))
        fig.update_layout(
            scene=dict(
                xaxis_title='Time Step',
                yaxis_title='Cell ID',
                zaxis_title='Telomere Length',
            ),
            title=f"Time Step {self.time_step}",
            width=900,
            height=700
        )
        clear_output(wait=True)
        fig.show()

# -------------------------------
# Train RL Agent
# -------------------------------
env = OrganoChipEnv()
agent = PPO("MlpPolicy", env, verbose=1)
agent.learn(total_timesteps=20000)

# -------------------------------
# Test & Real-Time Visualization
# -------------------------------
obs, _ = env.reset()
telomere_history = []
attention_history = []

for t in range(TIME_STEPS):
    action, _ = agent.predict(obs, deterministic=True)
    obs, reward, done, _, _ = env.step(action)
    telomere_history.append(obs[:NUM_CELLS].copy())
    attention_history.append(action.copy())
    env.render(attention=action)
    time.sleep(0.5)
    if done:
        break

# -------------------------------
# Optional: Static 3D Plot
# -------------------------------
fig = go.Figure()
for cell_id in range(NUM_CELLS):
    fig.add_trace(go.Scatter3d(
        x=np.arange(TIME_STEPS),
        y=[cell_id]*TIME_STEPS,
        z=[telomere_history[t][cell_id] for t in range(TIME_STEPS)],
        mode='lines+markers',
        marker=dict(
            size=4,
            color=[attention_history[t][cell_id] for t in range(TIME_STEPS)],
            colorscale='Viridis',
            cmin=0,
            cmax=1
        ),
        line=dict(width=2),
        name=f"Cell {cell_id}"
    ))
fig.update_layout(
    scene=dict(
        xaxis_title='Time Step',
        yaxis_title='Cell ID',
        zaxis_title='Telomere Length',
    ),
    title='3D Telomere Dynamics with Multi-Tissue, DNA Damage & Nutrients',
    width=900,
    height=700
)
fig.show()
```

```python
# Fully Integrated Multi-Tissue Organ-on-a-Chip RL Simulation
!pip install stable-baselines3[extra] gymnasium plotly moviepy opencv-python

import numpy as np
import gymnasium as gym
from gymnasium import spaces
from stable_baselines3 import PPO
import plotly.graph_objects as go
from PIL import Image
import tempfile
from moviepy.editor import ImageSequenceClip
import time

# -------------------------------
# Simulation Parameters
# -------------------------------
NUM_TISSUES = 3
CELLS_PER_TISSUE = 10
NUM_CELLS = NUM_TISSUES * CELLS_PER_TISSUE
TIME_STEPS = 50

MAX_TELOMERE = 1000
MIN_TELOMERE = 200
DIVISION_THRESHOLD = 800

STRESS_DECAY = 0.95
STRESS_NOISE = 0.05
NUTRIENT_DECAY = 0.98
NUTRIENT_NOISE = 0.05

PHAGOCYTOSIS_MEAN = 0.1
PHAGOCYTOSIS_STD = 0.05
DNA_DAMAGE_NOISE = 0.02

# -------------------------------
# Cell Class
# -------------------------------
class Cell:
    def __init__(self, cell_id, tissue_id, telomere_length=None):
        self.id = cell_id
        self.tissue_id = tissue_id
        self.telomere = telomere_length if telomere_length else np.random.randint(500,700)
        self.alive = True
        self.age = 0
        self.dna_damage = np.random.rand()*0.1
        self.phagocytosis_efficiency = np.clip(np.random.randn()*PHAGOCYTOSIS_STD + PHAGOCYTOSIS_MEAN,0,1)
        self.divisions = 0  # track number of divisions

    def apply_telomerase(self, factor):
        if self.alive:
            extension = np.random.randint(5,15) * factor
            self.telomere = min(self.telomere + extension, MAX_TELOMERE)

    def attempt_division(self):
        if self.alive and self.telomere >= DIVISION_THRESHOLD:
            self.telomere /= 2
            self.divisions += 1
            daughter = Cell(f"{self.id}_d{self.divisions}", self.tissue_id, telomere_length=self.telomere)
            return daughter
        return None

    def age_cell(self, stress_factor=0, nutrient_factor=1.0):
        self.age += 1
        damage_effect = self.dna_damage * 10
        self.telomere -= stress_factor*10 + damage_effect*(1-nutrient_factor)
        self.dna_damage = np.clip(self.dna_damage + np.random.randn()*DNA_DAMAGE_NOISE,0,1)
        if self.telomere < MIN_TELOMERE:
            self.alive = False

# -------------------------------
# Gym Environment
# -------------------------------
class OrganoChipEnv(gym.Env):
    metadata = {"render.modes":["human"]}

    def __init__(self):
        super().__init__()
        self.cells = [Cell(f"C{i}", tissue_id=i//CELLS_PER_TISSUE) for i in range(NUM_CELLS)]
        self.stress_signals = np.random.rand(NUM_TISSUES)*0.2
        self.nutrient_signals = np.ones(NUM_TISSUES)
        obs_len = NUM_CELLS + NUM_TISSUES*2
        self.observation_space = spaces.Box(low=0, high=MAX_TELOMERE, shape=(obs_len,), dtype=np.float32)
        self.action_space = spaces.Box(low=0, high=1, shape=(NUM_CELLS,), dtype=np.float32)
        self.time_step = 0
        self.frames = []

    def step(self, action):
        new_cells = []

        # Update stress & nutrients
        self.stress_signals = np.clip(self.stress_signals*STRESS_DECAY + np.random.randn(NUM_TISSUES)*STRESS_NOISE,0,1)
        self.nutrient_signals = np.clip(self.nutrient_signals*NUTRIENT_DECAY + np.random.randn(NUM_TISSUES)*NUTRIENT_NOISE,0,1)

        for i, cell in enumerate(self.cells):
            tissue_stress = self.stress_signals[cell.tissue_id]
            tissue_nutrient = self.nutrient_signals[cell.tissue_id]
            if cell.alive:
                cell.apply_telomerase(action[i])
                daughter = cell.attempt_division()
                if daughter:
                    new_cells.append(daughter)
                cell.age_cell(stress_factor=tissue_stress, nutrient_factor=tissue_nutrient)

        self.cells.extend(new_cells)
        # Phagocytosis: remove dead cells probabilistically
        self.cells = [c for c in self.cells if c.alive or np.random.rand() > c.phagocytosis_efficiency]

        # Observation: telomeres + stress + nutrient
        obs_telomeres = [c.telomere for c in self.cells[:NUM_CELLS]] + [0]*(NUM_CELLS-len(self.cells))
        obs = np.concatenate([obs_telomeres, self.stress_signals, self.nutrient_signals])
        reward = sum([min(c.telomere/MAX_TELOMERE,1.0) for c in self.cells if c.alive])
        self.time_step += 1
        done = self.time_step >= TIME_STEPS
        return obs.astype(np.float32), reward, done, False, {}

    def reset(self, seed=None, options=None):
        self.cells = [Cell(f"C{i}", tissue_id=i//CELLS_PER_TISSUE) for i in range(NUM_CELLS)]
        self.stress_signals = np.random.rand(NUM_TISSUES)*0.2
        self.nutrient_signals = np.ones(NUM_TISSUES)
        self.time_step = 0
        self.frames = []
        obs_telomeres = [c.telomere for c in self.cells]
        obs = np.concatenate([obs_telomeres, self.stress_signals, self.nutrient_signals])
        return obs.astype(np.float32), {}

    # -------------------------------
    # Render and save frames for video
    # -------------------------------
    def render_frame(self, action=None):
        fig = go.Figure()
        # Stress & nutrient planes
        for tissue_id in range(NUM_TISSUES):
            fig.add_trace(go.Mesh3d(
                x=[0,TIME_STEPS-1,TIME_STEPS-1,0],
                y=[tissue_id*CELLS_PER_TISSUE]*4,
                z=[self.stress_signals[tissue_id]]*4,
                color='red',
                opacity=0.2,
                name=f"Stress Tissue {tissue_id}"
            ))
            fig.add_trace(go.Mesh3d(
                x=[0,TIME_STEPS-1,TIME_STEPS-1,0],
                y=[tissue_id*CELLS_PER_TISSUE]*4,
                z=[self.nutrient_signals[tissue_id]]*4,
                color='green',
                opacity=0.2,
                name=f"Nutrient Tissue {tissue_id}"
            ))

        # Cell markers
        for i, cell in enumerate(self.cells[:NUM_CELLS]):
            if cell.alive:
                fig.add_trace(go.Scatter3d(
                    x=[self.time_step],
                    y=[i],
                    z=[cell.telomere],
                    mode='markers',
                    marker=dict(
                        size=5 + cell.dna_damage*15,
                        color=action[i] if action is not None else 0.5,
                        colorscale='Viridis',
                        cmin=0,
                        cmax=1,
                        opacity=0.9
                    ),
                    name=f"Cell {i}"
                ))
                # Division marker
                if cell.divisions > 0:
                    fig.add_trace(go.Scatter3d(
                        x=[self.time_step],
                        y=[i],
                        z=[cell.telomere+20],
                        mode='markers',
                        marker=dict(size=10, symbol='star', color='gold', opacity=1.0),
                        name=f"Division Cell {i}"
                    ))

        fig.update_layout(
            scene=dict(
                xaxis_title='Time Step',
                yaxis_title='Cell ID',
                zaxis_title='Telomere Length'
            ),
            width=900,
            height=700,
            title=f"Step {self.time_step}"
        )

        # Save frame
        img_bytes = fig.to_image(format="png")
        img = np.array(Image.open(tempfile.NamedTemporaryFile(delete=False, suffix=".png").name))
        self.frames.append(img)

# -------------------------------
# Initialize environment and RL agent
# -------------------------------
env = OrganoChipEnv()
agent = PPO("MlpPolicy", env, verbose=1)
agent.learn(total_timesteps=20000)

# -------------------------------
# Run simulation and capture frames
# -------------------------------
obs, _ = env.reset()
for t in range(TIME_STEPS):
    action, _ = agent.predict(obs, deterministic=True)
    obs, reward, done, _, _ = env.step(action)
    env.render_frame(action)
    time.sleep(0.2)
    if done:
        break

# -------------------------------
# Export video
# -------------------------------
video_path = "organ_chip_full_simulation.mp4"
clip = ImageSequenceClip([Image.fromarray(f) for f in env.frames], fps=2)
clip.write_videofile(video_path, codec="libx264")
print(f"Video saved to {video_path}")
```
