````python
"""
Colab-ready simulation skeleton: Virtual cell with synthetic telomere, telomerase controller,
phagocytic subsystem, environment, and simple visualization.

How to use in Colab:
1. Upload this file or paste into a notebook cell.
2. (Optional) Install dependencies:
   !pip install numpy matplotlib pandas tqdm imageio
   # For RL controller (optional):
   !pip install stable-baselines3[extra] torch gym
3. Run the demo at the bottom to execute a short simulation and produce plots/CSV.

This file intentionally keeps heavy ML components optional. The provided controllers:
- Simple PID-like controller (built-in)
- Optional RLController placeholder (requires stable-baselines3 + gym)

Outputs:
- Time-series CSV (telomere lengths, cell health, debris)
- Plots (matplotlib)
- Optional video export (imageio)

Author: Generated by ChatGPT â€” adapt as you like.
"""

import math
import random
import time
from dataclasses import dataclass, field
from typing import List, Dict, Any, Optional

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import trange

# -----------------------------
# Utility functions
# -----------------------------

def seed_everything(seed: int = 42):
    random.seed(seed)
    np.random.seed(seed)


# -----------------------------
# Telomere model
# -----------------------------
@dataclass
class Telomere:
    length: float  # abstract units (e.g., repeats)
    min_length: float = 5.0  # trigger senescence
    max_length: float = 200.0

    def shorten(self, amount: float):
        self.length = max(0.0, self.length - amount)

    def extend(self, amount: float):
        self.length = min(self.max_length, self.length + amount)

    def is_senescent(self) -> bool:
        return self.length < self.min_length


# -----------------------------
# Virtual Cell
# -----------------------------
@dataclass
class VirtualCell:
    id: int
    telomere: Telomere
    health: float = 1.0  # 0..1
    dna_damage: float = 0.0  # aggregated damage
    dividing: bool = False
    age: int = 0

    def step(self, env: "Environment", telomerase_activity: float):
        """Advance one timestep for this cell.

        - Telomere shortens during each division or due to background degradation.
        - Telomerase activity can extend telomere length.
        - DNA damage accumulates from environment stress; phagocytic cleanup in env reduces it.
        - Health is a function of DNA damage and telomere length.
        """
        # Background shortening / oxidative damage
        background_shortening = env.background_shorten_rate * (1.0 + env.oxidative_stress)
        self.telomere.shorten(background_shortening)

        # Telomerase-driven extension (bounded)
        extension = telomerase_activity * env.telomerase_efficiency
        self.telomere.extend(extension)

        # Accumulate DNA damage from environment
        damage_increase = env.stress_to_damage_rate * env.oxidative_stress
        self.dna_damage += damage_increase

        # Health update
        telomere_factor = min(1.0, self.telomere.length / (self.telomere.max_length * 0.5))
        damage_factor = max(0.0, 1.0 - self.dna_damage)
        self.health = max(0.0, 0.7 * telomere_factor + 0.3 * damage_factor)

        # Age and division decision
        self.age += 1
        self.dividing = self.should_divide(env)

    def should_divide(self, env: "Environment") -> bool:
        # Simple heuristic: divide if healthy and telomere above threshold and resources available
        return (
            self.health > 0.6
            and self.telomere.length > (self.telomere.min_length + 10)
            and env.nutrients > env.min_nutrients_to_divide
        )


# -----------------------------
# Phagocyte / cleanup subsystem (abstract)
# -----------------------------
@dataclass
class Phagocyte:
    efficiency: float = 0.5  # fraction of debris removed per timestep

    def cleanup(self, env: "Environment"):
        removed = min(env.debris, env.debris * self.efficiency)
        env.debris -= removed
        # convert removed debris to nutrient recycling (simple model)
        env.nutrients += removed * env.recycling_yield


# -----------------------------
# Environment
# -----------------------------
@dataclass
class Environment:
    nutrients: float = 1.0
    debris: float = 0.0
    oxidative_stress: float = 0.0  # 0..1
    background_shorten_rate: float = 0.1
    telomerase_efficiency: float = 1.0
    stress_to_damage_rate: float = 0.01
    min_nutrients_to_divide: float = 0.3
    recycling_yield: float = 0.5

    def step(self, phagocyte: Phagocyte):
        # Nutrient consumption proportional to cell count will be handled outside
        # Phagocyte cleans up debris
        phagocyte.cleanup(self)


# -----------------------------
# Controllers
# -----------------------------
class PIDTelomeraseController:
    """Simple PID-like controller that aims to keep mean telomere length at a target.

    This is intentionally simple and deterministic; it's useful as a baseline.
    """

    def __init__(self, target: float, kp: float = 0.1, ki: float = 0.01, kd: float = 0.0):
        self.target = target
        self.kp = kp
        self.ki = ki
        self.kd = kd
        self.integral = 0.0
        self.last_error = None

    def action(self, mean_telomere: float) -> float:
        error = self.target - mean_telomere
        self.integral += error
        derivative = 0.0 if (self.last_error is None) else (error - self.last_error)
        self.last_error = error
        # raw action value
        raw = self.kp * error + self.ki * self.integral + self.kd * derivative
        # map to [0,1]
        return float(np.clip(raw, 0.0, 1.0))


class RLControllerPlaceholder:
    """Placeholder for an RL controller. If you want to use SB3, replace this with a model.

    Expected interface: .action(mean_telomere, mean_health, env_state) -> telomerase_activity (0..1)
    """

    def __init__(self):
        pass

    def action(self, mean_telomere: float, mean_health: float, env_state: Dict[str, float]) -> float:
        # Simple learned heuristic fallback: increase telomerase when telomeres drop
        if mean_telomere < 50.0:
            return 0.8
        if mean_health < 0.5:
            return 0.4
        return 0.1


# -----------------------------
# Simulator
# -----------------------------
@dataclass
class Simulator:
    n_cells: int = 50
    steps: int = 1000
    env: Environment = field(default_factory=Environment)
    phagocyte: Phagocyte = field(default_factory=lambda: Phagocyte(efficiency=0.6))
    controller: Any = field(default_factory=lambda: PIDTelomeraseController(target=100.0))
    seed: int = 42

    # runtime traces
    traces: List[Dict[str, Any]] = field(default_factory=list)

    def __post_init__(self):
        seed_everything(self.seed)
        # initialize cells with slight heterogeneity
        self.cells: List[VirtualCell] = []
        for i in range(self.n_cells):
            initial_len = np.clip(np.random.normal(120.0, 10.0), 10.0, 160.0)
            tel = Telomere(length=initial_len)
            cell = VirtualCell(id=i, telomere=tel)
            self.cells.append(cell)

    def step(self, step_idx: int):
        # compute mean telomere length
        mean_tel = float(np.mean([c.telomere.length for c in self.cells]))
        mean_health = float(np.mean([c.health for c in self.cells]))

        # controller decides telomerase activity
        if hasattr(self.controller, "action"):
            # support both controllers with different signatures
            try:
                telomerase_activity = self.controller.action(mean_tel)
            except TypeError:
                telomerase_activity = self.controller.action(mean_tel, mean_health, {
                    "oxidative_stress": self.env.oxidative_stress,
                    "debris": self.env.debris,
                })
        else:
            telomerase_activity = 0.1

        # environment-driven nutrient consumption
        consumption_per_cell = 0.01
        total_consumption = consumption_per_cell * len(self.cells)
        self.env.nutrients = max(0.0, self.env.nutrients - total_consumption)

        # Step each cell
        new_cells: List[VirtualCell] = []
        for cell in self.cells:
            cell.step(self.env, telomerase_activity)

            # division event
            if cell.dividing:
                # Division consumes nutrients and shortens telomeres
                self.env.nutrients = max(0.0, self.env.nutrients - 0.05)
                cell.telomere.shorten(1.0 + np.random.rand() * 0.5)
                # daughter cell inherits telomere length with noise
                daughter_len = max(0.0, cell.telomere.length - np.random.normal(0.5, 0.2))
                daughter = VirtualCell(
                    id=max(c.id for c in self.cells + new_cells) + 1,
                    telomere=Telomere(length=daughter_len),
                )
                new_cells.append(daughter)

            # senescence or death
            if cell.telomere.is_senescent() or cell.health < 0.05:
                # cell becomes debris
                self.env.debris += 0.5 + random.random() * 0.5
                continue  # drop the cell (phagocyte will clean debris)
            else:
                new_cells.append(cell)

        # append new population
        self.cells = new_cells

        # environment step (phagocyte cleans up)
        self.env.step(self.phagocyte)

        # record trace
        trace = {
            "step": step_idx,
            "n_cells": len(self.cells),
            "mean_telomere": float(np.mean([c.telomere.length for c in self.cells])) if len(self.cells) > 0 else 0.0,
            "mean_health": float(np.mean([c.health for c in self.cells])) if len(self.cells) > 0 else 0.0,
            "nutrients": float(self.env.nutrients),
            "debris": float(self.env.debris),
            "oxidative_stress": float(self.env.oxidative_stress),
            "telomerase_activity": float(telomerase_activity),
        }
        self.traces.append(trace)

    def run(self, steps: Optional[int] = None, progress: bool = True):
        steps = steps or self.steps
        iterator = trange(steps) if progress else range(steps)
        for t in iterator:
            # optional: vary oxidative stress or nutrients over time to simulate environment
            if t % 200 == 0 and t > 0:
                # small pulsed oxidative event
                self.env.oxidative_stress = min(1.0, self.env.oxidative_stress + 0.2)
            # decay stress slowly
            self.env.oxidative_stress = max(0.0, self.env.oxidative_stress - 0.002)

            self.step(t)

    def to_dataframe(self) -> pd.DataFrame:
        return pd.DataFrame(self.traces)

    def save_csv(self, path: str = "simulation_traces.csv"):
        df = self.to_dataframe()
        df.to_csv(path, index=False)
        return path

    def plot(self, save_path: Optional[str] = None):
        df = self.to_dataframe()
        fig, axs = plt.subplots(3, 1, figsize=(8, 10), constrained_layout=True)

        axs[0].plot(df["step"], df["mean_telomere"])  # default matplotlib colors
        axs[0].set_title("Mean Telomere Length")

        axs[1].plot(df["step"], df["mean_health"])
        axs[1].set_title("Mean Cell Health")

        axs[2].plot(df["step"], df["n_cells"], label="n_cells")
        axs[2].plot(df["step"], df["nutrients"], label="nutrients")
        axs[2].plot(df["step"], df["debris"], label="debris")
        axs[2].set_title("Population / Environment")
        axs[2].legend()

        if save_path:
            fig.savefig(save_path)
        plt.show()


# -----------------------------
# Demo / main
# -----------------------------

def demo_run(seed: int = 123, steps: int = 800, use_rl: bool = False):
    seed_everything(seed)

    env = Environment(nutrients=5.0, debris=0.0, oxidative_stress=0.05)
    phag = Phagocyte(efficiency=0.7)

    if use_rl:
        controller = RLControllerPlaceholder()
    else:
        controller = PIDTelomeraseController(target=120.0, kp=0.02, ki=0.001, kd=0.0)

    sim = Simulator(n_cells=60, steps=steps, env=env, phagocyte=phag, controller=controller, seed=seed)

    sim.run(steps, progress=True)

    df = sim.to_dataframe()
    csv_path = sim.save_csv("colab_simulation_traces.csv")
    print(f"Saved CSV to: {csv_path}")

    sim.plot(save_path="colab_simulation_plots.png")

    return sim, df


# If run as a script in Colab, the user can call demo_run()
if __name__ == "__main__":
    sim, df = demo_run(seed=42, steps=400, use_rl=False)
    print(df.tail())
````

````python
"""
Colab-ready simulation skeleton: Virtual cell with synthetic telomere, telomerase controller,
phagocytic subsystem, environment, and simple visualization.

Extended with molecular-detail modules:
- TelomereMolecularModel: sequence-aware stability scoring and oxidative-susceptibility model
- TLoopModel: placeholder for short, GPU-free coarse MD steps or analytic approximations
- TelomeraseBindingModel: docking-like scoring function for small-molecule/peptide modifiers
- GNN placeholder: scaffold showing how to plug in a Graph Neural Network (PyTorch) to
  predict local telomere interaction propensities (requires additional dependencies)

How to use in Colab:
1. Upload this file or paste into a notebook cell.
2. (Optional) Install dependencies:
   !pip install numpy matplotlib pandas tqdm imageio
   # For advanced molecular/GNN features (optional):
   # !pip install torch torch-geometric rdkit-pypi
3. Run the demo at the bottom to execute a short simulation and produce plots/CSV.

Note: Molecular modules are coarse, computationally light placeholders intended for
integration and testing in an organ-on-chip scale simulator. Replace with high-fidelity
MD or docking tools (OpenMM, AutoDock Vina) if you need physically accurate results.

Author: Generated by ChatGPT â€” adapt as you like.
"""

import math
import random
import time
from dataclasses import dataclass, field
from typing import List, Dict, Any, Optional

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import trange

# -----------------------------
# Utilities
# -----------------------------

def seed_everything(seed: int = 42):
    random.seed(seed)
    np.random.seed(seed)


# -----------------------------
# Molecular-detail modules
# -----------------------------
class TelomereMolecularModel:
    """Sequence-aware telomere molecular model (coarse-grained).

    - Stores a repetitive synthetic telomere motif (e.g., 'TTAGGG' or custom)
    - Computes: base stability score, oxidation susceptibility, and predicted
      recognition score for telomerase based on motif similarity to canonical template.

    This is intentionally lightweight and deterministic so it can be used inside
    a Colab simulation without heavy dependencies.
    """

    def __init__(self, motif: str = "TTAGGG", repeat_count: int = 30):
        self.motif = motif.upper()
        self.repeat_count = repeat_count
        self.sequence = self.motif * self.repeat_count

    def stability_score(self) -> float:
        """Return a 0..1 stability estimate.

        Heuristics used:
        - GC fraction increases stability
        - Longer repeats are generally more stable (saturating)
        - Certain motifs with consecutive G runs may be more oxidation-prone
        """
        seq = self.sequence
        gc_frac = (seq.count("G") + seq.count("C")) / max(1, len(seq))
        length_factor = 1.0 - math.exp(-self.repeat_count / 50.0)
        # penalize long G runs (simple proxy for G-quadruplex prone -> oxidation)
        max_g_run = self._max_consecutive('G')
        g_penalty = min(1.0, max(0.0, (max_g_run - 3) * 0.1))
        score = np.clip(0.2 * gc_frac + 0.8 * length_factor - 0.3 * g_penalty, 0.0, 1.0)
        return float(score)

    def oxidation_susceptibility(self) -> float:
        """Higher value means more susceptible to oxidative shortening.

        Uses GC content and G-run heuristic as a proxy.
        """
        seq = self.sequence
        g_frac = seq.count('G') / max(1, len(seq))
        max_g_run = self._max_consecutive('G')
        return float(np.clip(0.1 + g_frac * 0.6 + max(0.0, (max_g_run - 4) * 0.05), 0.0, 1.0))

    def telomerase_recognition_score(self, telomerase_template: str = "TTAGGG") -> float:
        """Simple similarity-based recognition score between motif and telomerase template.

        Normalized to 0..1.
        """
        motif = self.motif
        # fraction of matching bases in one repeat (circular compare)
        matches = sum(1 for a, b in zip(motif, telomerase_template) if a == b)
        base_sim = matches / max(1, len(telomerase_template))
        # longer repeats slightly increase recognition by providing repeated binding sites
        length_boost = min(1.0, self.repeat_count / 50.0)
        return float(np.clip(0.3 * base_sim + 0.7 * length_boost, 0.0, 1.0))

    def _max_consecutive(self, base: str) -> int:
        maxrun = 0
        run = 0
        for c in self.sequence:
            if c == base:
                run += 1
                maxrun = max(maxrun, run)
            else:
                run = 0
        return maxrun


class TLoopModel:
    """Coarse T-loop model.

    Provides a protective effectiveness score based on telomere length and
    a `stiffness` parameter representing ability to sequester the chromosome end.

    Includes a placeholder 'mini-MD' integration step that perturbs the protective
    score by a small stochastic amount to emulate thermally-driven structural changes.
    """

    def __init__(self, stiffness: float = 0.5):
        self.stiffness = float(np.clip(stiffness, 0.0, 1.0))

    def protective_score(self, telomere_length: float, molecular_model: TelomereMolecularModel) -> float:
        # longer telomeres and higher stability improve protection
        length_component = np.tanh(telomere_length / 100.0)
        seq_stability = molecular_model.stability_score()
        base_score = 0.6 * length_component + 0.4 * seq_stability
        # stiffness increases resistance to unwinding
        score = float(np.clip(base_score * (0.5 + 0.5 * self.stiffness), 0.0, 1.0))
        return score

    def mini_md_step(self, current_score: float, temperature: float = 300.0) -> float:
        # small thermal perturbation (no external MD dependency)
        thermal_noise = np.random.normal(loc=0.0, scale=0.01 * (temperature / 300.0))
        return float(np.clip(current_score + thermal_noise, 0.0, 1.0))


class TelomeraseBindingModel:
    """Docking-like scoring for small molecules/peptides that modulate telomerase.

    Lightweight heuristic:
    - small molecules characterized by a dict of properties (hydrophobicity, charge, size)
    - scoring returns an activation-modulation in range [-1, +1] where +1 strongly
      activates telomerase, -1 strongly inhibits.
    """

    def __init__(self):
        pass

    def score(self, molecule_props: Dict[str, float], molecular_model: TelomereMolecularModel) -> float:
        # Example properties expected: {'hydrophobicity': -1..1, 'charge': -1..1, 'size': 0..1}
        hyd = float(molecule_props.get('hydrophobicity', 0.0))
        charge = float(molecule_props.get('charge', 0.0))
        size = float(molecule_props.get('size', 0.5))
        recog = molecular_model.telomerase_recognition_score()
        # heuristic: molecules with moderate size and matching polarity to telomerase template
        polarity_factor = 1.0 - abs(charge - (0.0))  # assume telomerase environment neutral-ish
        size_pref = 1.0 - abs(size - 0.5)
        raw = 0.6 * hyd + 0.3 * polarity_factor + 0.6 * size_pref * recog
        # normalize to -1..1 by subtracting 0.5 baseline
        mod = float(np.clip((raw - 0.5) * 2.0, -1.0, 1.0))
        return mod


# -----------------------------
# GNN placeholder (scaffold)
# -----------------------------
class GNNTelomereScaffold:
    """Scaffold for a Graph Neural Network predicting local binding hotspots or
    fragility along the telomere. This is a placeholder showing where to plug in
    a PyTorch Geometric model.

    To enable:
      pip install torch torch-geometric
    Then replace `predict` with a real model inference.
    """

    def __init__(self, device: str = 'cpu'):
        self.device = device
        self.model = None  # user can load a trained PyG model here

    def predict(self, telomere_sequence: str) -> List[float]:
        """Return a per-repeat fragility score (0..1). For now returns a simple
        heuristic based on G-runs and GC content per repeat.
        """
        motif_len = 6
        repeats = [telomere_sequence[i:i+motif_len] for i in range(0, len(telomere_sequence), motif_len)]
        scores = []
        for r in repeats:
            gc = (r.count('G') + r.count('C')) / max(1, len(r))
            max_g = 0
            run = 0
            for c in r:
                if c == 'G':
                    run += 1
                    max_g = max(max_g, run)
                else:
                    run = 0
            score = float(np.clip(0.2 + gc * 0.6 + max(0, (max_g - 3) * 0.1), 0.0, 1.0))
            scores.append(score)
        return scores


# -----------------------------
# Telomere model (runtime instance)
# -----------------------------
@dataclass
class Telomere:
    length: float  # abstract units (e.g., repeats)
    min_length: float = 5.0  # trigger senescence
    max_length: float = 200.0
    # molecular extensions
    molecular_model: TelomereMolecularModel = field(default_factory=lambda: TelomereMolecularModel())
    tloop: TLoopModel = field(default_factory=lambda: TLoopModel(stiffness=0.6))

    def shorten(self, amount: float, oxidative_factor: float = 1.0):
        # molecular susceptibility increases effective shortening
        ox_susc = self.molecular_model.oxidation_susceptibility()
        effective = amount * (1.0 + 0.5 * ox_susc * oxidative_factor)
        self.length = max(0.0, self.length - effective)

    def extend(self, amount: float):
        self.length = min(self.max_length, self.length + amount)

    def is_senescent(self) -> bool:
        return self.length < self.min_length

    def protective_effect(self) -> float:
        base = self.tloop.protective_score(self.length, self.molecular_model)
        # perform mini MD fluctuation
        return self.tloop.mini_md_step(base)


# -----------------------------
# Virtual Cell (integrates molecular modules)
# -----------------------------
@dataclass
class VirtualCell:
    id: int
    telomere: Telomere
    health: float = 1.0  # 0..1
    dna_damage: float = 0.0  # aggregated damage
    dividing: bool = False
    age: int = 0

    def step(self, env: "Environment", telomerase_activity: float):
        """Advance one timestep for this cell.

        Incorporates molecular protective effect to modulate shortening.
        """
        # Background shortening scaled by protective effect
        protective = self.telomere.protective_effect()
        protective_factor = 1.0 - 0.5 * protective

        background_shortening = env.background_shorten_rate * (1.0 + env.oxidative_stress) * protective_factor
        self.telomere.shorten(background_shortening, oxidative_factor=env.oxidative_stress)

        # Telomerase-driven extension (bounded) and modulated by telomerase recognition
        recog = self.telomere.molecular_model.telomerase_recognition_score()
        extension = telomerase_activity * env.telomerase_efficiency * (0.5 + 0.5 * recog)
        self.telomere.extend(extension)

        # Accumulate DNA damage from environment but protected by t-loop
        damage_increase = env.stress_to_damage_rate * env.oxidative_stress * (1.0 - protective)
        self.dna_damage += damage_increase

        # Health update
        telomere_factor = min(1.0, self.telomere.length / (self.telomere.max_length * 0.5))
        damage_factor = max(0.0, 1.0 - self.dna_damage)
        self.health = max(0.0, 0.7 * telomere_factor + 0.3 * damage_factor)

        # Age and division decision
        self.age += 1
        self.dividing = self.should_divide(env)

    def should_divide(self, env: "Environment") -> bool:
        # Add molecular checks: require minimal recognition score and protective score
        recog = self.telomere.molecular_model.telomerase_recognition_score()
        protective = self.telomere.protective_effect()
        return (
            self.health > 0.6
            and self.telomere.length > (self.telomere.min_length + 10)
            and recog > 0.2
            and protective > 0.2
            and env.nutrients > env.min_nutrients_to_divide
        )


# -----------------------------
# Phagocyte / cleanup subsystem (abstract)
# -----------------------------
@dataclass
class Phagocyte:
    efficiency: float = 0.5  # fraction of debris removed per timestep

    def cleanup(self, env: "Environment"):
        removed = min(env.debris, env.debris * self.efficiency)
        env.debris -= removed
        env.nutrients += removed * env.recycling_yield


# -----------------------------
# Environment
# -----------------------------
@dataclass
class Environment:
    nutrients: float = 1.0
    debris: float = 0.0
    oxidative_stress: float = 0.0  # 0..1
    background_shorten_rate: float = 0.1
    telomerase_efficiency: float = 1.0
    stress_to_damage_rate: float = 0.01
    min_nutrients_to_divide: float = 0.3
    recycling_yield: float = 0.5

    def step(self, phagocyte: Phagocyte):
        phagocyte.cleanup(self)


# -----------------------------
# Controllers (same as before)
# -----------------------------
class PIDTelomeraseController:
    def __init__(self, target: float, kp: float = 0.1, ki: float = 0.01, kd: float = 0.0):
        self.target = target
        self.kp = kp
        self.ki = ki
        self.kd = kd
        self.integral = 0.0
        self.last_error = None

    def action(self, mean_telomere: float) -> float:
        error = self.target - mean_telomere
        self.integral += error
        derivative = 0.0 if (self.last_error is None) else (error - self.last_error)
        self.last_error = error
        raw = self.kp * error + self.ki * self.integral + self.kd * derivative
        return float(np.clip(raw, 0.0, 1.0))


class RLControllerPlaceholder:
    def __init__(self):
        pass

    def action(self, mean_telomere: float, mean_health: float, env_state: Dict[str, float]) -> float:
        if mean_telomere < 50.0:
            return 0.8
        if mean_health < 0.5:
            return 0.4
        return 0.1


# -----------------------------
# Simulator (integrates molecular pieces)
# -----------------------------
@dataclass
class Simulator:
    n_cells: int = 50
    steps: int = 1000
    env: Environment = field(default_factory=Environment)
    phagocyte: Phagocyte = field(default_factory=lambda: Phagocyte(efficiency=0.6))
    controller: Any = field(default_factory=lambda: PIDTelomeraseController(target=100.0))
    seed: int = 42

    traces: List[Dict[str, Any]] = field(default_factory=list)

    def __post_init__(self):
        seed_everything(self.seed)
        self.cells: List[VirtualCell] = []
        for i in range(self.n_cells):
            initial_len = np.clip(np.random.normal(120.0, 10.0), 10.0, 160.0)
            tel = Telomere(length=initial_len,
                           molecular_model=TelomereMolecularModel(motif="TTAGGG", repeat_count=int(initial_len/4)),
                           tloop=TLoopModel(stiffness=np.clip(np.random.normal(0.6, 0.1), 0.2, 0.9)))
            cell = VirtualCell(id=i, telomere=tel)
            self.cells.append(cell)

    def step(self, step_idx: int):
        mean_tel = float(np.mean([c.telomere.length for c in self.cells]))
        mean_health = float(np.mean([c.health for c in self.cells]))

        if hasattr(self.controller, "action"):
            try:
                telomerase_activity = self.controller.action(mean_tel)
            except TypeError:
                telomerase_activity = self.controller.action(mean_tel, mean_health, {
                    "oxidative_stress": self.env.oxidative_stress,
                    "debris": self.env.debris,
                })
        else:
            telomerase_activity = 0.1

        consumption_per_cell = 0.01
        total_consumption = consumption_per_cell * len(self.cells)
        self.env.nutrients = max(0.0, self.env.nutrients - total_consumption)

        new_cells: List[VirtualCell] = []
        for cell in self.cells:
            cell.step(self.env, telomerase_activity)

            if cell.dividing:
                self.env.nutrients = max(0.0, self.env.nutrients - 0.05)
                cell.telomere.shorten(1.0 + np.random.rand() * 0.5)
                daughter_len = max(0.0, cell.telomere.length - np.random.normal(0.5, 0.2))
                # inherit molecular motif and tloop properties
                daughter = VirtualCell(
                    id=max(c.id for c in self.cells + new_cells) + 1,
                    telomere=Telomere(length=daughter_len,
                                      molecular_model=TelomereMolecularModel(motif=cell.telomere.molecular_model.motif, repeat_count=int(daughter_len/4)),
                                      tloop=TLoopModel(stiffness=cell.telomere.tloop.stiffness)),
                )
                new_cells.append(daughter)

            if cell.telomere.is_senescent() or cell.health < 0.05:
                self.env.debris += 0.5 + random.random() * 0.5
                continue
            else:
                new_cells.append(cell)

        self.cells = new_cells
        self.env.step(self.phagocyte)

        # record more molecular traces for the top-k fragile cells
        fragilities = []
        for c in self.cells:
            fragilities.append(np.mean(GNNTelomereScaffold().predict(c.telomere.molecular_model.sequence)))

        trace = {
            "step": step_idx,
            "n_cells": len(self.cells),
            "mean_telomere": float(np.mean([c.telomere.length for c in self.cells])) if len(self.cells) > 0 else 0.0,
            "mean_health": float(np.mean([c.health for c in self.cells])) if len(self.cells) > 0 else 0.0,
            "nutrients": float(self.env.nutrients),
            "debris": float(self.env.debris),
            "oxidative_stress": float(self.env.oxidative_stress),
            "telomerase_activity": float(telomerase_activity),
            "mean_fragility": float(np.mean(fragilities)) if len(fragilities)>0 else 0.0,
        }
        self.traces.append(trace)

    def run(self, steps: Optional[int] = None, progress: bool = True):
        steps = steps or self.steps
        iterator = trange(steps) if progress else range(steps)
        for t in iterator:
            if t % 200 == 0 and t > 0:
                self.env.oxidative_stress = min(1.0, self.env.oxidative_stress + 0.2)
            self.env.oxidative_stress = max(0.0, self.env.oxidative_stress - 0.002)
            self.step(t)

    def to_dataframe(self) -> pd.DataFrame:
        return pd.DataFrame(self.traces)

    def save_csv(self, path: str = "simulation_traces.csv"):
        df = self.to_dataframe()
        df.to_csv(path, index=False)
        return path

    def plot(self, save_path: Optional[str] = None):
        df = self.to_dataframe()
        fig, axs = plt.subplots(4, 1, figsize=(8, 12), constrained_layout=True)

        axs[0].plot(df["step"], df["mean_telomere"])
        axs[0].set_title("Mean Telomere Length")

        axs[1].plot(df["step"], df["mean_health"])
        axs[1].set_title("Mean Cell Health")

        axs[2].plot(df["step"], df["n_cells"], label="n_cells")
        axs[2].plot(df["step"], df["nutrients"], label="nutrients")
        axs[2].plot(df["step"], df["debris"], label="debris")
        axs[2].set_title("Population / Environment")
        axs[2].legend()

        axs[3].plot(df["step"], df["mean_fragility"])
        axs[3].set_title("Mean Molecular Fragility (GNN scaffold)")

        if save_path:
            fig.savefig(save_path)
        plt.show()


# -----------------------------
# Demo / main
# -----------------------------

def demo_run(seed: int = 123, steps: int = 400, use_rl: bool = False):
    seed_everything(seed)

    env = Environment(nutrients=5.0, debris=0.0, oxidative_stress=0.05)
    phag = Phagocyte(efficiency=0.7)

    if use_rl:
        controller = RLControllerPlaceholder()
    else:
        controller = PIDTelomeraseController(target=120.0, kp=0.02, ki=0.001, kd=0.0)

    sim = Simulator(n_cells=60, steps=steps, env=env, phagocyte=phag, controller=controller, seed=seed)

    sim.run(steps, progress=True)

    df = sim.to_dataframe()
    csv_path = sim.save_csv("colab_simulation_traces_molecular.csv")
    print(f"Saved CSV to: {csv_path}")

    sim.plot(save_path="colab_simulation_plots_molecular.png")

    return sim, df


if __name__ == "__main__":
    sim, df = demo_run(seed=42, steps=200, use_rl=False)
    print(df.tail())
````

````python
"""
Colab-ready simulation skeleton: Virtual cell with synthetic telomere, telomerase controller,
phagocytic subsystem, environment, and simple visualization.

Extended with molecular-detail modules and a runnable PyTorch example:
- TelomereMolecularModel: sequence-aware stability scoring and oxidative-susceptibility model
- TLoopModel: coarse protective model with mini-MD
- TelomeraseBindingModel: docking-like heuristic
- GNNTelomereScaffold: lightweight heuristic scaffold (replaced by a small PyTorch MLP trainer below)
- Runnable PyTorch example: synthetic dataset, MLP model, training loop, and inference wrapper

How to use in Colab:
1. Upload this file or paste into a notebook cell.
2. Install dependencies (if not present):
   !pip install numpy matplotlib pandas tqdm imageio torch
3. Run the demo at the bottom to execute a short simulation and produce plots/CSV.
   To run the PyTorch trainer before the simulation, call `train_fragility_model()`.

Notes:
- The PyTorch model here is a simple MLP trained on synthetic features derived from telomere repeats.
  It's a stand-in for a GNN and is intentionally lightweight so it runs quickly in Colab CPU.
- If you'd like a real GNN (PyG) integration, I can add a version that uses torch-geometricâ€”
  that requires additional installs and longer training.

Author: Generated by ChatGPT â€” adapt as you like.
"""

import math
import random
import time
from dataclasses import dataclass, field
from typing import List, Dict, Any, Optional

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import trange

# PyTorch imports (optional â€” install in Colab if needed)
try:
    import torch
    from torch import nn
    from torch.utils.data import Dataset, DataLoader
    TORCH_AVAILABLE = True
except Exception:
    TORCH_AVAILABLE = False

# -----------------------------
# Utilities
# -----------------------------

def seed_everything(seed: int = 42):
    random.seed(seed)
    np.random.seed(seed)
    if TORCH_AVAILABLE:
        torch.manual_seed(seed)


# -----------------------------
# Molecular-detail modules
# -----------------------------
class TelomereMolecularModel:
    def __init__(self, motif: str = "TTAGGG", repeat_count: int = 30):
        self.motif = motif.upper()
        self.repeat_count = repeat_count
        self.sequence = self.motif * self.repeat_count

    def stability_score(self) -> float:
        seq = self.sequence
        gc_frac = (seq.count("G") + seq.count("C")) / max(1, len(seq))
        length_factor = 1.0 - math.exp(-self.repeat_count / 50.0)
        max_g_run = self._max_consecutive('G')
        g_penalty = min(1.0, max(0.0, (max_g_run - 3) * 0.1))
        score = np.clip(0.2 * gc_frac + 0.8 * length_factor - 0.3 * g_penalty, 0.0, 1.0)
        return float(score)

    def oxidation_susceptibility(self) -> float:
        seq = self.sequence
        g_frac = seq.count('G') / max(1, len(seq))
        max_g_run = self._max_consecutive('G')
        return float(np.clip(0.1 + g_frac * 0.6 + max(0.0, (max_g_run - 4) * 0.05), 0.0, 1.0))

    def telomerase_recognition_score(self, telomerase_template: str = "TTAGGG") -> float:
        motif = self.motif
        matches = sum(1 for a, b in zip(motif, telomerase_template) if a == b)
        base_sim = matches / max(1, len(telomerase_template))
        length_boost = min(1.0, self.repeat_count / 50.0)
        return float(np.clip(0.3 * base_sim + 0.7 * length_boost, 0.0, 1.0))

    def _max_consecutive(self, base: str) -> int:
        maxrun = 0
        run = 0
        for c in self.sequence:
            if c == base:
                run += 1
                maxrun = max(maxrun, run)
            else:
                run = 0
        return maxrun


class TLoopModel:
    def __init__(self, stiffness: float = 0.5):
        self.stiffness = float(np.clip(stiffness, 0.0, 1.0))

    def protective_score(self, telomere_length: float, molecular_model: TelomereMolecularModel) -> float:
        length_component = np.tanh(telomere_length / 100.0)
        seq_stability = molecular_model.stability_score()
        base_score = 0.6 * length_component + 0.4 * seq_stability
        score = float(np.clip(base_score * (0.5 + 0.5 * self.stiffness), 0.0, 1.0))
        return score

    def mini_md_step(self, current_score: float, temperature: float = 300.0) -> float:
        thermal_noise = np.random.normal(loc=0.0, scale=0.01 * (temperature / 300.0))
        return float(np.clip(current_score + thermal_noise, 0.0, 1.0))


class TelomeraseBindingModel:
    def __init__(self):
        pass

    def score(self, molecule_props: Dict[str, float], molecular_model: TelomereMolecularModel) -> float:
        hyd = float(molecule_props.get('hydrophobicity', 0.0))
        charge = float(molecule_props.get('charge', 0.0))
        size = float(molecule_props.get('size', 0.5))
        recog = molecular_model.telomerase_recognition_score()
        polarity_factor = 1.0 - abs(charge - (0.0))
        size_pref = 1.0 - abs(size - 0.5)
        raw = 0.6 * hyd + 0.3 * polarity_factor + 0.6 * size_pref * recog
        mod = float(np.clip((raw - 0.5) * 2.0, -1.0, 1.0))
        return mod


# Lightweight heuristic scaffold (used for synthetic label generation)
class GNNTelomereScaffold:
    def __init__(self, device: str = 'cpu'):
        self.device = device

    def predict(self, telomere_sequence: str) -> List[float]:
        motif_len = 6
        repeats = [telomere_sequence[i:i+motif_len] for i in range(0, len(telomere_sequence), motif_len)]
        scores = []
        for r in repeats:
            gc = (r.count('G') + r.count('C')) / max(1, len(r))
            max_g = 0
            run = 0
            for c in r:
                if c == 'G':
                    run += 1
                    max_g = max(max_g, run)
                else:
                    run = 0
            score = float(np.clip(0.2 + gc * 0.6 + max(0, (max_g - 3) * 0.1), 0.0, 1.0))
            scores.append(score)
        return scores


# -----------------------------
# Telomere model (runtime instance)
# -----------------------------
@dataclass
class Telomere:
    length: float
    min_length: float = 5.0
    max_length: float = 200.0
    molecular_model: TelomereMolecularModel = field(default_factory=lambda: TelomereMolecularModel())
    tloop: TLoopModel = field(default_factory=lambda: TLoopModel(stiffness=0.6))

    def shorten(self, amount: float, oxidative_factor: float = 1.0):
        ox_susc = self.molecular_model.oxidation_susceptibility()
        effective = amount * (1.0 + 0.5 * ox_susc * oxidative_factor)
        self.length = max(0.0, self.length - effective)

    def extend(self, amount: float):
        self.length = min(self.max_length, self.length + amount)

    def is_senescent(self) -> bool:
        return self.length < self.min_length

    def protective_effect(self) -> float:
        base = self.tloop.protective_score(self.length, self.molecular_model)
        return self.tloop.mini_md_step(base)


# -----------------------------
# Virtual Cell (integrates molecular modules)
# -----------------------------
@dataclass
class VirtualCell:
    id: int
    telomere: Telomere
    health: float = 1.0
    dna_damage: float = 0.0
    dividing: bool = False
    age: int = 0

    def step(self, env: "Environment", telomerase_activity: float):
        protective = self.telomere.protective_effect()
        protective_factor = 1.0 - 0.5 * protective

        background_shortening = env.background_shorten_rate * (1.0 + env.oxidative_stress) * protective_factor
        self.telomere.shorten(background_shortening, oxidative_factor=env.oxidative_stress)

        recog = self.telomere.molecular_model.telomerase_recognition_score()
        extension = telomerase_activity * env.telomerase_efficiency * (0.5 + 0.5 * recog)
        self.telomere.extend(extension)

        damage_increase = env.stress_to_damage_rate * env.oxidative_stress * (1.0 - protective)
        self.dna_damage += damage_increase

        telomere_factor = min(1.0, self.telomere.length / (self.telomere.max_length * 0.5))
        damage_factor = max(0.0, 1.0 - self.dna_damage)
        self.health = max(0.0, 0.7 * telomere_factor + 0.3 * damage_factor)

        self.age += 1
        self.dividing = self.should_divide(env)

    def should_divide(self, env: "Environment") -> bool:
        recog = self.telomere.molecular_model.telomerase_recognition_score()
        protective = self.telomere.protective_effect()
        return (
            self.health > 0.6
            and self.telomere.length > (self.telomere.min_length + 10)
            and recog > 0.2
            and protective > 0.2
            and env.nutrients > env.min_nutrients_to_divide
        )


# -----------------------------
# Phagocyte / cleanup subsystem
# -----------------------------
@dataclass
class Phagocyte:
    efficiency: float = 0.5

    def cleanup(self, env: "Environment"):
        removed = min(env.debris, env.debris * self.efficiency)
        env.debris -= removed
        env.nutrients += removed * env.recycling_yield


# -----------------------------
# Environment
# -----------------------------
@dataclass
class Environment:
    nutrients: float = 1.0
    debris: float = 0.0
    oxidative_stress: float = 0.0
    background_shorten_rate: float = 0.1
    telomerase_efficiency: float = 1.0
    stress_to_damage_rate: float = 0.01
    min_nutrients_to_divide: float = 0.3
    recycling_yield: float = 0.5

    def step(self, phagocyte: Phagocyte):
        phagocyte.cleanup(self)


# -----------------------------
# Controllers
# -----------------------------
class PIDTelomeraseController:
    def __init__(self, target: float, kp: float = 0.1, ki: float = 0.01, kd: float = 0.0):
        self.target = target
        self.kp = kp
        self.ki = ki
        self.kd = kd
        self.integral = 0.0
        self.last_error = None

    def action(self, mean_telomere: float) -> float:
        error = self.target - mean_telomere
        self.integral += error
        derivative = 0.0 if (self.last_error is None) else (error - self.last_error)
        self.last_error = error
        raw = self.kp * error + self.ki * self.integral + self.kd * derivative
        return float(np.clip(raw, 0.0, 1.0))


class RLControllerPlaceholder:
    def __init__(self):
        pass

    def action(self, mean_telomere: float, mean_health: float, env_state: Dict[str, float]) -> float:
        if mean_telomere < 50.0:
            return 0.8
        if mean_health < 0.5:
            return 0.4
        return 0.1


# -----------------------------
# Runnable PyTorch fragility model
# -----------------------------
MODEL_PATH = "fragility_mlp.pth"

class FragilityDataset(Dataset):
    def __init__(self, X: np.ndarray, y: np.ndarray):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]


class FragilityMLP(nn.Module):
    def __init__(self, input_dim: int = 3, hidden: int = 64):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden),
            nn.ReLU(),
            nn.Linear(hidden, hidden),
            nn.ReLU(),
            nn.Linear(hidden, 1),
            nn.Sigmoid(),
        )

    def forward(self, x):
        return self.net(x).squeeze(-1)


def featurize_repeat(repeat: str) -> List[float]:
    # simple per-repeat features: gc_frac, max_g_run_norm, length_norm
    gc = (repeat.count('G') + repeat.count('C')) / max(1, len(repeat))
    max_g = 0
    run = 0
    for c in repeat:
        if c == 'G':
            run += 1
            max_g = max(max_g, run)
        else:
            run = 0
    max_g_norm = max_g / len(repeat)
    length_norm = len(repeat) / 6.0
    return [gc, max_g_norm, length_norm]


def generate_synthetic_fragility_dataset(n_sequences: int = 5000, motif_len: int = 6, repeats: int = 10):
    X_list = []
    y_list = []
    scaffold = GNNTelomereScaffold()
    for _ in range(n_sequences):
        # generate random motif with some bias
        base_choices = ['A', 'T', 'G', 'C']
        motif = ''.join(np.random.choice(base_choices, size=motif_len))
        seq = motif * repeats
        # create per-repeat features and synthetic target using scaffold
        per_repeat = [seq[i:i+motif_len] for i in range(0, len(seq), motif_len)]
        frag_scores = scaffold.predict(seq)
        for r, score in zip(per_repeat, frag_scores):
            feats = featurize_repeat(r)
            X_list.append(feats)
            y_list.append(score)
    X = np.array(X_list)
    y = np.array(y_list)
    return X, y


def train_fragility_model(epochs: int = 6, batch_size: int = 256, lr: float = 1e-3, small_dataset: bool = True):
    if not TORCH_AVAILABLE:
        raise RuntimeError("PyTorch is not available. Install torch in Colab to run the trainer.")
    X, y = generate_synthetic_fragility_dataset(n_sequences=200 if small_dataset else 2000, repeats=8)
    ds = FragilityDataset(X, y)
    dl = DataLoader(ds, batch_size=batch_size, shuffle=True)

    model = FragilityMLP(input_dim=X.shape[1], hidden=64)
    opt = torch.optim.Adam(model.parameters(), lr=lr)
    loss_fn = nn.MSELoss()

    model.train()
    for ep in range(epochs):
        running = 0.0
        for xb, yb in dl:
            pred = model(xb)
            loss = loss_fn(pred, yb)
            opt.zero_grad()
            loss.backward()
            opt.step()
            running += loss.item() * xb.size(0)
        avg = running / len(ds)
        print(f"Epoch {ep+1}/{epochs} â€” loss {avg:.6f}")

    torch.save(model.state_dict(), MODEL_PATH)
    print(f"Saved fragility model to {MODEL_PATH}")
    return model


def load_fragility_model():
    if not TORCH_AVAILABLE:
        return None
    model = FragilityMLP(input_dim=3, hidden=64)
    try:
        model.load_state_dict(torch.load(MODEL_PATH, map_location='cpu'))
        model.eval()
        return model
    except Exception:
        return None


def predict_fragility_with_model(model, telomere_sequence: str) -> List[float]:
    per_repeat = [telomere_sequence[i:i+6] for i in range(0, len(telomere_sequence), 6)]
    feats = np.array([featurize_repeat(r) for r in per_repeat], dtype=np.float32)
    with torch.no_grad():
        xb = torch.tensor(feats)
        preds = model(xb).numpy().tolist()
    return preds


# -----------------------------
# Simulator (integrates molecular pieces and optionally PyTorch model)
# -----------------------------
@dataclass
class Simulator:
    n_cells: int = 50
    steps: int = 1000
    env: Environment = field(default_factory=Environment)
    phagocyte: Phagocyte = field(default_factory=lambda: Phagocyte(efficiency=0.6))
    controller: Any = field(default_factory=lambda: PIDTelomeraseController(target=100.0))
    seed: int = 42
    fragility_model: Optional[Any] = None

    traces: List[Dict[str, Any]] = field(default_factory=list)

    def __post_init__(self):
        seed_everything(self.seed)
        self.cells: List[VirtualCell] = []
        for i in range(self.n_cells):
            initial_len = np.clip(np.random.normal(120.0, 10.0), 10.0, 160.0)
            tel = Telomere(length=initial_len,
                           molecular_model=TelomereMolecularModel(motif="TTAGGG", repeat_count=int(initial_len/4)),
                           tloop=TLoopModel(stiffness=np.clip(np.random.normal(0.6, 0.1), 0.2, 0.9)))
            cell = VirtualCell(id=i, telomere=tel)
            self.cells.append(cell)

    def step(self, step_idx: int):
        mean_tel = float(np.mean([c.telomere.length for c in self.cells]))
        mean_health = float(np.mean([c.health for c in self.cells]))

        if hasattr(self.controller, "action"):
            try:
                telomerase_activity = self.controller.action(mean_tel)
            except TypeError:
                telomerase_activity = self.controller.action(mean_tel, mean_health, {
                    "oxidative_stress": self.env.oxidative_stress,
                    "debris": self.env.debris,
                })
        else:
            telomerase_activity = 0.1

        consumption_per_cell = 0.01
        total_consumption = consumption_per_cell * len(self.cells)
        self.env.nutrients = max(0.0, self.env.nutrients - total_consumption)

        new_cells: List[VirtualCell] = []
        for cell in self.cells:
            cell.step(self.env, telomerase_activity)

            if cell.dividing:
                self.env.nutrients = max(0.0, self.env.nutrients - 0.05)
                cell.telomere.shorten(1.0 + np.random.rand() * 0.5)
                daughter_len = max(0.0, cell.telomere.length - np.random.normal(0.5, 0.2))
                daughter = VirtualCell(
                    id=max(c.id for c in self.cells + new_cells) + 1,
                    telomere=Telomere(length=daughter_len,
                                      molecular_model=TelomereMolecularModel(motif=cell.telomere.molecular_model.motif, repeat_count=int(daughter_len/4)),
                                      tloop=TLoopModel(stiffness=cell.telomere.tloop.stiffness)),
                )
                new_cells.append(daughter)

            if cell.telomere.is_senescent() or cell.health < 0.05:
                self.env.debris += 0.5 + random.random() * 0.5
                continue
            else:
                new_cells.append(cell)

        self.cells = new_cells
        self.env.step(self.phagocyte)

        # generate fragility trace using either trained model or heuristic scaffold
        fragilities = []
        model = self.fragility_model
        if model is not None and TORCH_AVAILABLE:
            try:
                for c in self.cells:
                    seq = c.telomere.molecular_model.sequence
                    preds = predict_fragility_with_model(model, seq)
                    fragilities.append(float(np.mean(preds)))
            except Exception:
                # fallback
                fragilities = [np.mean(GNNTelomereScaffold().predict(c.telomere.molecular_model.sequence)) for c in self.cells]
        else:
            fragilities = [np.mean(GNNTelomereScaffold().predict(c.telomere.molecular_model.sequence)) for c in self.cells]

        trace = {
            "step": step_idx,
            "n_cells": len(self.cells),
            "mean_telomere": float(np.mean([c.telomere.length for c in self.cells])) if len(self.cells) > 0 else 0.0,
            "mean_health": float(np.mean([c.health for c in self.cells])) if len(self.cells) > 0 else 0.0,
            "nutrients": float(self.env.nutrients),
            "debris": float(self.env.debris),
            "oxidative_stress": float(self.env.oxidative_stress),
            "telomerase_activity": float(telomerase_activity),
            "mean_fragility": float(np.mean(fragilities)) if len(fragilities)>0 else 0.0,
        }
        self.traces.append(trace)

    def run(self, steps: Optional[int] = None, progress: bool = True):
        steps = steps or self.steps
        iterator = trange(steps) if progress else range(steps)
        for t in iterator:
            if t % 200 == 0 and t > 0:
                self.env.oxidative_stress = min(1.0, self.env.oxidative_stress + 0.2)
            self.env.oxidative_stress = max(0.0, self.env.oxidative_stress - 0.002)
            self.step(t)

    def to_dataframe(self) -> pd.DataFrame:
        return pd.DataFrame(self.traces)

    def save_csv(self, path: str = "simulation_traces.csv"):
        df = self.to_dataframe()
        df.to_csv(path, index=False)
        return path

    def plot(self, save_path: Optional[str] = None):
        df = self.to_dataframe()
        fig, axs = plt.subplots(4, 1, figsize=(8, 12), constrained_layout=True)

        axs[0].plot(df["step"], df["mean_telomere"])
        axs[0].set_title("Mean Telomere Length")

        axs[1].plot(df["step"], df["mean_health"])
        axs[1].set_title("Mean Cell Health")

        axs[2].plot(df["step"], df["n_cells"], label="n_cells")
        axs[2].plot(df["step"], df["nutrients"], label="nutrients")
        axs[2].plot(df["step"], df["debris"], label="debris")
        axs[2].set_title("Population / Environment")
        axs[2].legend()

        axs[3].plot(df["step"], df["mean_fragility"])
        axs[3].set_title("Mean Molecular Fragility (model)")

        if save_path:
            fig.savefig(save_path)
        plt.show()


# -----------------------------
# Demo / main
# -----------------------------

def demo_run(seed: int = 123, steps: int = 400, use_rl: bool = False, use_frag_model: bool = False):
    seed_everything(seed)

    env = Environment(nutrients=5.0, debris=0.0, oxidative_stress=0.05)
    phag = Phagocyte(efficiency=0.7)

    if use_rl:
        controller = RLControllerPlaceholder()
    else:
        controller = PIDTelomeraseController(target=120.0, kp=0.02, ki=0.001, kd=0.0)

    frag_model = None
    if use_frag_model and TORCH_AVAILABLE:
        frag_model = load_fragility_model()
        if frag_model is None:
            print("No trained fragility model found â€” training a small one now...")
            frag_model = train_fragility_model(epochs=4, small_dataset=True)

    sim = Simulator(n_cells=60, steps=steps, env=env, phagocyte=phag, controller=controller, seed=seed, fragility_model=frag_model)

    sim.run(steps, progress=True)

    df = sim.to_dataframe()
    csv_path = sim.save_csv("colab_simulation_traces_molecular_pytorch.csv")
    print(f"Saved CSV to: {csv_path}")

    sim.plot(save_path="colab_simulation_plots_molecular_pytorch.png")

    return sim, df


if __name__ == "__main__":
    # Example: train model first (requires torch) then run simulation using it
    if TORCH_AVAILABLE:
        print("PyTorch available â€” training a tiny fragility model for demo (this may take ~30s on CPU)")
        train_fragility_model(epochs=3, small_dataset=True)
        sim, df = demo_run(seed=42, steps=200, use_rl=False, use_frag_model=True)
    else:
        print("PyTorch not available â€” run demo without model. To enable, install torch in Colab.")
        sim, df = demo_run(seed=42, steps=200, use_rl=False, use_frag_model=False)
    print(df.tail())
````

````python
"""
Colab-ready simulation skeleton: Virtual cell with synthetic telomere, telomerase controller,
phagocytic subsystem, environment, and simple visualization.

Extended with molecular-detail modules and a runnable PyTorch example:
- TelomereMolecularModel: sequence-aware stability scoring and oxidative-susceptibility model
- TLoopModel: coarse protective model with mini-MD
- TelomeraseBindingModel: docking-like heuristic
- GNNTelomereScaffold: lightweight heuristic scaffold (replaced by a small PyTorch MLP trainer below)
- Runnable PyTorch example: synthetic dataset, MLP model, training loop, and inference wrapper

How to use in Colab:
1. Upload this file or paste into a notebook cell.
2. Install dependencies (if not present):
   !pip install numpy matplotlib pandas tqdm imageio torch
3. Run the demo at the bottom to execute a short simulation and produce plots/CSV.
   To run the PyTorch trainer before the simulation, call `train_fragility_model()`.

Notes:
- The PyTorch model here is a simple MLP trained on synthetic features derived from telomere repeats.
  It's a stand-in for a GNN and is intentionally lightweight so it runs quickly in Colab CPU.
- If you'd like a real GNN (PyG) integration, I can add a version that uses torch-geometric-
  that requires additional installs and longer training.

Author: Generated by ChatGPT - adapt as you like.
"""

import math
import random
import time
from dataclasses import dataclass, field
from typing import List, Dict, Any, Optional

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import trange

# PyTorch imports (optional - install in Colab if needed)
try:
    import torch
    from torch import nn
    from torch.utils.data import Dataset, DataLoader
    TORCH_AVAILABLE = True
except Exception:
    TORCH_AVAILABLE = False

# -----------------------------
# Utilities
# -----------------------------

def seed_everything(seed: int = 42):
    random.seed(seed)
    np.random.seed(seed)
    if TORCH_AVAILABLE:
        torch.manual_seed(seed)


# -----------------------------
# Molecular-detail modules
# -----------------------------
class TelomereMolecularModel:
    def __init__(self, motif: str = "TTAGGG", repeat_count: int = 30):
        self.motif = motif.upper()
        self.repeat_count = repeat_count
        self.sequence = self.motif * self.repeat_count

    def stability_score(self) -> float:
        seq = self.sequence
        gc_frac = (seq.count("G") + seq.count("C")) / max(1, len(seq))
        length_factor = 1.0 - math.exp(-self.repeat_count / 50.0)
        max_g_run = self._max_consecutive('G')
        g_penalty = min(1.0, max(0.0, (max_g_run - 3) * 0.1))
        score = np.clip(0.2 * gc_frac + 0.8 * length_factor - 0.3 * g_penalty, 0.0, 1.0)
        return float(score)

    def oxidation_susceptibility(self) -> float:
        seq = self.sequence
        g_frac = seq.count('G') / max(1, len(seq))
        max_g_run = self._max_consecutive('G')
        return float(np.clip(0.1 + g_frac * 0.6 + max(0.0, (max_g_run - 4) * 0.05), 0.0, 1.0))

    def telomerase_recognition_score(self, telomerase_template: str = "TTAGGG") -> float:
        motif = self.motif
        matches = sum(1 for a, b in zip(motif, telomerase_template) if a == b)
        base_sim = matches / max(1, len(telomerase_template))
        length_boost = min(1.0, self.repeat_count / 50.0)
        return float(np.clip(0.3 * base_sim + 0.7 * length_boost, 0.0, 1.0))

    def _max_consecutive(self, base: str) -> int:
        maxrun = 0
        run = 0
        for c in self.sequence:
            if c == base:
                run += 1
                maxrun = max(maxrun, run)
            else:
                run = 0
        return maxrun


class TLoopModel:
    def __init__(self, stiffness: float = 0.5):
        self.stiffness = float(np.clip(stiffness, 0.0, 1.0))

    def protective_score(self, telomere_length: float, molecular_model: TelomereMolecularModel) -> float:
        length_component = np.tanh(telomere_length / 100.0)
        seq_stability = molecular_model.stability_score()
        base_score = 0.6 * length_component + 0.4 * seq_stability
        score = float(np.clip(base_score * (0.5 + 0.5 * self.stiffness), 0.0, 1.0))
        return score

    def mini_md_step(self, current_score: float, temperature: float = 300.0) -> float:
        thermal_noise = np.random.normal(loc=0.0, scale=0.01 * (temperature / 300.0))
        return float(np.clip(current_score + thermal_noise, 0.0, 1.0))


class TelomeraseBindingModel:
    def __init__(self):
        pass

    def score(self, molecule_props: Dict[str, float], molecular_model: TelomereMolecularModel) -> float:
        hyd = float(molecule_props.get('hydrophobicity', 0.0))
        charge = float(molecule_props.get('charge', 0.0))
        size = float(molecule_props.get('size', 0.5))
        recog = molecular_model.telomerase_recognition_score()
        polarity_factor = 1.0 - abs(charge - (0.0))
        size_pref = 1.0 - abs(size - 0.5)
        raw = 0.6 * hyd + 0.3 * polarity_factor + 0.6 * size_pref * recog
        mod = float(np.clip((raw - 0.5) * 2.0, -1.0, 1.0))
        return mod


# Lightweight heuristic scaffold (used for synthetic label generation)
class GNNTelomereScaffold:
    def __init__(self, device: str = 'cpu'):
        self.device = device

    def predict(self, telomere_sequence: str) -> List[float]:
        motif_len = 6
        repeats = [telomere_sequence[i:i+motif_len] for i in range(0, len(telomere_sequence), motif_len)]
        scores = []
        for r in repeats:
            gc = (r.count('G') + r.count('C')) / max(1, len(r))
            max_g = 0
            run = 0
            for c in r:
                if c == 'G':
                    run += 1
                    max_g = max(max_g, run)
                else:
                    run = 0
            score = float(np.clip(0.2 + gc * 0.6 + max(0, (max_g - 3) * 0.1), 0.0, 1.0))
            scores.append(score)
        return scores


# -----------------------------
# Telomere model (runtime instance)
# -----------------------------
@dataclass
class Telomere:
    length: float
    min_length: float = 5.0
    max_length: float = 200.0
    molecular_model: TelomereMolecularModel = field(default_factory=lambda: TelomereMolecularModel())
    tloop: TLoopModel = field(default_factory=lambda: TLoopModel(stiffness=0.6))

    def shorten(self, amount: float, oxidative_factor: float = 1.0):
        ox_susc = self.molecular_model.oxidation_susceptibility()
        effective = amount * (1.0 + 0.5 * ox_susc * oxidative_factor)
        self.length = max(0.0, self.length - effective)

    def extend(self, amount: float):
        self.length = min(self.max_length, self.length + amount)

    def is_senescent(self) -> bool:
        return self.length < self.min_length

    def protective_effect(self) -> float:
        base = self.tloop.protective_score(self.length, self.molecular_model)
        return self.tloop.mini_md_step(base)


# -----------------------------
# Virtual Cell (integrates molecular modules)
# -----------------------------
@dataclass
class VirtualCell:
    id: int
    telomere: Telomere
    health: float = 1.0
    dna_damage: float = 0.0
    dividing: bool = False
    age: int = 0

    def step(self, env: "Environment", telomerase_activity: float):
        protective = self.telomere.protective_effect()
        protective_factor = 1.0 - 0.5 * protective

        background_shortening = env.background_shorten_rate * (1.0 + env.oxidative_stress) * protective_factor
        self.telomere.shorten(background_shortening, oxidative_factor=env.oxidative_stress)

        recog = self.telomere.molecular_model.telomerase_recognition_score()
        extension = telomerase_activity * env.telomerase_efficiency * (0.5 + 0.5 * recog)
        self.telomere.extend(extension)

        damage_increase = env.stress_to_damage_rate * env.oxidative_stress * (1.0 - protective)
        self.dna_damage += damage_increase

        telomere_factor = min(1.0, self.telomere.length / (self.telomere.max_length * 0.5))
        damage_factor = max(0.0, 1.0 - self.dna_damage)
        self.health = max(0.0, 0.7 * telomere_factor + 0.3 * damage_factor)

        self.age += 1
        self.dividing = self.should_divide(env)

    def should_divide(self, env: "Environment") -> bool:
        recog = self.telomere.molecular_model.telomerase_recognition_score()
        protective = self.telomere.protective_effect()
        return (
            self.health > 0.6
            and self.telomere.length > (self.telomere.min_length + 10)
            and recog > 0.2
            and protective > 0.2
            and env.nutrients > env.min_nutrients_to_divide
        )


# -----------------------------
# Phagocyte / cleanup subsystem
# -----------------------------
@dataclass
class Phagocyte:
    efficiency: float = 0.5

    def cleanup(self, env: "Environment"):
        removed = min(env.debris, env.debris * self.efficiency)
        env.debris -= removed
        env.nutrients += removed * env.recycling_yield


# -----------------------------
# Environment
# -----------------------------
@dataclass
class Environment:
    nutrients: float = 1.0
    debris: float = 0.0
    oxidative_stress: float = 0.0
    background_shorten_rate: float = 0.1
    telomerase_efficiency: float = 1.0
    stress_to_damage_rate: float = 0.01
    min_nutrients_to_divide: float = 0.3
    recycling_yield: float = 0.5

    def step(self, phagocyte: Phagocyte):
        phagocyte.cleanup(self)


# -----------------------------
# Controllers
# -----------------------------
class PIDTelomeraseController:
    def __init__(self, target: float, kp: float = 0.1, ki: float = 0.01, kd: float = 0.0):
        self.target = target
        self.kp = kp
        self.ki = ki
        self.kd = kd
        self.integral = 0.0
        self.last_error = None

    def action(self, mean_telomere: float) -> float:
        error = self.target - mean_telomere
        self.integral += error
        derivative = 0.0 if (self.last_error is None) else (error - self.last_error)
        self.last_error = error
        raw = self.kp * error + self.ki * self.integral + self.kd * derivative
        return float(np.clip(raw, 0.0, 1.0))


class RLControllerPlaceholder:
    def __init__(self):
        pass

    def action(self, mean_telomere: float, mean_health: float, env_state: Dict[str, float]) -> float:
        if mean_telomere < 50.0:
            return 0.8
        if mean_health < 0.5:
            return 0.4
        return 0.1


# -----------------------------
# Runnable PyTorch fragility model
# -----------------------------
MODEL_PATH = "fragility_mlp.pth"

class FragilityDataset(Dataset):
    def __init__(self, X: np.ndarray, y: np.ndarray):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]


class FragilityMLP(nn.Module):
    def __init__(self, input_dim: int = 3, hidden: int = 64):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden),
            nn.ReLU(),
            nn.Linear(hidden, hidden),
            nn.ReLU(),
            nn.Linear(hidden, 1),
            nn.Sigmoid(),
        )

    def forward(self, x):
        return self.net(x).squeeze(-1)


def featurize_repeat(repeat: str) -> List[float]:
    # simple per-repeat features: gc_frac, max_g_run_norm, length_norm
    gc = (repeat.count('G') + repeat.count('C')) / max(1, len(repeat))
    max_g = 0
    run = 0
    for c in repeat:
        if c == 'G':
            run += 1
            max_g = max(max_g, run)
        else:
            run = 0
    max_g_norm = max_g / len(repeat)
    length_norm = len(repeat) / 6.0
    return [gc, max_g_norm, length_norm]


def generate_synthetic_fragility_dataset(n_sequences: int = 5000, motif_len: int = 6, repeats: int = 10):
    X_list = []
    y_list = []
    scaffold = GNNTelomereScaffold()
    for _ in range(n_sequences):
        # generate random motif with some bias
        base_choices = ['A', 'T', 'G', 'C']
        motif = ''.join(np.random.choice(base_choices, size=motif_len))
        seq = motif * repeats
        # create per-repeat features and synthetic target using scaffold
        per_repeat = [seq[i:i+motif_len] for i in range(0, len(seq), motif_len)]
        frag_scores = scaffold.predict(seq)
        for r, score in zip(per_repeat, frag_scores):
            feats = featurize_repeat(r)
            X_list.append(feats)
            y_list.append(score)
    X = np.array(X_list)
    y = np.array(y_list)
    return X, y


def train_fragility_model(epochs: int = 6, batch_size: int = 256, lr: float = 1e-3, small_dataset: bool = True):
    if not TORCH_AVAILABLE:
        raise RuntimeError("PyTorch is not available. Install torch in Colab to run the trainer.")
    X, y = generate_synthetic_fragility_dataset(n_sequences=200 if small_dataset else 2000, repeats=8)
    ds = FragilityDataset(X, y)
    dl = DataLoader(ds, batch_size=batch_size, shuffle=True)

    model = FragilityMLP(input_dim=X.shape[1], hidden=64)
    opt = torch.optim.Adam(model.parameters(), lr=lr)
    loss_fn = nn.MSELoss()

    model.train()
    for ep in range(epochs):
        running = 0.0
        for xb, yb in dl:
            pred = model(xb)
            loss = loss_fn(pred, yb)
            opt.zero_grad()
            loss.backward()
            opt.step()
            running += loss.item() * xb.size(0)
        avg = running / len(ds)
        print(f"Epoch {ep+1}/{epochs} - loss {avg:.6f}")

    torch.save(model.state_dict(), MODEL_PATH)
    print(f"Saved fragility model to {MODEL_PATH}")
    return model


def load_fragility_model():
    if not TORCH_AVAILABLE:
        return None
    model = FragilityMLP(input_dim=3, hidden=64)
    try:
        model.load_state_dict(torch.load(MODEL_PATH, map_location='cpu'))
        model.eval()
        return model
    except Exception:
        return None


def predict_fragility_with_model(model, telomere_sequence: str) -> List[float]:
    per_repeat = [telomere_sequence[i:i+6] for i in range(0, len(telomere_sequence), 6)]
    feats = np.array([featurize_repeat(r) for r in per_repeat], dtype=np.float32)
    with torch.no_grad():
        xb = torch.tensor(feats)
        preds = model(xb).numpy().tolist()
    return preds


# -----------------------------
# Simulator (integrates molecular pieces and optionally PyTorch model)
# -----------------------------
@dataclass
class Simulator:
    n_cells: int = 50
    steps: int = 1000
    env: Environment = field(default_factory=Environment)
    phagocyte: Phagocyte = field(default_factory=lambda: Phagocyte(efficiency=0.6))
    controller: Any = field(default_factory=lambda: PIDTelomeraseController(target=100.0))
    seed: int = 42
    fragility_model: Optional[Any] = None

    traces: List[Dict[str, Any]] = field(default_factory=list)

    def __post_init__(self):
        seed_everything(self.seed)
        self.cells: List[VirtualCell] = []
        for i in range(self.n_cells):
            initial_len = np.clip(np.random.normal(120.0, 10.0), 10.0, 160.0)
            tel = Telomere(length=initial_len,
                           molecular_model=TelomereMolecularModel(motif="TTAGGG", repeat_count=int(initial_len/4)),
                           tloop=TLoopModel(stiffness=np.clip(np.random.normal(0.6, 0.1), 0.2, 0.9)))
            cell = VirtualCell(id=i, telomere=tel)
            self.cells.append(cell)

    def step(self, step_idx: int):
        mean_tel = float(np.mean([c.telomere.length for c in self.cells]))
        mean_health = float(np.mean([c.health for c in self.cells]))

        if hasattr(self.controller, "action"):
            try:
                telomerase_activity = self.controller.action(mean_tel)
            except TypeError:
                telomerase_activity = self.controller.action(mean_tel, mean_health, {
                    "oxidative_stress": self.env.oxidative_stress,
                    "debris": self.env.debris,
                })
        else:
            telomerase_activity = 0.1

        consumption_per_cell = 0.01
        total_consumption = consumption_per_cell * len(self.cells)
        self.env.nutrients = max(0.0, self.env.nutrients - total_consumption)

        new_cells: List[VirtualCell] = []
        for cell in self.cells:
            cell.step(self.env, telomerase_activity)

            if cell.dividing:
                self.env.nutrients = max(0.0, self.env.nutrients - 0.05)
                cell.telomere.shorten(1.0 + np.random.rand() * 0.5)
                daughter_len = max(0.0, cell.telomere.length - np.random.normal(0.5, 0.2))
                daughter = VirtualCell(
                    id=max(c.id for c in self.cells + new_cells) + 1,
                    telomere=Telomere(length=daughter_len,
                                      molecular_model=TelomereMolecularModel(motif=cell.telomere.molecular_model.motif, repeat_count=int(daughter_len/4)),
                                      tloop=TLoopModel(stiffness=cell.telomere.tloop.stiffness)),
                )
                new_cells.append(daughter)

            if cell.telomere.is_senescent() or cell.health < 0.05:
                self.env.debris += 0.5 + random.random() * 0.5
                continue
            else:
                new_cells.append(cell)

        self.cells = new_cells
        self.env.step(self.phagocyte)

        # generate fragility trace using either trained model or heuristic scaffold
        fragilities = []
        model = self.fragility_model
        if model is not None and TORCH_AVAILABLE:
            try:
                for c in self.cells:
                    seq = c.telomere.molecular_model.sequence
                    preds = predict_fragility_with_model(model, seq)
                    fragilities.append(float(np.mean(preds)))
            except Exception:
                # fallback
                fragilities = [np.mean(GNNTelomereScaffold().predict(c.telomere.molecular_model.sequence)) for c in self.cells]
        else:
            fragilities = [np.mean(GNNTelomereScaffold().predict(c.telomere.molecular_model.sequence)) for c in self.cells]

        trace = {
            "step": step_idx,
            "n_cells": len(self.cells),
            "mean_telomere": float(np.mean([c.telomere.length for c in self.cells])) if len(self.cells) > 0 else 0.0,
            "mean_health": float(np.mean([c.health for c in self.cells])) if len(self.cells) > 0 else 0.0,
            "nutrients": float(self.env.nutrients),
            "debris": float(self.env.debris),
            "oxidative_stress": float(self.env.oxidative_stress),
            "telomerase_activity": float(telomerase_activity),
            "mean_fragility": float(np.mean(fragilities)) if len(fragilities)>0 else 0.0,
        }
        self.traces.append(trace)

    def run(self, steps: Optional[int] = None, progress: bool = True):
        steps = steps or self.steps
        iterator = trange(steps) if progress else range(steps)
        for t in iterator:
            if t % 200 == 0 and t > 0:
                self.env.oxidative_stress = min(1.0, self.env.oxidative_stress + 0.2)
            self.env.oxidative_stress = max(0.0, self.env.oxidative_stress - 0.002)
            self.step(t)

    def to_dataframe(self) -> pd.DataFrame:
        return pd.DataFrame(self.traces)

    def save_csv(self, path: str = "simulation_traces.csv"):
        df = self.to_dataframe()
        df.to_csv(path, index=False)
        return path

    def plot(self, save_path: Optional[str] = None):
        df = self.to_dataframe()
        fig, axs = plt.subplots(4, 1, figsize=(8, 12), constrained_layout=True)

        axs[0].plot(df["step"], df["mean_telomere"])
        axs[0].set_title("Mean Telomere Length")

        axs[1].plot(df["step"], df["mean_health"])
        axs[1].set_title("Mean Cell Health")

        axs[2].plot(df["step"], df["n_cells"], label="n_cells")
        axs[2].plot(df["step"], df["nutrients"], label="nutrients")
        axs[2].plot(df["step"], df["debris"], label="debris")
        axs[2].set_title("Population / Environment")
        axs[2].legend()

        axs[3].plot(df["step"], df["mean_fragility"])
        axs[3].set_title("Mean Molecular Fragility (model)")

        if save_path:
            fig.savefig(save_path)
        plt.show()


# -----------------------------
# Demo / main
# -----------------------------

def demo_run(seed: int = 123, steps: int = 400, use_rl: bool = False, use_frag_model: bool = False):
    seed_everything(seed)

    env = Environment(nutrients=5.0, debris=0.0, oxidative_stress=0.05)
    phag = Phagocyte(efficiency=0.7)

    if use_rl:
        controller = RLControllerPlaceholder()
    else:
        controller = PIDTelomeraseController(target=120.0, kp=0.02, ki=0.001, kd=0.0)

    frag_model = None
    if use_frag_model and TORCH_AVAILABLE:
        frag_model = load_fragility_model()
        if frag_model is None:
            print("No trained fragility model found - training a small one now...")
            frag_model = train_fragility_model(epochs=4, small_dataset=True)

    sim = Simulator(n_cells=60, steps=steps, env=env, phagocyte=phag, controller=controller, seed=seed, fragility_model=frag_model)

    sim.run(steps, progress=True)

    df = sim.to_dataframe()
    csv_path = sim.save_csv("colab_simulation_traces_molecular_pytorch.csv")
    print(f"Saved CSV to: {csv_path}")

    sim.plot(save_path="colab_simulation_plots_molecular_pytorch.png")

    return sim, df


if __name__ == "__main__":
    # Example: train model first (requires torch) then run simulation using it
    if TORCH_AVAILABLE:
        print("PyTorch available - training a tiny fragility model for demo (this may take ~30s on CPU)")
        train_fragility_model(epochs=3, small_dataset=True)
        sim, df = demo_run(seed=42, steps=200, use_rl=False, use_frag_model=True)
    else:
        print("PyTorch not available - run demo without model. To enable, install torch in Colab.")
        sim, df = demo_run(seed=42, steps=200, use_rl=False, use_frag_model=False)
    print(df.tail())
````

````python
def simulate_live(telomerase_target, stress_level, phago_efficiency, div_threshold, steps=100, delay=0.05):
    telomere_length = 1.0
    fragility = 0.0
    model = load_fragility_model()
    results = []

    plt.figure(figsize=(8,4))
    for t in range(steps):
        x = torch.tensor([[telomere_length, stress_level, phago_efficiency, fragility]], dtype=torch.float32)
        fragility = float(model(x).item())
        telomere_length += (telomerase_target - fragility*0.2 - stress_level*0.05)
        telomere_length = np.clip(telomere_length, 0, 2)
        division_ready = telomere_length > div_threshold
        results.append((t, telomere_length, fragility, division_ready))

        # Live update
        clear_output(wait=True)
        plt.clf()
        plt.plot([r[0] for r in results], [r[1] for r in results], label='Telomere Length', color='tab:blue')
        plt.plot([r[0] for r in results], [r[2] for r in results], label='Fragility', color='tab:red')
        plt.title(f"Step {t+1}/{steps} â€” {'Dividing' if division_ready else 'Stable'}")
        plt.xlabel('Step')
        plt.ylabel('Value')
        plt.legend()
        plt.grid(True)
        plt.pause(0.001)
        time.sleep(delay)

    plt.show()
    df = pd.DataFrame(results, columns=['Step','Telomere','Fragility','Divide'])
    return df
````
