```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
from copy import deepcopy
import torch
import torch.nn as nn
import torch.optim as optim
from typing import List, Dict, Tuple

# ----------------------------- Utilities ---------------------------------

def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)

set_seed(42)

# -------------------------- Virtual Cell Model ---------------------------

class VirtualCell:
    """
    Abstract virtual cell with an internal telomere "length" state,
    a telomerase activity scalar (0..1), a simple damage accumulator and a
    phagocytosis activity indicator. All units are abstract and nondimensional.
    """
    def __init__(self, tel_len: float=1.0):
        self.tel_len = tel_len  # abstract telomere metric
        self.tertiary_state = 0.0  # placeholder for additional state
        self.age = 0
        self.alive = True
        self.damage = 0.0
        self.phag_activity = 0.0

    def step(self, telomerase_pulse: float, microenv_signal: float,
             damage_input: float=0.0, phag_trigger: float=0.0):
        """Advance 1 timestep. telomerase_pulse ∈ [0,1].
        microenv_signal is an abstract growth-factor-like scalar.
        damage_input adds acute damage.
        phag_trigger boosts phagocytic cleanup."""
        if not self.alive:
            return

        # accumulate damage
        self.damage += max(0.0, damage_input - self.phag_activity*0.5)

        # telomere update (abstract): growth from telomerase; shrinkage from division
        growth = telomerase_pulse * 0.1 * (1.0 - math.tanh(self.tel_len - 2.0))
        shrink = 0.02 * float(microenv_signal > 0.5)  # division-related shortening
        # self-correcting trimming if telomeres too long
        trimming = 0.05 * max(0.0, self.tel_len - 3.0)
        self.tel_len = max(0.1, self.tel_len + growth - shrink - trimming)

        # phagocytosis dynamics (abstract)
        self.phag_activity = max(0.0, min(1.0, self.phag_activity + 0.1*phag_trigger - 0.02*self.phag_activity))

        # damage influences viability
        if self.damage > 2.5:
            self.alive = False

        self.age += 1

    def copy(self):
        return deepcopy(self)

# ------------------------ Virtual OOC Environment ------------------------

class VirtualOOC:
    """
    Minimal virtual organ-on-chip abstraction: a 1D array of cells with
    nutrient/flow field and events like injury or pathogen pulses.
    """
    def __init__(self, n_cells=64):
        self.n = n_cells
        self.cells = [VirtualCell(tel_len=1.0 + 0.1*random.random()) for _ in range(n_cells)]
        self.nutrient = np.ones(n_cells)
        self.flow = np.zeros(n_cells)
        self.time = 0

    def step(self, tel_pulses: np.ndarray, nutrient_in: float=1.0, pathogen_pulse: float=0.0, injury_mask: np.ndarray=None):
        # simple flow — shift nutrients along
        self.flow = np.roll(self.flow, 1)
        self.flow[0] = nutrient_in
        self.nutrient = 0.9*self.nutrient + 0.1*self.flow

        # apply events
        for i, cell in enumerate(self.cells):
            damage_input = float(pathogen_pulse * (0.5 + 0.5*random.random()))
            phag = float(pathogen_pulse > 0.1)  # if pathogen exists, stimulate phag
            if injury_mask is not None and injury_mask[i]:
                damage_input += 0.5

            microenv_signal = self.nutrient[i]
            cell.step(telomerase_pulse=float(tel_pulses[i]), microenv_signal=microenv_signal,
                      damage_input=damage_input, phag_trigger=phag)

        self.time += 1

    def snapshot(self) -> Dict[str, np.ndarray]:
        tel = np.array([c.tel_len for c in self.cells])
        alive = np.array([1.0 if c.alive else 0.0 for c in self.cells])
        damage = np.array([c.damage for c in self.cells])
        phag = np.array([c.phag_activity for c in self.cells])
        return {"tel_len": tel, "alive": alive, "damage": damage, "phag": phag, "nutrient": self.nutrient.copy()}

    def copy(self):
        return deepcopy(self)

# ------------------------- Data Generator --------------------------------

def generate_simulation_sequence(length: int=50, n_cells: int=32, seed: int=None,
                                 injury_time: int=20, pathogen_times: List[int]=None,
                                 instability_rate: float=0.0) -> Dict[str, np.ndarray]:
    """Run a short simulation and record observations and oracle telomerase pulses.
    We create an "oracle" controller that pulses telomerase based on simple rules
    (this oracle is what we try to learn). instability_rate randomly corrupts cells' telomeres to simulate genomic instability."""
    if seed is not None:
        set_seed(seed)
    env = VirtualOOC(n_cells=n_cells)

    X = []  # observations per timestep
    Y = []  # telomerase pulses (oracle)

    for t in range(length):
        # define injury mask at injury_time
        injury_mask = None
        if t == injury_time:
            injury_mask = np.zeros(n_cells, dtype=bool)
            center = n_cells // 2
            span = max(1, n_cells // 8)
            injury_mask[center-span:center+span] = True

        pathogen_pulse = 0.0
        if pathogen_times and t in pathogen_times:
            pathogen_pulse = 1.0

        snap = env.snapshot()
        # oracle telomerase policy: if local telomere < threshold and local nutrient high and cell alive -> pulse
        tel = snap["tel_len"]
        nutrient = snap["nutrient"]
        alive = snap["alive"]
        oracle_pulse = np.clip( (0.5 - tel)*2.0 + 0.5*(nutrient - 0.8), 0.0, 1.0)
        oracle_pulse = oracle_pulse * alive

        # apply small random interventions simulating regulation noise
        oracle_pulse = np.clip(oracle_pulse + 0.1*(np.random.rand(n_cells)-0.5), 0.0, 1.0)

        # instability: randomly increase damage/telomere disruptions
        if instability_rate > 0.0:
            for i in range(n_cells):
                if random.random() < instability_rate:
                    env.cells[i].damage += 0.5*(random.random())
                    env.cells[i].tel_len = max(0.1, env.cells[i].tel_len * (0.5 + random.random()))

        # record
        obs = np.stack([snap["tel_len"], snap["damage"], snap["phag"], snap["nutrient"], snap["alive"]], axis=1)
        X.append(obs)
        Y.append(oracle_pulse.reshape(-1,1))

        # step environment
        env.step(tel_pulses=oracle_pulse, nutrient_in=1.0, pathogen_pulse=pathogen_pulse, injury_mask=injury_mask)

    X = np.array(X)  # (T, n_cells, features)
    Y = np.array(Y)  # (T, n_cells, 1)
    return {"X": X, "Y": Y}

# -------------------------- PyTorch Model --------------------------------

class TelomereController(nn.Module):
    """Temporal model that predicts per-cell telomerase pulse given recent history.
    Input: (batch, seq_len, n_cells, features) -> we flatten cells into batch* n_cells
    For simplicity we treat each cell independently across space (no graph conv).
    """
    def __init__(self, in_features=5, hidden=64, seq_len=8):
        super().__init__()
        self.seq_len = seq_len
        self.rnn = nn.GRU(input_size=in_features, hidden_size=hidden, batch_first=True)
        self.fc = nn.Sequential(
            nn.Linear(hidden, 64),
            nn.ReLU(),
            nn.Linear(64, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        # x shape: (B, seq_len, in_features)
        out, _ = self.rnn(x)
        # take last timestep
        h = out[:, -1, :]
        return self.fc(h)

# --------------------------- Training Utils -------------------------------

def prepare_batches(sim_data_list: List[Dict], seq_len=8) -> Tuple[torch.Tensor, torch.Tensor]:
    """Turn list of sim runs into training samples. Each cell becomes a sample.
    Returns X: (N_samples, seq_len, features), Y: (N_samples,1) predicting next-step pulse."""
    X_samples = []
    Y_samples = []
    for sim in sim_data_list:
        X = sim['X']  # (T, n_cells, feat)
        Y = sim['Y']  # (T, n_cells, 1)
        T, n_cells, feat = X.shape
        for t in range(seq_len, T-1):
            xwin = X[t-seq_len:t, :, :].reshape(seq_len, n_cells, feat)
            ynext = Y[t+1, :, :]
            # create per-cell samples
            for c in range(n_cells):
                X_samples.append(xwin[:, c, :])
                Y_samples.append(ynext[c, 0])
    X_arr = np.stack(X_samples, axis=0)
    Y_arr = np.array(Y_samples).reshape(-1,1)
    return torch.tensor(X_arr, dtype=torch.float32), torch.tensor(Y_arr, dtype=torch.float32)

# ------------------------------ Demo / Main -------------------------------

def train_demo(epochs=6, small_data=True):
    # generate dataset
    sims = []
    for i in range(20 if not small_data else 8):
        sims.append(generate_simulation_sequence(length=50, n_cells=16, seed=100+i,
                                                 injury_time=20, pathogen_times=[30] if i%3==0 else [],
                                                 instability_rate=0.02 if i%5==0 else 0.0))

    seq_len = 8
    X, Y = prepare_batches(sims, seq_len=seq_len)
    print('Prepared dataset shapes:', X.shape, Y.shape)

    # tiny train/val split
    N = X.shape[0]
    idx = np.arange(N)
    np.random.shuffle(idx)
    train_idx = idx[:int(0.8*N)]
    val_idx = idx[int(0.8*N):]

    X_train = X[train_idx]
    Y_train = Y[train_idx]
    X_val = X[val_idx]
    Y_val = Y[val_idx]

    model = TelomereController(in_features=X.shape[-1], hidden=64, seq_len=seq_len)
    opt = optim.Adam(model.parameters(), lr=1e-3)
    loss_fn = nn.MSELoss()

    for ep in range(epochs):
        model.train()
        perm = torch.randperm(X_train.shape[0])
        total_loss = 0.0
        bs = 256
        for i in range(0, X_train.shape[0], bs):
            batch_idx = perm[i:i+bs]
            xb = X_train[batch_idx]
            yb = Y_train[batch_idx]
            pred = model(xb)
            loss = loss_fn(pred, yb)
            opt.zero_grad()
            loss.backward()
            opt.step()
            total_loss += loss.item() * xb.shape[0]
        total_loss /= X_train.shape[0]

        # val
        model.eval()
        with torch.no_grad():
            val_pred = model(X_val)
            val_loss = loss_fn(val_pred, Y_val).item()
        print(f'Epoch {ep+1}/{epochs} train_loss={total_loss:.5f} val_loss={val_loss:.5f}')

    return model, sims

# --------------------------- Interpretability -----------------------------

def gradient_saliency(model: nn.Module, sample: torch.Tensor) -> np.ndarray:
    """Compute gradient w.r.t. input features for a single sample to get saliency.
    sample shape: (1, seq_len, features)
    Returns saliency array shape (seq_len, features)
    """
    model.eval()
    x = sample.clone().detach().requires_grad_(True)
    out = model(x)
    out_sum = out.sum()
    out_sum.backward()
    sal = x.grad.detach().numpy()[0]
    return sal

# --------------------------- Example Experiments -------------------------

def run_experiment(model: nn.Module, experiment: str='damage_regen'):
    if experiment == 'damage_regen':
        # Define injury_time explicitly for the experiment
        experiment_injury_time = 20
        sim = generate_simulation_sequence(length=80, n_cells=32, seed=999,
                                           injury_time=experiment_injury_time, pathogen_times=None)
        # run with learned controller in closed loop
        env = VirtualOOC(n_cells=32)
        seq_len = 8
        # initialize with warmup oracle steps
        history = []
        for t in range(seq_len):
            snap = env.snapshot()
            obs = np.stack([snap['tel_len'], snap['damage'], snap['phag'], snap['nutrient'], snap['alive']], axis=1)
            history.append(obs)
            env.step(tel_pulses=np.zeros(env.n), nutrient_in=1.0)

        history = np.array(history)
        records = []
        for t in range(seq_len, 80):
            # prepare per-cell samples for model
            X_batch = torch.tensor(history[-seq_len:], dtype=torch.float32).unsqueeze(0) # (1,seq_len,n_cells,feat)
            # reshape to (n_cells, seq_len, feat)
            n_cells = X_batch.shape[2]
            samples = X_batch.squeeze(0).transpose(0,1).reshape(n_cells, seq_len, X_batch.shape[3])
            with torch.no_grad():
                preds = model(samples).numpy().squeeze()
            tel_pulses = preds.clip(0.0,1.0)

            # Create injury mask as a numpy array if at injury_time, else None
            current_injury_mask = None
            if t == experiment_injury_time:
                current_injury_mask = np.zeros(env.n, dtype=bool)
                center = env.n // 2
                span = max(1, env.n // 8)
                current_injury_mask[center-span:center+span] = True

            env.step(tel_pulses=tel_pulses, nutrient_in=1.0, pathogen_pulse=0.0, injury_mask=current_injury_mask)
            snap = env.snapshot()
            records.append(snap)
            obs = np.stack([snap['tel_len'], snap['damage'], snap['phag'], snap['nutrient'], snap['alive']], axis=1)
            history = np.concatenate([history, obs[np.newaxis,:,:]], axis=0)

        # plot average telomere length over time
        avg_tel = [r['tel_len'].mean() for r in records]
        plt.figure()
        plt.plot(avg_tel)
        plt.title('Damage & Regeneration: Avg telomere length over time')
        plt.xlabel('t')
        plt.ylabel('avg tel len')
        plt.show()

    elif experiment == 'pathogen_challenge':
        sim = generate_simulation_sequence(length=80, n_cells=32, seed=111,
                                           injury_time=20, pathogen_times=[25,40,55])
        print('Generated pathogen-challenge sim (not run in closed loop in this example).')

    elif experiment == 'instability_test':
        sim = generate_simulation_sequence(length=80, n_cells=32, seed=222,
                                           injury_time=20, pathogen_times=None, instability_rate=0.05)
        print('Generated instability-challenge sim (inspect sim data for failures).')

    else:
        print('Unknown experiment')

# --------------------------------- Run ----------------------------------

if __name__ == '__main__':
    print('Running short training demo...')
    model, sims = train_demo(epochs=6, small_data=True)

    # saliency on a random sample
    X, Y = prepare_batches(sims, seq_len=8)
    sample = X[10:11]
    sal = gradient_saliency(model, sample)
    print('Saliency shape:', sal.shape)

    # run a closed-loop experiment
    run_experiment(model, experiment='damage_regen')

    print('''Demo complete. Inspect and extend the script for longer runs, graph-based
analysis, or more complex controllers (actor-critic, multi-agent, graph nets).''')

# ------------------------- Reinforcement Learning Agent -------------------------
# Lightweight actor-critic (PPO-style clipped objective) to learn telomerase pulses
# The RL agent treats each cell as an independent actor for simplicity (no spatial coupling)
# and learns a stochastic policy producing telomerase pulses in [0,1]. The environment
# remains the VirtualOOC defined above. This is a minimal, educational RL loop; tune
# hyperparameters and add stability features for production research.

class ActorCritic(nn.Module):
    def __init__(self, in_features=5, hidden=64):
        super().__init__()
        # shared base
        self.base = nn.GRU(input_size=in_features, hidden_size=hidden, batch_first=True)
        # policy head outputs mean and logstd
        self.policy = nn.Sequential(
            nn.Linear(hidden, 64), nn.ReLU(), nn.Linear(64, 1)
        )
        # value head
        self.value = nn.Sequential(
            nn.Linear(hidden, 64), nn.ReLU(), nn.Linear(64, 1)
        )
        # learnable log std
        self.log_std = nn.Parameter(torch.tensor(-1.0))

    def forward(self, x):
        # x shape (B, seq_len, features)
        out, _ = self.base(x)
        h = out[:, -1, :]
        mean = self.policy(h)
        value = self.value(h)
        std = torch.exp(self.log_std)
        return mean.squeeze(-1), std, value.squeeze(-1)


def run_rl_agent(episodes=200, seq_len=8, n_cells=16, gamma=0.99, clip_eps=0.2):
    set_seed(123)
    env = VirtualOOC(n_cells=n_cells)
    agent = ActorCritic(in_features=5, hidden=64)
    optim_ac = optim.Adam(agent.parameters(), lr=3e-4)

    # helper to compute returns
    def compute_returns(rewards, dones, next_value, gamma):
        R = next_value
        returns = []
        for r, d in zip(rewards[::-1], dones[::-1]):
            R = r + gamma * R * (1.0 - d)
            returns.insert(0, R)
        return returns

    for ep in range(episodes):
        # reset env
        env = VirtualOOC(n_cells=n_cells)
        # warmup history
        history = []
        for t in range(seq_len):
            snap = env.snapshot()
            obs = np.stack([snap['tel_len'], snap['damage'], snap['phag'], snap['nutrient'], snap['alive']], axis=1)
            history.append(obs)
            env.step(tel_pulses=np.zeros(env.n), nutrient_in=1.0)
        history = np.array(history)

        ep_rewards = []
        log_probs = []
        values = []
        entropies = []
        observations = []
        actions = []
        dones = []

        for t in range(200):
            # prepare per-cell samples
            # shape (n_cells, seq_len, feat)
            seq = torch.tensor(history[-seq_len:], dtype=torch.float32)  # (seq_len, n_cells, feat)
            seq = seq.transpose(0,1)  # (n_cells, seq_len, feat)
            mean, std, value = agent(seq)
            # sample actions from Normal(mean, std) and squash to [0,1]
            normal = torch.distributions.Normal(mean, std)
            raw_action = normal.rsample()
            action = torch.sigmoid(raw_action)
            # log prob with change-of-variable correction for sigmoid
            logp_raw = normal.log_prob(raw_action)
            # derivative of sigmoid = sigmoid*(1-sigmoid)
            log_det = - (torch.log(action) + torch.log(1.0 - action) + 1e-8)
            logp = (logp_raw - log_det).sum(dim=-1)
            entropy = normal.entropy().sum(dim=-1)

            tel_pulses = action.detach().numpy()
            # step env
            env.step(tel_pulses=tel_pulses, nutrient_in=1.0)
            snap = env.snapshot()

            # define reward: encourage repair (increase alive cells), penalize damage and overextending telomeres
            alive_bonus = snap['alive'].sum() / n_cells
            avg_damage = snap['damage'].mean()
            avg_tel = snap['tel_len'].mean()
            reward = alive_bonus - 0.3*avg_damage - 0.05*max(0.0, avg_tel - 2.5)
            # per-step scalar reward
            r = float(reward)

            # record
            ep_rewards.append(r)
            log_probs.append(logp)
            values.append(value)
            entropies.append(entropy)
            observations.append(seq)
            actions.append(action)
            dones.append(0.0 if snap['alive'].sum()>0 else 1.0)

            # append new obs to history
            obs = np.stack([snap['tel_len'], snap['damage'], snap['phag'], snap['nutrient'], snap['alive']], axis=1)
            history = np.concatenate([history, obs[np.newaxis,:,:]], axis=0)

            if all(c==0.0 for c in snap['alive']):
                break

        # compute returns and advantages
        with torch.no_grad():
            next_val = agent(torch.tensor(observations[-1], dtype=torch.float32))[2] if len(observations)>0 else torch.tensor(0.0)
        returns = compute_returns(ep_rewards, dones, float(next_val.mean()), gamma)
        returns = torch.tensor(returns, dtype=torch.float32)

        # flatten tensors for PPO update
        logp_tensor = torch.stack([lp.mean() if lp.dim()>0 else lp for lp in log_probs])
        value_tensor = torch.stack([v.mean() for v in values])
        advantage = returns - value_tensor.detach()

        # policy loss (clipped)
        ratio = torch.exp(logp_tensor - logp_tensor.detach())
        surr1 = ratio * advantage
        surr2 = torch.clamp(ratio, 1.0-clip_eps, 1.0+clip_eps) * advantage
        policy_loss = -torch.min(surr1, surr2).mean()
        value_loss = nn.MSELoss()(value_tensor, returns)
        entropy_loss = -torch.stack(entropies).mean()

        loss = policy_loss + 0.5*value_loss + 0.01*entropy_loss
        optim_ac.zero_grad()
        loss.backward()
        optim_ac.step()

        if ep % 10 == 0:
            print(f'RL Episode {ep} total_reward={sum(ep_rewards):.3f} steps={len(ep_rewards)} avg_reward={np.mean(ep_rewards):.4f}')

    print('RL run complete. Inspect agent behavior and tune reward or architecture.')

# Example: run a short RL training (disabled by default to avoid long runs)
if __name__ == '__main__':
    # Uncomment to run RL training
    # run_rl_agent(episodes=100, seq_len=8, n_cells=16)
    pass
```

```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
from copy import deepcopy
import torch
import torch.nn as nn
import torch.optim as optim
from typing import List, Dict, Tuple

# ----------------------------- Utilities ---------------------------------

def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)

set_seed(42)

# -------------------------- Virtual Cell Model ---------------------------

class VirtualCell:
    """
    Abstract virtual cell with an internal telomere "length" state,
    a telomerase activity scalar (0..1), a simple damage accumulator and a
    phagocytosis activity indicator. All units are abstract and nondimensional.
    """
    def __init__(self, tel_len: float=1.0):
        self.tel_len = tel_len  # abstract telomere metric
        self.tertiary_state = 0.0  # placeholder for additional state
        self.age = 0
        self.alive = True
        self.damage = 0.0
        self.phag_activity = 0.0

    def step(self, telomerase_pulse: float, microenv_signal: float,
             damage_input: float=0.0, phag_trigger: float=0.0):
        """Advance 1 timestep. telomerase_pulse ∈ [0,1].
        microenv_signal is an abstract growth-factor-like scalar.
        damage_input adds acute damage.
        phag_trigger boosts phagocytic cleanup."""
        if not self.alive:
            return

        # accumulate damage
        self.damage += max(0.0, damage_input - self.phag_activity*0.5)

        # telomere update (abstract): growth from telomerase; shrinkage from division
        growth = telomerase_pulse * 0.1 * (1.0 - math.tanh(self.tel_len - 2.0))
        shrink = 0.02 * float(microenv_signal > 0.5)  # division-related shortening
        # self-correcting trimming if telomeres too long
        trimming = 0.05 * max(0.0, self.tel_len - 3.0)
        self.tel_len = max(0.1, self.tel_len + growth - shrink - trimming)

        # phagocytosis dynamics (abstract)
        self.phag_activity = max(0.0, min(1.0, self.phag_activity + 0.1*phag_trigger - 0.02*self.phag_activity))

        # damage influences viability
        if self.damage > 2.5:
            self.alive = False

        self.age += 1

    def copy(self):
        return deepcopy(self)

# ------------------------ Virtual OOC Environment ------------------------

class VirtualOOC:
    """
    Minimal virtual organ-on-chip abstraction: a 1D array of cells with
    nutrient/flow field and events like injury or pathogen pulses.
    """
    def __init__(self, n_cells=64):
        self.n = n_cells
        self.cells = [VirtualCell(tel_len=1.0 + 0.1*random.random()) for _ in range(n_cells)]
        self.nutrient = np.ones(n_cells)
        self.flow = np.zeros(n_cells)
        self.time = 0

    def step(self, tel_pulses: np.ndarray, nutrient_in: float=1.0, pathogen_pulse: float=0.0, injury_mask: np.ndarray=None):
        # simple flow — shift nutrients along
        self.flow = np.roll(self.flow, 1)
        self.flow[0] = nutrient_in
        self.nutrient = 0.9*self.nutrient + 0.1*self.flow

        # apply events
        for i, cell in enumerate(self.cells):
            damage_input = float(pathogen_pulse * (0.5 + 0.5*random.random()))
            phag = float(pathogen_pulse > 0.1)  # if pathogen exists, stimulate phag
            if injury_mask is not None and injury_mask[i]:
                damage_input += 0.5

            microenv_signal = self.nutrient[i]
            cell.step(telomerase_pulse=float(tel_pulses[i]), microenv_signal=microenv_signal,
                      damage_input=damage_input, phag_trigger=phag)

        self.time += 1

    def snapshot(self) -> Dict[str, np.ndarray]:
        tel = np.array([c.tel_len for c in self.cells])
        alive = np.array([1.0 if c.alive else 0.0 for c in self.cells])
        damage = np.array([c.damage for c in self.cells])
        phag = np.array([c.phag_activity for c in self.cells])
        return {"tel_len": tel, "alive": alive, "damage": damage, "phag": phag, "nutrient": self.nutrient.copy()}

    def copy(self):
        return deepcopy(self)

# ------------------------- Data Generator --------------------------------

def generate_simulation_sequence(length: int=50, n_cells: int=32, seed: int=None,
                                 injury_time: int=20, pathogen_times: List[int]=None,
                                 instability_rate: float=0.0) -> Dict[str, np.ndarray]:
    """Run a short simulation and record observations and oracle telomerase pulses.
    We create an "oracle" controller that pulses telomerase based on simple rules
    (this oracle is what we try to learn). instability_rate randomly corrupts cells' telomeres to simulate genomic instability."""
    if seed is not None:
        set_seed(seed)
    env = VirtualOOC(n_cells=n_cells)

    X = []  # observations per timestep
    Y = []  # telomerase pulses (oracle)

    for t in range(length):
        # define injury mask at injury_time
        injury_mask = None
        if t == injury_time:
            injury_mask = np.zeros(n_cells, dtype=bool)
            center = n_cells // 2
            span = max(1, n_cells // 8)
            injury_mask[center-span:center+span] = True

        pathogen_pulse = 0.0
        if pathogen_times and t in pathogen_times:
            pathogen_pulse = 1.0

        snap = env.snapshot()
        # oracle telomerase policy: if local telomere < threshold and local nutrient high and cell alive -> pulse
        tel = snap["tel_len"]
        nutrient = snap["nutrient"]
        alive = snap["alive"]
        oracle_pulse = np.clip( (0.5 - tel)*2.0 + 0.5*(nutrient - 0.8), 0.0, 1.0)
        oracle_pulse = oracle_pulse * alive

        # apply small random interventions simulating regulation noise
        oracle_pulse = np.clip(oracle_pulse + 0.1*(np.random.rand(n_cells)-0.5), 0.0, 1.0)

        # instability: randomly increase damage/telomere disruptions
        if instability_rate > 0.0:
            for i in range(n_cells):
                if random.random() < instability_rate:
                    env.cells[i].damage += 0.5*(random.random())
                    env.cells[i].tel_len = max(0.1, env.cells[i].tel_len * (0.5 + random.random()))

        # record
        obs = np.stack([snap["tel_len"], snap["damage"], snap["phag"], snap["nutrient"], snap["alive"]], axis=1)
        X.append(obs)
        Y.append(oracle_pulse.reshape(-1,1))

        # step environment
        env.step(tel_pulses=oracle_pulse, nutrient_in=1.0, pathogen_pulse=pathogen_pulse, injury_mask=injury_mask)

    X = np.array(X)  # (T, n_cells, features)
    Y = np.array(Y)  # (T, n_cells, 1)
    return {"X": X, "Y": Y}

# -------------------------- PyTorch Model --------------------------------

class TelomereController(nn.Module):
    """Temporal model that predicts per-cell telomerase pulse given recent history.
    Input: (batch, seq_len, n_cells, features) -> we flatten cells into batch* n_cells
    For simplicity we treat each cell independently across space (no graph conv).
    """
    def __init__(self, in_features=5, hidden=64, seq_len=8):
        super().__init__()
        self.seq_len = seq_len
        self.rnn = nn.GRU(input_size=in_features, hidden_size=hidden, batch_first=True)
        self.fc = nn.Sequential(
            nn.Linear(hidden, 64),
            nn.ReLU(),
            nn.Linear(64, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        # x shape: (B, seq_len, in_features)
        out, _ = self.rnn(x)
        # take last timestep
        h = out[:, -1, :]
        return self.fc(h)

# --------------------------- Training Utils -------------------------------

def prepare_batches(sim_data_list: List[Dict], seq_len=8) -> Tuple[torch.Tensor, torch.Tensor]:
    """Turn list of sim runs into training samples. Each cell becomes a sample.
    Returns X: (N_samples, seq_len, features), Y: (N_samples,1) predicting next-step pulse."""
    X_samples = []
    Y_samples = []
    for sim in sim_data_list:
        X = sim['X']  # (T, n_cells, feat)
        Y = sim['Y']  # (T, n_cells, 1)
        T, n_cells, feat = X.shape
        for t in range(seq_len, T-1):
            xwin = X[t-seq_len:t, :, :].reshape(seq_len, n_cells, feat)
            ynext = Y[t+1, :, :]
            # create per-cell samples
            for c in range(n_cells):
                X_samples.append(xwin[:, c, :])
                Y_samples.append(ynext[c, 0])
    X_arr = np.stack(X_samples, axis=0)
    Y_arr = np.array(Y_samples).reshape(-1,1)
    return torch.tensor(X_arr, dtype=torch.float32), torch.tensor(Y_arr, dtype=torch.float32)

# ------------------------------ Demo / Main -------------------------------

def train_demo(epochs=6, small_data=True):
    # generate dataset
    sims = []
    for i in range(20 if not small_data else 8):
        sims.append(generate_simulation_sequence(length=50, n_cells=16, seed=100+i,
                                                 injury_time=20, pathogen_times=[30] if i%3==0 else [],
                                                 instability_rate=0.02 if i%5==0 else 0.0))

    seq_len = 8
    X, Y = prepare_batches(sims, seq_len=seq_len)
    print('Prepared dataset shapes:', X.shape, Y.shape)

    # tiny train/val split
    N = X.shape[0]
    idx = np.arange(N)
    np.random.shuffle(idx)
    train_idx = idx[:int(0.8*N)]
    val_idx = idx[int(0.8*N):]

    X_train = X[train_idx]
    Y_train = Y[train_idx]
    X_val = X[val_idx]
    Y_val = Y[val_idx]

    model = TelomereController(in_features=X.shape[-1], hidden=64, seq_len=seq_len)
    opt = optim.Adam(model.parameters(), lr=1e-3)
    loss_fn = nn.MSELoss()

    for ep in range(epochs):
        model.train()
        perm = torch.randperm(X_train.shape[0])
        total_loss = 0.0
        bs = 256
        for i in range(0, X_train.shape[0], bs):
            batch_idx = perm[i:i+bs]
            xb = X_train[batch_idx]
            yb = Y_train[batch_idx]
            pred = model(xb)
            loss = loss_fn(pred, yb)
            opt.zero_grad()
            loss.backward()
            opt.step()
            total_loss += loss.item() * xb.shape[0]
        total_loss /= X_train.shape[0]

        # val
        model.eval()
        with torch.no_grad():
            val_pred = model(X_val)
            val_loss = loss_fn(val_pred, Y_val).item()
        print(f'Epoch {ep+1}/{epochs} train_loss={total_loss:.5f} val_loss={val_loss:.5f}')

    return model, sims

# --------------------------- Interpretability -----------------------------

def gradient_saliency(model: nn.Module, sample: torch.Tensor) -> np.ndarray:
    """Compute gradient w.r.t. input features for a single sample to get saliency.
    sample shape: (1, seq_len, features)
    Returns saliency array shape (seq_len, features)
    """
    model.eval()
    x = sample.clone().detach().requires_grad_(True)
    out = model(x)
    out_sum = out.sum()
    out_sum.backward()
    sal = x.grad.detach().numpy()[0]
    return sal

# --------------------------- Example Experiments -------------------------

def run_experiment(model: nn.Module, experiment: str='damage_regen'):
    if experiment == 'damage_regen':
        # Define injury_time explicitly for the experiment
        experiment_injury_time = 20
        sim = generate_simulation_sequence(length=80, n_cells=32, seed=999,
                                           injury_time=experiment_injury_time, pathogen_times=None)
        # run with learned controller in closed loop
        env = VirtualOOC(n_cells=32)
        seq_len = 8
        # initialize with warmup oracle steps
        history = []
        for t in range(seq_len):
            snap = env.snapshot()
            obs = np.stack([snap['tel_len'], snap['damage'], snap['phag'], snap['nutrient'], snap['alive']], axis=1)
            history.append(obs)
            env.step(tel_pulses=np.zeros(env.n), nutrient_in=1.0)

        history = np.array(history)
        records = []
        for t in range(seq_len, 80):
            # prepare per-cell samples for model
            X_batch = torch.tensor(history[-seq_len:], dtype=torch.float32).unsqueeze(0) # (1,seq_len,n_cells,feat)
            # reshape to (n_cells, seq_len, feat)
            n_cells = X_batch.shape[2]
            samples = X_batch.squeeze(0).transpose(0,1).reshape(n_cells, seq_len, X_batch.shape[3])
            with torch.no_grad():
                preds = model(samples).numpy().squeeze()
            tel_pulses = preds.clip(0.0,1.0)

            # Create injury mask as a numpy array if at injury_time, else None
            current_injury_mask = None
            if t == experiment_injury_time:
                current_injury_mask = np.zeros(env.n, dtype=bool)
                center = env.n // 2
                span = max(1, env.n // 8)
                current_injury_mask[center-span:center+span] = True

            env.step(tel_pulses=tel_pulses, nutrient_in=1.0, pathogen_pulse=0.0, injury_mask=current_injury_mask)
            snap = env.snapshot()
            records.append(snap)
            obs = np.stack([snap['tel_len'], snap['damage'], snap['phag'], snap['nutrient'], snap['alive']], axis=1)
            history = np.concatenate([history, obs[np.newaxis,:,:]], axis=0)

        # plot average telomere length over time
        avg_tel = [r['tel_len'].mean() for r in records]
        plt.figure()
        plt.plot(avg_tel)
        plt.title('Damage & Regeneration: Avg telomere length over time')
        plt.xlabel('t')
        plt.ylabel('avg tel len')
        plt.show()

    elif experiment == 'pathogen_challenge':
        sim = generate_simulation_sequence(length=80, n_cells=32, seed=111,
                                           injury_time=20, pathogen_times=[25,40,55])
        print('Generated pathogen-challenge sim (not run in closed loop in this example).')

    elif experiment == 'instability_test':
        sim = generate_simulation_sequence(length=80, n_cells=32, seed=222,
                                           injury_time=20, pathogen_times=None, instability_rate=0.05)
        print('Generated instability-challenge sim (inspect sim data for failures).')

    else:
        print('Unknown experiment')

# --------------------------------- Run ----------------------------------

if __name__ == '__main__':
    print('Running short training demo...')
    model, sims = train_demo(epochs=6, small_data=True)

    # saliency on a random sample
    X, Y = prepare_batches(sims, seq_len=8)
    sample = X[10:11]
    sal = gradient_saliency(model, sample)
    print('Saliency shape:', sal.shape)

    # run a closed-loop experiment
    run_experiment(model, experiment='damage_regen')

    print('''Demo complete. Inspect and extend the script for longer runs, graph-based
analysis, or more complex controllers (actor-critic, multi-agent, graph nets).''')

# ------------------------- Reinforcement Learning Agent -------------------------
# Lightweight actor-critic (PPO-style clipped objective) to learn telomerase pulses
# The RL agent treats each cell as an independent actor for simplicity (no spatial coupling)
# and learns a stochastic policy producing telomerase pulses in [0,1]. The environment
# remains the VirtualOOC defined above. This is a minimal, educational RL loop; tune
# hyperparameters and add stability features for production research.

class ActorCritic(nn.Module):
    def __init__(self, in_features=5, hidden=64):
        super().__init__()
        # shared base
        self.base = nn.GRU(input_size=in_features, hidden_size=hidden, batch_first=True)
        # policy head outputs mean and logstd
        self.policy = nn.Sequential(
            nn.Linear(hidden, 64), nn.ReLU(), nn.Linear(64, 1)
        )
        # value head
        self.value = nn.Sequential(
            nn.Linear(hidden, 64), nn.ReLU(), nn.Linear(64, 1)
        )
        # learnable log std
        self.log_std = nn.Parameter(torch.tensor(-1.0))

    def forward(self, x):
        # x shape (B, seq_len, features)
        out, _ = self.base(x)
        h = out[:, -1, :]
        mean = self.policy(h)
        value = self.value(h)
        std = torch.exp(self.log_std)
        return mean.squeeze(-1), std, value.squeeze(-1)


def run_rl_agent(episodes=200, seq_len=8, n_cells=16, gamma=0.99, clip_eps=0.2):
    set_seed(123)
    env = VirtualOOC(n_cells=n_cells)
    agent = ActorCritic(in_features=5, hidden=64)
    optim_ac = optim.Adam(agent.parameters(), lr=3e-4)

    # helper to compute returns
    def compute_returns(rewards, dones, next_value, gamma):
        R = next_value
        returns = []
        for r, d in zip(rewards[::-1], dones[::-1]):
            R = r + gamma * R * (1.0 - d)
            returns.insert(0, R)
        return returns

    for ep in range(episodes):
        # reset env
        env = VirtualOOC(n_cells=n_cells)
        # warmup history
        history = []
        for t in range(seq_len):
            snap = env.snapshot()
            obs = np.stack([snap['tel_len'], snap['damage'], snap['phag'], snap['nutrient'], snap['alive']], axis=1)
            history.append(obs)
            env.step(tel_pulses=np.zeros(env.n), nutrient_in=1.0)
        history = np.array(history)

        ep_rewards = []
        log_probs = []
        values = []
        entropies = []
        observations = []
        actions = []
        dones = []

        for t in range(200):
            # prepare per-cell samples
            # shape (n_cells, seq_len, feat)
            seq = torch.tensor(history[-seq_len:], dtype=torch.float32)  # (seq_len, n_cells, feat)
            seq = seq.transpose(0,1)  # (n_cells, seq_len, feat)
            mean, std, value = agent(seq)
            # sample actions from Normal(mean, std) and squash to [0,1]
            normal = torch.distributions.Normal(mean, std)
            raw_action = normal.rsample()
            action = torch.sigmoid(raw_action)
            # log prob with change-of-variable correction for sigmoid
            logp_raw = normal.log_prob(raw_action)
            # derivative of sigmoid = sigmoid*(1-sigmoid)
            log_det = - (torch.log(action) + torch.log(1.0 - action) + 1e-8)
            logp = (logp_raw - log_det).sum(dim=-1)
            entropy = normal.entropy().sum(dim=-1)

            tel_pulses = action.detach().numpy()
            # step env
            env.step(tel_pulses=tel_pulses, nutrient_in=1.0)
            snap = env.snapshot()

            # define reward: encourage repair (increase alive cells), penalize damage and overextending telomeres
            alive_bonus = snap['alive'].sum() / n_cells
            avg_damage = snap['damage'].mean()
            avg_tel = snap['tel_len'].mean()
            reward = alive_bonus - 0.3*avg_damage - 0.05*max(0.0, avg_tel - 2.5)
            # per-step scalar reward
            r = float(reward)

            # record
            ep_rewards.append(r)
            log_probs.append(logp)
            values.append(value)
            entropies.append(entropy)
            observations.append(seq)
            actions.append(action)
            dones.append(0.0 if snap['alive'].sum()>0 else 1.0)

            # append new obs to history
            obs = np.stack([snap['tel_len'], snap['damage'], snap['phag'], snap['nutrient'], snap['alive']], axis=1)
            history = np.concatenate([history, obs[np.newaxis,:,:]], axis=0)

            if all(c==0.0 for c in snap['alive']):
                break

        # compute returns and advantages
        with torch.no_grad():
            next_val = agent(torch.tensor(observations[-1], dtype=torch.float32))[2] if len(observations)>0 else torch.tensor(0.0)
        returns = compute_returns(ep_rewards, dones, float(next_val.mean()), gamma)
        returns = torch.tensor(returns, dtype=torch.float32)

        # flatten tensors for PPO update
        logp_tensor = torch.stack([lp.mean() if lp.dim()>0 else lp for lp in log_probs])
        value_tensor = torch.stack([v.mean() for v in values])
        advantage = returns - value_tensor.detach()

        # policy loss (clipped)
        ratio = torch.exp(logp_tensor - logp_tensor.detach())
        surr1 = ratio * advantage
        surr2 = torch.clamp(ratio, 1.0-clip_eps, 1.0+clip_eps) * advantage
        policy_loss = -torch.min(surr1, surr2).mean()
        value_loss = nn.MSELoss()(value_tensor, returns)
        entropy_loss = -torch.stack(entropies).mean()

        loss = policy_loss + 0.5*value_loss + 0.01*entropy_loss
        optim_ac.zero_grad()
        loss.backward()
        optim_ac.step()

        if ep % 10 == 0:
            print(f'RL Episode {ep} total_reward={sum(ep_rewards):.3f} steps={len(ep_rewards)} avg_reward={np.mean(ep_rewards):.4f}')

    print('RL run complete. Inspect agent behavior and tune reward or architecture.')

# Example: run a short RL training (disabled by default to avoid long runs)
if __name__ == '__main__':
    # Uncomment to run RL training
    # run_rl_agent(episodes=100, seq_len=8, n_cells=16)

    # ------------------------- Multi-Agent Extension -------------------------
    # This extension adds spatial coupling between neighboring cells and treats
    # the tissue as a multi-agent system. Each agent (cell) has its own policy
    # but policies share parameters (parameter-sharing). The environment now
    # includes local neighborhood aggregation (average neighbor telomere, damage,
    # phag activity) which is included in each agent's observation to allow
    # coordinated behavior like localized repair without global knowledge.

    def run_multi_agent(episodes=200, seq_len=8, n_cells=32, gamma=0.98, clip_eps=0.2):
        set_seed(42)
        env = VirtualOOC(n_cells=n_cells)

        # Agent architecture re-used but now observation includes neighbor aggregates
        class MultiAgentActorCritic(nn.Module):
            def __init__(self, in_features=8, hidden=128):
                super().__init__()
                self.base = nn.GRU(input_size=in_features, hidden_size=hidden, batch_first=True)
                self.policy = nn.Sequential(nn.Linear(hidden, 64), nn.ReLU(), nn.Linear(64, 1))
                self.value = nn.Sequential(nn.Linear(hidden, 64), nn.ReLU(), nn.Linear(64, 1))
                self.log_std = nn.Parameter(torch.tensor(-1.0))

            def forward(self, x):
                out, _ = self.base(x)
                h = out[:, -1, :]
                mean = self.policy(h)
                value = self.value(h)
                std = torch.exp(self.log_std)
                return mean.squeeze(-1), std, value.squeeze(-1)

        agent = MultiAgentActorCritic(in_features=8, hidden=128)
        optim_ac = optim.Adam(agent.parameters(), lr=2e-4)

        def neighbor_aggregate(snapshot):
            # compute per-cell neighbor averages (left and right; circular)
            tel = snapshot['tel_len']
            dmg = snapshot['damage']
            ph = snapshot['phag']
            n = len(tel)
            left = np.roll(tel, 1)
            right = np.roll(tel, -1)
            tel_neigh = 0.5*(left + right)
            left_d = np.roll(dmg, 1); right_d = np.roll(dmg, -1)
            dmg_neigh = 0.5*(left_d + right_d)
            left_p = np.roll(ph, 1); right_p = np.roll(ph, -1)
            ph_neigh = 0.5*(left_p + right_p)
            return tel_neigh, dmg_neigh, ph_neigh

        for ep in range(episodes):
            env = VirtualOOC(n_cells=n_cells)
            # warmup
            history = []
            for t in range(seq_len):
                snap = env.snapshot()
                tel_n, dmg_n, ph_n = neighbor_aggregate(snap)
                obs = np.stack([snap['tel_len'], snap['damage'], snap['phag'], snap['nutrient'], snap['alive'],
                                tel_n, dmg_n, ph_n], axis=1)
                history.append(obs)
                env.step(tel_pulses=np.zeros(env.n), nutrient_in=1.0)
            history = np.array(history)

            ep_rewards = []
            log_probs = []
            values = []
            entropies = []
            observations = []
            dones = []

            for t in range(300):
                seq = torch.tensor(history[-seq_len:], dtype=torch.float32)
                seq = seq.transpose(0,1)  # (n_cells, seq_len, feat)
                mean, std, value = agent(seq)
                normal = torch.distributions.Normal(mean, std)
                raw_action = normal.rsample()
                action = torch.sigmoid(raw_action)
                logp_raw = normal.log_prob(raw_action)
                log_det = - (torch.log(action) + torch.log(1.0 - action) + 1e-8)
                logp = (logp_raw - log_det).sum(dim=-1)
                entropy = normal.entropy().sum(dim=-1)

                tel_pulses = action.detach().numpy()
                # spatial smoothing of actions to encourage locality (small diffusion)
                tel_pulses = 0.8*tel_pulses + 0.1*np.roll(tel_pulses, 1) + 0.1*np.roll(tel_pulses, -1)
                tel_pulses = np.clip(tel_pulses, 0.0, 1.0)

                env.step(tel_pulses=tel_pulses, nutrient_in=1.0)
                snap = env.snapshot()

                # reward encourages local repair: per-cell reward combined into scalar
                alive_frac = snap['alive'].sum() / n_cells
                avg_damage = snap['damage'].mean()
                avg_tel = snap['tel_len'].mean()
                # additional locality bonus: lower variance of neighbor differences
                tel_var = np.mean((snap['tel_len'] - np.roll(snap['tel_len'],1))**2)
                reward = alive_frac - 0.35*avg_damage - 0.04*max(0.0, avg_tel - 2.5) - 0.1*tel_var
                r = float(reward)

                ep_rewards.append(r)
                log_probs.append(logp)
                values.append(value)
                entropies.append(entropy)
                observations.append(seq)
                dones.append(0.0 if snap['alive'].sum()>0 else 1.0)

                tel_n, dmg_n, ph_n = neighbor_aggregate(snap)
                obs = np.stack([snap['tel_len'], snap['damage'], snap['phag'], snap['nutrient'], snap['alive'],
                                tel_n, dmg_n, ph_n], axis=1)
                history = np.concatenate([history, obs[np.newaxis,:,:]], axis=0)

                if all(c==0.0 for c in snap['alive']):
                    break

            # simple policy/value update (one-shot) for brevity; replace with PPO minibatch loops
            with torch.no_grad():
                next_val = agent(torch.tensor(observations[-1], dtype=torch.float32))[2] if len(observations)>0 else torch.tensor(0.0)
            returns = []
            R = float(next_val.mean())
            for rew in ep_rewards[::-1]:
                R = rew + gamma * R
                returns.insert(0, R)
            returns = torch.tensor(returns, dtype=torch.float32)

            logp_tensor = torch.stack([lp.mean() if lp.dim()>0 else lp for lp in log_probs])
            value_tensor = torch.stack([v.mean() for v in values])
            advantage = returns - value_tensor.detach()

            ratio = torch.exp(logp_tensor - logp_tensor.detach())
            surr1 = ratio * advantage
            surr2 = torch.clamp(ratio, 1.0-clip_eps, 1.0+clip_eps) * advantage
            policy_loss = -torch.min(surr1, surr2).mean()
            value_loss = nn.MSELoss()(value_tensor, returns)
            entropy_loss = -torch.stack(entropies).mean()

            loss = policy_loss + 0.5*value_loss + 0.01*entropy_loss
            optim_ac.zero_grad()
            loss.backward()
            optim_ac.step()

            if ep % 10 == 0:
                print(f'Multi-Agent Episode {ep} total_reward={sum(ep_rewards):.3f} steps={len(ep_rewards)} avg_reward={np.mean(ep_rewards):.4f}')

        print('Multi-agent run complete. Tune neighbor observations, reward terms, and update scheme for stability.')

    # To run multi-agent training, uncomment below
    # run_multi_agent(episodes=100, seq_len=8, n_cells=32)
    pass
```

```python
"""
multi_agent_ooc_demo.py

Compact runnable multi-agent demo of a virtual organ-on-a-chip telomere controller.

- Simulation-only (abstract dynamics).
- Parameter-sharing multi-agent actor-critic (GRU-based).
- Produces three separate matplotlib figures:
    1) Total episode reward per episode
    2) Final alive fraction per episode
    3) Final telomere std (stability) per episode

Usage:
    python multi_agent_ooc_demo.py

Tweak hyperparams at the top of `main()`.
"""

import math, random, numpy as np, matplotlib.pyplot as plt
import torch, torch.nn as nn, torch.optim as optim
import pandas as pd
from copy import deepcopy

# ---------------------------
# Abstract Virtual Cell + OOC
# ---------------------------
class VirtualCell:
    def __init__(self, tel_len=1.0):
        self.tel_len = tel_len
        self.age = 0
        self.alive = True
        self.damage = 0.0
        self.phag_activity = 0.0

    def step(self, telomerase_pulse, microenv_signal, damage_input=0.0, phag_trigger=0.0):
        if not self.alive:
            return
        # damage and phagocytosis (abstract)
        self.damage += max(0.0, damage_input - self.phag_activity * 0.5)
        # telomere growth from telomerase, shrink on division-like signal, trimming if too long
        growth = float(telomerase_pulse) * 0.1 * (1.0 - math.tanh(self.tel_len - 2.0))
        shrink = 0.02 * float(microenv_signal > 0.5)
        trimming = 0.05 * max(0.0, self.tel_len - 3.0)
        self.tel_len = max(0.1, self.tel_len + growth - shrink - trimming)
        # phagocytosis dynamics (abstract)
        self.phag_activity = max(0.0, min(1.0, self.phag_activity + 0.1 * phag_trigger - 0.02 * self.phag_activity))
        # failure threshold
        if self.damage > 2.5:
            self.alive = False
        self.age += 1

class VirtualOOC:
    def __init__(self, n_cells=24, seed=None):
        if seed is not None:
            random.seed(seed); np.random.seed(seed)
        self.n = n_cells
        self.cells = [VirtualCell(tel_len=1.0 + 0.12 * random.random()) for _ in range(n_cells)]
        self.nutrient = np.ones(n_cells)
        self.flow = np.zeros(n_cells)
        self.time = 0

    def step(self, tel_pulses, nutrient_in=1.0, pathogen_pulse=0.0, injury_mask=None):
        # simple flow model
        self.flow = np.roll(self.flow, 1)
        self.flow[0] = nutrient_in
        self.nutrient = 0.9 * self.nutrient + 0.1 * self.flow
        for i, cell in enumerate(self.cells):
            damage_input = float(pathogen_pulse * (0.5 + 0.5 * random.random()))
            ph = float(pathogen_pulse > 0.1)
            if injury_mask is not None and injury_mask[i]:
                damage_input += 0.5
            microenv_signal = self.nutrient[i]
            cell.step(telomerase_pulse=float(tel_pulses[i]), microenv_signal=microenv_signal,
                      damage_input=damage_input, phag_trigger=ph)
        self.time += 1

    def snapshot(self):
        tel = np.array([c.tel_len for c in self.cells])
        alive = np.array([1.0 if c.alive else 0.0 for c in self.cells])
        damage = np.array([c.damage for c in self.cells])
        phag = np.array([c.phag_activity for c in self.cells])
        return {"tel_len": tel, "alive": alive, "damage": damage, "phag": phag, "nutrient": self.nutrient.copy()}

# ---------------------------
# Multi-agent Actor-Critic
# ---------------------------
class MultiAgentActorCritic(nn.Module):
    def __init__(self, in_features=8, hidden=96):
        super().__init__()
        self.base = nn.GRU(input_size=in_features, hidden_size=hidden, batch_first=True)
        self.policy = nn.Sequential(nn.Linear(hidden, 64), nn.ReLU(), nn.Linear(64, 1))
        self.value = nn.Sequential(nn.Linear(hidden, 64), nn.ReLU(), nn.Linear(64, 1))
        self.log_std = nn.Parameter(torch.tensor(-1.0))

    def forward(self, x):
        # x: (n_agents, seq_len, in_features)
        out, _ = self.base(x)
        h = out[:, -1, :]
        mean = self.policy(h)          # (n_agents, 1)
        value = self.value(h)          # (n_agents, 1)
        std = torch.exp(self.log_std)
        return mean.squeeze(-1), std, value.squeeze(-1)

# ---------------------------
# Helpers (neighbor aggregates)
# ---------------------------
def neighbor_aggregate(snapshot):
    tel = snapshot['tel_len']; dmg = snapshot['damage']; ph = snapshot['phag']
    left = np.roll(tel, 1); right = np.roll(tel, -1)
    tel_neigh = 0.5 * (left + right)
    left_d = np.roll(dmg, 1); right_d = np.roll(dmg, -1)
    dmg_neigh = 0.5 * (left_d + right_d)
    left_p = np.roll(ph, 1); right_p = np.roll(ph, -1)
    ph_neigh = 0.5 * (left_p + right_p)
    return tel_neigh, dmg_neigh, ph_neigh

# ---------------------------
# Training loop (trimmed PPO-style minibatch single-update)
# ---------------------------
def run_multi_agent_demo(episodes=80, seq_len=8, n_cells=24, seed=0,
                         gamma=0.98, clip_eps=0.2, max_steps=200, lr=2e-4, verbose_every=10):
    # deterministic seeds for reproducibility
    random.seed(seed); np.random.seed(seed); torch.manual_seed(seed)

    agent = MultiAgentActorCritic(in_features=8, hidden=96)
    optim_ac = optim.Adam(agent.parameters(), lr=lr)

    ep_rewards_hist = []
    alive_frac_hist = []
    tel_std_hist = []

    for ep in range(episodes):
        env = VirtualOOC(n_cells=n_cells, seed=seed+ep)
        # warmup sequence
        history = []
        for t in range(seq_len):
            snap = env.snapshot()
            tel_n, dmg_n, ph_n = neighbor_aggregate(snap)
            obs = np.stack([snap['tel_len'], snap['damage'], snap['phag'], snap['nutrient'], snap['alive'],
                            tel_n, dmg_n, ph_n], axis=1)  # (n_cells, features)
            history.append(obs)
            env.step(tel_pulses=np.zeros(env.n), nutrient_in=1.0)
        history = np.array(history)  # (seq_len, n_cells, features)

        ep_rewards = []
        log_probs = []
        values = []
        entropies = []
        observations = []

        for step in range(max_steps):
            seq = torch.tensor(history[-seq_len:], dtype=torch.float32).transpose(0, 1)  # (n_cells, seq_len, feat)
            mean, std, value = agent(seq)   # mean: (n_cells,), std: scalar, value: (n_cells,)
            normal = torch.distributions.Normal(mean, std)
            raw_action = normal.rsample()   # reparameterized sample
            action = torch.sigmoid(raw_action)
            logp_raw = normal.log_prob(raw_action)
            # change-of-variable correction for sigmoid
            log_det = - (torch.log(action + 1e-8) + torch.log(1.0 - action + 1e-8))
            logp = (logp_raw - log_det).sum(dim=-1)
            entropy = normal.entropy().sum(dim=-1)

            tel_pulses = action.detach().numpy()
            # spatial smoothing (encourage local coordination)
            tel_pulses = 0.8 * tel_pulses + 0.1 * np.roll(tel_pulses, 1) + 0.1 * np.roll(tel_pulses, -1)
            tel_pulses = np.clip(tel_pulses, 0.0, 1.0)

            env.step(tel_pulses=tel_pulses, nutrient_in=1.0)
            snap = env.snapshot()

            alive_frac = snap['alive'].sum() / n_cells
            avg_damage = float(snap['damage'].mean())
            avg_tel = float(snap['tel_len'].mean())
            tel_var = float(np.mean((snap['tel_len'] - np.roll(snap['tel_len'], 1)) ** 2))

            # reward: keep cells alive, minimize damage and telomere dispersion/overlength
            reward = alive_frac - 0.35 * avg_damage - 0.04 * max(0.0, avg_tel - 2.5) - 0.1 * tel_var
            r = float(reward)

            ep_rewards.append(r)
            log_probs.append(logp)
            values.append(value)
            entropies.append(entropy)
            observations.append(seq)

            # append next observation to history
            tel_n, dmg_n, ph_n = neighbor_aggregate(snap)
            obs = np.stack([snap['tel_len'], snap['damage'], snap['phag'], snap['nutrient'], snap['alive'],
                            tel_n, dmg_n, ph_n], axis=1)
            history = np.concatenate([history, obs[np.newaxis, :, :]], axis=0)

            # termination if all dead
            if all(c == 0.0 for c in snap['alive']):
                break

        # Compute simple returns (discounted) and do one PPO-like update (single minibatch)
        with torch.no_grad():
            next_val = agent(torch.tensor(observations[-1], dtype=torch.float32))[2] if len(observations) > 0 else torch.tensor(0.0)

        # discounted returns
        R = float(next_val.mean()) if isinstance(next_val, torch.Tensor) else 0.0
        returns = []
        for rew in ep_rewards[::-1]:
            R = rew + gamma * R
            returns.insert(0, R)
        returns = torch.tensor(returns, dtype=torch.float32)

        if len(log_probs) == 0:
            # nothing to update
            ep_rewards_hist.append(float(0.0))
            alive_frac_hist.append(float(alive_frac))
            tel_std_hist.append(float(np.std(snap['tel_len'])))
            continue

        logp_tensor = torch.stack([lp.mean() if lp.dim() > 0 else lp for lp in log_probs])
        value_tensor = torch.stack([v.mean() for v in values])
        advantage = returns - value_tensor.detach()

        ratio = torch.exp(logp_tensor - logp_tensor.detach())
        surr1 = ratio * advantage
        surr2 = torch.clamp(ratio, 1.0 - clip_eps, 1.0 + clip_eps) * advantage
        policy_loss = -torch.min(surr1, surr2).mean()
        value_loss = nn.MSELoss()(value_tensor, returns)
        entropy_loss = -torch.stack(entropies).mean()

        loss = policy_loss + 0.5 * value_loss + 0.01 * entropy_loss

        optim_ac.zero_grad()
        loss.backward()
        optim_ac.step()

        ep_rewards_hist.append(float(sum(ep_rewards)))
        alive_frac_hist.append(float(alive_frac))
        tel_std_hist.append(float(np.std(snap['tel_len'])))

        if (ep + 1) % verbose_every == 0:
            print(f"Episode {ep+1}/{episodes}  reward={ep_rewards_hist[-1]:.3f}  alive_frac={alive_frac_hist[-1]:.3f}  tel_std={tel_std_hist[-1]:.3f}")

    # return metrics and a tiny dataframe summary for convenience
    df = pd.DataFrame({
        "episode": list(range(1, len(ep_rewards_hist) + 1)),
        "total_reward": ep_rewards_hist,
        "alive_frac": alive_frac_hist,
        "tel_std": tel_std_hist
    })
    return ep_rewards_hist, alive_frac_hist, tel_std_hist, df

# ---------------------------
# Plotting helpers
# ---------------------------
def plot_metrics(ep_rewards_hist, alive_frac_hist, tel_std_hist, save_prefix=None):
    episodes = np.arange(1, len(ep_rewards_hist) + 1)

    plt.figure()
    plt.plot(episodes, ep_rewards_hist)
    plt.xlabel("Episode")
    plt.ylabel("Total Episode Reward")
    plt.title("Multi-Agent: Total Episode Reward")
    plt.grid(True)
    if save_prefix: plt.savefig(f"{save_prefix}_reward.png", dpi=150)
    plt.show()

    plt.figure()
    plt.plot(episodes, alive_frac_hist)
    plt.xlabel("Episode")
    plt.ylabel("Final Alive Fraction (episode end)")
    plt.title("Multi-Agent: Alive Fraction per Episode")
    plt.grid(True)
    if save_prefix: plt.savefig(f"{save_prefix}_alive.png", dpi=150)
    plt.show()

    plt.figure()
    plt.plot(episodes, tel_std_hist)
    plt.xlabel("Episode")
    plt.ylabel("Std of Telomere Length (final)")
    plt.title("Multi-Agent: Telomere Std per Episode")
    plt.grid(True)
    if save_prefix: plt.savefig(f"{save_prefix}_telstd.png", dpi=150)
    plt.show()

# ---------------------------
# Main: run & plot
# ---------------------------
def main():
    # change these to tune run length / fidelity
    EPISODES = 80          # medium (set to e.g. 20 for faster runs, 200+ for heavier)
    SEQ_LEN = 8
    N_CELLS = 24
    SAVE_PREFIX = "multi_agent_demo"  # set None to avoid saving files

    ep_rewards_hist, alive_frac_hist, tel_std_hist, df = run_multi_agent_demo(
        episodes=EPISODES, seq_len=SEQ_LEN, n_cells=N_CELLS, seed=1, gamma=0.98, clip_eps=0.2, max_steps=200, lr=2e-4, verbose_every=10
    )

    plot_metrics(ep_rewards_hist, alive_frac_hist, tel_std_hist, save_prefix=SAVE_PREFIX)

    print("\nLast 8 episodes (summary):")
    print(df.tail(8).to_string(index=False))
    if SAVE_PREFIX:
        df.to_csv(f"{SAVE_PREFIX}_metrics.csv", index=False)
        print(f"\nSaved metrics CSV as {SAVE_PREFIX}_metrics.csv and PNGs with prefix {SAVE_PREFIX}_*.png")

if __name__ == "__main__":
    main()
```

```python
import math, random, os
import numpy as np
import matplotlib.pyplot as plt
import torch, torch.nn as nn, torch.optim as optim
import pandas as pd

# ---------------------------
# Config / Toggle features
# ---------------------------
USE_PPO = True           # if False uses single-update simplified loop
SAVE_HEATMAPS = True     # save per-episode telomere heatmap PNGs
OUTDIR = "ppo_heatmaps_out"
os.makedirs(OUTDIR, exist_ok=True)

EPISODES = 80            # medium run; lower to 20 for quick tests
SEQ_LEN = 8
N_CELLS = 24
MAX_STEPS = 200
SEED = 1
LR = 2e-4

# PPO hyperparams
PPO_EPOCHS = 4
MINIBATCH_SIZE = 64      # across time-steps (flattened across time)
GAMMA = 0.98
GAE_LAMBDA = 0.95
CLIP_EPS = 0.2
ENTROPY_COEF = 0.01
VALUE_COEF = 0.5

# ---------------------------
# Environment (same as before)
# ---------------------------
class VirtualCell:
    def __init__(self, tel_len=1.0):
        self.tel_len = tel_len
        self.age = 0
        self.alive = True
        self.damage = 0.0
        self.phag_activity = 0.0

    def step(self, telomerase_pulse, microenv_signal, damage_input=0.0, phag_trigger=0.0):
        if not self.alive:
            return
        self.damage += max(0.0, damage_input - self.phag_activity * 0.5)
        growth = float(telomerase_pulse) * 0.1 * (1.0 - math.tanh(self.tel_len - 2.0))
        shrink = 0.02 * float(microenv_signal > 0.5)
        trimming = 0.05 * max(0.0, self.tel_len - 3.0)
        self.tel_len = max(0.1, self.tel_len + growth - shrink - trimming)
        self.phag_activity = max(0.0, min(1.0, self.phag_activity + 0.1 * phag_trigger - 0.02 * self.phag_activity))
        if self.damage > 2.5:
            self.alive = False
        self.age += 1

class VirtualOOC:
    def __init__(self, n_cells=24, seed=None):
        if seed is not None:
            random.seed(seed); np.random.seed(seed)
        self.n = n_cells
        self.cells = [VirtualCell(tel_len=1.0 + 0.12 * random.random()) for _ in range(n_cells)]
        self.nutrient = np.ones(n_cells)
        self.flow = np.zeros(n_cells)
        self.time = 0

    def step(self, tel_pulses, nutrient_in=1.0, pathogen_pulse=0.0, injury_mask=None):
        self.flow = np.roll(self.flow, 1)
        self.flow[0] = nutrient_in
        self.nutrient = 0.9 * self.nutrient + 0.1 * self.flow
        for i, cell in enumerate(self.cells):
            damage_input = float(pathogen_pulse * (0.5 + 0.5 * random.random()))
            ph = float(pathogen_pulse > 0.1)
            if injury_mask is not None and injury_mask[i]:
                damage_input += 0.5
            microenv_signal = self.nutrient[i]
            cell.step(telomerase_pulse=float(tel_pulses[i]), microenv_signal=microenv_signal,
                      damage_input=damage_input, phag_trigger=ph)
        self.time += 1

    def snapshot(self):
        tel = np.array([c.tel_len for c in self.cells])
        alive = np.array([1.0 if c.alive else 0.0 for c in self.cells])
        damage = np.array([c.damage for c in self.cells])
        phag = np.array([c.phag_activity for c in self.cells])
        return {"tel_len": tel, "alive": alive, "damage": damage, "phag": phag, "nutrient": self.nutrient.copy()}

# ---------------------------
# Agent
# ---------------------------
class MultiAgentActorCritic(nn.Module):
    def __init__(self, in_features=8, hidden=96):
        super().__init__()
        self.base = nn.GRU(input_size=in_features, hidden_size=hidden, batch_first=True)
        self.policy = nn.Sequential(nn.Linear(hidden, 64), nn.ReLU(), nn.Linear(64, 1))
        self.value = nn.Sequential(nn.Linear(hidden, 64), nn.ReLU(), nn.Linear(64, 1))
        self.log_std = nn.Parameter(torch.tensor(-1.0))

    def forward(self, x):
        out, _ = self.base(x)
        h = out[:, -1, :]
        mean = self.policy(h)
        value = self.value(h)
        std = torch.exp(self.log_std)
        return mean.squeeze(-1), std, value.squeeze(-1)

# ---------------------------
# Helpers
# ---------------------------
def neighbor_aggregate(snapshot):
    tel = snapshot['tel_len']; dmg = snapshot['damage']; ph = snapshot['phag']
    left = np.roll(tel, 1); right = np.roll(tel, -1)
    tel_neigh = 0.5 * (left + right)
    left_d = np.roll(dmg, 1); right_d = np.roll(dmg, -1)
    dmg_neigh = 0.5 * (left_d + right_d)
    left_p = np.roll(ph, 1); right_p = np.roll(ph, -1)
    ph_neigh = 0.5 * (left_p + right_p)
    return tel_neigh, dmg_neigh, ph_neigh

def save_tel_heatmap(tel_history, ep_idx):
    """
    tel_history: list of arrays shape (n_cells,) over time-steps in episode
    Saves a heatmap image with time on y-axis and cell index on x-axis.
    """
    arr = np.vstack(tel_history)  # (timesteps, n_cells)
    fig, ax = plt.subplots(figsize=(6, 4))
    im = ax.imshow(arr, aspect='auto', origin='lower')
    ax.set_xlabel("Cell index")
    ax.set_ylabel("Timestep")
    ax.set_title(f"Episode {ep_idx} telomere lengths (time x cell)")
    plt.colorbar(im, ax=ax, label='tel len')
    fname = os.path.join(OUTDIR, f"tel_heatmap_ep{ep_idx:03d}.png")
    plt.tight_layout()
    fig.savefig(fname, dpi=150)
    plt.close(fig)

# ---------------------------
# GAE + PPO utilities
# ---------------------------
def compute_gae(rewards, values, dones, last_value, gamma=GAMMA, lam=GAE_LAMBDA):
    """
    rewards: list length T (scalars)
    values: list length T (scalars)
    dones: list length T (0/1)
    last_value: scalar bootstrap value
    returns: list of discounted returns (length T)
    advantages: list of advantages (length T)
    """
    T = len(rewards)
    adv = np.zeros(T, dtype=np.float32)
    lastgaelam = 0.0
    for t in reversed(range(T)):
        nonterminal = 1.0 - dones[t]
        nextvalue = last_value if t == T - 1 else values[t + 1]
        delta = rewards[t] + gamma * nextvalue * nonterminal - values[t]
        adv[t] = lastgaelam = delta + gamma * lam * nonterminal * lastgaelam
    returns = adv + np.array(values, dtype=np.float32)
    return returns.tolist(), adv.tolist()

# ---------------------------
# Main training loop (PPO-capable)
# ---------------------------
def run_training(episodes=EPISODES, seq_len=SEQ_LEN, n_cells=N_CELLS):
    random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
    agent = MultiAgentActorCritic(in_features=8, hidden=96)
    optim_ac = optim.Adam(agent.parameters(), lr=LR)

    ep_rewards_hist, alive_frac_hist, tel_std_hist = [], [], []
    df_rows = []

    for ep in range(episodes):
        env = VirtualOOC(n_cells=n_cells, seed=SEED + ep)
        history = []
        for t in range(seq_len):
            snap = env.snapshot()
            tel_n, dmg_n, ph_n = neighbor_aggregate(snap)
            obs = np.stack([snap['tel_len'], snap['damage'], snap['phag'], snap['nutrient'], snap['alive'],
                            tel_n, dmg_n, ph_n], axis=1)
            history.append(obs)
            env.step(tel_pulses=np.zeros(env.n), nutrient_in=1.0)
        history = np.array(history)

        # buffers (per time-step, storing PyTorch tensors directly)
        tel_history_for_heatmap = [] # for plotting
        observations_list = [] # list of (n_cells, seq_len, feat) tensors
        raw_actions_list = [] # list of (n_cells,) tensors
        log_probs_behav_list = [] # list of (n_cells,) tensors (log_prob of action taken by behavior policy)
        values_behav_list = [] # list of (n_cells,) tensors (value from behavior policy)
        rewards_list = [] # list of scalar rewards
        dones_list = [] # list of scalar dones

        for step in range(MAX_STEPS):
            seq = torch.tensor(history[-seq_len:], dtype=torch.float32).transpose(0, 1)
            observations_list.append(seq.detach()) # Store observations for later re-evaluation

            mean, std, value = agent(seq)
            normal = torch.distributions.Normal(mean, std)
            raw_action = normal.rsample() # Sample raw action
            action = torch.sigmoid(raw_action)
            logp_raw = normal.log_prob(raw_action)
            # change-of-variable correction for sigmoid
            log_det = - (torch.log(action + 1e-8) + torch.log(1.0 - action + 1e-8))
            logp = (logp_raw - log_det) # (n_cells,) tensor

            tel_pulses = action.detach().numpy()
            # spatial smoothing (encourage local coordination)
            tel_pulses = 0.8 * tel_pulses + 0.1 * np.roll(tel_pulses, 1) + 0.1 * np.roll(tel_pulses, -1)
            tel_pulses = np.clip(tel_pulses, 0.0, 1.0)

            env.step(tel_pulses=tel_pulses, nutrient_in=1.0)
            snap = env.snapshot()

            alive_frac = snap['alive'].sum() / n_cells
            avg_damage = float(snap['damage'].mean())
            avg_tel = float(snap['tel_len'].mean())
            tel_var = float(np.mean((snap['tel_len'] - np.roll(snap['tel_len'], 1)) ** 2))
            reward = alive_frac - 0.35 * avg_damage - 0.04 * max(0.0, avg_tel - 2.5) - 0.1 * tel_var
            r = float(reward)

            # store buffers (as PyTorch tensors)
            raw_actions_list.append(raw_action.detach()) # Detach for use as 'old_action'
            log_probs_behav_list.append(logp.detach()) # Detach for use as 'old_logp'
            values_behav_list.append(value.detach()) # Detach for use as 'old_value'
            rewards_list.append(r)
            dones_list.append(0.0 if snap['alive'].sum() > 0 else 1.0)

            tel_history_for_heatmap.append(snap['tel_len'].copy())

            tel_n, dmg_n, ph_n = neighbor_aggregate(snap)
            obs = np.stack([snap['tel_len'], snap['damage'], snap['phag'], snap['nutrient'], snap['alive'],
                            tel_n, dmg_n, ph_n], axis=1)
            history = np.concatenate([history, obs[np.newaxis, :, :]], axis=0)

            if all(c == 0.0 for c in snap['alive']):
                break

        # optional heatmap save
        if SAVE_HEATMAPS:
            save_tel_heatmap(tel_history_for_heatmap, ep + 1)

        # Handle empty trajectory (e.g., all cells die immediately)
        T_traj = len(rewards_list)
        if T_traj == 0:
            ep_rewards_hist.append(0.0)
            alive_frac_hist.append(0.0)
            tel_std_hist.append(0.0)
            df_rows.append({"episode": ep + 1, "total_reward": 0.0, "alive_frac": 0.0, "tel_std": 0.0})
            continue

        # Stack collected tensors and flatten for PPO processing
        all_obs = torch.cat(observations_list, dim=0) # (T*n_cells, seq_len, feat)
        all_raw_actions = torch.cat(raw_actions_list, dim=0) # (T*n_cells,)
        all_old_logps = torch.cat(log_probs_behav_list, dim=0) # (T*n_cells,)
        all_old_values = torch.cat(values_behav_list, dim=0) # (T*n_cells,)

        # Expand rewards and dones to match (T*n_cells) for GAE
        all_rewards_flat = torch.tensor(rewards_list, dtype=torch.float32).unsqueeze(1).repeat(1, n_cells).view(-1)
        all_dones_flat = torch.tensor(dones_list, dtype=torch.float32).unsqueeze(1).repeat(1, n_cells).view(-1)

        # Compute GAE
        # last_value for GAE should be the value prediction of the *current* policy on the *last state*.
        # Using last stored value (from behavior policy) detached as an approximation for the demo.
        last_value_for_gae = all_old_values[-n_cells:].mean().item() if len(all_old_values) > 0 else 0.0
        returns_flat, adv_flat = compute_gae(all_rewards_flat.tolist(), all_old_values.tolist(), all_dones_flat.tolist(), last_value_for_gae, gamma=GAMMA, lam=GAE_LAMBDA)
        returns_flat = torch.tensor(returns_flat, dtype=torch.float32)
        adv_flat = torch.tensor(adv_flat, dtype=torch.float32)

        # Normalize advantages
        adv_flat = (adv_flat - adv_flat.mean()) / (adv_flat.std() + 1e-8)

        # PPO update: multiple epochs, minibatches
        if USE_PPO:
            N_total_samples = all_obs.shape[0]
            idxs = np.arange(N_total_samples)
            for _epoch in range(PPO_EPOCHS):
                np.random.shuffle(idxs)
                for start in range(0, N_total_samples, MINIBATCH_SIZE):
                    mb_idx = idxs[start:start + MINIBATCH_SIZE]
                    if len(mb_idx) == 0:
                        continue

                    mb_obs = all_obs[mb_idx] # (minibatch_size, seq_len, feat)
                    mb_raw_actions = all_raw_actions[mb_idx]
                    mb_old_logps = all_old_logps[mb_idx]
                    mb_returns = returns_flat[mb_idx]
                    mb_advantages = adv_flat[mb_idx]

                    # Recompute policy and value for current minibatch using the current agent
                    new_mean, new_std, new_value = agent(mb_obs) # These will have requires_grad=True

                    # Calculate new_log_prob for the *taken actions* under the *current* policy
                    new_normal = torch.distributions.Normal(new_mean, new_std)
                    new_log_prob_raw = new_normal.log_prob(mb_raw_actions)
                    new_action = torch.sigmoid(mb_raw_actions)
                    new_log_det = - (torch.log(new_action + 1e-8) + torch.log(1.0 - new_action + 1e-8))
                    new_log_prob = (new_log_prob_raw - new_log_det) # (minibatch_size,)

                    # PPO Policy Loss (clipped objective)
                    ratio = torch.exp(new_log_prob - mb_old_logps) # mb_old_logps is detached
                    surr1 = ratio * mb_advantages
                    surr2 = torch.clamp(ratio, 1.0 - CLIP_EPS, 1.0 + CLIP_EPS) * mb_advantages
                    policy_loss = -torch.min(surr1, surr2).mean()

                    # PPO Value Loss
                    value_loss = VALUE_COEF * nn.MSELoss()(new_value, mb_returns) # new_value requires_grad=True

                    # Entropy Loss
                    entropy_loss = -ENTROPY_COEF * new_normal.entropy().mean() # new_normal.entropy() requires_grad=True

                    loss = policy_loss + value_loss + entropy_loss

                    optim_ac.zero_grad()
                    loss.backward()
                    optim_ac.step()
        else:
            # Simplified single update (if PPO is disabled)
            # Re-run agent on all_obs_flat to get new_mean, new_std, new_value that require grad
            new_mean_all, new_std_all, new_value_all = agent(all_obs_flat)

            new_normal_all = torch.distributions.Normal(new_mean_all, new_std_all)
            new_log_prob_raw_all = new_normal_all.log_prob(all_raw_actions)
            new_action_all = torch.sigmoid(all_raw_actions)
            new_log_det_all = - (torch.log(new_action_all + 1e-8) + torch.log(1.0 - new_action_all + 1e-8))
            new_log_prob_all = (new_log_prob_raw_all - new_log_det_all)

            ratio = torch.exp(new_log_prob_all - all_old_logps)
            surr1 = ratio * adv_flat
            surr2 = torch.clamp(ratio, 1.0 - CLIP_EPS, 1.0 + CLIP_EPS) * adv_flat
            policy_loss = -torch.min(surr1, surr2).mean()
            value_loss = VALUE_COEF * nn.MSELoss()(new_value_all, returns_flat)
            entropy_loss = -ENTROPY_COEF * new_normal_all.entropy().mean()

            loss = policy_loss + value_loss + entropy_loss

            optim_ac.zero_grad()
            loss.backward()
            optim_ac.step()

        ep_rewards_hist.append(float(sum(rewards_list)))
        alive_frac_hist.append(float(alive_frac))
        tel_std_hist.append(float(np.std(snap['tel_len'])))
        df_rows.append({"episode": ep + 1, "total_reward": ep_rewards_hist[-1], "alive_frac": alive_frac_hist[-1], "tel_std": tel_std_hist[-1]})

        if (ep + 1) % 10 == 0:
            print(f"Ep {ep+1}/{episodes} reward={ep_rewards_hist[-1]:.3f} alive_frac={alive_frac_hist[-1]:.3f} tel_std={tel_std_hist[-1]:.3f}")

    df = pd.DataFrame(df_rows)
    return ep_rewards_hist, alive_frac_hist, tel_std_hist, df

# ---------------------------
# Plot helpers
# ---------------------------
def plot_and_save(ep_rewards_hist, alive_frac_hist, tel_std_hist, prefix="ppo_demo"):
    eps = np.arange(1, len(ep_rewards_hist) + 1)
    plt.figure(); plt.plot(eps, ep_rewards_hist); plt.xlabel("Episode"); plt.ylabel("Total Reward"); plt.title("Total Episode Reward"); plt.grid(True); plt.savefig(f"{prefix}_reward.png", dpi=150); plt.show()
    plt.figure(); plt.plot(eps, alive_frac_hist); plt.xlabel("Episode"); plt.ylabel("Alive Fraction"); plt.title("Alive Fraction"); plt.grid(True); plt.savefig(f"{prefix}_alive.png", dpi=150); plt.show()
    plt.figure(); plt.plot(eps, tel_std_hist); plt.xlabel("Episode"); plt.ylabel("Telomere Std"); plt.title("Telomere Std"); plt.grid(True); plt.savefig(f"{prefix}_telstd.png", dpi=150); plt.show()

# ---------------------------
# Run
# ---------------------------
if __name__ == "__main__":
    rewards, alive, telstd, df_metrics = run_training(episodes=EPISODES, seq_len=SEQ_LEN, n_cells=N_CELLS)
    plot_and_save(rewards, alive, telstd, prefix=os.path.join(OUTDIR, "ppo_demo"))
    print("\nLast 8 episodes summary:")
    print(df_metrics.tail(8).to_string(index=False))
    df_metrics.to_csv(os.path.join(OUTDIR, "ppo_metrics.csv"), index=False)
    print(f"\nSaved metrics and heatmaps to {OUTDIR}/")
```
