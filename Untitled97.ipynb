```python
# Simulacin conceptual (Colab-ready)  Modelo seguro y abstracto de dinmica de telmeros
# Notas importantes:
# - Este cdigo es **puramente computacional** y **no** contiene instrucciones para experimentos biolgicos reales.
# - Objetivo: modelar abstractamente la longitud de telmeros, dao celular y un controlador adaptativo (PID-like)
#   que regula la actividad de 'telomerasa' para mantener una longitud objetivo.
# - El modelo es simple y diseado para exploracin y enseanza, no para uso clnico.
#
# Ejecuta este bloque en Colab/Jupyter para ver grficos y una tabla resumen.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import exp
from dataclasses import dataclass
np.random.seed(42)
@dataclass
class CellState:
    telomere: float
    damage: float
    telomerase_activity: float  # 0..1 internal parameter (abstract)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def simulate(num_cells=50, divisions=800,
             base_shorten=0.03,  # telomere loss per division (abstract units)
             noise_std=0.005,
             max_elong_per_activity=0.12, # max telomere elongation when activity=1
             damage_accum_per_div=0.01,
             phagocytosis_rate=0.008, # removal of damage per timestep
             setpoint=1.0,
             pid_params=(0.8, 0.03, 0.15),
             activity_safety_cap=0.9):
    """
    Simula mltiples clulas. El controlador PID ajusta actividad de 'telomerasa'
    para mantener telomere ~ setpoint.
    """
    Kp, Ki, Kd = pid_params
    # Inicializacin
    cells = [CellState(telomere=np.random.normal(setpoint, 0.05),
                       damage=np.random.uniform(0,0.02),
                       telomerase_activity=0.05) for _ in range(num_cells)]
    # Historial para graficar
    history = {
        "step": [], "cell": [], "telomere": [], "damage": [], "activity": [], "risk_metric": []
    }
    # PID states per cell
    integral = np.zeros(num_cells)
    prev_error = np.zeros(num_cells)
    for t in range(divisions):
        for i, c in enumerate(cells):
            # Measurement (with small noise)
            measured = c.telomere + np.random.normal(0, noise_std)
            error = setpoint - measured
            integral[i] += error
            derivative = error - prev_error[i]
            prev_error[i] = error
            # PID-like output
            control = Kp*error + Ki*integral[i] + Kd*derivative
            # map control to activity 0..activity_safety_cap with logistic smoothness
            raw_activity = sigmoid(control) * activity_safety_cap
            # apply some inertia / limited rate of change
            c.telomerase_activity += 0.2 * (raw_activity - c.telomerase_activity)
            # internal telomere dynamics per division
            elong = c.telomerase_activity * max_elong_per_activity
            loss = base_shorten + np.random.normal(0, noise_std)
            c.telomere += (elong - loss)
            # Telomere cannot be negative (abstract)
            if c.telomere < 0.05:
                c.telomere = 0.05
            # Damage accumulation & phagocytosis
            c.damage += damage_accum_per_div * (1 - c.telomerase_activity*0.2)  # better telomerase slightly reduces damage accumulation hypothetically
            # Phagocytosis removes some damage; we model it as a function of phagocytosis_rate and 'activity' (not realistic)
            c.damage -= phagocytosis_rate * (0.5 + 0.5*c.telomerase_activity)
            if c.damage < 0:
                c.damage = 0.0
            # Risk metric (abstract)  non-actionable; just a normalized score
            risk = sigmoid((c.damage - 0.25) * 8.0)  # maps higher damage -> risk closer to 1
            # Save snapshot
            history["step"].append(t)
            history["cell"].append(i)
            history["telomere"].append(c.telomere)
            history["damage"].append(c.damage)
            history["activity"].append(c.telomerase_activity)
            history["risk_metric"].append(risk)
    df = pd.DataFrame(history)
    return df
# Ejecutar simulacin
df = simulate(num_cells=60, divisions=1000)
# Crear resmenes y grficas
# Promedios por paso
avg = df.groupby("step").mean().reset_index()
# Mostrar una tabla con las ltimas condiciones promedio
last_row = avg.tail(1).copy()
last_row = last_row.rename(columns={"telomere":"avg_telomere","damage":"avg_damage","activity":"avg_activity","risk_metric":"avg_risk"})
# Mostrar tabla a usuario con caas helper (si est disponible)
try:
    import caas_jupyter_tools as cjt
    cjt.display_dataframe_to_user("Resumen final (promedio por paso)", last_row)
except Exception:
    # si no est disponible, imprimimos la tabla
    display(last_row)
# Graficar: telmeros promedio, dao promedio y actividad promedio
plt.figure(figsize=(10,4))
plt.plot(avg["step"], avg["telomere"])
plt.title("Longitud promedio de telmeros vs paso (simulacin abstracta)")
plt.xlabel("Divisin / paso de tiempo")
plt.ylabel("Telmero (unidades abstractas)")
plt.grid(True)
plt.show()
plt.figure(figsize=(10,4))
plt.plot(avg["step"], avg["damage"])
plt.title("Dao promedio por clula vs paso (simulacin abstracta)")
plt.xlabel("Divisin / paso de tiempo")
plt.ylabel("Dao (unidades abstractas)")
plt.grid(True)
plt.show()
plt.figure(figsize=(10,4))
plt.plot(avg["step"], avg["activity"])
plt.title("Actividad promedio de 'telomerasa' regulada por PID-like controller")
plt.xlabel("Divisin / paso de tiempo")
plt.ylabel("Actividad (0..1)")
plt.grid(True)
plt.show()
# Guardar CSV con resultados (descargable en Colab)
df.to_csv("/content/ivc_simulation_results.csv", index=False)
print("\nCSV guardado en: /content/ivc_simulation_results.csv")
```

```python
# Simulacin vectorizada (2000 clulas, 1000 divisiones)
# - Guardamos promedios por paso y el estado final por clula (para histogramas/violin)
# - Cdigo optimizado para memoria (no guardamos historial por clula en cada paso)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import exp
from dataclasses import dataclass
np.random.seed(42)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def simulate_vectorized(num_cells=2000, divisions=1000,
                        base_shorten=0.03,
                        noise_std=0.005,
                        max_elong_per_activity=0.12,
                        damage_accum_per_div=0.01,
                        phagocytosis_rate=0.008,
                        setpoint=1.0,
                        pid_params=(0.8, 0.03, 0.15),
                        activity_safety_cap=0.9):
    Kp, Ki, Kd = pid_params
    # Inicializar arrays por clula
    telomere = np.random.normal(setpoint, 0.05, size=num_cells)
    damage = np.random.uniform(0, 0.02, size=num_cells)
    activity = np.full(num_cells, 0.05)
    integral = np.zeros(num_cells)
    prev_error = np.zeros(num_cells)
    # Historial de promedios por paso
    steps = np.arange(divisions)
    avg_tel = np.empty(divisions)
    avg_dmg = np.empty(divisions)
    avg_act = np.empty(divisions)
    avg_risk = np.empty(divisions)
    for t in range(divisions):
        measured = telomere + np.random.normal(0, noise_std, size=num_cells)
        error = setpoint - measured
        integral += error
        derivative = error - prev_error
        prev_error = error.copy()
        control = Kp*error + Ki*integral + Kd*derivative
        raw_activity = sigmoid(control) * activity_safety_cap
        # inercia en el cambio de actividad
        activity += 0.2 * (raw_activity - activity)
        # Dinmica telmero
        elong = activity * max_elong_per_activity
        loss = base_shorten + np.random.normal(0, noise_std, size=num_cells)
        telomere += (elong - loss)
        telomere = np.clip(telomere, 0.05, None)
        # Dao
        damage += damage_accum_per_div * (1 - activity*0.2)
        damage -= phagocytosis_rate * (0.5 + 0.5*activity)
        damage = np.clip(damage, 0.0, None)
        # Riesgo (heurstico)
        risk = 1 / (1 + np.exp(- (damage - 0.25) * 8.0))
        # Guardar promedios
        avg_tel[t] = telomere.mean()
        avg_dmg[t] = damage.mean()
        avg_act[t] = activity.mean()
        avg_risk[t] = risk.mean()
    # DataFrames de salida
    df_avg = pd.DataFrame({
        "step": steps,
        "avg_telomere": avg_tel,
        "avg_damage": avg_dmg,
        "avg_activity": avg_act,
        "avg_risk": avg_risk
    })
    df_final = pd.DataFrame({
        "cell_id": np.arange(num_cells),
        "final_telomere": telomere,
        "final_damage": damage,
        "final_activity": activity,
        "final_risk": risk
    })
    return df_avg, df_final
# Ejecutar simulacin con 2000 clulas
df_avg, df_final = simulate_vectorized(num_cells=2000, divisions=1000)
# Guardar CSVs
csv_avg_path = "/content/ivc_simulation_2000_avg.csv"
csv_final_path = "/content/ivc_simulation_2000_final.csv"
df_avg.to_csv(csv_avg_path, index=False)
df_final.to_csv(csv_final_path, index=False)
# Mostrar resumen final (promedios ltimos pasos)
last = df_avg.tail(1).rename(columns={
    "avg_telomere":"avg_telomere_last",
    "avg_damage":"avg_damage_last",
    "avg_activity":"avg_activity_last",
    "avg_risk":"avg_risk_last"
})
try:
    import caas_jupyter_tools as cjt
    cjt.display_dataframe_to_user("Resumen final (promedios)", last)
except Exception:
    display(last)
# Graficar promedios por paso
plt.figure(figsize=(10,4))
plt.plot(df_avg["step"], df_avg["avg_telomere"])
plt.title("Telmero promedio por paso (2000 clulas)")
plt.xlabel("Divisin / paso de tiempo")
plt.ylabel("Telmero (unidades abstractas)")
plt.grid(True)
plt.show()
plt.figure(figsize=(10,4))
plt.plot(df_avg["step"], df_avg["avg_damage"])
plt.title("Dao promedio por clula por paso (2000 clulas)")
plt.xlabel("Divisin / paso de tiempo")
plt.ylabel("Dao (unidades abstractas)")
plt.grid(True)
plt.show()
plt.figure(figsize=(10,4))
plt.plot(df_avg["step"], df_avg["avg_activity"])
plt.title("Actividad promedio de 'telomerasa' por paso (2000 clulas)")
plt.xlabel("Divisin / paso de tiempo")
plt.ylabel("Actividad (0..1)")
plt.grid(True)
plt.show()
# Distribuciones finales: histogramas y violin plots
plt.figure(figsize=(8,4))
plt.hist(df_final["final_telomere"], bins=60)
plt.title("Histograma - Telmero final (2000 clulas)")
plt.xlabel("Telmero (unidades abstractas)")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
plt.figure(figsize=(8,4))
plt.hist(df_final["final_damage"], bins=60)
plt.title("Histograma - Dao final (2000 clulas)")
plt.xlabel("Dao (unidades abstractas)")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
plt.figure(figsize=(8,4))
plt.violinplot(df_final["final_activity"], showmeans=True)
plt.title("Violin plot - Actividad final de 'telomerasa' (2000 clulas)")
plt.xlabel("Distribucin de actividad (cada punto = clula)")
plt.grid(True)
plt.show()
# Guardar un CSV combinado pequeo (final por clula + medias finales)
combined_path = "/content/ivc_simulation_2000_combined_summary.csv"
combined = df_final.copy()
combined["avg_telomere_last"] = float(df_avg["avg_telomere"].iloc[-1])
combined["avg_damage_last"] = float(df_avg["avg_damage"].iloc[-1])
combined["avg_activity_last"] = float(df_avg["avg_activity"].iloc[-1])
combined.to_csv(combined_path, index=False)
print(f"\nCSV promedios: {csv_avg_path}")
print(f"CSV final por clula: {csv_final_path}")
print(f"CSV combinado resumen: {combined_path}")
```

```python
# Ejecutando 3 escenarios alternativos (comparativa)
# Escenarios:
# A - setpoint aumentado a 1.2 (objetivo ms alto)
# B - phagocytosis_rate reducido a 0.003 (peor limpieza)
# C - ruido de sensor aumentado (noise_std=0.02) simulando fallas de medicin
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(42)
def simulate_vectorized(num_cells=2000, divisions=1000,
                        base_shorten=0.03,
                        noise_std=0.005,
                        max_elong_per_activity=0.12,
                        damage_accum_per_div=0.01,
                        phagocytosis_rate=0.008,
                        setpoint=1.0,
                        pid_params=(0.8, 0.03, 0.15),
                        activity_safety_cap=0.9):
    Kp, Ki, Kd = pid_params
    telomere = np.random.normal(setpoint, 0.05, size=num_cells)
    damage = np.random.uniform(0, 0.02, size=num_cells)
    activity = np.full(num_cells, 0.05)
    integral = np.zeros(num_cells)
    prev_error = np.zeros(num_cells)
    steps = np.arange(divisions)
    avg_tel = np.empty(divisions)
    avg_dmg = np.empty(divisions)
    avg_act = np.empty(divisions)
    avg_risk = np.empty(divisions)
    for t in range(divisions):
        measured = telomere + np.random.normal(0, noise_std, size=num_cells)
        error = setpoint - measured
        integral += error
        derivative = error - prev_error
        prev_error = error.copy()
        control = Kp*error + Ki*integral + Kd*derivative
        raw_activity = 1/(1+np.exp(-control)) * activity_safety_cap
        activity += 0.2 * (raw_activity - activity)
        elong = activity * max_elong_per_activity
        loss = base_shorten + np.random.normal(0, noise_std, size=num_cells)
        telomere += (elong - loss)
        telomere = np.clip(telomere, 0.05, None)
        damage += damage_accum_per_div * (1 - activity*0.2)
        damage -= phagocytosis_rate * (0.5 + 0.5*activity)
        damage = np.clip(damage, 0.0, None)
        risk = 1 / (1 + np.exp(- (damage - 0.25) * 8.0))
        avg_tel[t] = telomere.mean()
        avg_dmg[t] = damage.mean()
        avg_act[t] = activity.mean()
        avg_risk[t] = risk.mean()
    df_avg = pd.DataFrame({"step": steps, "avg_telomere": avg_tel, "avg_damage": avg_dmg, "avg_activity": avg_act, "avg_risk": avg_risk})
    df_final = pd.DataFrame({"cell_id": np.arange(num_cells), "final_telomere": telomere, "final_damage": damage, "final_activity": activity, "final_risk": risk})
    return df_avg, df_final
# Run scenarios
scenarios = {
    "A_setpoint_1.2": {"setpoint":1.2},
    "B_low_phago_0.003": {"phagocytosis_rate":0.003},
    "C_high_noise_0.02": {"noise_std":0.02}
}
results = {}
for name, params in scenarios.items():
    cfg = {"num_cells":2000, "divisions":1000}
    cfg.update(params)
    df_avg_s, df_final_s = simulate_vectorized(**cfg)
    results[name] = {"avg": df_avg_s, "final": df_final_s}
    # save CSVs
    df_avg_s.to_csv(f"/content/{name}_avg.csv", index=False)
    df_final_s.to_csv(f"/content/{name}_final.csv", index=False)
# Plot comparative average telomere and damage
plt.figure(figsize=(10,4))
for name, r in results.items():
    plt.plot(r["avg"]["step"], r["avg"]["avg_telomere"], label=name)
plt.title("Comparativa - Telmero promedio por paso (escenarios)")
plt.xlabel("Divisin / paso de tiempo")
plt.ylabel("Telmero (unidades abstractas)")
plt.legend()
plt.grid(True)
plt.show()
plt.figure(figsize=(10,4))
for name, r in results.items():
    plt.plot(r["avg"]["step"], r["avg"]["avg_damage"], label=name)
plt.title("Comparativa - Dao promedio por paso (escenarios)")
plt.xlabel("Divisin / paso de tiempo")
plt.ylabel("Dao (unidades abstractas)")
plt.legend()
plt.grid(True)
plt.show()
# Show final histograms of telomere for each scenario (subplot)
plt.figure(figsize=(12,4))
for i,(name,r) in enumerate(results.items(),1):
    plt.subplot(1,3,i)
    plt.hist(r["final"]["final_telomere"], bins=40)
    plt.title(name)
    plt.xlabel("Telmero final")
    plt.ylabel("Frecuencia")
plt.tight_layout()
plt.show()
# Print paths of saved CSVs
for name in results.keys():
    print(f"/content/{name}_avg.csv  |  /content/{name}_final.csv")
```
