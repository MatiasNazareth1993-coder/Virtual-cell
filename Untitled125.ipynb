```python
# Prototype virtual-cell simulator + deep learning (monitoring + PPO controller)
# This notebook provides a runnable, **purely in-silico** system:
# - TelomereEnv: gym-like environment modeling telomere length, damage, stemness, debris, metabolism
# - MonitoringNet: LSTM model that predicts next-step telomere length and damage probability
# - ActorCritic + PPO-style training: trains a policy to control virtual telomerase & phagocytosis
# - Demo run and simple plots
#
# Notes:
# - This is a compact, educational prototype. You'll want to scale models, hyperparams, and
#   vectorize environments for real experiments.
# - No biological wet-lab instructions anywhere; all units are simulation units.
#
# Run this cell to train a short example policy (kept small so it runs quickly in the notebook).
# After execution you will see training logs and plots.
import os, math, random, tempfile, json, time
from dataclasses import dataclass
from typing import Tuple, Dict, Any, List
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.distributions import Normal, Categorical
from torch.utils.tensorboard import SummaryWriter
import matplotlib.pyplot as plt

# --- Environment ---
class TelomereEnv:
    """Simple stochastic telomere simulator (vectorized single-cell).

    State vector: [L, D, S, M, A, B]
      - L: telomere length (float)
      - D: damage score (0..1)
      - S: stemness (0..1)
      - M: metabolic health (0..1)
      - A: age / division count (int as float)
      - B: debris load (0..1)

    Actions:
      - a_tel: continuous [-1,1] -> telomerase intensity (negative = suppress, positive = extend)
      - a_phag: continuous [-1,1] -> phagocytosis intensity (positive = cleanup)
    Reward:
      - encourages L to stay near target window and low D/B, penalizes excessive telomerase use and runaway divisions.
    """
    def __init__(self, seed=None, fast_mode=False):
        self.rng = np.random.RandomState(seed)
        self.t = 0
        self.max_steps = 200 if not fast_mode else 80
        self.reset()

        # dynamics hyperparams (tunable / domain randomized during training optionally)
        self.params = {
            "shortening_mean": 0.05,   # L units lost per division event baseline
            "shortening_noise": 0.02,
            "damage_rate_base": 0.01,  # baseline per-step damage prob scaler
            "repair_efficacy": 0.6,    # how effective repair actions are
            "telomerase_scale": 0.2,   # how much L changes per telomerase intensity unit
            "phagocytosis_scale": 0.4, # how much debris decreases per phag intensity unit
            "damage_to_debris": 0.3,   # scale converting new damage to debris load
            "metabolic_decay": 0.003,  # metabolic health declines slowly under stress
        }

    def seed(self, s):
        self.rng = np.random.RandomState(s)

    def reset(self, randomize=False):
        # Initialize state with some variation
        self.t = 0
        if randomize and self.rng is not None:
            L = self.rng.uniform(5.0, 10.0)
            D = self.rng.uniform(0.0, 0.05)
            S = self.rng.uniform(0.2, 1.0)
            M = self.rng.uniform(0.8, 1.0)
            B = self.rng.uniform(0.0, 0.05)
            A = 0.0
        else:
            L, D, S, M, B, A = 8.0, 0.02, 0.8, 0.95, 0.01, 0.0
        self.state = np.array([L, D, S, M, A, B], dtype=np.float32)
        return self._get_obs()

    def _get_obs(self):
        return self.state.copy()

    def step(self, action: np.ndarray):
        # action: 2-dim continuous [-1,1]
        a_tel = float(np.clip(action[0], -1.0, 1.0))
        a_phag = float(np.clip(action[1], -1.0, 1.0))
        L, D, S, M, A, B = self.state
        p = self.params

        # Simulate division event probability (higher S and M increases division chance)
        div_prob = max(0.01, 0.2 * S * M)
        division_happened = self.rng.rand() < div_prob

        # Telomere shortening on division
        if division_happened:
            short = self.rng.normal(p["shortening_mean"], p["shortening_noise"])
            L -= max(0.0, short)

        # Telomerase action: extend L scaled by telomerase_scale and by S (more effect in stem-like cells)
        tel_effect = a_tel * p["telomerase_scale"] * (0.5 + 0.5 * S)
        L += tel_effect
        L = max(0.0, L)  # clamp

        # Damage accrual depends on metabolic health and random bursts
        damage_prob = p["damage_rate_base"] * (1.0 - M) * (1.0 + D * 2.0)
        # occasional oxidative burst:
        if self.rng.rand() < 0.01:
            damage_prob += 0.2
        new_damage = self.rng.rand() < damage_prob
        if new_damage:
            D += self.rng.uniform(0.02, 0.12)

        # Repair via telomerase as well (repair efficacy reduces damage)
        repair_strength = max(0.0, a_tel) * p["repair_efficacy"]
        D = max(0.0, D - repair_strength * 0.05)

        # Phagocytosis reduces debris B
        phag_effect = max(0.0, a_phag) * p["phagocytosis_scale"]
        B = max(0.0, B - phag_effect * 0.1)

        # convert some damage into debris
        if new_damage:
            B += p["damage_to_debris"] * D * 0.1

        # metabolic health declines slowly with damage & high telomerase overuse
        M = M - (D * 0.002) - (max(0.0, a_tel) * 0.0008)
        M = np.clip(M, 0.0, 1.0)

        # age increases when divisions happen
        if division_happened:
            A += 1.0

        # Update state
        self.state = np.array([L, D, S, M, A, B], dtype=np.float32)
        self.t += 1

        # Reward shaping: encourage L in window [6,10], low D and low B, discourage excessive telomerase > 0.8
        r_L = -((L - 8.0) ** 2) / 4.0  # peak at L=8
        r_D = - (D * 5.0)
        r_B = - (B * 3.0)
        r_tel_pen = - max(0.0, a_tel - 0.8) * 2.0  # penalize very high telomerase intensity
        reward = r_L + r_D + r_B + r_tel_pen

        # Done condition: end when too many steps or telomere collapsed or metabolic failure
        done = False
        info = {"division": division_happened}
        if self.t >= self.max_steps:
            done = True
        if L <= 0.0 or M <= 0.01:
            done = True
            reward -= 10.0  # catastrophic penalty

        return self._get_obs(), float(reward), done, info

# --- Simple monitoring model ---
class MonitoringNet(nn.Module):
    """LSTM-based predictor: given a short history, predict next telomere length and damage score."""
    def __init__(self, input_dim=6, hidden=64, n_layers=1, out_dim=2):
        super().__init__()
        self.lstm = nn.LSTM(input_dim, hidden, num_layers=n_layers, batch_first=True)
        self.head = nn.Sequential(
            nn.Linear(hidden, hidden//2),
            nn.ReLU(),
            nn.Linear(hidden//2, out_dim)
        )

    def forward(self, x):  # x: (B, T, input_dim)
        out, _ = self.lstm(x)
        h = out[:, -1, :]
        return self.head(h)  # predicts [L_next, D_next]

# --- Actor-Critic for continuous 2-dim action ---
class ActorCritic(nn.Module):
    def __init__(self, obs_dim=6, hidden=128, action_dim=2):
        super().__init__()
        self.shared = nn.Sequential(
            nn.Linear(obs_dim, hidden),
            nn.ReLU(),
            nn.Linear(hidden, hidden),
            nn.ReLU(),
        )
        # actor outputs mean and logstd for each action dim (Gaussian)
        self.mu = nn.Linear(hidden, action_dim)
        self.logstd = nn.Parameter(torch.zeros(action_dim))
        # critic
        self.v = nn.Linear(hidden, 1)

    def forward(self, x):
        h = self.shared(x)
        mu = self.mu(h)
        std = torch.exp(self.logstd)
        v = self.v(h).squeeze(-1)
        return mu, std, v

# PPO utilities (lightweight)
@dataclass
class RolloutBatch:
    obs: List[np.ndarray]
    actions: List[np.ndarray]
    rewards: List[float]
    dones: List[bool]
    values: List[float]
    logps: List[float]

def compute_gae(rewards, values, dones, gamma=0.99, lam=0.95):
    # rewards, values are lists per time step; values has last value appended externally
    adv = []
    gae = 0.0
    for i in reversed(range(len(rewards))):
        mask = 0.0 if dones[i] else 1.0
        delta = rewards[i] + gamma * values[i+1] * mask - values[i]
        gae = delta + gamma * lam * mask * gae
        adv.insert(0, gae)
    return adv

# --- Training loop ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# Hyperparams (small for demo)
env = TelomereEnv(seed=42, fast_mode=True)
obs_dim = 6
action_dim = 2
ac = ActorCritic(obs_dim=obs_dim, hidden=128, action_dim=action_dim).to(device)
monitor = MonitoringNet(input_dim=obs_dim, hidden=64).to(device)

ac_opt = optim.Adam(ac.parameters(), lr=3e-4, weight_decay=1e-5)
mon_opt = optim.Adam(monitor.parameters(), lr=1e-3)

writer = SummaryWriter(log_dir="./tb_logs_virtual_cell_demo")

def select_action(ac_model, obs_np):
    obs = torch.tensor(obs_np, dtype=torch.float32, device=device).unsqueeze(0)
    mu, std, v = ac_model(obs)
    dist = Normal(mu, std)
    a = dist.sample()
    logp = dist.log_prob(a).sum(axis=-1)
    return a.squeeze(0).cpu().numpy(), float(logp.item()), float(v.item())

# Short training for demo
n_epochs = 120
steps_per_epoch = 12  # small episodes for demo
gamma = 0.99
lam = 0.95
clip_eps = 0.2

for epoch in range(n_epochs):
    # Collect trajectories
    batch = RolloutBatch([], [], [], [], [], [])
    for ep in range(steps_per_epoch):
        obs_hist = []
        acts = []
        rewards = []
        dones = []
        values = []
        logps = []

        obs = env.reset(randomize=True)
        # warm-up history for monitoring model (use repeated obs)
        hist_buffer = [obs.copy() for _ in range(8)]
        done = False
        ep_reward = 0.0
        while not done and len(rewards) < 40:
            obs_hist.append(obs.copy())
            # policy uses current obs only (could be extended to history)
            action, logp, val = select_action(ac, obs)
            next_obs, r, done, info = env.step(action)
            acts.append(action.copy())
            rewards.append(r)
            dones.append(done)
            values.append(val)
            logps.append(logp)
            obs = next_obs
            ep_reward += r

        # append last value
        # estimate last value with critic
        obs_t = torch.tensor(obs, dtype=torch.float32, device=device).unsqueeze(0)
        with torch.no_grad():
            _, _, last_v = ac(obs_t)
        values.append(float(last_v.item()))

        # extend batch lists
        batch.obs.extend(obs_hist)
        batch.actions.extend(acts)
        batch.rewards.extend(rewards)
        batch.dones.extend(dones)
        batch.values.extend(values)
        batch.logps.extend(logps)

        writer.add_scalar("episode/reward", ep_reward, epoch * steps_per_epoch + ep)

    # Convert batch to arrays
    obs_arr = np.stack(batch.obs).astype(np.float32)
    actions_arr = np.stack(batch.actions).astype(np.float32)
    rewards_arr = batch.rewards
    dones_arr = batch.dones
    values_arr = batch.values  # values length = T+1 per episode aggregated sequentially; we'll compute GAE per-episode later

    # For simplicity compute advantages per-episode by splitting at dones
    advantages = []
    returns = []
    idx = 0
    # iterate through collected trajectories (we don't store episode boundaries separately, reconstruct by dones)
    # find episodes by splitting where dones True
    episode_indices = []
    for i, d in enumerate(dones_arr):
        if d:
            episode_indices.append(i)

    # --- FIX START ---
    # If no episode completed (i.e., 'done' was never True) within the collected batch,
    # treat the entire batch as a single, continuous segment for GAE calculation.
    # This ensures that 'advantages' and 'returns' are always populated.
    if not episode_indices and len(rewards_arr) > 0:
        episode_indices.append(len(rewards_arr) - 1) # Mark the end of the batch as an effective episode end
    # --- FIX END ---

    start = 0
    ep_ptr = 0
    for end in episode_indices:
        r_slice = rewards_arr[start:end+1]
        v_slice = values_arr[start: end+2]  # values has +1
        d_slice = dones_arr[start:end+1]
        adv = compute_gae(r_slice, v_slice, d_slice, gamma=gamma, lam=lam)
        ret = [a + b for a,b in zip(adv, v_slice[:-1])]
        advantages.extend(adv)
        returns.extend(ret)
        start = end+1
        ep_ptr += 1

    # Convert to tensors
    obs_tensor = torch.tensor(obs_arr, dtype=torch.float32, device=device)
    actions_tensor = torch.tensor(actions_arr, dtype=torch.float32, device=device)
    advantages_tensor = torch.tensor(advantages, dtype=torch.float32, device=device)
    returns_tensor = torch.tensor(returns, dtype=torch.float32, device=device)
    old_logps_tensor = torch.tensor(batch.logps, dtype=torch.float32, device=device)

    # Normalize advantages
    advantages_tensor = (advantages_tensor - advantages_tensor.mean()) / (advantages_tensor.std() + 1e-8)

    # PPO update (few epochs, small minibatch)
    ppo_epochs = 6
    batch_size = max(32, len(advantages_tensor))
    for _ in range(ppo_epochs):
        mu, std, v = ac(obs_tensor)
        dist = Normal(mu, std)
        logps = dist.log_prob(actions_tensor).sum(axis=-1)
        ratio = torch.exp(logps - old_logps_tensor)

        clip_adv = torch.clamp(ratio, 1.0 - clip_eps, 1.0 + clip_eps) * advantages_tensor
        policy_loss = -torch.min(ratio * advantages_tensor, clip_adv).mean()
        value_loss = ((v - returns_tensor) ** 2).mean()
        entropy = dist.entropy().sum(axis=-1).mean()

        loss = policy_loss + 0.5 * value_loss - 0.01 * entropy
        ac_opt.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(ac.parameters(), 0.5)
        ac_opt.step()

    writer.add_scalar("train/policy_loss", policy_loss.item(), epoch)
    writer.add_scalar("train/value_loss", value_loss.item(), epoch)
    writer.add_scalar("train/entropy", entropy.item(), epoch)

    # Train monitoring model briefly on small windows sampled from collected observations
    # create sequences of length T=8
    T = 8
    seqs = []
    targets = []
    if len(obs_arr) >= T + 1:
        for i in range(0, len(obs_arr) - T):
            seqs.append(obs_arr[i:i+T])
            # target next L and D
            next_state = obs_arr[i+T]
            targets.append([next_state[0], next_state[1]])
        seqs = np.stack(seqs).astype(np.float32)
        targets = np.stack(targets).astype(np.float32)
        # train for a few steps
        mon_dataset_idx = list(range(len(seqs)))
        for _ in range(8):
            np.random.shuffle(mon_dataset_idx)
            for j in range(0, len(mon_dataset_idx), 64):
                batch_idx = mon_dataset_idx[j:j+64]
                xb = torch.tensor(seqs[batch_idx], dtype=torch.float32, device=device)
                yb = torch.tensor(targets[batch_idx], dtype=torch.float32, device=device)
                pred = monitor(xb)
                loss_mon = ((pred - yb) ** 2).mean()
                mon_opt.zero_grad()
                loss_mon.backward()
                mon_opt.step()
        writer.add_scalar("train/mon_loss", loss_mon.item(), epoch)

    if epoch % 10 == 0:
        print(f"Epoch {epoch}/{n_epochs} - policy_loss {policy_loss.item():.4f}, value_loss {value_loss.item():.4f}, mon_loss {loss_mon.item():.4f}")

# --- Demonstration run with trained policy ---
env_demo = TelomereEnv(seed=1234, fast_mode=False)
obs = env_demo.reset(randomize=True)
history = [obs.copy()]
actions = []
rewards = []
for step in range(200):
    with torch.no_grad():
        a, _, _ = select_action(ac, obs)
    obs, r, done, info = env_demo.step(a)
    history.append(obs.copy())
    actions.append(a.copy())
    rewards.append(r)
    if done:
        break

history = np.stack(history)
# Save small artifacts
os.makedirs("./models", exist_ok=True)
torch.save(ac.state_dict(), "./models/actor_critic.pth")
torch.save(monitor.state_dict(), "./models/monitor.pth")
np.save("./models/demo_history.npy", history)
with open("./models/metadata.json", "w") as f:
    json.dump({"demo_steps": len(history), "final_state": history[-1].tolist()}, f)

print("Demo finished: steps:", len(history))
print("Final state:", history[-1])

# --- Plots ---
plt.figure(figsize=(9,4))
plt.plot(history[:,0], label="L (telomere length)")
plt.plot(history[:,1], label="D (damage)")
plt.plot(history[:,5], label="B (debris)")
plt.legend()
plt.title("Demo trajectory (telomere env)")
plt.xlabel("step")
plt.grid(True)
plt.tight_layout()
plt.show()

# Provide model & artifact paths to user
print("\nSaved models and artifacts:")
for fn in os.listdir("./models"):
    print(" -", os.path.join("./models", fn))

print("\nTensorBoard logs directory: ./tb_logs_virtual_cell_demo")
print("To view tensorboard run locally: tensorboard --logdir ./tb_logs_virtual_cell_demo")

# End of notebook cell.
```

```python
# Full PyTorch Monitoring Model using refined hyperparameters
# This notebook cell provides a self-contained implementation of:
# - config (fallback if you don't have external file)
# - synthetic virtual-cell sequence generator (in-silico, safe)
# - PyTorch Dataset and DataLoader
# - Transformer-based MonitoringModel (time-series)
# - training loop with basic logging and a small demo run
#
# The model predicts next-step: [telomere_length, damage, debris, stemness, stability_flag]
# Designed to be clear, efficient, and easy to extend into your RL/control stack.
#
# Run this cell to train a short example (kept small to run quickly).
import math, random, time, os
from dataclasses import dataclass
from typing import Tuple, List, Optional
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import matplotlib.pyplot as plt

# ---------- Config (fallback / self-contained) ----------
DEFAULT_CFG = {
    "model": {
        "hidden_dim": 256,
        "num_layers": 3,
        "dropout": 0.12,
        "attention_heads": 4,
        "embedding_dim": 64,
        "sequence_window": 64,   # shorter window for demo speed
        "output_dim": 5,         # [L, D, B, S, stability]
    },
    "training": {
        "learning_rate": 3e-4,
        "weight_decay": 1e-5,
        "batch_size": 32,
        "max_epochs": 12,        # short for demo
        "gradient_clip": 1.0,
        "warmup_steps": 50,
        "device": "cuda" if torch.cuda.is_available() else "cpu",
    },
    "simulation": {
        "BASE_TELOMERE_LENGTH": 10000.0,
        "shortening_per_div": 5.0,
        "telomerase_effect_scale": 3.0,
        "damage_base": 0.001,
        "debris_scale": 0.02,
    }
}

cfg = DEFAULT_CFG

# ---------- Simple synthetic virtual-cell sequence generator ----------
def simulate_virtual_cell_sequence(seq_len: int, seed: Optional[int]=None):
    """Generate a synthetic multivariate time-series with shape (seq_len, 6)
    Features: [L, D, S, M, A, B]
      - L: telomere length (float)
      - D: damage score (0..1)
      - S: stemness (0..1)
      - M: metabolic health (0..1)
      - A: age / division count (float)
      - B: debris load (0..1)
    This is a simple stochastic generator, not biological.
    """
    rng = np.random.RandomState(seed)
    L = cfg["simulation"]["BASE_TELOMERE_LENGTH"] + rng.randn() * 200.0
    D = max(0.0, 0.02 + rng.randn() * 0.01)
    S = min(1.0, max(0.0, 0.8 + rng.randn() * 0.1))
    M = min(1.0, max(0.0, 0.95 + rng.randn() * 0.03))
    A = 0.0
    B = max(0.0, 0.01 + rng.randn() * 0.01)

    seq = []
    for t in range(seq_len):
        # division chance
        div_p = 0.05 * (0.5 + S * 0.5) * (0.5 + M * 0.5)
        if rng.rand() < div_p:
            L -= cfg["simulation"]["shortening_per_div"] + rng.randn() * 0.5
            A += 1.0
        # random oxidative bursts
        if rng.rand() < 0.01:
            D += rng.uniform(0.02, 0.08)
            B += cfg["simulation"]["debris_scale"] * D
        # slow drift
        D = min(1.0, max(0.0, D + rng.randn() * 0.001 - 0.0005))
        S = min(1.0, max(0.0, S - 0.0008 + rng.randn() * 0.0005))
        M = min(1.0, max(0.0, M - D * 0.0002 + rng.randn() * 0.0002))
        B = min(1.0, max(0.0, B + rng.randn() * 0.001 + 0.0001 * D))

        # occasional telomerase-like random extension (simulated external intervention)
        if rng.rand() < 0.03:
            L += cfg["simulation"]["telomerase_effect_scale"] * rng.uniform(1.0, 3.0)

        seq.append([L, D, S, M, A, B])

    return np.array(seq, dtype=np.float32)


# ---------- Dataset ----------
class VirtualCellDataset(Dataset):
    def __init__(self, n_seq: int, seq_len: int, window: int):
        self.n_seq = n_seq
        self.seq_len = seq_len
        self.window = window
        self.data = [simulate_virtual_cell_sequence(seq_len) for _ in range(n_seq)]

    def __len__(self):
        # number of training windows available across sequences
        return sum(max(0, self.seq_len - self.window) for _ in self.data)

    def __getitem__(self, idx):
        # map global idx to sequence and window start
        cum = 0
        for seq in self.data:
            avail = max(0, len(seq) - self.window)
            if idx < cum + avail:
                start = idx - cum
                x = seq[start:start+self.window]
                y = seq[start+self.window]  # predict next step state
                # target: [L, D, B, S, stability_flag]
                L = y[0]
                D = y[1]
                B = y[5]
                S = y[2]
                stability = 1.0 if (L > 5000 and D < 0.5 and B < 0.5) else 0.0
                target = np.array([L, D, B, S, stability], dtype=np.float32)
                return x, target
            cum += avail
        raise IndexError("idx out of range")

# ---------- Model: Transformer-based MonitoringNet ----------
class PositionalEncoding(nn.Module):
    def __init__(self, d_model: int, max_len: int = 5000):
        super().__init__()
        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0)  # (1, max_len, d_model)
        self.register_buffer('pe', pe)

    def forward(self, x):
        # x shape: (B, T, D)
        x = x + self.pe[:, :x.size(1)]
        return x

class MonitoringModel(nn.Module):
    def __init__(self, input_dim=6, cfg_model=None):
        super().__init__()
        if cfg_model is None:
            cfg_model = cfg["model"]
        d_model = cfg_model["embedding_dim"]
        self.input_proj = nn.Linear(input_dim, d_model)
        self.pos_enc = PositionalEncoding(d_model, max_len=cfg_model["sequence_window"])
        encoder_layers = nn.TransformerEncoderLayer(d_model=d_model,
                                                    nhead=cfg_model["attention_heads"],
                                                    dim_feedforward=cfg_model["hidden_dim"],
                                                    dropout=cfg_model["dropout"],
                                                    batch_first=True)
        self.transformer = nn.TransformerEncoder(encoder_layers, num_layers=cfg_model["num_layers"])
        self.decoder = nn.Sequential(
            nn.Linear(d_model, cfg_model["hidden_dim"]//2),
            nn.ReLU(),
            nn.Linear(cfg_model["hidden_dim"]//2, cfg_model["output_dim"])
        )

    def forward(self, x):
        # x: (B, T, input_dim)
        x = self.input_proj(x)  # (B, T, d_model)
        x = self.pos_enc(x)
        x = self.transformer(x)  # (B, T, d_model)
        x = x[:, -1, :]  # take final timestep embedding
        out = self.decoder(x)
        return out  # (B, output_dim)

# ---------- Training utilities ----------
def collate_batch(batch):
    xs = [torch.tensor(b[0], dtype=torch.float32) for b in batch]
    ys = [torch.tensor(b[1], dtype=torch.float32) for b in batch]
    xs = torch.stack(xs)
    ys = torch.stack(ys)
    return xs, ys

def train_monitoring_model(device=None, demo_plot=True):
    device = device or cfg["training"]["device"]
    seq_len = 200
    window = cfg["model"]["sequence_window"]
    dataset = VirtualCellDataset(n_seq=120, seq_len=seq_len, window=window)
    loader = DataLoader(dataset, batch_size=cfg["training"]["batch_size"], shuffle=True, collate_fn=collate_batch)

    model = MonitoringModel(input_dim=6, cfg_model=cfg["model"]).to(device)
    optimizer = optim.AdamW(model.parameters(), lr=cfg["training"]["learning_rate"], weight_decay=cfg["training"]["weight_decay"])
    scaler = torch.cuda.amp.GradScaler() if device.startswith("cuda") else None
    loss_fn = nn.MSELoss()

    model.train()
    losses = []
    for epoch in range(cfg["training"]["max_epochs"]):
        epoch_loss = 0.0
        t0 = time.time()
        for xb, yb in loader:
            xb = xb.to(device)
            yb = yb.to(device)
            optimizer.zero_grad()
            if scaler is not None:
                with torch.cuda.amp.autocast():
                    pred = model(xb)
                    loss = loss_fn(pred, yb)
                scaler.scale(loss).backward()
                scaler.unscale_(optimizer)
                torch.nn.utils.clip_grad_norm_(model.parameters(), cfg["training"]["gradient_clip"])
                scaler.step(optimizer)
                scaler.update()
            else:
                pred = model(xb)
                loss = loss_fn(pred, yb)
                loss.backward()
                torch.nn.utils.clip_grad_norm_(model.parameters(), cfg["training"]["gradient_clip"])
                optimizer.step()
            epoch_loss += float(loss.item()) * xb.size(0)
        epoch_loss /= len(dataset)
        losses.append(epoch_loss)
        t1 = time.time()
        print(f"Epoch {epoch+1}/{cfg['training']['max_epochs']} - loss: {epoch_loss:.6f} - time: {t1-t0:.2f}s")

    # demo evaluation on a fresh sequence
    model.eval()
    with torch.no_grad():
        test_seq = simulate_virtual_cell_sequence(300, seed=999)
        # build sliding windows and predict next-step
        preds = []
        trues = []
        for i in range(len(test_seq) - window):
            x = torch.tensor(test_seq[i:i+window], dtype=torch.float32).unsqueeze(0).to(device)
            y = test_seq[i+window]
            out = model(x).cpu().numpy().squeeze(0)
            preds.append(out)
            L = y[0]; D=y[1]; B=y[5]; S=y[2]; stability = 1.0 if (L>5000 and D<0.5 and B<0.5) else 0.0
            trues.append([L, D, B, S, stability])
        preds = np.array(preds)
        trues = np.array(trues)

    # Plot loss and a few predicted vs true time-series components
    if demo_plot:
        plt.figure(figsize=(8,3))
        plt.plot(losses, label="train_loss")
        plt.xlabel("epoch"); plt.ylabel("MSE"); plt.grid(True); plt.legend(); plt.tight_layout()
        plt.show()

        # plot telomere predictions vs true on first 150 timesteps
        t_plot = min(150, len(preds))
        fig, axs = plt.subplots(2,2, figsize=(10,6))
        axs = axs.flatten()
        axs[0].plot(trues[:t_plot,0], label="true L"); axs[0].plot(preds[:t_plot,0], label="pred L"); axs[0].legend(); axs[0].set_title("Telomere length (L)")
        axs[1].plot(trues[:t_plot,1], label="true D"); axs[1].plot(preds[:t_plot,1], label="pred D"); axs[1].legend(); axs[1].set_title("Damage (D)")
        axs[2].plot(trues[:t_plot,2], label="true B"); axs[2].plot(preds[:t_plot,2], label="pred B"); axs[2].legend(); axs[2].set_title("Debris (B)")
        axs[3].plot(trues[:t_plot,3], label="true S"); axs[3].plot(preds[:t_plot,3], label="pred S"); axs[3].legend(); axs[3].set_title("Stemness (S)")
        plt.tight_layout(); plt.show()

    return model, losses, preds, trues

# ---------- Run training demo ----------
if __name__ == "__main__":
    device = cfg["training"]["device"]
    print("Using device:", device)
    model, losses, preds, trues = train_monitoring_model(device=device, demo_plot=True)
    # Save model
    os.makedirs("models", exist_ok=True)
    torch.save(model.state_dict(), "models/monitoring_model_demo.pth")
    print("Saved model to models/monitoring_model_demo.pth")

```

```python
# virtual_tissue_repair.py
# Purely in-silico tissue repair simulator + monitoring model + PPO-style controller.
# SAFE: No wet-lab instructions. Only simulation code useful for computational research.
#
# Usage:
#   python virtual_tissue_repair.py
#
# Requirements:
#   pip install torch numpy matplotlib

import os
import math
import time
import random
from typing import Tuple, List
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

# ----------------------------
# Configuration / Hyperparams
# ----------------------------
CFG = {
    "seed": 42,
    "device": "cuda" if torch.cuda.is_available() else "cpu",

    # Tissue / simulation
    "n_cells": 64,                 # number of cells in tissue
    "max_steps": 400,              # steps per episode
    "init_telomere": 9000.0,
    "tel_short_per_div": 6.0,
    "telomerase_eff_scale": 3.5,
    "damage_base": 0.001,
    "debris_scale": 0.02,
    "division_base_prob": 0.02,

    # Controller action limits (global budgets)
    "max_telomerase_budget": 1000.0,   # total "telomerase units" available per step
    "max_phagocytosis_budget": 60.0,   # total cleanup units per step

    # RL training
    "policy_hidden": 256,
    "lr_policy": 3e-4,
    "lr_value": 5e-4,
    "ppo_clip": 0.15,
    "ppo_epochs": 6,
    "gamma": 0.98,
    "gae_lambda": 0.95,
    "batch_size": 256,
    "train_epochs": 120,

    # Safety thresholds
    "max_population": 200,  # if tissue cell count exceeds this, heavy penalty
    "target_tel_lower": 7000.0,
    "target_tel_upper": 11000.0,
    "max_safe_divisions_per_step": 12,  # budget per step: avoid mass divisions
}

torch.manual_seed(CFG["seed"])
np.random.seed(CFG["seed"])
random.seed(CFG["seed"])

# ----------------------------
# Tissue Environment
# ----------------------------
class TissueEnv:
    """
    Simulates a tissue composed of N virtual cells.
    Each cell state: [L, D, S, M, A, B]
      L: telomere length (float)
      D: damage (0..1)
      S: stemness (0..1)
      M: metabolic health (0..1)
      A: age/division count (float)
      B: debris (0..1)
    Action (global): 2-dim continuous in [0,1]:
      - a_tel_frac: fraction of max_telomerase_budget to spend on telomerase this step
      - a_phag_frac: fraction of max_phagocytosis_budget to spend on cleanup this step
    The env distributes telomerase budget preferentially to cells that need it (stemness & short L & low damage),
    and distributes phagocytosis to highest-debris cells.
    """
    def __init__(self, cfg):
        self.cfg = cfg
        self.n = cfg["n_cells"]
        self.max_steps = cfg["max_steps"]
        self.step_count = 0
        self.reset()

    def reset(self, randomize=True):
        self.step_count = 0
        # Initialize N cells
        baseL = self.cfg["init_telomere"]
        self.cells = np.zeros((self.n, 6), dtype=np.float32)
        for i in range(self.n):
            L = baseL + np.random.randn() * 200.0
            D = max(0.0, 0.03 + np.random.randn() * 0.01)
            S = min(1.0, max(0.0, 0.6 + np.random.randn() * 0.2))
            M = min(1.0, max(0.0, 0.95 + np.random.randn() * 0.03))
            A = 0.0
            B = max(0.0, 0.01 + np.random.randn() * 0.01)
            self.cells[i] = np.array([L, D, S, M, A, B], dtype=np.float32)

        # tissue-level metrics
        self.population = self.n
        obs = self._get_obs()
        return obs

    def _get_obs(self):
        # Observations: per-cell summary aggregated + global stats
        # For efficiency, return aggregated stats: means and variances of core states
        L = self.cells[:,0]
        D = self.cells[:,1]
        S = self.cells[:,2]
        M = self.cells[:,3]
        B = self.cells[:,5]
        obs = np.concatenate([
            [L.mean(), L.std(), D.mean(), D.std(), S.mean(), S.std(),
             M.mean(), M.std(), B.mean(), B.std(), float(self.population)]
        ])
        return obs.astype(np.float32)

    def step(self, action: np.ndarray) -> Tuple[np.ndarray, float, bool, dict]:
        """
        action: [a_tel_frac, a_phag_frac] in [-1,1] (clipped to 0..1)
        returns obs, reward, done, info
        """
        self.step_count += 1
        a_tel_frac = float(np.clip(action[0], 0.0, 1.0))
        a_phag_frac = float(np.clip(action[1], 0.0, 1.0))

        # compute budgets
        tel_budget = a_tel_frac * self.cfg["max_telomerase_budget"]
        phag_budget = a_phag_frac * self.cfg["max_phagocytosis_budget"]

        # Priority lists: compute need scores for telomerase and phagocytosis
        L = self.cells[:,0]
        D = self.cells[:,1]
        S = self.cells[:,2]
        M = self.cells[:,3]
        B = self.cells[:,5]

        # Telomerase need score: short telomere, stemness high, low damage favours extension
        need_tel = np.maximum(0.0, (self.cfg["target_tel_lower"] - L) / (self.cfg["target_tel_lower"]))  # >0 if L < lower target
        tel_priority = need_tel * (0.5 + 0.5 * S) * (1.0 - D)  # prefer stem + low damage
        tel_priority = np.clip(tel_priority, 0.0, 1.0)
        if tel_priority.sum() > 0:
            # allocate tel budget proportionally to priority
            alloc = (tel_priority / tel_priority.sum()) * tel_budget
        else:
            alloc = np.zeros_like(tel_priority)

        # Apply telomerase effect as extension and minor repair
        tel_effect = (alloc / (1.0 + alloc)) * self.cfg["telomerase_eff_scale"]  # saturating
        self.cells[:,0] += tel_effect  # extend telomere
        # repair damage slightly proportional to tel effect
        self.cells[:,1] = np.maximum(0.0, self.cells[:,1] - 0.01 * (tel_effect / (1.0 + tel_effect)))

        # Phagocytosis allocation: target highest debris cells
        debris_priority = B.copy()
        if debris_priority.sum() > 0:
            alloc_phag = (debris_priority / debris_priority.sum()) * phag_budget
        else:
            alloc_phag = np.zeros_like(debris_priority)
        # reduce debris (saturating)
        phag_effect = (alloc_phag / (1.0 + alloc_phag)) * 0.5  # per-cell debris reduction
        self.cells[:,5] = np.maximum(0.0, self.cells[:,5] - phag_effect)

        # Now simulate natural dynamics for each cell: divisions, damage accrual, ageing
        divisions = np.zeros(self.n, dtype=np.int32)
        for i in range(self.n):
            # division probability depends on stemness and metabolic health
            div_p = self.cfg["division_base_prob"] * (0.5 + 0.5 * self.cells[i,2]) * (0.5 + 0.5 * self.cells[i,3])
            # reduce division probability if telomere too short or damage high
            if self.cells[i,0] < self.cfg["target_tel_lower"]:
                div_p *= 0.5
            if self.cells[i,1] > 0.5:
                div_p *= 0.4
            # clamp
            div_p = float(np.clip(div_p, 0.0, 0.6))
            if np.random.rand() < div_p:
                divisions[i] = 1
                # perform division: telomere shortening, age increment
                loss = max(0.0, np.random.normal(self.cfg["tel_short_per_div"], 0.8))
                self.cells[i,0] -= loss
                self.cells[i,4] += 1.0  # age/division counter
                # some damage created by division
                self.cells[i,1] = np.minimum(1.0, self.cells[i,1] + 0.01 * loss / 5.0)

            # random oxidative event
            if np.random.rand() < 0.01:
                self.cells[i,1] = min(1.0, self.cells[i,1] + np.random.uniform(0.02, 0.08))
                self.cells[i,5] = min(1.0, self.cells[i,5] + self.cfg["debris_scale"] * self.cells[i,1])

            # metabolic decline with damage
            self.cells[i,3] = np.clip(self.cells[i,3] - 0.0005 * self.cells[i,1], 0.0, 1.0)

            # spontaneous small telomerase-like noise events (repair-like)
            if np.random.rand() < 0.02:
                self.cells[i,0] += np.random.uniform(0.0, 2.0)

        # Potentially generate new cells via division births (simplified: keep fixed population, treat each division as attempt to
        # create a new cell with same parameters; to avoid runaway proliferation we enforce a global cap and probabilistic birth)
        attempted_births = divisions.sum()
        births_allowed = int(max(0, min(self.cfg["max_population"] - self.population, attempted_births)))
        # choose births_allowed cells among divisions to produce new cells
        if births_allowed > 0:
            idxs = np.where(divisions==1)[0]
            np.random.shuffle(idxs)
            for j in idxs[:births_allowed]:
                # create new daughter cell with slight variations
                parent = self.cells[j].copy()
                # daughter has slightly shorter telomere, small reset to age and debris
                daughter = parent.copy()
                daughter[0] = max(0.0, parent[0] - 1.5)  # small inherited shortening
                daughter[4] = 0.0  # new age counter for daughter
                daughter[5] = max(0.0, parent[5] * 0.6)
                # append to cell list if capacity allows
                self.cells = np.vstack([self.cells, daughter.reshape(1,6)])
                self.population += 1

        # If population > max_population impose mortality pressure: remove worst-off cells (highest damage & debris)
        if self.population > self.cfg["max_population"]:
            # determine how many to remove
            over = int(self.population - self.cfg["max_population"])
            # score cells by (damage + debris - stemness) to remove senescent/damaged ones first
            scores = self.cells[:,1] + self.cells[:,5] - 0.5 * self.cells[:,2]
            remove_idx = np.argsort(scores)[-over:]
            self.cells = np.delete(self.cells, remove_idx, axis=0)
            self.population = self.cells.shape[0]

        # Bound state variables
        self.cells[:,0] = np.clip(self.cells[:,0], 0.0, 20000.0)
        self.cells[:,1:] = np.clip(self.cells[:,1:], 0.0, 1.0)

        # compute reward: encourage raising mean telomere toward target, reducing mean damage and debris,
        # maintain stemness pool, discourage excessive births and population growth
        L_mean = self.cells[:,0].mean()
        D_mean = self.cells[:,1].mean()
        B_mean = self.cells[:,5].mean()
        S_mean = self.cells[:,2].mean()

        # reward components
        r_tel = -abs(L_mean - (self.cfg["target_tel_lower"] + self.cfg["target_tel_upper"]) / 2.0) / 1000.0
        r_damage = - (D_mean * 2.0)
        r_debris = - (B_mean * 1.5)
        r_stem = + (S_mean * 0.5)
        # penalty for large population (safety)
        pop_penalty = -2.0 * max(0.0, (self.population - self.n) / max(1.0, self.n))
        # heavy penalty for exceeding max population to prevent runaway replication
        heavy_pop_pen = -50.0 if self.population > self.cfg["max_population"] else 0.0
        # penalty for too many divisions this step (encourages conservative division)
        divs = divisions.sum()
        div_penalty = -1.0 * max(0, divs - self.cfg["max_safe_divisions_per_step"]) / max(1.0, self.cfg["max_safe_divisions_per_step"])

        reward = float(r_tel + r_damage + r_debris + r_stem + pop_penalty + heavy_pop_pen + div_penalty)

        # Additional info
        info = {
            "L_mean": float(L_mean),
            "D_mean": float(D_mean),
            "B_mean": float(B_mean),
            "S_mean": float(S_mean),
            "population": int(self.population),
            "divisions": int(divs),
            "tel_budget_used": float(tel_budget),
            "phag_budget_used": float(phag_budget),
        }

        done = False
        if self.step_count >= self.max_steps:
            done = True
        # catastrophic terminal if tissue collapses (all cells dead/zero meta)
        if self.population == 0 or (self.cells[:,3].mean() < 0.02 and self.cells[:,0].mean() < 100.0):
            done = True
            reward -= 100.0

        obs = self._get_obs()
        return obs, reward, done, info

# -----------------------------------
# Monitoring network (lightweight)
# -----------------------------------
class MonitoringNet(nn.Module):
    """
    Small LSTM-based predictor that consumes aggregated obs windows and predicts next-step tissue stats.
    """
    def __init__(self, input_dim=11, hidden=128, n_layers=1, out_dim=11):
        super().__init__()
        self.lstm = nn.LSTM(input_dim, hidden, num_layers=n_layers, batch_first=True)
        self.fc = nn.Sequential(
            nn.Linear(hidden, hidden//2),
            nn.ReLU(),
            nn.Linear(hidden//2, out_dim)
        )

    def forward(self, x):  # x: (B, T, input_dim)
        out, _ = self.lstm(x)
        h = out[:, -1, :]
        return self.fc(h)


# -----------------------------------
# Actor-Critic policy (global actions)
# -----------------------------------
class ActorCritic(nn.Module):
    def __init__(self, obs_dim, hidden=CFG["policy_hidden"], action_dim=2):
        super().__init__()
        self.shared = nn.Sequential(
            nn.Linear(obs_dim, hidden),
            nn.ReLU(),
            nn.Linear(hidden, hidden),
            nn.ReLU(),
        )
        self.mu = nn.Linear(hidden, action_dim)
        self.logstd = nn.Parameter(torch.zeros(action_dim))
        self.v = nn.Linear(hidden, 1)

    def forward(self, x):
        h = self.shared(x)
        mu = torch.sigmoid(self.mu(h))  # output in (0,1) for our fraction actions
        std = torch.exp(self.logstd)
        v = self.v(h).squeeze(-1)
        return mu, std, v

# ------------------------------
# PPO-style training utilities
# ------------------------------
def compute_gae(rewards, values, dones, gamma=0.99, lam=0.95):
    adv = []
    gae = 0.0
    next_value = values[-1]
    for i in reversed(range(len(rewards))):
        mask = 0.0 if dones[i] else 1.0
        delta = rewards[i] + gamma * next_value * mask - values[i]
        gae = delta + gamma * lam * mask * gae
        adv.insert(0, gae)
        next_value = values[i]
    return adv

# ------------------------------
# Training loop (compact demo)
# ------------------------------
def train():
    device = CFG["device"]
    env = TissueEnv(CFG)
    obs_dim = 11  # size of aggregated observation vector in env._get_obs()
    ac = ActorCritic(obs_dim=obs_dim).to(device)
    monitor = MonitoringNet(input_dim=obs_dim, hidden=128, out_dim=obs_dim).to(device)

    opt_policy = optim.Adam(list(ac.shared.parameters()) + [ac.mu.weight, ac.mu.bias, ac.logstd], lr=CFG["lr_policy"])
    opt_value = optim.Adam(list(ac.v.parameters()) + list(ac.shared.parameters()), lr=CFG["lr_value"])
    opt_mon = optim.Adam(monitor.parameters(), lr=1e-3)

    # Experience buffers
    obs_buf = []
    action_buf = []
    reward_buf = []
    value_buf = []
    logp_buf = []
    done_buf = []

    all_rewards = []
    for epoch in range(CFG["train_epochs"]):
        obs = env.reset()
        # initial value estimate
        obs_t = torch.tensor(obs, dtype=torch.float32, device=device).unsqueeze(0)
        with torch.no_grad():
            mu, std, v = ac(obs_t)
        ep_rewards = 0.0
        done = False

        # collect one episode of rollouts
        while not done:
            # select action
            obs_t = torch.tensor(obs, dtype=torch.float32, device=device).unsqueeze(0)
            mu, std, v = ac(obs_t)
            dist = torch.distributions.Normal(mu, std)
            a = dist.sample().cpu().numpy().squeeze(0)
            # actions expected as fractions in [0,1], mu is sigmoid already; clip
            a = np.clip(a, 0.0, 1.0)
            logp = float(dist.log_prob(torch.tensor(a, device=device)).sum().item())

            next_obs, reward, done, info = env.step(a)
            obs_buf.append(obs.copy())
            action_buf.append(a.copy())
            reward_buf.append(float(reward))
            value_buf.append(float(v.item()))
            logp_buf.append(logp)
            done_buf.append(done)

            obs = next_obs
            ep_rewards += reward

            # small online train of monitoring net using last few obs windows
            # (we simulate sequence of length 8)
            # collect if we have enough obs history
            if len(obs_buf) >= 9:
                seq = np.array(obs_buf[-9:-1], dtype=np.float32)  # shape (8, obs_dim)
                target = np.array(obs_buf[-1], dtype=np.float32)  # predict next obs
                xb = torch.tensor(seq.reshape(1, seq.shape[0], seq.shape[1]), device=device)
                yb = torch.tensor(target.reshape(1, -1), device=device)
                pred = monitor(xb)
                loss_mon = ((pred - yb) ** 2).mean()
                opt_mon.zero_grad()
                loss_mon.backward()
                opt_mon.step()

        all_rewards.append(ep_rewards)

        # finalize value array by appending last value estimate
        obs_t = torch.tensor(obs, dtype=torch.float32, device=device).unsqueeze(0)
        with torch.no_grad():
            _, _, last_v = ac(obs_t)
        value_buf.append(float(last_v.item()))

        # compute advantages & returns
        adv = compute_gae(reward_buf, value_buf, done_buf, gamma=CFG["gamma"], lam=CFG["gae_lambda"])
        returns = [a + v for a, v in zip(adv, value_buf[:-1])]

        # convert buffers to tensors
        obs_tensor = torch.tensor(np.array(obs_buf, dtype=np.float32), device=device)
        actions_tensor = torch.tensor(np.array(action_buf, dtype=np.float32), device=device)
        returns_tensor = torch.tensor(np.array(returns, dtype=np.float32), device=device)
        adv_tensor = torch.tensor(np.array(adv, dtype=np.float32), device=device)
        old_logp = torch.tensor(np.array(logp_buf, dtype=np.float32), device=device)

        # normalize adv
        adv_tensor = (adv_tensor - adv_tensor.mean()) / (adv_tensor.std() + 1e-8)

        # PPO updates
        for _ in range(CFG["ppo_epochs"]):
            mu, std, values = ac(obs_tensor)
            dist = torch.distributions.Normal(mu, std)
            logps = dist.log_prob(actions_tensor).sum(dim=-1)
            ratio = torch.exp(logps - old_logp)
            surr1 = ratio * adv_tensor
            surr2 = torch.clamp(ratio, 1.0 - CFG["ppo_clip"], 1.0 + CFG["ppo_clip"]) * adv_tensor
            policy_loss = -torch.min(surr1, surr2).mean()
            value_loss = ((values - returns_tensor) ** 2).mean()
            entropy = dist.entropy().sum(dim=-1).mean()

            # combined loss
            loss = policy_loss + 0.5 * value_loss - 0.01 * entropy

            opt_policy.zero_grad()
            opt_value.zero_grad()
            loss.backward()
            nn.utils.clip_grad_norm_(ac.parameters(), 0.5)
            opt_policy.step()
            opt_value.step()

        # clear buffers
        obs_buf.clear(); action_buf.clear(); reward_buf.clear(); value_buf.clear(); logp_buf.clear(); done_buf.clear()

        # logging
        avg_reward = np.mean(all_rewards[-20:]) if len(all_rewards) >= 1 else 0.0
        print(f"Epoch {epoch+1}/{CFG['train_epochs']}: ep_reward {ep_rewards:.2f}, avg_recent_reward {avg_reward:.2f}, population {env.population}")

        # occasional save
        if (epoch+1) % 20 == 0:
            os.makedirs("models", exist_ok=True)
            torch.save(ac.state_dict(), f"models/ac_epoch_{epoch+1}.pth")
            torch.save(monitor.state_dict(), f"models/monitor_epoch_{epoch+1}.pth")

    # final save
    os.makedirs("models", exist_ok=True)
    torch.save(ac.state_dict(), "models/ac_final.pth")
    torch.save(monitor.state_dict(), "models/monitor_final.pth")
    print("Training complete. Models saved in ./models")
    return ac, monitor, all_rewards

# ------------------------------
# Demo run and visualization
# ------------------------------
def demo_run(ac_model, steps=300):
    device = CFG["device"]
    env = TissueEnv(CFG)
    obs = env.reset()
    history = []
    for t in range(steps):
        obs_t = torch.tensor(obs, dtype=torch.float32, device=device).unsqueeze(0)
        with torch.no_grad():
            mu, std, v = ac_model(obs_t)
            a = mu.cpu().numpy().squeeze(0)
        a = np.clip(a, 0.0, 1.0)
        obs, r, done, info = env.step(a)
        history.append([info["L_mean"], info["D_mean"], info["B_mean"], info["S_mean"], info["population"], info["divisions"]])
        if done:
            break
    history = np.array(history)
    # plots
    fig, axs = plt.subplots(2,2, figsize=(10,6))
    axs = axs.flatten()
    axs[0].plot(history[:,0]); axs[0].set_title("Mean Telomere Length")
    axs[1].plot(history[:,1]); axs[1].set_title("Mean Damage")
    axs[2].plot(history[:,2]); axs[2].set_title("Mean Debris")
    axs[3].plot(history[:,4]); axs[3].set_title("Population")
    plt.tight_layout(); plt.show()

if __name__ == "__main__":
    ac_model, monitor, rewards = train()
    demo_run(ac_model, steps=300)
```
